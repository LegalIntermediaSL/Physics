# Laboratorio Especializado: Nuclear Especializado 2

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def bateman_equations(t, N):
    l1, l2, l3 = 0.5, 0.2, 0.05
    dN1 = -l1 * N[0]
    dN2 = l1 * N[0] - l2 * N[1]
    dN3 = l2 * N[1] - l3 * N[2]
    return [dN1, dN2, dN3]

if __name__ == "__main__":
    sol = solve_ivp(bateman_equations, [0, 20], [100.0, 0.0, 0.0])
    plt.plot(sol.t, sol.y[0], label='N1')
    plt.plot(sol.t, sol.y[1], label='N2')
    plt.plot(sol.t, sol.y[2], label='N3')
    plt.legend()
    plt.title('Radioactive Decay Chain')
    plt.savefig('decay_chain.png')
```
