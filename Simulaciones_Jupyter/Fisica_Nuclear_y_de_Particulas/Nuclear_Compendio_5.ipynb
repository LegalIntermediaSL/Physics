# Compendio Subtemático: Nuclear Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    r = np.linspace(1, 10, 100)
    V_coulomb = 10 / r # Example Coulomb potential
    E_alpha = np.full_like(r, 2.5) # Alpha particle energy
    
    plt.plot(r, V_coulomb, label='Coulomb Barrier', color='orange')
    plt.plot(r, E_alpha, label='Alpha Particle Energy', linestyle='--', color='red')
    plt.fill_between(r, E_alpha, V_coulomb, where=(V_coulomb > E_alpha), color='gray', alpha=0.3, label='Tunneling Region')
    plt.title('Alpha Decay (Quantum Tunneling)')
    plt.xlabel('Distance r')
    plt.ylabel('Potential V(r)')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
