# Fisica Nuclear Galactica 15

Simulación matemática generada paramétricamente para el compendio galáctico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# Physics simulation 15
t = np.linspace(0, 10, 100)
plt.plot(np.cos(t*15), np.sin(t*15))
plt.title('Galactic Nuclear Physics 15')
```
