# Fisica Nuclear Galactica 21

Simulación matemática generada paramétricamente para el compendio galáctico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# Physics simulation 21
t = np.linspace(0, 10, 100)
plt.plot(np.cos(t*21), np.sin(t*21))
plt.title('Galactic Nuclear Physics 21')
```
