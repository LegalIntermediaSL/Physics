# Compendio Subtemático: Nuclear Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 100)
    lambda_decay = 0.5
    R = 100 # Production rate
    # Activity buildup
    A_t = R * (1 - np.exp(-lambda_decay * t))
    
    plt.plot(t, A_t, color='indigo')
    plt.title('Neutron Activation (Activity Buildup)')
    plt.xlabel('Irradiation Time')
    plt.ylabel('Activity')
    plt.axhline(R, color='r', linestyle='--', label='Saturation Activity')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
