# Compendio Subtemático: Nuclear Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Random walk moderation energy loss
    steps = 20
    energy = [2.0] # Initial energy in MeV
    for _ in range(steps):
        # Average log energy decrement
        energy.append(energy[-1] * np.random.uniform(0.3, 0.9))
        
    plt.plot(range(steps+1), energy, marker='o', color='brown')
    plt.yscale('log')
    plt.title('Neutron Moderation (Energy Loss per Collision)')
    plt.xlabel('Collision Number')
    plt.ylabel('Neutron Energy (MeV)')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
