# Compendio Subtemático: Nuclear Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    levels = [0, 1.33, 2.5] # Example energy levels in MeV
    plt.hlines(levels, 0, 1, colors='k')
    
    plt.annotate('', xy=(0.5, 1.33), xytext=(0.5, 2.5), arrowprops=dict(arrowstyle="->", color='r'))
    plt.annotate('', xy=(0.3, 0), xytext=(0.3, 1.33), arrowprops=dict(arrowstyle="->", color='b'))
    
    plt.title('Gamma Ray Emission (Nuclear Energy Levels)')
    plt.ylabel('Energy (MeV)')
    plt.xticks([])
    plt.xlim(0, 1)
    plt.show()

if __name__ == '__main__':
    main()
```
