# Compendio Subtemático: Nuclear Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    Z = np.arange(1, 100)
    # Valley of stability approximation N ~ Z + 0.006 * Z^2
    N_stable = Z + 0.006 * Z**2
    
    plt.plot(Z, Z, 'k--', label='N = Z')
    plt.plot(Z, N_stable, 'g-', lw=2, label='Valley of Stability')
    plt.title('Chart of Nuclides (N vs Z)')
    plt.xlabel('Proton Number (Z)')
    plt.ylabel('Neutron Number (N)')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
