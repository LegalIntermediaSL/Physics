# Compendio Subtemático: Nuclear Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    theta = np.linspace(0, np.pi, 100)
    E0 = 0.511 # Photon energy in MeV (equal to electron rest mass for illustration)
    m_e_c2 = 0.511
    
    E_prime = E0 / (1 + (E0 / m_e_c2) * (1 - np.cos(theta)))
    
    plt.plot(np.degrees(theta), E_prime, color='blue')
    plt.title('Compton Scattering (Scattered Photon Energy)')
    plt.xlabel('Scattering Angle (Degrees)')
    plt.ylabel("Scattered Energy E' (MeV)")
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
