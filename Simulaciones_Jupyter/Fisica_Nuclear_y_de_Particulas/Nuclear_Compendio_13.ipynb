# Compendio Subtemático: Nuclear Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def main():
    mu = 15 # Average counts per interval
    x = np.arange(0, 30)
    prob = poisson.pmf(x, mu)
    
    plt.bar(x, prob, color='orange', alpha=0.7)
    plt.title('Geiger-Muller Counter Statistics (Poisson)')
    plt.xlabel('Counts per Interval')
    plt.ylabel('Probability')
    plt.grid(axis='y')
    plt.show()

if __name__ == '__main__':
    main()
```
