# Laboratorio Especializado: Nuclear Especializado 4

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def wkb_tunneling_prob(E, V_func, a, b):
    # WKB approximation integral
    x = np.linspace(a, b, 1000)
    dx = x[1] - x[0]
    integral = np.sum(np.sqrt(np.abs(V_func(x) - E))) * dx
    return np.exp(-2 * integral)

def coulomb_potential(r):
    return 1.0 / r

if __name__ == "__main__":
    E = 0.5
    prob = wkb_tunneling_prob(E, coulomb_potential, 2.0, 10.0)
    print(f"Alpha decay tunneling probability: {prob:.4e}")
```
