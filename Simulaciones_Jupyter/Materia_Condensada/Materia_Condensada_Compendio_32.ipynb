# Compendio Subtemático: Materia Condensada Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Electron in a Magnetic Field (Cyclotron Motion)
t = np.linspace(0, 20, 1000)
B = 1.0 # Magnetic field
m = 1.0 # Mass
q = 1.0 # Charge
v_perp = 1.0 # Velocity perpendicular to B
v_parallel = 0.2 # Velocity parallel to B (z-axis)

# Cyclotron frequency
omega_c = q * B / m

# Trajectory
x = (v_perp / omega_c) * np.sin(omega_c * t)
y = (v_perp / omega_c) * (1 - np.cos(omega_c * t))
z = v_parallel * t

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='Electron Trajectory', color='blue', lw=2)
ax.set_title("Cyclotron Motion in a Magnetic Field")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z (Parallel to B)')
plt.legend()
plt.savefig("cyclotron_motion.png")
```
