# Compendio Subtemático: Materia Condensada Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Random Walk Model for Diffusion (1D)
steps = 1000
num_particles = 5

plt.figure(figsize=(8, 5))
for i in range(num_particles):
    # Random steps: +1 or -1
    walk = np.random.choice([-1, 1], size=steps)
    position = np.cumsum(walk)
    plt.plot(position, lw=1, alpha=0.8, label=f'Particle {i+1}')

# Theoretical expected standard deviation: sqrt(N)
std_dev = np.sqrt(np.arange(steps))
plt.plot(std_dev, color='black', lw=2, linestyle='--', label='Expected $\pm \sqrt{N}$')
plt.plot(-std_dev, color='black', lw=2, linestyle='--')

plt.title("1D Random Walk (Brownian Motion Model)")
plt.xlabel("Time Step N")
plt.ylabel("Position x")
plt.legend()
plt.grid(True)
plt.savefig("random_walk.png")
```
