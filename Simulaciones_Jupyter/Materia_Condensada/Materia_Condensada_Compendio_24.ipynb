Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Spin waves (Magnons) in a 1D Ferromagnet
k = np.linspace(-np.pi, np.pi, 200)
J = 1.0 # Exchange interaction (J > 0 for Ferromagnet)
S = 1/2 # Spin

# Magnon dispersion for Ferromagnet: E(k) = 2 J S (1 - cos(ka))
E_k = 2 * J * S * (1 - np.cos(k))

plt.figure(figsize=(8, 5))
plt.plot(k, E_k, lw=2, color='firebrick')
plt.title("Magnon Dispersion in 1D Ferromagnet")
plt.xlabel("Wavevector $k a$")
plt.ylabel("Energy $E(k)$")
plt.xticks([-np.pi, 0, np.pi], [r'$-\pi$', '0', r'$\pi$'])
plt.grid(True)
plt.savefig("ferro_spin_waves.png")
```
