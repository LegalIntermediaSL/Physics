# Compendio Subtemático: Materia Condensada Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# BCS Superconducting Gap (Approximate temperature dependence)
T_Tc = np.linspace(0, 1, 100)
# Approximation: \Delta(T) / \Delta(0) \approx 1.74 \sqrt{1 - T/T_c} near Tc
# For full range, a numerical solution is typically used, but we use an empirical fit here:
delta_T = np.tanh(1.74 * np.sqrt(1.0 / T_Tc - 1.0))
delta_T[0] = 1.0 # fix limit

plt.figure(figsize=(8, 5))
plt.plot(T_Tc, delta_T, lw=2, color='navy')
plt.title("BCS Theory: Superconducting Gap vs Temperature")
plt.xlabel(r"Reduced Temperature $T/T_c$")
plt.ylabel(r"Reduced Gap $\Delta(T)/\Delta(0)$")
plt.xlim(0, 1)
plt.ylim(0, 1.1)
plt.grid(True)
plt.savefig("bcs_gap.png")
```
