# Compendio Subtemático: Materia Condensada Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Spin waves (Magnons) in a 1D Antiferromagnet
k = np.linspace(-np.pi, np.pi, 200)
J = 1.0 # Exchange interaction (J > 0 for Antiferromagnet magnitude)
S = 1/2 # Spin

# Magnon dispersion for Antiferromagnet: E(k) = 2 J S |sin(ka)|
E_k = 2 * J * S * np.abs(np.sin(k))

plt.figure(figsize=(8, 5))
plt.plot(k, E_k, lw=2, color='navy')
plt.title("Magnon Dispersion in 1D Antiferromagnet")
plt.xlabel("Wavevector $k a$")
plt.ylabel("Energy $E(k)$")
plt.xticks([-np.pi, 0, np.pi], [r'$-\pi$', '0', r'$\pi$'])
plt.grid(True)
plt.savefig("antiferro_spin_waves.png")
```
