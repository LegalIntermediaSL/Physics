# Compendio Subtemático: Materia Condensada Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Density of States (DOS) for 1D, 2D, 3D free electron gas
E = np.linspace(0.1, 10, 200)

dos_1d = 1 / np.sqrt(E)
dos_2d = np.ones_like(E)
dos_3d = np.sqrt(E)

plt.figure(figsize=(10, 6))
plt.plot(E, dos_1d, label='1D ($E^{-1/2}$)', lw=2)
plt.plot(E, dos_2d, label='2D (Constant)', lw=2)
plt.plot(E, dos_3d, label='3D ($E^{1/2}$)', lw=2)
plt.title("Density of States for Free Electron Gas")
plt.xlabel("Energy E")
plt.ylabel("Density of States g(E)")
plt.ylim(0, 4)
plt.legend()
plt.grid(True)
plt.savefig("dos_123.png")
```
