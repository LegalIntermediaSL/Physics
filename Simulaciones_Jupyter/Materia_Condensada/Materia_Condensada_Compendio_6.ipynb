# Compendio Subtemático: Materia Condensada Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Kronig-Penney Model - Graphical Solution
def kronig_penney(P, alpha_a):
    return P * np.sin(alpha_a) / alpha_a + np.cos(alpha_a)

alpha_a = np.linspace(0.01, 4 * np.pi, 500)
P_val = 3.0 * np.pi / 2.0

f_alpha = kronig_penney(P_val, alpha_a)

plt.figure(figsize=(8, 5))
plt.plot(alpha_a, f_alpha, lw=2, color='b', label='f($\\alpha a$)')
plt.axhline(1, color='r', linestyle='--')
plt.axhline(-1, color='r', linestyle='--')
plt.fill_between(alpha_a, -1, 1, color='gray', alpha=0.2, label='Allowed bands')
plt.ylim(-3, 3)
plt.title("Kronig-Penney Model")
plt.xlabel(r'$\alpha a$')
plt.ylabel('f($\\alpha a$)')
plt.legend()
plt.grid(True)
plt.savefig("kronig_penney.png")
```
