# Compendio Subtemático: Materia Condensada Compendio 22

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Peierls Transition: 1D chain distortion opening a gap
k = np.linspace(-np.pi, np.pi, 300)
t = 1.0 # hopping
delta = 0.2 # distortion parameter

# Undistorted bands
E_undistorted = -2 * t * np.cos(k)

# Distorted bands (folding the Brillouin zone)
# Gap opens at k = +/- pi/2
E_lower = -np.sqrt((2 * t * np.cos(k))**2 + delta**2)
E_upper = np.sqrt((2 * t * np.cos(k))**2 + delta**2)

plt.figure(figsize=(8, 5))
plt.plot(k, E_undistorted, '--', color='gray', label='Undistorted Chain')
plt.plot(k, E_lower, lw=2, color='red', label='Lower Band (Distorted)')
plt.plot(k, E_upper, lw=2, color='blue', label='Upper Band (Distorted)')
plt.axvline(-np.pi/2, color='k', linestyle=':', label='Zone Boundary')
plt.axvline(np.pi/2, color='k', linestyle=':')
plt.title("Peierls Transition: Band Gap Opening")
plt.xlabel("Wavevector $k$")
plt.ylabel("Energy $E(k)$")
plt.legend()
plt.grid(True)
plt.savefig("peierls_transition.png")
```
