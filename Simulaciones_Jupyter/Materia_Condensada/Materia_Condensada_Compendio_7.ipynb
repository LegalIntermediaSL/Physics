# Compendio Subtemático: Materia Condensada Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Hall effect simulation: Hall voltage vs Magnetic field
B = np.linspace(-2.0, 2.0, 100) # Magnetic field in Tesla
I = 1.0 # Current in Amps
n = 8.5e28 # Carrier density (e.g. Copper)
e = 1.602e-19
d = 1e-3 # Thickness in meters

V_H = (I * B) / (n * e * d)

plt.figure(figsize=(8, 5))
plt.plot(B, V_H * 1e6, lw=2, color='orange')
plt.title("Hall Effect: Hall Voltage vs Magnetic Field")
plt.xlabel("Magnetic Field B (T)")
plt.ylabel(r"Hall Voltage $V_H$ ($\mu$V)")
plt.grid(True)
plt.savefig("hall_effect.png")
```
