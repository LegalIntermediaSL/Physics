# Compendio Subtemático: Materia Condensada Compendio 26

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Coupled Quantum Harmonic Oscillator Array (1D)
# Models phonons or excitations in a crystalline lattice
k = np.linspace(-np.pi, np.pi, 200)
omega_0 = 1.0 # Natural frequency of each oscillator
K_coupling = 0.5 # Coupling strength

# Dispersion relation: omega^2 = omega_0^2 + 2 * K_coupling * (1 - cos(ka))
omega_k = np.sqrt(omega_0**2 + 2 * K_coupling * (1 - np.cos(k)))

plt.figure(figsize=(8, 5))
plt.plot(k, omega_k, lw=2, color='darkgreen')
plt.title("Coupled Quantum Harmonic Oscillators: Dispersion")
plt.xlabel("Wavevector $k a$")
plt.ylabel(r"Frequency $\omega(k)$")
plt.grid(True)
plt.savefig("coupled_qho.png")
```
