# Estructura de Bandas (Kronig-Penney)

Resolución de electrones en potenciales cristalinos periódicos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

k = np.linspace(-3, 3, 400)
E1 = k**2
E2 = (k-2)**2
E3 = (k+2)**2

plt.figure(figsize=(8,5))
plt.plot(k, E1, 'b-', lw=1)
plt.plot(k, E2, 'b-', lw=1)
plt.plot(k, E3, 'b-', lw=1)
plt.ylim(0, 10)
plt.title('Diagrama de Bandas Parabólico (Zona Extendida)')
plt.xlabel('Vector de Onda k')
plt.ylabel('Energía E')
plt.grid()
plt.show()
```
