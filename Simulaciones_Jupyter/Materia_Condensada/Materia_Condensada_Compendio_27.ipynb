# Compendio Subtemático: Materia Condensada Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Mott Insulator Transition (Hubbard Model - Simplified Mean Field View)
U_t = np.linspace(0, 10, 200) # U/t ratio (Coulomb repulsion / hopping)

# Simplified density of states gap opening
# Below critical U/t, it's a metal (gap = 0)
# Above critical U/t, gap opens linearly with U
Uc_t = 3.0 # Critical ratio
gap = np.maximum(0, U_t - Uc_t)

plt.figure(figsize=(8, 5))
plt.plot(U_t, gap, lw=2, color='purple')
plt.axvline(Uc_t, color='r', linestyle='--', label=f'Mott Transition ($U/t = {Uc_t}$)')
plt.fill_between(U_t, 0, gap, color='purple', alpha=0.2)
plt.title("Mott Insulator Transition: Charge Gap vs Interaction")
plt.xlabel("Interaction Strength $U/t$")
plt.ylabel(r"Charge Gap $\Delta$")
plt.text(1.5, 2, "Metal", fontsize=12, ha='center')
plt.text(7, 2, "Mott Insulator", fontsize=12, ha='center')
plt.legend()
plt.grid(True)
plt.savefig("mott_insulator.png")
```
