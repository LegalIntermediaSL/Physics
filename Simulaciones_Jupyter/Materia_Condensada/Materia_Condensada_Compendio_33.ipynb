# Compendio Subtemático: Materia Condensada Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Anderson Localization (1D Tight Binding with Disorder)
N = 100 # Number of sites
t = 1.0 # Hopping parameter
W = 2.5 # Disorder strength (width of uniform distribution)

# Construct Hamiltonian
np.random.seed(42)
onsite_energies = np.random.uniform(-W/2, W/2, N)
H = np.diag(onsite_energies)

# Add hopping (off-diagonal elements)
for i in range(N - 1):
    H[i, i+1] = -t
    H[i+1, i] = -t

# Diagonalize
eigvals, eigvecs = np.linalg.eigh(H)

# Plot the probability density of a state near the center of the band
target_state_idx = N // 2
prob_density = np.abs(eigvecs[:, target_state_idx])**2

plt.figure(figsize=(10, 4))
plt.bar(range(N), prob_density, color='darkred')
plt.title(f"Anderson Localization in 1D (W/t = {W/t})")
plt.xlabel("Lattice Site Index")
plt.ylabel(r"Probability Density $|\psi|^2$")
plt.grid(True, alpha=0.3)
plt.savefig("anderson_localization.png")
```
