# Laboratorio Especializado: Materia Condensada Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def ising_model():
    L = 50
    T = 2.269 # Critical temperature for 2D Ising
    J = 1.0
    steps = 100000
    
    spins = np.random.choice([-1, 1], size=(L, L))
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(spins, cmap='coolwarm')
    axes[0].set_title('Initial State')
    axes[0].axis('off')
    
    magnetization = []
    
    for step in range(steps):
        i, j = np.random.randint(0, L, size=2)
        S = spins[i, j]
        neighbors = spins[(i+1)%L, j] + spins[(i-1)%L, j] + spins[i, (j+1)%L] + spins[i, (j-1)%L]
        dE = 2 * J * S * neighbors
        
        if dE < 0 or np.random.rand() < np.exp(-dE / T):
            spins[i, j] *= -1
            
        if step % 1000 == 0:
            magnetization.append(np.abs(np.mean(spins)))
            
        if step == steps // 2:
            axes[1].imshow(spins, cmap='coolwarm')
            axes[1].set_title('Intermediate State')
            axes[1].axis('off')
            
    axes[2].imshow(spins, cmap='coolwarm')
    axes[2].set_title('Final State (near $T_c$)')
    axes[2].axis('off')
    
    plt.tight_layout()
    plt.savefig('ising_2d.png')
    
    plt.figure()
    plt.plot(magnetization)
    plt.title('Absolute Magnetization over Time')
    plt.xlabel('MC Steps (x1000)')
    plt.ylabel('|M|')
    plt.savefig('ising_mag.png')

if __name__ == "__main__":
    ising_model()
```
