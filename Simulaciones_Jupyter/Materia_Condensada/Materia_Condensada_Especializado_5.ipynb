# Laboratorio Especializado: Materia Condensada Especializado 5

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft2, ifft2

def gross_pitaevskii():
    N = 128
    L = 10.0
    dx = L / N
    x = np.linspace(-L/2, L/2, N)
    y = np.linspace(-L/2, L/2, N)
    X, Y = np.meshgrid(x, y)
    
    # Harmonic trap potential
    V = 0.5 * (X**2 + Y**2)
    
    # Initial wavefunction
    psi = np.exp(-(X**2 + Y**2) / 2) + 0.0j
    psi /= np.sqrt(np.sum(np.abs(psi)**2) * dx**2)
    
    # Interaction parameter
    g = 100.0
    
    # Imaginary time evolution for ground state
    dt = -1j * 0.01 
    steps = 1000
    
    kx = 2 * np.pi * np.fft.fftfreq(N, d=dx)
    ky = 2 * np.pi * np.fft.fftfreq(N, d=dx)
    KX, KY = np.meshgrid(kx, ky)
    K2 = KX**2 + KY**2
    
    for _ in range(steps):
        # Half step in real space (potential + interaction)
        psi *= np.exp(-1j * (V + g * np.abs(psi)**2) * dt / 2)
        
        # Full step in momentum space (kinetic)
        psi_k = fft2(psi)
        psi_k *= np.exp(-1j * (K2 / 2) * dt)
        psi = ifft2(psi_k)
        
        # Half step in real space
        psi *= np.exp(-1j * (V + g * np.abs(psi)**2) * dt / 2)
        
        # Normalize (imaginary time evolution reduces norm)
        psi /= np.sqrt(np.sum(np.abs(psi)**2) * dx**2)
        
    plt.figure(figsize=(8, 6))
    plt.contourf(X, Y, np.abs(psi)**2, cmap='viridis', levels=50)
    plt.colorbar(label='Density $|\psi|^2$')
    plt.title('Bose-Einstein Condensate Ground State (GPE)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.savefig('bec_ground_state.png')

if __name__ == "__main__":
    gross_pitaevskii()
```
