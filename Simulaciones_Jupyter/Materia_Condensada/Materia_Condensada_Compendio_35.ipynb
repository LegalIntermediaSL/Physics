# Compendio Subtemático: Materia Condensada Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Wigner Crystal (2D Triangular Lattice of Electrons)
# At very low densities, electrons crystallize to minimize Coulomb repulsion.

def generate_triangular(n_rows, n_cols):
    points = []
    a = 1.0
    for row in range(n_rows):
        for col in range(n_cols):
            x = col * a + (0.5 * a if row % 2 != 0 else 0)
            y = row * a * np.sqrt(3)/2
            points.append([x, y])
    return np.array(points)

pts = generate_triangular(10, 10)

plt.figure(figsize=(8, 8))
# Add small random perturbations to simulate zero-point motion or finite temperature
perturb = np.random.normal(0, 0.05, pts.shape)
pts_perturbed = pts + perturb

plt.scatter(pts_perturbed[:,0], pts_perturbed[:,1], c='cyan', edgecolors='blue', s=100, label='Electrons')
plt.title("2D Wigner Crystal (Triangular Lattice)")
plt.xlabel("X")
plt.ylabel("Y")
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig("wigner_crystal.png")
```
