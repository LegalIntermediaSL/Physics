# Compendio Subtemático: Materia Condensada Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Quantum Dot Energy Levels (3D Particle in a Box)
L = 10e-9 # Box side length (10 nm)
m = 9.11e-31 # Electron mass
h = 6.626e-34 # Planck constant
eV = 1.602e-19 # Joules per eV

def energy(nx, ny, nz):
    E_joules = (h**2 / (8 * m * L**2)) * (nx**2 + ny**2 + nz**2)
    return E_joules / eV

# Calculate first few energy levels
levels = []
for nx in range(1, 4):
    for ny in range(1, 4):
        for nz in range(1, 4):
            levels.append(energy(nx, ny, nz))

levels = sorted(levels)

plt.figure(figsize=(4, 8))
for i, E in enumerate(levels[:15]):
    plt.hlines(E, 0, 1, color='blue', lw=2)
    plt.text(1.05, E, f"{E:.3f} eV", va='center')

plt.xlim(-0.2, 1.5)
plt.xticks([])
plt.ylabel("Energy (eV)")
plt.title("Quantum Dot Energy Levels\n(10 nm Cubic Box)")
plt.grid(axis='y')
plt.savefig("quantum_dot.png")
```
