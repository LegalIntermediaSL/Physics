# Compendio Subtemático: Materia Condensada Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Mean Field Theory of Ferromagnetism
T_vals = np.linspace(0.1, 1.5, 100) # T/Tc
M_vals = []

for T in T_vals:
    if T < 1.0:
        # Solve m = tanh(m/T)
        m = fsolve(lambda m: m - np.tanh(m / T), 1.0)[0]
    else:
        m = 0.0
    M_vals.append(m)

plt.figure(figsize=(8, 5))
plt.plot(T_vals, M_vals, lw=2, color='darkgreen')
plt.axvline(1.0, color='r', linestyle='--', label='$T_c$')
plt.title("Mean Field Theory: Magnetization vs Temperature")
plt.xlabel(r"Reduced Temperature $T/T_c$")
plt.ylabel("Magnetization $m$")
plt.legend()
plt.grid(True)
plt.savefig("mean_field.png")
```
