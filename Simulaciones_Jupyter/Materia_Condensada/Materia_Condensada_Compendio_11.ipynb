# Compendio Subtemático: Materia Condensada Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 1D Diatomic chain phonon dispersion
a = 1.0 # lattice constant
C = 1.0 # force constant
M1 = 1.0 # mass 1
M2 = 2.0 # mass 2

k = np.linspace(-np.pi/a, np.pi/a, 200)

term1 = C * (M1 + M2) / (M1 * M2)
term2 = C * np.sqrt((M1 - M2)**2 + 4 * M1 * M2 * np.cos(k * a / 2)**2) / (M1 * M2)

omega_optical = np.sqrt(term1 + term2)
omega_acoustic = np.sqrt(term1 - term2)

plt.figure(figsize=(8, 5))
plt.plot(k, omega_optical, lw=2, label="Optical Branch", color='r')
plt.plot(k, omega_acoustic, lw=2, label="Acoustic Branch", color='b')
plt.title("Phonon Dispersion: 1D Diatomic Chain")
plt.xlabel("Wavevector k")
plt.ylabel(r"Angular Frequency $\omega$")
plt.xticks([-np.pi/a, 0, np.pi/a], [r'$-\pi/a$', '0', r'$\pi/a$'])
plt.legend()
plt.grid(True)
plt.savefig("diatomic_phonon.png")
```
