# Supercomputación: Materia Condensada HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

def dft_1d():
    N = 100
    x = np.linspace(-5, 5, N)
    dx = x[1] - x[0]
    
    V = 0.5 * x**2
    T = -0.5 * (np.diag(-2 * np.ones(N)) + np.diag(np.ones(N-1), 1) + np.diag(np.ones(N-1), -1)) / dx**2
    H = T + np.diag(V)
    
    E, psi = la.eigh(H)
    
    plt.plot(x, V, 'k--', label='Potential')
    plt.plot(x, np.abs(psi[:, 0])**2 + E[0], label='Ground State Density')
    plt.title('1D Density Functional Theory (Mock)')
    plt.legend()
    plt.show()

if __name__ == '__main__':
    dft_1d()
```
