# Supercomputación: Materia Condensada HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gp_3d_vortex():
    N = 32
    x = np.linspace(-5, 5, N)
    X, Y, Z = np.meshgrid(x, x, x)
    R = np.sqrt(X**2 + Y**2)
    Theta = np.arctan2(Y, X)
    
    psi = R / np.sqrt(R**2 + 1) * np.exp(1j * Theta) * np.exp(-Z**2 / 10)
    density = np.abs(psi)**2
    
    plt.imshow(density[:, :, N//2], cmap='inferno')
    plt.title('Gross-Pitaevskii Vortex Density Cross-Section')
    plt.show()

if __name__ == '__main__':
    gp_3d_vortex()
```
