# Compendio Subtemático: Materia Condensada Compendio 20

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Tight-binding band structure of Graphene (nearest neighbor)
kx = np.linspace(-np.pi, np.pi, 200)
ky = 0 # cross section along ky = 0
a = 1.0 # lattice constant
t = 2.7 # hopping parameter in eV

# Dispersion relation
def E_graphene(kx, ky):
    f = 1 + 4 * np.cos(kx * a * np.sqrt(3)/2) * np.cos(ky * a / 2) + 4 * np.cos(ky * a / 2)**2
    return t * np.sqrt(f)

E_plus = E_graphene(kx, ky)
E_minus = -E_graphene(kx, ky)

plt.figure(figsize=(8, 5))
plt.plot(kx, E_plus, lw=2, color='black')
plt.plot(kx, E_minus, lw=2, color='black')
plt.title("Graphene Band Structure (Cross-section)")
plt.xlabel(r"$k_x$")
plt.ylabel("Energy (eV)")
plt.grid(True)
plt.savefig("graphene_bands.png")
```
