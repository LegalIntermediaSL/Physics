# Compendio Subtemático: Materia Condensada Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Landau Theory of Second Order Phase Transitions
m = np.linspace(-2.5, 2.5, 200)
b = 1.0 # quartic coefficient

plt.figure(figsize=(8, 5))
# Varying 'a' which is proportional to (T - Tc)
for a in [2.0, 0.0, -1.0, -2.0]:
    f = a * m**2 + b * m**4
    plt.plot(m, f, lw=2, label=f'$a \propto (T-T_c) = {a}$')

plt.title("Landau Free Energy (Second Order Transition)")
plt.xlabel("Order Parameter $m$")
plt.ylabel("Free Energy Density $f(m)$")
plt.ylim(-2, 10)
plt.legend()
plt.grid(True)
plt.savefig("landau_theory.png")
```
