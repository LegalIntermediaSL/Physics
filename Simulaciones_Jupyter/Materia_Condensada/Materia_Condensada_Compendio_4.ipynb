# Compendio Subtemático: Materia Condensada Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def einstein_cv(T, TE):
    x = TE / T
    return 3 * x**2 * np.exp(x) / (np.exp(x) - 1)**2

T_vals = np.linspace(1, 500, 200)
TE = 300 # Einstein temperature in K

cv_vals = [einstein_cv(T, TE) if T > 0 else 0 for T in T_vals]

plt.figure(figsize=(8, 5))
plt.plot(T_vals, cv_vals, 'g-', lw=2)
plt.title("Einstein Model: Specific Heat vs Temperature")
plt.xlabel("Temperature (K)")
plt.ylabel("C_V / N k_B")
plt.axvline(TE, color='r', linestyle='--', label="Einstein Temp")
plt.legend()
plt.grid(True)
plt.savefig("einstein_model.png")
```
