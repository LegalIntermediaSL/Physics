# Compendio Subtemático: Materia Condensada Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Sommerfeld free electron model - Fermi energy vs density
def fermi_energy(n):
    hbar = 1.054e-34
    m_e = 9.11e-31
    return (hbar**2 / (2 * m_e)) * (3 * np.pi**2 * n)**(2/3)

densities = np.logspace(27, 29, 100) # electrons per m^3
E_F_joules = fermi_energy(densities)
E_F_eV = E_F_joules / 1.602e-19

plt.figure(figsize=(8, 5))
plt.semilogx(densities, E_F_eV, lw=2, color='blue')
plt.title("Sommerfeld Model: Fermi Energy vs Electron Density")
plt.xlabel("Electron Density $n$ (m$^{-3}$)")
plt.ylabel("Fermi Energy (eV)")
plt.grid(True, which="both", ls="--")
plt.savefig("sommerfeld_model.png")
```
