# Supercomputación: Fisica Atomica HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la

# 1D TDDFT toy model (LDA) for 2 particles
N, L = 500, 10.0
x = np.linspace(-L, L, N)
dx = x[1] - x[0]
V_ext = 0.5 * x**2  # Harmonic trap

T = -0.5 / dx**2 * (np.diag(-2 * np.ones(N)) + np.diag(np.ones(N-1), 1) + np.diag(np.ones(N-1), -1))

def lda_potential(n):
    return - (3.0 / np.pi)**(1.0/3.0) * n**(1.0/3.0)

n = np.exp(-x**2)
n /= np.sum(n) * dx

for i in range(50):
    V_hartree = np.zeros(N) # simplistic hartree
    V_xc = lda_potential(n)
    H = T + np.diag(V_ext + V_hartree + V_xc)
    evals, evecs = la.eigh(H)
    psi = evecs[:, 0]
    psi /= np.sqrt(np.sum(np.abs(psi)**2) * dx)
    n = 2.0 * np.abs(psi)**2  # 2 particles

plt.plot(x, n)
plt.title('TDDFT Ground State Density')
plt.savefig('tddft_density.png')
```
