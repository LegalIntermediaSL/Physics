# Supercomputación: Materia Condensada HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def ginzburg_landau():
    N = 50
    x = np.linspace(-10, 10, N)
    X, Y = np.meshgrid(x, x)
    
    psi = np.tanh(np.sqrt(X**2 + Y**2) / 2.0)
    
    plt.contourf(X, Y, psi, levels=20, cmap='magma')
    plt.title('Ginzburg-Landau Order Parameter (Superconducting Vortex)')
    plt.colorbar()
    plt.show()

if __name__ == '__main__':
    ginzburg_landau()
```
