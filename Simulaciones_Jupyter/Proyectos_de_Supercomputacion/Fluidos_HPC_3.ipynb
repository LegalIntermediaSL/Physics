# Supercomputación: Fluidos HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def shallow_water(Nx=50, Ny=50, steps=100):
    h = np.ones((Nx, Ny))
    h[20:30, 20:30] = 2.0
    u = np.zeros((Nx, Ny))
    v = np.zeros((Nx, Ny))
    g = 9.81
    dt, dx = 0.01, 1.0
    
    for _ in range(steps):
        h[1:-1, 1:-1] -= dt * (h[1:-1, 1:-1] * (u[2:, 1:-1] - u[:-2, 1:-1])/(2*dx) + h[1:-1, 1:-1] * (v[1:-1, 2:] - v[1:-1, :-2])/(2*dx))
        u[1:-1, 1:-1] -= dt * g * (h[2:, 1:-1] - h[:-2, 1:-1])/(2*dx)
        v[1:-1, 1:-1] -= dt * g * (h[1:-1, 2:] - h[1:-1, :-2])/(2*dx)
        
    plt.imshow(h, cmap='ocean')
    plt.title('Shallow Water Equations')
    plt.show()

if __name__ == '__main__':
    shallow_water()
```
