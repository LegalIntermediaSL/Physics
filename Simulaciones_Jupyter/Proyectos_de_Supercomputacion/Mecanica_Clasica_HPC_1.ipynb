# Supercomputación: Mecanica Clasica HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def accel(pos, mass, G=1.0, softening=0.1):
    x, y, z = pos[:, 0], pos[:, 1], pos[:, 2]
    dx = x.reshape(-1, 1) - x.reshape(1, -1)
    dy = y.reshape(-1, 1) - y.reshape(1, -1)
    dz = z.reshape(-1, 1) - z.reshape(1, -1)
    inv_r3 = (dx**2 + dy**2 + dz**2 + softening**2)**(-1.5)
    ax = G * (dx * inv_r3) @ mass
    ay = G * (dy * inv_r3) @ mass
    az = G * (dz * inv_r3) @ mass
    return np.vstack((ax, ay, az)).T

def simulate(N=100, tEnd=10.0, dt=0.01):
    mass = 20.0 * np.ones(N) / N
    pos = np.random.randn(N, 3)
    vel = np.random.randn(N, 3)
    
    t = 0
    while t < tEnd:
        vel += accel(pos, mass) * dt / 2
        pos += vel * dt
        vel += accel(pos, mass) * dt / 2
        t += dt
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(pos[:,0], pos[:,1], pos[:,2], s=mass*10)
    plt.savefig("nbody_final.png")
    
if __name__ == "__main__":
    simulate()
```
