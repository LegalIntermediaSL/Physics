# Supercomputación: Materia Condensada HPC 2

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def hubbard_qmc():
    N_sites = 16
    N_tau = 20
    U = 4.0
    
    # Hubbard-Stratonovich auxiliary field
    hs_field = np.random.choice([-1, 1], size=(N_sites, N_tau))
    
    plt.imshow(hs_field, cmap='Greys', aspect='auto')
    plt.title('Hubbard Model Auxiliary Field Configuration')
    plt.ylabel('Sites')
    plt.xlabel('Imaginary Time')
    plt.show()

if __name__ == '__main__':
    hubbard_qmc()
```
