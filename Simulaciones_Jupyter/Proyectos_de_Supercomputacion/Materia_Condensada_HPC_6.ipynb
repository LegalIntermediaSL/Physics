# Supercomputación: Materia Condensada HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def graphene_nanoribbon():
    k = np.linspace(-np.pi, np.pi, 100)
    t = 2.7
    
    E1 = t * np.sqrt(1 + 4*np.cos(k/2)**2 + 4*np.cos(k/2))
    E2 = -E1
    
    plt.plot(k, E1, 'k')
    plt.plot(k, E2, 'k')
    plt.title('Graphene Nanoribbon Band Structure (Tight-Binding)')
    plt.xlabel('Wave vector k')
    plt.ylabel('Energy (eV)')
    plt.show()

if __name__ == '__main__':
    graphene_nanoribbon()
```
