# Supercomputación: Optica HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Metasurface Phase Profile (Metalens)
N = 1000
x = np.linspace(-50, 50, N)
f = 100.0
lam = 1.0

# hyperboloidal phase profile
phase = 2 * np.pi / lam * (f - np.sqrt(x**2 + f**2))
phase = np.mod(phase, 2*np.pi)

# diffraction via angular spectrum method
E = np.exp(1j * phase)
E_f = np.fft.fftshift(np.fft.fft(E))
fx = np.fft.fftshift(np.fft.fftfreq(N, d=(x[1]-x[0])))

H = np.exp(1j * 2 * np.pi / lam * f * np.sqrt(1 - (lam * fx)**2 + 0j))
E_focus = np.fft.ifft(np.fft.ifftshift(E_f * H))

plt.plot(x, np.abs(E_focus)**2)
plt.title('Metalens Focal Spot')
plt.savefig('metalens.png')
```
