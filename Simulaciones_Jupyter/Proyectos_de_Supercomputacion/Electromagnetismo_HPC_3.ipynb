# Supercomputación: Electromagnetismo HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import spherical_jn, spherical_yn

def simulate():
    wavelengths = np.linspace(300, 800, 100)
    eps_m = 1.0
    radius = 20.0
    
    wp = 1.37e16
    gamma = 1.05e14
    c = 3e8
    
    C_sca = []
    for wl in wavelengths:
        w = 2 * np.pi * c / (wl * 1e-9)
        eps_Au = 1 - wp**2 / (w**2 + 1j * gamma * w)
        
        m = np.sqrt(eps_Au / eps_m)
        x = 2 * np.pi * radius * np.sqrt(eps_m) / wl
        
        j1 = spherical_jn(1, x)
        y1 = spherical_yn(1, x)
        h1 = j1 + 1j*y1
        jm1 = spherical_jn(1, m*x)
        
        a1 = (m**2 * jm1 * (x*j1 + j1) - j1 * (m*x*jm1 + jm1)) / (m**2 * jm1 * (x*h1 + h1) - h1 * (m*x*jm1 + jm1))
        C_sca.append((2 * np.pi / (2 * np.pi / wl)**2) * 3 * np.abs(a1)**2)
        
    plt.plot(wavelengths, C_sca)
    plt.savefig("plasmonics.png")

if __name__ == '__main__': simulate()
```
