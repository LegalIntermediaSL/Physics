# Supercomputación: Mecanica Clasica HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def lj_force(pos, L):
    N = pos.shape[0]
    f = np.zeros_like(pos)
    for i in range(N):
        for j in range(i+1, N):
            dr = pos[i] - pos[j]
            dr = dr - L * np.round(dr / L)
            r2 = np.dot(dr, dr)
            if r2 < 2.5**2:
                r6 = r2**3
                fij = 48 * (1/r6**2 - 0.5/r6) / r2 * dr
                f[i] += fij
                f[j] -= fij
    return f

def simulate():
    N, L = 64, 10.0
    pos = np.random.rand(N, 3) * L
    vel = np.random.randn(N, 3) - 0.5
    f = lj_force(pos, L)
    dt = 0.005
    
    kin = []
    for _ in range(500):
        pos = pos + vel * dt + 0.5 * f * dt**2
        pos = pos % L
        f_new = lj_force(pos, L)
        vel = vel + 0.5 * (f + f_new) * dt
        f = f_new
        kin.append(0.5 * np.sum(vel**2))
        
    plt.plot(kin)
    plt.savefig("md_energy.png")

if __name__ == '__main__': simulate()
```
