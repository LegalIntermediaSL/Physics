# Supercomputación: Mecanica Cuantica HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np

def truncate(rho, m):
    evals, evecs = np.linalg.eigh(rho)
    idx = np.argsort(evals)[::-1][:m]
    return evals[idx], evecs[:, idx]

# Heisenberg model toy DMRG step
Sz = np.array([[0.5, 0], [0, -0.5]])
Sp = np.array([[0, 1], [0, 0]])
Sm = np.array([[0, 0], [1, 0]])
I = np.eye(2)

H_block = np.kron(Sz, Sz) + 0.5 * (np.kron(Sp, Sm) + np.kron(Sm, Sp))
block_op = {'Sz': Sz, 'Sp': Sp, 'Sm': Sm, 'I': I, 'H': H_block}

# Simulate one step of infinite system DMRG
m_states = 4
H_super = np.kron(block_op['H'], np.eye(2)) + np.kron(np.eye(2), block_op['H'])
H_int = np.kron(block_op['Sz'], block_op['Sz']) + 0.5 * (np.kron(block_op['Sp'], block_op['Sm']) + np.kron(block_op['Sm'], block_op['Sp']))
H_super += H_int

evals, evecs = np.linalg.eigh(H_super)
psi0 = evecs[:, 0]

dim_block = H_super.shape[0] // 2
psi0_mat = psi0.reshape((dim_block, -1))
rho_system = psi0_mat @ psi0_mat.T

w, v = truncate(rho_system, m_states)
print("Truncation error:", 1 - np.sum(w))
print("DMRG Toy Step Completed Successfully")
```
