# Supercomputación: Astrofisica HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 0, 5000)
t = t[t < -0.01]
M_c = 1.0
f = (5 / (256 * t * M_c**(5/3)))**(-3/8)
phi = -2 * (1 / (256 * t * M_c**(5/3)))**(5/8)
h = np.cos(phi) / (-t)**0.25

plt.plot(t, h, color='indigo')
plt.title('Gravitational Wave Strain (Binary Inspiral)')
plt.xlabel('Time to merger')
plt.ylabel('Strain h')
plt.show()
```
