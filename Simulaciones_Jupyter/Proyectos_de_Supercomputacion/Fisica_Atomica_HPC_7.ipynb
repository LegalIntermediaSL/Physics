# Supercomputación: Fisica Atomica HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Electromagnetically Induced Transparency (EIT) Density Matrix
rho = np.zeros((3, 3), dtype=complex)
rho[0, 0] = 1.0
gamma_21 = 0.1
gamma_31 = 1.0
Omega_p = 0.01
Omega_c = 1.0
delta_p = np.linspace(-5, 5, 200)
absorption = []

for dp in delta_p:
    H = np.array([[0, -Omega_p/2, 0],
                  [-Omega_p/2, -dp - 1j*gamma_21, -Omega_c/2],
                  [0, -Omega_c/2, -dp - 1j*gamma_31]])
    
    # steady state approximation for rho_21
    chi = Omega_p * (dp + 1j*gamma_31) / (Omega_c**2 / 4 - (dp + 1j*gamma_21)*(dp + 1j*gamma_31))
    absorption.append(np.imag(chi))

plt.plot(delta_p, absorption)
plt.title('EIT Absorption Profile')
plt.savefig('eit_absorption.png')
```
