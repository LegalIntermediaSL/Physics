# Supercomputación: Materia Condensada HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def phase_field():
    N = 100
    phi = np.random.rand(N, N) * 0.1
    phi[N//2-2:N//2+2, N//2-2:N//2+2] = 1.0
    
    plt.imshow(phi, cmap='bone')
    plt.title('Phase Field Dendritic Solidification Seed')
    plt.show()

if __name__ == '__main__':
    phase_field()
```
