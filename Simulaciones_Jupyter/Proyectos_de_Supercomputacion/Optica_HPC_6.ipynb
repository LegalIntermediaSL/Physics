# Supercomputación: Optica HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Rigorous Coupled-Wave Analysis (RCWA) toy model
orders = 5
N = 2*orders + 1
kx = np.arange(-orders, orders+1) * 2 * np.pi / 1.0 # grating period 1um
k0 = 2 * np.pi / 0.5 # lambda 0.5um

eps_m = np.zeros((N, N), dtype=complex)
for i in range(N):
    for j in range(N):
        if i == j: eps_m[i, j] = 2.0
        elif abs(i-j) == 1: eps_m[i, j] = 0.5

Kx = np.diag(kx / k0)
A = Kx**2 - eps_m
evals, evecs = np.linalg.eig(A)

plt.plot(np.real(evals), 'o')
plt.title('RCWA eigenvalues')
plt.savefig('rcwa.png')
```
