# Supercomputación: Optica HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 1D FDTD for Photonic Crystal
N = 1000
steps = 2000
E = np.zeros(N)
H = np.zeros(N)
eps = np.ones(N)

# periodic dielectric structure
for i in range(400, 600):
    if (i // 10) % 2 == 0:
        eps[i] = 4.0

source_pos = 100
signal = []

for t in range(steps):
    E[source_pos] += np.sin(2 * np.pi * t / 20.0)
    
    for i in range(1, N):
        E[i] += (H[i-1] - H[i]) / eps[i]
    
    for i in range(N-1):
        H[i] += E[i] - E[i+1]
        
    signal.append(E[800])

plt.plot(signal)
plt.title('FDTD Transmission')
plt.savefig('fdtd_1d.png')
```
