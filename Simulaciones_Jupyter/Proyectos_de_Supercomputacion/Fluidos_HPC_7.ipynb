# Supercomputación: Fluidos HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def sph_dam_break():
    N = 500
    pos = np.random.rand(N, 2)
    pos[:, 0] *= 0.3
    pos[:, 1] *= 0.6
    
    plt.scatter(pos[:, 0], pos[:, 1], s=5)
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.title('SPH Dam Break Initial Configuration')
    plt.show()

if __name__ == '__main__':
    sph_dam_break()
```
