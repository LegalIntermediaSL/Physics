# Supercomputación: Optica HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Beam Propagation Method (BPM) for a Directional Coupler
N = 500
x = np.linspace(-20, 20, N)
dx = x[1] - x[0]
dz = 0.1
steps = 500
k0 = 2 * np.pi

n_bg = 1.5
n_core = 1.51
n_dist = np.ones(N) * n_bg
n_dist[(x > -5) & (x < -3)] = n_core
n_dist[(x > 3) & (x < 5)] = n_core

E = np.exp(-(x + 4)**2) + 0j
E_hist = []

for i in range(steps):
    phase = np.exp(1j * k0 * (n_dist - n_bg) * dz)
    E = E * phase
    
    E_f = np.fft.fft(E)
    kx = np.fft.fftfreq(N, d=dx) * 2 * np.pi
    prop = np.exp(-1j * kx**2 / (2 * k0 * n_bg) * dz)
    E = np.fft.ifft(E_f * prop)
    
    if i % 10 == 0:
        E_hist.append(np.abs(E)**2)

plt.imshow(np.array(E_hist), aspect='auto')
plt.title('Directional Coupler BPM')
plt.savefig('bpm_coupler.png')
```
