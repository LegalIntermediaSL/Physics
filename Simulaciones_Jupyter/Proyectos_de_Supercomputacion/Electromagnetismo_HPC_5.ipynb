# Supercomputación: Electromagnetismo HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def simulate():
    N, ng = 10000, 64
    L = 4 * np.pi
    dx = L / ng
    dt = 0.1
    
    pos = np.random.rand(N) * L
    vel = np.random.randn(N) + 0.1 * np.sin(2 * np.pi * pos / L)
    
    for _ in range(100):
        rho = np.zeros(ng)
        for p in pos:
            idx = int(p / dx) % ng
            rho[idx] += 1
        rho = rho / (N/ng) - 1.0
        
        phi_hat = np.fft.rfft(rho)
        k = 2 * np.pi * np.fft.rfftfreq(ng, d=dx)
        k[0] = 1.0
        E_hat = -1j * phi_hat / k
        E_hat[0] = 0
        E = np.fft.irfft(E_hat, n=ng)
        
        for i in range(N):
            idx = int(pos[i] / dx) % ng
            vel[i] -= E[idx] * dt
            pos[i] = (pos[i] + vel[i] * dt) % L
            
    plt.hist(vel, bins=50)
    plt.savefig("landau_damping.png")

if __name__ == '__main__': simulate()
```
