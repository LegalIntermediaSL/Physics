# Supercomputación: Mecanica Cuantica HPC 2

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import scipy.sparse as sparse
import scipy.sparse.linalg as spla
import matplotlib.pyplot as plt

N = 500
L = 10.0
x = np.linspace(-L, L, N)
dx = x[1] - x[0]
dt = 0.01
t_max = 5.0

V = 0.5 * x**2
H_kin = sparse.diags([-1, 2, -1], [-1, 0, 1], shape=(N, N)) / (2 * dx**2)
H_pot = sparse.diags([V], [0], shape=(N, N))
H = H_kin + H_pot

I = sparse.eye(N)
U_plus = I + 1j * 0.5 * dt * H
U_minus = I - 1j * 0.5 * dt * H

psi = np.exp(-(x - 2)**2)
psi = psi / np.linalg.norm(psi) * np.sqrt(1/dx)

t = 0
while t < t_max:
    psi = spla.spsolve(U_plus, U_minus.dot(psi))
    t += dt

plt.plot(x, np.abs(psi)**2)
plt.title('Wavepacket in Harmonic Oscillator')
plt.xlabel('x')
plt.ylabel('|psi|^2')
plt.savefig('Mecanica_Cuantica_HPC_2_output.png')
```
