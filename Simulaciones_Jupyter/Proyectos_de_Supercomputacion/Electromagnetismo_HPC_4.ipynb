# Supercomputación: Electromagnetismo HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def simulate():
    theta = np.linspace(0, 2*np.pi, 500)
    beta = 0.95
    gamma = 1 / np.sqrt(1 - beta**2)
    
    dP_dOmega = (1 - beta * np.cos(theta))**(-5) * ((1 - beta * np.cos(theta))**2 - (1 - beta**2) * np.sin(theta)**2)
    
    plt.polar(theta, dP_dOmega)
    plt.savefig("synchrotron.png")

if __name__ == '__main__': simulate()
```
