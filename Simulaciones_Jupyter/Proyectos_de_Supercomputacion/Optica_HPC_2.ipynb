# Supercomputación: Optica HPC 2

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Split-Step Fourier Method for Non-linear Schrodinger Equation (Optical Solitons)
N = 1024
L = 40.0
dz = 0.01
z_steps = 1000
T = np.linspace(-L/2, L/2, N)
omega = np.fft.fftfreq(N, d=L/N) * 2 * np.pi

beta2 = -1.0 # anomalous dispersion
gamma = 1.0  # nonlinearity

U = 1.0 / np.cosh(T)
U_history = []

for i in range(z_steps):
    U = U * np.exp(1j * gamma * np.abs(U)**2 * dz)
    U_f = np.fft.fft(U)
    U_f = U_f * np.exp(-1j * 0.5 * beta2 * omega**2 * dz)
    U = np.fft.ifft(U_f)
    
    if i % 100 == 0:
        U_history.append(np.abs(U)**2)

plt.imshow(np.array(U_history), aspect='auto', extent=[-L/2, L/2, 0, z_steps*dz])
plt.title('Soliton Propagation')
plt.savefig('soliton.png')
```
