# Supercomputación: Electromagnetismo HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def simulate():
    N = 16
    d = 0.5
    theta = np.linspace(-np.pi/2, np.pi/2, 500)
    theta_0 = np.pi / 6
    
    AF = np.zeros_like(theta, dtype=complex)
    for n in range(N):
        phase = 2 * np.pi * d * n * (np.sin(theta) - np.sin(theta_0))
        AF += np.exp(1j * phase)
        
    plt.plot(np.degrees(theta), 20 * np.log10(np.abs(AF)/N))
    plt.ylim(-40, 0)
    plt.savefig("phased_array.png")

if __name__ == '__main__': simulate()
```
