# Supercomputación: Electromagnetismo HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def simulate():
    t = np.linspace(0, 2*np.pi, 100)
    R, r = 2.0, 0.5
    N_coils = 5
    
    x_c = (R + r * np.cos(N_coils * t)) * np.cos(t)
    y_c = (R + r * np.cos(N_coils * t)) * np.sin(t)
    z_c = r * np.sin(N_coils * t)
    
    dlx, dly, dlz = np.gradient(x_c), np.gradient(y_c), np.gradient(z_c)
    
    bx, by, bz = 0, 0, 0
    for i in range(len(t)):
        r2 = x_c[i]**2 + y_c[i]**2 + z_c[i]**2
        r3 = r2**1.5
        bx += (dly[i]*(-z_c[i]) - dlz[i]*(-y_c[i])) / r3
        by += (dlz[i]*(-x_c[i]) - dlx[i]*(-z_c[i])) / r3
        bz += (dlx[i]*(-y_c[i]) - dly[i]*(-x_c[i])) / r3
        
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x_c, y_c, z_c)
    plt.savefig("stellarator_coil.png")

if __name__ == '__main__': simulate()
```
