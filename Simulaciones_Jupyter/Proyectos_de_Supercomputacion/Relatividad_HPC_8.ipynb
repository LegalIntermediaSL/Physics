# Supercomputación: Relatividad HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

v = 0.8 # v/c
theta_rest = np.linspace(0, np.pi, 100)

# Aberration
theta_obs = np.arccos((np.cos(theta_rest) + v) / (1 + v * np.cos(theta_rest)))

# Doppler shift frequency ratio
gamma = 1 / np.sqrt(1 - v**2)
f_ratio = 1 / (gamma * (1 - v * np.cos(theta_obs)))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.plot(np.degrees(theta_rest), np.degrees(theta_obs))
ax1.set_title('Aberration of Light')
ax1.set_xlabel('Theta Rest (deg)')
ax1.set_ylabel('Theta Obs (deg)')

ax2.plot(np.degrees(theta_obs), f_ratio)
ax2.set_title('Relativistic Doppler Shift')
ax2.set_xlabel('Theta Obs (deg)')
ax2.set_ylabel('Frequency Ratio')
plt.savefig('Relatividad_HPC_8_output.png')
```
