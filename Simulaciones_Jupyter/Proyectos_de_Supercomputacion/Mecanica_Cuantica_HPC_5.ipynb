# Supercomputación: Mecanica Cuantica HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 256
L = 10.0
x = np.linspace(-L, L, N)
dx = x[1] - x[0]
V = 0.5 * x**2
g = 10.0
dt = 0.01

psi = np.exp(-x**2)
psi /= np.sqrt(np.sum(np.abs(psi)**2)*dx)

for _ in range(1000):
    # Imaginary time split-step
    psi = psi * np.exp(-dt * (V + g * np.abs(psi)**2))
    
    psi_k = np.fft.fft(psi)
    k = np.fft.fftfreq(N, d=dx) * 2 * np.pi
    psi_k = psi_k * np.exp(-dt * k**2 / 2)
    psi = np.fft.ifft(psi_k)
    
    psi /= np.sqrt(np.sum(np.abs(psi)**2)*dx)

plt.plot(x, np.abs(psi)**2, label='BEC Density')
plt.plot(x, V*0.1, '--', label='Scaled Potential')
plt.legend()
plt.title('Gross-Pitaevskii Ground State')
plt.savefig('Mecanica_Cuantica_HPC_5_output.png')
```
