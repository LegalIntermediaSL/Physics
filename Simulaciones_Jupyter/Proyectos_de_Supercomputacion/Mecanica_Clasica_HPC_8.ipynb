# Supercomputación: Mecanica Clasica HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def cr3bp(t, Y, mu):
    x, y, vx, vy = Y
    r1 = np.sqrt((x + mu)**2 + y**2)
    r2 = np.sqrt((x - 1 + mu)**2 + y**2)
    ax = 2*vy + x - (1 - mu)*(x + mu)/r1**3 - mu*(x - 1 + mu)/r2**3
    ay = -2*vx + y - (1 - mu)*y/r1**3 - mu*y/r2**3
    return [vx, vy, ax, ay]

def simulate():
    mu = 0.012277471
    Y0 = [0.994, 0, 0, -2.001585106]
    t_span = (0, 20)
    t_eval = np.linspace(0, 20, 1000)
    sol = solve_ivp(cr3bp, t_span, Y0, args=(mu,), t_eval=t_eval, rtol=1e-8, atol=1e-8)
    
    plt.plot(sol.y[0], sol.y[1])
    plt.plot(-mu, 0, 'ro')
    plt.plot(1-mu, 0, 'go')
    plt.savefig("cr3bp.png")

if __name__ == '__main__': simulate()
```
