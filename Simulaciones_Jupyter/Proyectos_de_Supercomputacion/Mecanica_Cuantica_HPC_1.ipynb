# Supercomputación: Mecanica Cuantica HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

def lindblad_evolution(rho0, H, L_ops, gamma, t_max, dt):
    t_steps = int(t_max / dt)
    rho_t = [rho0]
    rho = rho0
    
    for _ in range(t_steps):
        # Dissipator
        D = np.zeros_like(rho, dtype=complex)
        for L in L_ops:
            D += gamma * (L @ rho @ L.conj().T - 0.5 * (L.conj().T @ L @ rho + rho @ L.conj().T @ L))
        
        # Unitary evolution
        drho = -1j * (H @ rho - rho @ H) + D
        rho = rho + drho * dt
        rho /= np.trace(rho) # normalize
        rho_t.append(rho)
        
    return np.array(rho_t)

H = np.array([[1, 0], [0, -1]])
L = np.array([[0, 1], [0, 0]]) # Decay
rho0 = np.array([[0.5, 0.5], [0.5, 0.5]])

t_max, dt = 10, 0.01
rho_t = lindblad_evolution(rho0, H, [L], 0.1, t_max, dt)

plt.plot(np.linspace(0, t_max, len(rho_t)), [r[1,1].real for r in rho_t], label='Excited state population')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Lindblad Master Equation - Qubit Decay')
plt.legend()
plt.savefig('Mecanica_Cuantica_HPC_1_output.png')
```
