# Supercomputación: Optica HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# FDTD 2D for Surface Plasmon Polaritons
Nx, Ny = 200, 200
Ex = np.zeros((Nx, Ny))
Ey = np.zeros((Nx, Ny))
Hz = np.zeros((Nx, Ny))
eps = np.ones((Nx, Ny))

eps[:, Ny//2:] = -2.0 + 0.1j # Drude metal

steps = 300
for t in range(steps):
    Ex[:, 1:] += (Hz[:, 1:] - Hz[:, :-1]) / eps[:, 1:]
    Ey[1:, :] -= (Hz[1:, :] - Hz[:-1, :]) / eps[1:, :]
    Hz[:-1, :-1] += Ex[:-1, 1:] - Ex[:-1, :-1] - (Ey[1:, :-1] - Ey[:-1, :-1])
    
    Hz[Nx//2, Ny//2 - 5] += np.sin(2 * np.pi * t / 20.0) # SPP excitation

plt.imshow(np.abs(Hz))
plt.title('SPP FDTD 2D')
plt.savefig('spp_2d.png')
```
