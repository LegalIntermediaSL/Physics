# Supercomputación: Fluidos HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def spectral_navier_stokes(N=64, steps=100, dt=1e-3, nu=1e-4):
    k = np.fft.fftfreq(N) * N
    kx, ky = np.meshgrid(k, k)
    k2 = kx**2 + ky**2
    k2[0, 0] = 1.0
    
    w = np.random.randn(N, N)
    w_hat = np.fft.fft2(w)
    
    for _ in range(steps):
        psi_hat = -w_hat / k2
        u = np.real(np.fft.ifft2(1j * ky * psi_hat))
        v = np.real(np.fft.ifft2(-1j * kx * psi_hat))
        w = np.real(np.fft.ifft2(w_hat))
        
        adv = u * np.real(np.fft.ifft2(1j * kx * w_hat)) + v * np.real(np.fft.ifft2(1j * ky * w_hat))
        w_hat -= dt * (np.fft.fft2(adv) + nu * k2 * w_hat)
        
    plt.imshow(w, cmap='RdBu')
    plt.title('2D Spectral Navier-Stokes Turbulence')
    plt.show()

if __name__ == '__main__':
    spectral_navier_stokes()
```
