# Supercomputación: Mecanica Clasica HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def euler_eqs(t, w, I1, I2, I3):
    dw1 = (I2 - I3) / I1 * w[1] * w[2]
    dw2 = (I3 - I1) / I2 * w[2] * w[0]
    dw3 = (I1 - I2) / I3 * w[0] * w[1]
    return [dw1, dw2, dw3]

def simulate():
    I1, I2, I3 = 1.0, 2.0, 3.0
    w0 = [0.1, 5.0, 0.1]
    t_span = (0, 20)
    sol = solve_ivp(euler_eqs, t_span, w0, args=(I1, I2, I3), dense_output=True)
    
    t = np.linspace(0, 20, 500)
    w = sol.sol(t)
    plt.plot(t, w[0], label='w1')
    plt.plot(t, w[1], label='w2')
    plt.plot(t, w[2], label='w3')
    plt.legend()
    plt.savefig("dzhanibekov.png")

if __name__ == '__main__': simulate()
```
