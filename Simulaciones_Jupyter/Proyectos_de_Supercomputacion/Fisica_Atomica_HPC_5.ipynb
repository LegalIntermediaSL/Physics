# Supercomputación: Fisica Atomica HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Quantum Monte Carlo for 1D Harmonic Oscillator
N_walkers = 1000
steps = 5000
dt = 0.01
x = np.random.randn(N_walkers)
E_T = 0.5
energy = []

for step in range(steps):
    x += np.random.randn(N_walkers) * np.sqrt(dt)
    V = 0.5 * x**2
    weights = np.exp(-(V - E_T) * dt)
    
    new_x = []
    for i in range(N_walkers):
        n_copies = int(weights[i] + np.random.rand())
        new_x.extend([x[i]] * n_copies)
    
    x = np.array(new_x)
    if len(x) > 0:
        E_T += 0.01 * np.log(N_walkers / len(x))
    N_walkers = len(x)
    energy.append(E_T)

plt.plot(energy)
plt.title('QMC Energy')
plt.savefig('qmc_energy.png')
```
