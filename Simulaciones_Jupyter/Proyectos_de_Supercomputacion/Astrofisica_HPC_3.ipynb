# Supercomputación: Astrofisica HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

r_s = 2.0
impact_params = np.linspace(3, 8, 20)
phi = np.linspace(0, 4*np.pi, 500)

plt.figure(figsize=(6,6))
circle = plt.Circle((0,0), r_s, color='black')
plt.gca().add_patch(circle)
plt.gca().add_patch(plt.Circle((0,0), 1.5*r_s, color='red', fill=False, ls='--', label='Photon sphere'))

for b in impact_params:
    u = np.zeros_like(phi)
    u[0] = 0.0
    u[1] = 1.0 / b * (phi[1] - phi[0])
    for i in range(1, len(phi)-1):
        dphi = phi[i] - phi[i-1]
        u[i+1] = 2*u[i] - u[i-1] + dphi**2 * (1.5 * r_s * u[i]**2 - u[i])
    
    r = 1.0 / (u + 1e-6)
    r[r < r_s] = np.nan
    x, y = r * np.cos(phi), r * np.sin(phi)
    plt.plot(x, y, color='blue', alpha=0.5)

plt.xlim(-10, 10); plt.ylim(-10, 10)
plt.title('Photon Orbits in Schwarzschild Spacetime')
plt.show()
```
