# Supercomputación: Materia Condensada HPC 9

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def quantum_spin_liquid():
    N_spins = 4
    dim = 2**N_spins
    H = np.zeros((dim, dim))
    # Heisenberg interaction mock on 4 sites
    
    E = np.linalg.eigvalsh(H)
    
    plt.bar(range(dim), E)
    plt.title('Quantum Spin Liquid (Exact Diagonalization Spectrum)')
    plt.xlabel('State index')
    plt.ylabel('Energy')
    plt.show()

if __name__ == '__main__':
    quantum_spin_liquid()
```
