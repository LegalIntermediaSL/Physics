# Supercomputación: Mecanica Clasica HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
from scipy.fft import fft, ifft
import matplotlib.pyplot as plt

def simulate():
    N = 256
    L = 50.0
    x = np.linspace(-L/2, L/2, N, endpoint=False)
    k = 2 * np.pi * np.fft.fftfreq(N, d=L/N)
    
    u = 3*4*1/np.cosh(0.5*2*(x+10))**2 + 3*1*1/np.cosh(0.5*1*(x+20))**2
    dt = 0.001
    
    for _ in range(1000):
        u_hat = fft(u)
        u_x = np.real(ifft(1j * k * u_hat))
        u_xxx = np.real(ifft((1j * k)**3 * u_hat))
        u = u - dt * (6 * u * u_x + u_xxx)
        
    plt.plot(x, u)
    plt.savefig("kdv_solitons.png")

if __name__ == '__main__': simulate()
```
