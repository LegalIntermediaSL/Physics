# Supercomputación: Fluidos HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def euler_1d_sod(N=200, steps=100):
    rho = np.ones(N); rho[N//2:] = 0.125
    u = np.zeros(N)
    p = np.ones(N); p[N//2:] = 0.1
    gamma = 1.4
    e = p / ((gamma - 1) * rho)
    
    plt.plot(rho, label='Density')
    plt.plot(p, label='Pressure')
    plt.title('Euler Equations (Sod Shock Tube Initial State)')
    plt.legend()
    plt.show()

if __name__ == '__main__':
    euler_1d_sod()
```
