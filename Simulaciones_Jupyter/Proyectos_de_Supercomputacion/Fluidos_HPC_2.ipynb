# Supercomputación: Fluidos HPC 2

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def lbm_multiphase(Nx=100, Ny=100, steps=200):
    rho = np.ones((Nx, Ny)) + np.random.rand(Nx, Ny) * 0.1
    # Simplified pseudopotential multiphase LBM
    G = -1.2
    cs2 = 1/3
    
    for _ in range(steps):
        # Force computation
        psi = 1.0 - np.exp(-rho)
        force_x = -G * psi * np.roll(psi, 1, axis=0)
        force_y = -G * psi * np.roll(psi, 1, axis=1)
        
        # Update density (mock step for brevity)
        rho += 0.01 * (force_x + force_y)
        rho = np.clip(rho, 0.1, 2.0)
        
    plt.imshow(rho, cmap='viridis')
    plt.title('Lattice Boltzmann Multiphase (Mock)')
    plt.show()

if __name__ == '__main__':
    lbm_multiphase()
```
