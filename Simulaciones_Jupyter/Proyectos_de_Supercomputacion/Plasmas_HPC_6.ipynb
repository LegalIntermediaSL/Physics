# Supercomputación: Plasmas HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np, matplotlib.pyplot as plt
def langmuir():
    t = np.linspace(0, 10, 100)
    waves = np.sin(2*np.pi*t)
    plt.plot(t, waves); plt.savefig("langmuir.png")
if __name__=="__main__": langmuir()
```
