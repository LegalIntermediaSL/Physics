# Supercomputación: Electromagnetismo HPC 9

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def simulate():
    X, Y = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
    beta = 0.9
    gamma = 1 / np.sqrt(1 - beta**2)
    
    r3 = (gamma**2 * X**2 + Y**2)**1.5
    Ex = gamma * X / r3
    Ey = gamma * Y / r3
    
    plt.quiver(X, Y, Ex, Ey)
    plt.savefig("relativistic_efield.png")

if __name__ == '__main__': simulate()
```
