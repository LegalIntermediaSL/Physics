# Supercomputación: Fisica Atomica HPC 2

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 1D TDSE High Harmonic Generation
N, L = 2000, 50.0
x = np.linspace(-L, L, N)
dx = x[1] - x[0]
V = -1.0 / np.sqrt(x**2 + 1.0)  # Soft-core potential

# Crank-Nicolson matrices
H0 = -0.5 / dx**2 * (np.diag(-2 * np.ones(N)) + np.diag(np.ones(N-1), 1) + np.diag(np.ones(N-1), -1)) + np.diag(V)
dt = 0.05
steps = 2000
omega, E0 = 0.057, 0.05

psi = np.exp(-x**2)
psi /= np.linalg.norm(psi)

dipole = []
for t in range(steps):
    time = t * dt
    E_field = E0 * np.sin(omega * time) * np.sin(time * np.pi / (steps*dt))**2
    H = H0 + np.diag(x * E_field)
    I = np.eye(N)
    A = I + 0.5j * dt * H
    B = I - 0.5j * dt * H
    psi = np.linalg.solve(A, np.dot(B, psi))
    dipole.append(np.sum(np.abs(psi)**2 * x))

spectrum = np.abs(np.fft.fft(dipole))**2
plt.semilogy(spectrum[:200])
plt.title('HHG Spectrum')
plt.savefig('hhg_spectrum.png')
```
