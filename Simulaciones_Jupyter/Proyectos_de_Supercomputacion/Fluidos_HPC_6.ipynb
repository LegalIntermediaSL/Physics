# Supercomputación: Fluidos HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def mhd_orszag_tang():
    N = 64
    x = np.linspace(0, 2*np.pi, N)
    y = np.linspace(0, 2*np.pi, N)
    X, Y = np.meshgrid(x, y)
    
    vx = -np.sin(Y)
    vy = np.sin(X)
    Bx = -np.sin(Y)
    By = np.sin(2*X)
    
    plt.streamplot(X, Y, Bx, By, color=np.sqrt(Bx**2 + By**2), cmap='plasma')
    plt.title('MHD Orszag-Tang Vortex Initial Magnetic Field')
    plt.show()

if __name__ == '__main__':
    mhd_orszag_tang()
```
