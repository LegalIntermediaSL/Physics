# Supercomputación: Plasmas HPC 2

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np, matplotlib.pyplot as plt
def vlasov():
    k = np.fft.fftfreq(256)
    f = np.exp(-k**2)
    plt.plot(k, f); plt.savefig("vlasov.png")
if __name__=="__main__": vlasov()
```
