# Supercomputación: Mecanica Cuantica HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
# GOE
H = np.random.randn(N, N)
H = (H + H.T) / 2
evals = np.linalg.eigvalsh(H)
spacings = np.diff(evals)
spacings = spacings / np.mean(spacings)

# GUE
H_gue = np.random.randn(N, N) + 1j * np.random.randn(N, N)
H_gue = (H_gue + H_gue.conj().T) / 2
evals_gue = np.linalg.eigvalsh(H_gue)
spacings_gue = np.diff(evals_gue)
spacings_gue = spacings_gue / np.mean(spacings_gue)

s = np.linspace(0, 3, 100)
p_goe = (np.pi / 2) * s * np.exp(-np.pi * s**2 / 4)
p_gue = (32 / np.pi**2) * s**2 * np.exp(-4 * s**2 / np.pi)

plt.hist(spacings, bins=50, density=True, alpha=0.5, label='GOE samples')
plt.plot(s, p_goe, label='Wigner Surmise GOE')
plt.hist(spacings_gue, bins=50, density=True, alpha=0.5, label='GUE samples')
plt.plot(s, p_gue, label='Wigner Surmise GUE')
plt.legend()
plt.title('Quantum Chaos - Level Spacing Statistics')
plt.savefig('Mecanica_Cuantica_HPC_6_output.png')
```
