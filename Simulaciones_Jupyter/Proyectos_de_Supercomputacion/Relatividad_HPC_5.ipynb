# Supercomputación: Relatividad HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# a'' / a = - (4 pi G / 3) (rho + 3P)
# Matter + Dark Energy
Omega_m = 0.3
Omega_L = 0.7
H0 = 70.0

def hubble(a, t):
    return H0 * np.sqrt(Omega_m/a**3 + Omega_L)

def a_dot(a, t):
    return a * hubble(a, t)

t = np.linspace(0, 20/H0, 1000)
a = odeint(a_dot, 1e-3, t)

plt.plot(t, a)
plt.title('Friedmann Equation - Universe Expansion')
plt.xlabel('Time')
plt.ylabel('Scale Factor a(t)')
plt.savefig('Relatividad_HPC_5_output.png')
```
