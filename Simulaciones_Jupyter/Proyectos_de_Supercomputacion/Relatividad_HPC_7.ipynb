# Supercomputación: Relatividad HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np

# A trivial snippet of GRMHD variables manipulation
rho = np.random.rand(10, 10)
P = 0.1 * rho
u = np.zeros((10, 10, 4)) # 4-velocity
u[..., 0] = 1.0 # Static fluid

B = np.zeros((10, 10, 3)) # Magnetic field 3-vector
B[..., 2] = 1.0 # Bz

b_mu = np.zeros((10, 10, 4))
b_mu[..., 0] = np.sum(B * u[..., 1:], axis=-1)
for i in range(1, 4):
    b_mu[..., i] = B[..., i-1] / u[..., 0] + b_mu[..., 0] * u[..., i]

b_sq = np.sum(-b_mu[..., 0]**2 + np.sum(b_mu[..., 1:]**2, axis=-1))

T_00 = (rho + P + b_sq) * u[..., 0]**2 + (P + b_sq/2) - b_mu[..., 0]**2
print("GRMHD Toy T_00 computed successfully. Mean T_00:", np.mean(T_00))
```
