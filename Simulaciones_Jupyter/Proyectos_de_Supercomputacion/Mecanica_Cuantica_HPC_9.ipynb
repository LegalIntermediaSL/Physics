# Supercomputación: Mecanica Cuantica HPC 9

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1024
L = 20.0
x = np.linspace(-L, L, N)
dx = x[1] - x[0]
k = np.fft.fftfreq(N, d=dx) * 2 * np.pi

# Barrier
V = np.zeros_like(x)
V[(x > 0) & (x < 1)] = 10.0

dt = 0.05
# Initial wavepacket
p0 = 5.0
x0 = -5.0
psi = np.exp(-(x - x0)**2) * np.exp(1j * p0 * x)
psi /= np.linalg.norm(psi)

U_v = np.exp(-1j * V * dt / 2)
U_k = np.exp(-1j * k**2 / 2 * dt)

for _ in range(200):
    psi = U_v * psi
    psi = np.fft.fft(psi)
    psi = U_k * psi
    psi = np.fft.ifft(psi)
    psi = U_v * psi

plt.plot(x, np.abs(psi)**2, label='Tunneling Wavepacket')
plt.plot(x, V*0.01, label='Barrier')
plt.title('Quantum Tunneling')
plt.legend()
plt.savefig('Mecanica_Cuantica_HPC_9_output.png')
```
