# Supercomputación: Astrofisica HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
q = np.random.rand(N, 2) * 100
k = np.fft.fftfreq(int(np.sqrt(N)))
kx, ky = np.meshgrid(k, k)
P_k = 1 / (kx**2 + ky**2 + 1e-5)**1.5
phi_k = np.sqrt(P_k) * np.exp(2j * np.pi * np.random.rand(*kx.shape))
dx = np.fft.ifft2(1j * kx * phi_k).real.flatten()
dy = np.fft.ifft2(1j * ky * phi_k).real.flatten()

D = 5.0
x = (q[:,0] + D * dx) % 100
y = (q[:,1] + D * dy) % 100

plt.scatter(x, y, s=1, color='black')
plt.title('Cosmic Web Formation (Zel\'dovich Approximation)')
plt.show()
```
