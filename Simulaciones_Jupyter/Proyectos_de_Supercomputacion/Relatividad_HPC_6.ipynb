# Supercomputación: Relatividad HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 1000)
tc = 1.05
f = 10 * (tc - t)**(-3/8)
amp = (tc - t)**(-1/4)
strain = amp * np.cos(2 * np.pi * np.cumsum(f) * (t[1]-t[0]))

plt.plot(t, strain)
plt.title('Gravitational Wave - Binary Inspiral')
plt.xlabel('Time')
plt.ylabel('Strain')
plt.savefig('Relatividad_HPC_6_output.png')
```
