# Supercomputación: Termodinamica HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def ising_2d(L=100, steps=2000, T=2.269):
    lattice = np.random.choice([-1, 1], size=(L, L))
    beta = 1.0 / T
    for _ in range(steps):
        for _ in range(L * L):
            x, y = np.random.randint(0, L, 2)
            s = lattice[x, y]
            neighbors = lattice[(x+1)%L, y] + lattice[(x-1)%L, y] + lattice[x, (y+1)%L] + lattice[x, (y-1)%L]
            dE = 2 * s * neighbors
            if dE < 0 or np.random.rand() < np.exp(-dE * beta):
                lattice[x, y] = -s
    return lattice

res = ising_2d()
plt.imshow(res, cmap='inferno')
plt.title('2D Ising Model near Critical Temperature')
plt.show()
```
