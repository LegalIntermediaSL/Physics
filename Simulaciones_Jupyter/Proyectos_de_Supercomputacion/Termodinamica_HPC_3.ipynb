# Supercomputación: Termodinamica HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform

N = 100
pos = np.random.rand(N, 2) * 10
vel = (np.random.rand(N, 2) - 0.5)
dt = 0.01

def lj_force(r):
    r6 = r**6
    r13 = r**13
    return 48 * (1/r13 - 0.5/r6)

for step in range(200):
    dist = squareform(pdist(pos))
    np.fill_diagonal(dist, np.inf)
    force_mag = lj_force(dist)
    dx = pos[:, 0][:, np.newaxis] - pos[:, 0]
    dy = pos[:, 1][:, np.newaxis] - pos[:, 1]
    fx = np.sum(force_mag * dx / dist, axis=1)
    fy = np.sum(force_mag * dy / dist, axis=1)
    vel += np.column_stack((fx, fy)) * dt
    pos += vel * dt
    pos = pos % 10.0

plt.scatter(pos[:,0], pos[:,1])
plt.title('Molecular Dynamics: Lennard-Jones Gas')
plt.show()
```
