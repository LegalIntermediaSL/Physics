# Supercomputación: Astrofisica HPC 9

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

l = np.arange(2, 2000)
c_l = 1.0 / (l * (l + 1)) * (1 + 2*np.exp(-((l-200)/100)**2) + 0.5*np.exp(-((l-500)/50)**2))

plt.plot(l, l*(l+1)*c_l, color='green')
plt.title('CMB Temperature Power Spectrum $D_l^{TT}$')
plt.xlabel('Multipole Moment l')
plt.ylabel('$D_l$')
plt.xscale('log')
plt.show()
```
