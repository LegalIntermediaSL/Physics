# Supercomputación: Fisica Atomica HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 2D Hartree-Fock for a simple two-electron system (Quantum Dot)
grid_size = 50
x = np.linspace(-5, 5, grid_size)
y = np.linspace(-5, 5, grid_size)
X, Y = np.meshgrid(x, y)
V_ext = 0.5 * (X**2 + Y**2)

psi = np.exp(-(X**2 + Y**2))
psi /= np.sqrt(np.sum(np.abs(psi)**2))

def hartree_potential(n):
    return n * 0.1

for _ in range(20):
    n = 2 * np.abs(psi)**2
    V = V_ext + hartree_potential(n)
    
    H_psi = -0.5 * (np.roll(psi, 1, axis=0) + np.roll(psi, -1, axis=0) + np.roll(psi, 1, axis=1) + np.roll(psi, -1, axis=1) - 4*psi) + V * psi
    psi = psi - 0.01 * H_psi
    psi /= np.sqrt(np.sum(np.abs(psi)**2))

plt.imshow(np.abs(psi)**2)
plt.title('Quantum Dot Density')
plt.savefig('quantum_dot.png')
```
