# Supercomputación: Astrofisica HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 200
x = np.linspace(0, 1, N)
dx = 1.0 / N
dt = 0.001
gamma = 1.4

rho = np.where(x < 0.5, 1.0, 0.125)
v = np.zeros(N)
P = np.where(x < 0.5, 1.0, 0.1)
E = P / (gamma - 1) + 0.5 * rho * v**2

def flux(u1, u2, u3):
    v = u2 / u1
    P = (gamma - 1) * (u3 - 0.5 * u1 * v**2)
    return np.array([u2, u2*v + P, v*(u3 + P)])

U = np.array([rho, rho*v, E])

for _ in range(200):
    F = flux(U[0], U[1], U[2])
    U_new = np.zeros_like(U)
    for i in range(1, N-1):
        U_new[:, i] = 0.5*(U[:, i+1] + U[:, i-1]) - dt/(2*dx) * (F[:, i+1] - F[:, i-1])
    U_new[:, 0] = U[:, 0]
    U_new[:, -1] = U[:, -1]
    U = U_new

rho_out = U[0]
plt.plot(x, rho_out, label='Density')
plt.title('Sod Shock Tube (Astrophysical Jets Proxy)')
plt.legend()
plt.show()
```
