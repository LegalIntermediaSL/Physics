# Supercomputación: Info Cuantica HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np, scipy.linalg as la, matplotlib.pyplot as plt
def vqe():
    H = np.array([[1, 0.5], [0.5, -1]])
    plt.plot(la.eigvalsh(H)); plt.savefig("vqe.png")
if __name__=="__main__": vqe()
```
