# Supercomputación: Electromagnetismo HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def simulate():
    nx, ny = 100, 100
    ez = np.zeros((nx, ny))
    hx = np.zeros((nx, ny))
    hy = np.zeros((nx, ny))
    
    for t in range(200):
        hx[:, :-1] -= 0.5 * (ez[:, 1:] - ez[:, :-1])
        hy[:-1, :] += 0.5 * (ez[1:, :] - ez[:-1, :])
        ez[1:, 1:] += 0.5 * ((hy[1:, 1:] - hy[:-1, 1:]) - (hx[1:, 1:] - hx[1:, :-1]))
        
        # Source
        ez[nx//4, ny//2] = np.sin(0.1 * t)
        
    plt.imshow(ez, cmap='RdBu')
    plt.savefig("fdtd.png")

if __name__ == '__main__': simulate()
```
