# Supercomputación: Relatividad HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Effective potential approach for r, phi
# u = 1/r
# d^2 u / d phi^2 + u = M/L^2 + 3 M u^2

M = 1.0
L = 4.0

def geo_derivs(phi, y):
    u, du = y
    return [du, M/L**2 + 3*M*u**2 - u]

sol = solve_ivp(geo_derivs, [0, 20*np.pi], [1/10.0, 0], max_step=0.1)

r = 1 / sol.y[0]
phi = sol.t

x = r * np.cos(phi)
y = r * np.sin(phi)

plt.plot(x, y)
circle = plt.Circle((0, 0), 2*M, color='black', fill=True)
plt.gca().add_patch(circle)
plt.title('Schwarzschild Geodesic')
plt.axis('equal')
plt.savefig('Relatividad_HPC_1_output.png')
```
