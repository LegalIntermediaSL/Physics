# Supercomputación: Relatividad HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# TOV equations for Neutron Star
# Simplified polytropic EOS: P = K * rho^Gamma
K = 100.0
Gamma = 2.0

def tov_eqs(r, y):
    P, m = y
    if P <= 0:
        return [0, 0]
    rho = (P / K)**(1/Gamma)
    dP = - (rho + P) * (m + 4*np.pi*r**3*P) / (r * (r - 2*m))
    dm = 4*np.pi*r**2 * rho
    return [dP, dm]

r_span = [1e-5, 20.0]
y0 = [1e-3, 0.0] # Central pressure, mass

sol = solve_ivp(tov_eqs, r_span, y0, events=lambda r, y: y[0], rtol=1e-6)

plt.plot(sol.t, sol.y[0])
plt.title('TOV Equation - Pressure Profile')
plt.xlabel('Radius')
plt.ylabel('Pressure')
plt.savefig('Relatividad_HPC_4_output.png')
```
