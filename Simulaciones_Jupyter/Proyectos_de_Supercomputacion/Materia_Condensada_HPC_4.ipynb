# Supercomputación: Materia Condensada HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def ising_3d():
    L = 16
    spins = np.random.choice([-1, 1], size=(L, L, L))
    
    plt.imshow(spins[:, :, L//2], cmap='coolwarm')
    plt.title('3D Ising Model Slice (T > Tc)')
    plt.show()

if __name__ == '__main__':
    ising_3d()
```
