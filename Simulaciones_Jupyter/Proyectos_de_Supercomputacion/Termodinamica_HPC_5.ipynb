# Supercomputación: Termodinamica HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 100
phi = np.random.uniform(-0.1, 0.1, (N, N))
dx = 1.0
dt = 0.01
M, gamma = 1.0, 1.0

for _ in range(200):
    mu = phi**3 - phi - gamma * (np.roll(phi, 1, 0) + np.roll(phi, -1, 0) + np.roll(phi, 1, 1) + np.roll(phi, -1, 1) - 4*phi) / dx**2
    lap_mu = (np.roll(mu, 1, 0) + np.roll(mu, -1, 0) + np.roll(mu, 1, 1) + np.roll(mu, -1, 1) - 4*mu) / dx**2
    phi += dt * M * lap_mu

plt.imshow(phi, cmap='coolwarm')
plt.title('Cahn-Hilliard Phase Separation')
plt.show()
```
