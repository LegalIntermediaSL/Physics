# Supercomputación: Optica HPC 5

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Ray tracing spherical aberration
y_rays = np.linspace(-5, 5, 20)
R = 10.0 # radius of curvature
n_lens = 1.5

plt.figure()
for y0 in y_rays:
    x0 = -np.sqrt(R**2 - y0**2)
    sin_theta_i = y0 / R
    theta_i = np.arcsin(sin_theta_i)
    theta_t = np.arcsin(sin_theta_i / n_lens)
    
    slope = np.tan(theta_t - theta_i)
    x_f = 20
    y_f = y0 + slope * (x_f - x0)
    plt.plot([x0, x_f], [y0, y_f], 'b', alpha=0.5)
    
plt.title('Spherical Aberration')
plt.savefig('ray_tracing.png')
```
