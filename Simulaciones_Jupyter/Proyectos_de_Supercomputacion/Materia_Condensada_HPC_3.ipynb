# Supercomputación: Materia Condensada HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def md_argon():
    N = 100
    pos = np.random.rand(N, 2) * 10
    
    plt.scatter(pos[:, 0], pos[:, 1], color='b')
    plt.title('Molecular Dynamics Argon (Lennard-Jones Setup)')
    plt.show()

if __name__ == '__main__':
    md_argon()
```
