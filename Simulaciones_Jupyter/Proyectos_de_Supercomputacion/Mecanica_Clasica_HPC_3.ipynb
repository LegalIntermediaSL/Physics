# Supercomputación: Mecanica Clasica HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def derivs(t, state):
    dydx = np.zeros_like(state)
    dydx[0] = state[1]
    del_ = state[2] - state[0]
    den1 = (1 + np.sin(del_)**2)
    dydx[1] = (state[3]*np.sin(del_)*np.cos(del_) + np.sin(state[2])*np.cos(del_) - 2*np.sin(state[0])) / den1
    dydx[2] = state[3]
    dydx[3] = (-state[3]*np.sin(del_)*np.cos(del_) - 2*np.sin(state[2]) + 2*np.sin(state[0])*np.cos(del_)) / den1
    return dydx

def simulate():
    t_span = (0, 50)
    t_eval = np.linspace(0, 50, 5000)
    sol = solve_ivp(derivs, t_span, [np.pi/2, 0, np.pi/2, 0], t_eval=t_eval)
    
    plt.plot(np.sin(sol.y[0]), -np.cos(sol.y[0]), label='m1')
    plt.plot(np.sin(sol.y[0])+np.sin(sol.y[2]), -np.cos(sol.y[0])-np.cos(sol.y[2]), label='m2')
    plt.legend()
    plt.savefig("double_pendulum.png")

if __name__ == '__main__': simulate()
```
