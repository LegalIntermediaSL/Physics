# Supercomputación: Optica HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Optical Tweezers potential force (Rayleigh regime)
r = np.linspace(-2, 2, 100)
z = np.linspace(-5, 5, 100)
R, Z = np.meshgrid(r, z)

w0 = 1.0
zR = 5.0
w = w0 * np.sqrt(1 + (Z/zR)**2)
I = (w0/w)**2 * np.exp(-2 * R**2 / w**2)

# Gradient force
dIdR = np.gradient(I, axis=1)
dIdZ = np.gradient(I, axis=0)

plt.streamplot(r, z, dIdR, dIdZ)
plt.title('Optical Tweezer Force Field')
plt.savefig('tweezers.png')
```
