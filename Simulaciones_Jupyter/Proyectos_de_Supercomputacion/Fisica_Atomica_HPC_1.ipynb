# Supercomputación: Fisica Atomica HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
import matplotlib.pyplot as plt

# 1D Gross-Pitaevskii Equation for Bose-Einstein Condensate
# Imaginary time propagation for ground state
N = 1000
L = 10.0
x = np.linspace(-L, L, N)
dx = x[1] - x[0]
dt = 0.01j  # Imaginary time
V = 0.5 * x**2  # Harmonic trap
g = 10.0  # Interaction strength

# Kinetic energy operator (finite difference)
T = -0.5 / dx**2 * sp.diags([1, -2, 1], [-1, 0, 1], shape=(N, N))

psi = np.exp(-x**2/2)
psi /= np.sqrt(np.sum(np.abs(psi)**2) * dx)

for i in range(1000):
    H = T + sp.diags(V + g * np.abs(psi)**2)
    psi = spla.expm_multiply(-1j * H * dt, psi)
    psi /= np.sqrt(np.sum(np.abs(psi)**2) * dx)

plt.plot(x, np.abs(psi)**2)
plt.title('BEC Ground State Density')
plt.savefig('bec_density.png')
```
