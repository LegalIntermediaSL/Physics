# Supercomputación: Termodinamica HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def lorenz(state, t, sigma, rho, beta):
    x, y, z = state
    return sigma * (y - x), x * (rho - z) - y, x * y - beta * z

t = np.arange(0.0, 40.0, 0.01)
state0 = [1.0, 1.0, 1.0]
res = odeint(lorenz, state0, t, args=(10.0, 28.0, 8.0/3.0))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(res[:,0], res[:,1], res[:,2], lw=0.8)
ax.set_title('Rayleigh-Bénard Convection Attractor (Lorenz)')
plt.show()
```
