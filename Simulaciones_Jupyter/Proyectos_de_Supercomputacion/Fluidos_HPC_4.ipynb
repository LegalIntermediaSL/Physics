# Supercomputación: Fluidos HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rayleigh_benard(Nx=64, Ny=32, steps=100):
    T = np.zeros((Nx, Ny))
    T[:, 0] = 1.0
    T[:, -1] = 0.0
    
    for _ in range(steps):
        T[1:-1, 1:-1] += 0.01 * (T[2:, 1:-1] + T[:-2, 1:-1] + T[1:-1, 2:] + T[1:-1, :-2] - 4*T[1:-1, 1:-1])
        T[:, 0] = 1.0
        T[:, -1] = 0.0
        
    plt.imshow(T.T, cmap='hot')
    plt.title('Rayleigh-Bénard Convection (Thermal only)')
    plt.show()

if __name__ == '__main__':
    rayleigh_benard()
```
