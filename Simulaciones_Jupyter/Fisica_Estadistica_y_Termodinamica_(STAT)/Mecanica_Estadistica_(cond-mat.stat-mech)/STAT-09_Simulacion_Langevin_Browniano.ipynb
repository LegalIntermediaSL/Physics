# [STAT-09] Mecánica Estadística: Ecuación Estocástica de Langevin

```python
import numpy as np
import matplotlib.pyplot as plt

# Mecánica Estadística: Ecuación Estocástica de Langevin y Movimiento Browniano
# Simulamos miles de partículas bombareadas aleatoriamente por moléculas térmicas de agua.
# Ecuación: dv/dt = -gamma*v + Ruido_Térmico

N_particles = 2000
N_steps = 1000
dt = 0.05
gamma = 1.0     # Fricción viscosa
kT = 2.0        # Energía térmica (Temperatura)

# Coeficiente del ruido térmico (Fluctuation-Dissipation Theorem)
sigma = np.sqrt(2 * gamma * kT / dt)

positions = np.zeros(N_particles)
velocities = np.zeros(N_particles)
variance = np.zeros(N_steps)

plt.figure(figsize=(10, 6))

for i in range(N_steps):
    # El ruido térmico es ruido blanco Gaussiano
    noise = np.random.normal(0, 1, N_particles) * sigma
    
    # Ecuación Diferencial Estocástica (Método de Euler-Maruyama)
    velocities += (-gamma * velocities + noise) * dt
    positions += velocities * dt
    
    variance[i] = np.var(positions)
    
    # Ploteamos las 50 primeras partículas para ver su camino caótico aleatorio
    if i % 10 == 0:
        for p in range(50):
            plt.plot(i * dt, positions[p], 'o', color='blue', alpha=0.05, markersize=2)

plt.title("Termodinámica del No-Equilibrio: Movimiento Browniano Estocástico", fontsize=14)
plt.xlabel("Tiempo (Pasos de Langevin)")
plt.ylabel("Posición $x$ de las Partículas")
plt.grid(True)
plt.show()

# Gráfico del paso del caos a la ley física
plt.figure(figsize=(10, 5))
time_axis = np.arange(N_steps) * dt
plt.plot(time_axis, variance, 'r-', lw=3, label="Varianza Microscópica Calculada $\sigma^2(t)$")
plt.plot(time_axis, 2 * (kT / gamma) * time_axis, 'k--', lw=2, label="Ley Macroscópica Teórica de Difusión (Einstein)")

plt.title("De la Estocástica Caótica al Orden Macroscópico (Relación de Einstein)")
plt.xlabel("Tiempo $t$")
plt.ylabel("Varianza (Difusión Cuadrática Media)")
plt.legend()
plt.grid(True)
plt.show()
# FÍSICA: La Relación de Fluctuación-Disipación de Einstein (1905) se cumple.
# Las colisiones individuales son ruido puro indescifrable, pero el conjunto macroscópico 
# se difunde perfectamente de forma proporcional al tiempo.

```
