# [BIO-02] Biología Cuántica: Mutación por Efecto Túnel

```python
import numpy as np
import matplotlib.pyplot as plt

# Biología Cuántica: Efecto Túnel en un Puente de Hidrógeno (Mutagenesis Espontánea)
# Simulamos cómo la probabilidad de un Protón (paquete Gaussiano) penetra
# la barrera energética entre dos bases nucleotídicas.

N_x = 400
L = 10.0
x = np.linspace(0, L, N_x)
dx = x[1] - x[0]

# Barrera de Potencial Doble Pozo (Los dos lados del puente de hidrógeno en el ADN)
V = np.zeros(N_x)
for i in range(N_x):
    # Barrera central de potencial V_0
    if 4.5 < x[i] < 5.5: V[i] = 150.0 
    else:                V[i] = 0.5 * 100.0 * (x[i] - 5)**2 # Confinamiento parabólico
V = np.clip(V, 0, 150.0)

# Paquete de Ondas Inicial (Protón anclado rígidamente en el lado izquierdo)
x0 = 3.5
sigma = 0.3
k0 = 20.0 # Energía cinética térmica empujando contra la barrera
psi = np.exp(-(x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k0 * x)
psi /= np.linalg.norm(psi) * np.sqrt(dx) # Normalización

# Integrador de Schrödinger (Split-Step Crank-Nicolson simple / Euler Implícito modificado)
# Para simplificar la visualización rápida, aplicaremos evolución temporal discreta.
hbar = 1.0
m_proton = 2.0 
dt = 0.005

prob_inicial = np.abs(psi)**2

# Operador Hamiltoniano Cinético (Diferencias Finitas D2)
D2 = (np.diag(-2 * np.ones(N_x)) + np.diag(np.ones(N_x-1), 1) + np.diag(np.ones(N_x-1), -1)) / dx**2
H = -(hbar**2 / (2 * m_proton)) * D2 + np.diag(V)

# Evolución Temporal Unitaria U = exp(-iHt/hbar). Aproximación de Taylor a primer orden.
for _ in range(100):
    psi = psi - (1j * dt / hbar) * (H @ psi)
    psi /= np.linalg.norm(psi) * np.sqrt(dx) # Renormalización en cada paso

prob_final = np.abs(psi)**2

plt.figure(figsize=(10, 6))
plt.plot(x, V / np.max(V) * 0.1, 'k-', lw=2, label="Barrera de Potencial (Puente Hidrógeno ADN)")
plt.fill_between(x, 0, prob_inicial, color='blue', alpha=0.3, label="Protón Inicial (Lado Izquierdo)")
plt.fill_between(x, 0, prob_final, color='red', alpha=0.5, label="Protón Evolucionado (Probabilidad Cruzada)")

# Zoom y ejes
plt.xlim(2, 8)
plt.ylim(0, 0.15)
plt.title("Biología Cuántica: Mutación por Efecto Túnel en el ADN", fontsize=14)
plt.xlabel("Posición Espacial $x$ (Amstrongs)")
plt.ylabel("Densidad de Probabilidad $|\psi|^2$")
plt.legend()
plt.grid(True)
plt.show()

# FÍSICA: Observa la densidad de probabilidad roja en el lado derecho de la barrera negra.
# La mecánica cuántica indica que existe una probabilidad no nula de que el protón se 
# "teletransporte" al lado opuesto, induciendo un fallo en la réplica celular genética.

```
