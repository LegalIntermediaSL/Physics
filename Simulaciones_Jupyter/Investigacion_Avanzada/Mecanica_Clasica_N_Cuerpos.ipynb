# Investigación Avanzada: Mecanica Clasica N Cuerpos

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def compute_accelerations(positions, masses, G=1.0):
    N = len(masses)
    acc = np.zeros((N, 3))
    for i in range(N):
        for j in range(N):
            if i != j:
                r_vec = positions[j] - positions[i]
                r_mag = np.linalg.norm(r_vec)
                # Softening parameter to prevent singularities
                softening = 0.01
                acc[i] += G * masses[j] * r_vec / (r_mag**3 + softening**3)
    return acc

def verlet_step(pos, vel, acc, masses, dt, G=1.0):
    pos_new = pos + vel * dt + 0.5 * acc * dt**2
    acc_new = compute_accelerations(pos_new, masses, G)
    vel_new = vel + 0.5 * (acc + acc_new) * dt
    return pos_new, vel_new, acc_new

def main():
    N = 50
    np.random.seed(42)
    masses = np.random.uniform(0.5, 2.0, N)
    positions = np.random.randn(N, 3) * 10.0
    velocities = np.random.randn(N, 3) * 0.5
    
    G = 1.0
    dt = 0.01
    steps = 1000
    
    acc = compute_accelerations(positions, masses, G)
    
    history_pos = np.zeros((steps, N, 3))
    
    for t in range(steps):
        history_pos[t] = positions
        positions, velocities, acc = verlet_step(positions, velocities, acc, masses, dt, G)
        
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    def update(num, history_pos, ax):
        ax.clear()
        ax.set_xlim(-20, 20)
        ax.set_ylim(-20, 20)
        ax.set_zlim(-20, 20)
        ax.scatter(history_pos[num, :, 0], history_pos[num, :, 1], history_pos[num, :, 2], s=masses*10, c='b')
        ax.set_title(f"N-Body Simulation (Verlet), Step {num}")
        
    ani = animation.FuncAnimation(fig, update, frames=steps, fargs=(history_pos, ax), interval=20, blit=False)
    plt.show()

if __name__ == "__main__":
    main()
```
