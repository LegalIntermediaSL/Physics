# Investigación Avanzada: Astrofisica SPH Colapso

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def sph_collapse():
    N = 400
    np.random.seed(42)
    pos = np.random.randn(N, 2) * 5
    vel = np.zeros((N, 2))
    mass = np.ones(N)
    
    dt = 0.05
    G = 1.0
    
    for step in range(100):
        acc = np.zeros_like(pos)
        for i in range(N):
            dx = pos[:, 0] - pos[i, 0]
            dy = pos[:, 1] - pos[i, 1]
            r2 = dx**2 + dy**2 + 0.1
            r = np.sqrt(r2)
            f = G * mass[i] / r2
            acc[:, 0] -= f * dx / r
            acc[:, 1] -= f * dy / r
        
        vel += acc * dt
        pos += vel * dt
        
    plt.figure(figsize=(8,8))
    plt.scatter(pos[:,0], pos[:,1], s=5, color='darkred', alpha=0.7)
    plt.title('SPH Gas Collapse (2D projection)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.savefig('SPH_Collapse.png')

if __name__ == '__main__':
    sph_collapse()
```
