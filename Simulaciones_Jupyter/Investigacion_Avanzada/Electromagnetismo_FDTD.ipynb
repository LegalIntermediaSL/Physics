# Investigación Avanzada: Electromagnetismo FDTD

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def main():
    # 1D FDTD Simulation
    c = 3e8
    dx = 1e-3
    dt = dx / (2 * c) # Courant condition
    
    nx = 200
    ex = np.zeros(nx)
    hy = np.zeros(nx)
    
    steps = 400
    history_ex = np.zeros((steps, nx))
    
    for t in range(steps):
        # Update H field
        for i in range(nx - 1):
            hy[i] = hy[i] + (dt / (4 * np.pi * 1e-7 * dx)) * (ex[i+1] - ex[i])
            
        # Update E field
        for i in range(1, nx):
            ex[i] = ex[i] + (dt / (8.85e-12 * dx)) * (hy[i] - hy[i-1])
            
        # Gaussian pulse source
        ex[nx//4] += np.exp(-0.5 * ((t - 30) / 10)**2)
        
        history_ex[t, :] = ex.copy()
        
    fig, ax = plt.subplots(figsize=(8, 6))
    line, = ax.plot(history_ex[0, :])
    ax.set_ylim(-1.5, 1.5)
    ax.set_title("1D FDTD Simulation of EM Wave")
    ax.set_xlabel("x position")
    ax.set_ylabel("Electric Field Ex")
    
    def update(frame):
        line.set_ydata(history_ex[frame, :])
        return line,
        
    ani = animation.FuncAnimation(fig, update, frames=steps, interval=20, blit=True)
    plt.show()

if __name__ == "__main__":
    main()
```
