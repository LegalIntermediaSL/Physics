# Investigación Avanzada: Mecanica Clasica Pendulo Doble

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def derivs(state, t, L1, L2, m1, m2, g):
    theta1, z1, theta2, z2 = state
    c, s = np.cos(theta1-theta2), np.sin(theta1-theta2)
    theta1dot = z1
    z1dot = (m2*g*np.sin(theta2)*c - m2*s*(L1*z1**2*c + L2*z2**2) - (m1+m2)*g*np.sin(theta1)) / L1 / (m1 + m2*s**2)
    theta2dot = z2
    z2dot = ((m1+m2)*(L1*z1**2*s - g*np.sin(theta2) + g*np.sin(theta1)*c) + m2*L2*z2**2*s*c) / L2 / (m1 + m2*s**2)
    return theta1dot, z1dot, theta2dot, z2dot

def main():
    L1, L2 = 1.0, 1.0
    m1, m2 = 1.0, 1.0
    g = 9.81
    t = np.linspace(0, 50, 5000)
    
    # Grid of initial conditions to visualize fractality in flip times (time to flip over)
    th1_vals = np.linspace(-np.pi, np.pi, 50)
    th2_vals = np.linspace(-np.pi, np.pi, 50)
    flip_times = np.zeros((50, 50))
    
    for i, th1 in enumerate(th1_vals):
        for j, th2 in enumerate(th2_vals):
            state0 = [th1, 0.0, th2, 0.0]
            sol = odeint(derivs, state0, t, args=(L1, L2, m1, m2, g))
            # Find first time angle > pi or < -pi (flip)
            flipped_th1 = np.abs(sol[:,0]) > np.pi
            flipped_th2 = np.abs(sol[:,2]) > np.pi
            flipped = flipped_th1 | flipped_th2
            if np.any(flipped):
                flip_times[i, j] = t[np.argmax(flipped)]
            else:
                flip_times[i, j] = 50.0
                
    plt.figure(figsize=(8, 6))
    plt.imshow(flip_times, extent=[-np.pi, np.pi, -np.pi, np.pi], origin='lower', cmap='hot')
    plt.colorbar(label='Time to flip')
    plt.xlabel('Initial Theta 1')
    plt.ylabel('Initial Theta 2')
    plt.title('Fractal Flip Times for Double Pendulum')
    plt.show()

if __name__ == "__main__":
    main()
```
