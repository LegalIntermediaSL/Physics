# Investigación Avanzada: Ondas Optica Fourier

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft2, ifft2, fftshift, ifftshift

def generate_image(N):
    x = np.linspace(-1, 1, N)
    y = np.linspace(-1, 1, N)
    X, Y = np.meshgrid(x, y)
    img = np.sin(20 * np.pi * X) + np.sin(50 * np.pi * Y) + np.sin(100 * np.pi * (X + Y))
    # Add some noise
    img += 0.5 * np.random.randn(N, N)
    return img

def spatial_filtering():
    N = 512
    img = generate_image(N)
    
    # Fourier transform
    F = fftshift(fft2(img))
    
    # Create filters
    radius_low = 30
    radius_high = 10
    
    Y, X = np.ogrid[-N//2:N//2, -N//2:N//2]
    R_sq = X**2 + Y**2
    
    # Low pass filter
    H_low = R_sq <= radius_low**2
    
    # High pass filter
    H_high = R_sq >= radius_high**2
    
    # Apply filters
    F_low = F * H_low
    F_high = F * H_high
    
    # Inverse transform
    img_low = np.real(ifft2(ifftshift(F_low)))
    img_high = np.real(ifft2(ifftshift(F_high)))
    
    # Plotting
    fig, axs = plt.subplots(2, 2, figsize=(12, 12))
    
    axs[0, 0].imshow(img, cmap='gray')
    axs[0, 0].set_title('Original Image')
    axs[0, 0].axis('off')
    
    axs[0, 1].imshow(np.log1p(np.abs(F)), cmap='gray')
    axs[0, 1].set_title('Fourier Spectrum (Log Scale)')
    axs[0, 1].axis('off')
    
    axs[1, 0].imshow(img_low, cmap='gray')
    axs[1, 0].set_title('Low Pass Filtered')
    axs[1, 0].axis('off')
    
    axs[1, 1].imshow(img_high, cmap='gray')
    axs[1, 1].set_title('High Pass Filtered')
    axs[1, 1].axis('off')
    
    plt.tight_layout()
    plt.savefig('Ondas_Optica_Fourier.png')
    plt.show()

if __name__ == '__main__':
    spatial_filtering()
```
