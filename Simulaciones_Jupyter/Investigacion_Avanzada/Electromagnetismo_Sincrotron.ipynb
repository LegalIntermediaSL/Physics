# Investigación Avanzada: Electromagnetismo Sincrotron

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def synchrotron_radiation(gamma, theta, omega, omega_c):
    # Simplified approximation for synchrotron radiation power spectrum
    # Using modified Bessel functions K_v
    from scipy.special import kv
    xi = omega / (2 * omega_c) * (1 + (gamma * theta)**2)**1.5
    if xi > 100:
        return 0.0 # prevent overflow
    p_parallel = (1 + (gamma * theta)**2)**2 * kv(2/3, xi)**2
    p_perp = (gamma * theta)**2 * (1 + (gamma * theta)**2) * kv(1/3, xi)**2
    return p_parallel + p_perp

def main():
    gamma = 1000.0  # Lorentz factor
    omega_c = 1e15  # Critical frequency
    theta_vals = np.linspace(-0.01, 0.01, 400)
    omega = omega_c * 0.1
    
    power = np.array([synchrotron_radiation(gamma, th, omega, omega_c) for th in theta_vals])
    
    plt.figure(figsize=(8,6))
    plt.plot(theta_vals * gamma, power)
    plt.xlabel(r'$\gamma \theta$')
    plt.ylabel('Radiation Intensity (arb. units)')
    plt.title(r'Synchrotron Radiation Angular Distribution ($\omega = 0.1 \omega_c$)')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
