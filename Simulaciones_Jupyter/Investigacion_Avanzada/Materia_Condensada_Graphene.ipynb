# Investigación Avanzada: Materia Condensada Graphene

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def graphene_bands(kx, ky, t=2.7):
    # Nearest neighbor vectors in graphene
    a = 1.42 # Carbon-carbon distance in Angstroms
    delta1 = np.array([0, a])
    delta2 = np.array([np.sqrt(3)/2 * a, -a/2])
    delta3 = np.array([-np.sqrt(3)/2 * a, -a/2])
    
    # Off-diagonal element of the Hamiltonian
    f = np.exp(1j * (kx * delta1[0] + ky * delta1[1])) + \
        np.exp(1j * (kx * delta2[0] + ky * delta2[1])) + \
        np.exp(1j * (kx * delta3[0] + ky * delta3[1]))
    
    # Energy eigenvalues
    E = t * np.abs(f)
    return E, -E

kx = np.linspace(-3, 3, 200)
ky = np.linspace(-3, 3, 200)
KX, KY = np.meshgrid(kx, ky)

E_con, E_val = np.zeros_like(KX), np.zeros_like(KX)
for i in range(KX.shape[0]):
    for j in range(KX.shape[1]):
        E_con[i, j], E_val[i, j] = graphene_bands(KX[i, j], KY[i, j])

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(KX, KY, E_con, cmap='hot', alpha=0.9)
ax.plot_surface(KX, KY, E_val, cmap='cool', alpha=0.9)
ax.set_title('Graphene Tight-Binding Band Structure')
ax.set_xlabel('$k_x (1/\AA)$')
ax.set_ylabel('$k_y (1/\AA)$')
ax.set_zlabel('Energy (eV)')
plt.savefig('graphene_bands.png')
```
