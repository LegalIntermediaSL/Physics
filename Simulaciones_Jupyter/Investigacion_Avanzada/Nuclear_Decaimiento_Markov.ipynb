# Investigación Avanzada: Nuclear Decaimiento Markov

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt

# Simulate radioactive decay chains A -> B -> C (stable)
# using Markov Chains / Master Equation approach

def decay_derivatives(t, y, lambda_A, lambda_B):
    N_A, N_B, N_C = y
    dN_A = -lambda_A * N_A
    dN_B = lambda_A * N_A - lambda_B * N_B
    dN_C = lambda_B * N_B
    return [dN_A, dN_B, dN_C]

if __name__ == "__main__":
    lambda_A = 0.1
    lambda_B = 0.05
    
    t_span = (0, 100)
    t_eval = np.linspace(t_span[0], t_span[1], 1000)
    
    y0 = [1000, 0, 0]  # Initial populations
    
    sol = integrate.solve_ivp(decay_derivatives, t_span, y0, t_eval=t_eval, args=(lambda_A, lambda_B))
    
    plt.figure(figsize=(8, 5))
    plt.plot(sol.t, sol.y[0], label='Isotope A', color='blue')
    plt.plot(sol.t, sol.y[1], label='Isotope B', color='green')
    plt.plot(sol.t, sol.y[2], label='Isotope C (Stable)', color='red')
    
    plt.title('Nuclear Decay Chain A -> B -> C (Markov/Master Equation)')
    plt.xlabel('Time')
    plt.ylabel('Number of Nuclei')
    plt.legend()
    plt.grid(True)
    plt.show()
```
