# Radiación de Carga Acelerada (Larmor)

Patrón de radiación de antena dipolar clásica.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 200)
P = np.sin(theta)**2

plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, P, color='purple', lw=2)
plt.title('Patrón de Radiación Dipolar (Fórmula de Larmor)')
plt.show()
```
