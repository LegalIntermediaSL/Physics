# Compendio Subtemático: Mecanica Clasica Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    m1 = 5.0
    m2 = 3.0
    g = 9.81
    a = g * (m1 - m2) / (m1 + m2)
    t = np.linspace(0, 5, 100)
    y1 = -0.5 * a * t**2
    y2 = 0.5 * a * t**2
    plt.plot(t, y1, label='Mass 1')
    plt.plot(t, y2, label='Mass 2')
    plt.title('Atwood Machine Kinematics')
    plt.xlabel('Time (s)')
    plt.ylabel('Displacement (m)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
