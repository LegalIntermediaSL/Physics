# Laboratorio Especializado: Mecanica Clasica Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def three_body(t, y, m1, m2, m3, G=1):
    r1, r2, r3 = y[0:2], y[2:4], y[4:6]
    v1, v2, v3 = y[6:8], y[8:10], y[10:12]
    d12 = np.linalg.norm(r2 - r1)**3
    d13 = np.linalg.norm(r3 - r1)**3
    d23 = np.linalg.norm(r3 - r2)**3
    a1 = G * m2 * (r2 - r1) / d12 + G * m3 * (r3 - r1) / d13
    a2 = G * m1 * (r1 - r2) / d12 + G * m3 * (r3 - r2) / d23
    a3 = G * m1 * (r1 - r3) / d13 + G * m2 * (r2 - r3) / d23
    return np.concatenate((v1, v2, v3, a1, a2, a3))

y0 = [-1, 0, 1, 0, 0, 0, 0, -1, 0, 1, 0, 0] # Figure-8 roughly
m1, m2, m3 = 1, 1, 1
t_span = (0, 10)
t_eval = np.linspace(*t_span, 1000)

sol = solve_ivp(three_body, t_span, y0, t_eval=t_eval, args=(m1, m2, m3), rtol=1e-9, atol=1e-9)

plt.figure(figsize=(6,6))
plt.plot(sol.y[0], sol.y[1], label='Body 1')
plt.plot(sol.y[2], sol.y[3], label='Body 2')
plt.plot(sol.y[4], sol.y[5], label='Body 3')
plt.title('Three-Body Problem')
plt.legend()
plt.show()
```
