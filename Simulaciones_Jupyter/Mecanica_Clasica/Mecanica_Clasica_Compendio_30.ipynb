# Compendio Subtemático: Mecanica Clasica Compendio 30

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x = np.linspace(0, 2*np.pi, 200)
    t = 1.0
    k = 2.0
    omega = 3.0
    y_traveling = np.sin(k*x - omega*t)
    
    plt.plot(x, y_traveling)
    plt.title('Snapshot of a Traveling Wave (t=1s)')
    plt.xlabel('Position x (m)')
    plt.ylabel('Displacement y(x,t)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
