# Supercomputación: Mecanica Clasica HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def simulate():
    N = 100
    pos = np.random.rand(N, 2) * 10
    vel = np.zeros((N, 2))
    radius = 0.2
    dt = 0.01
    g = np.array([0, -9.8])
    
    for _ in range(300):
        force = np.zeros((N, 2))
        force += g
        for i in range(N):
            for j in range(i+1, N):
                dr = pos[i] - pos[j]
                r = np.linalg.norm(dr)
                if r < 2*radius:
                    overlap = 2*radius - r
                    force[i] += 1000 * overlap * dr/r
                    force[j] -= 1000 * overlap * dr/r
            if pos[i, 1] < radius:
                force[i, 1] += 1000 * (radius - pos[i, 1])
                
        vel += force * dt
        pos += vel * dt
        
    plt.scatter(pos[:, 0], pos[:, 1])
    plt.savefig("granular_dem.png")

if __name__ == '__main__': simulate()
```
