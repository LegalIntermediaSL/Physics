# Compendio Subtemático: Mecanica Clasica Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    theta = np.linspace(0, np.pi/2, 100)
    mu_s = 0.5
    f_max = mu_s * np.cos(theta)
    f_gravity = np.sin(theta)
    plt.plot(np.degrees(theta), f_max, label='Max Static Friction')
    plt.plot(np.degrees(theta), f_gravity, label='Component of Gravity')
    plt.title('Friction on Inclined Plane')
    plt.xlabel('Angle (degrees)')
    plt.ylabel('Force / mg')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
