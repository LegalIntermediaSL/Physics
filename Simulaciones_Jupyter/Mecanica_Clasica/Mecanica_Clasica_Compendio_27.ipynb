# Compendio Subtemático: Mecanica Clasica Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Physical Pendulum: T = 2*pi*sqrt(I/mgd)
    d = np.linspace(0.1, 1.0, 100)
    M = 1.0
    L = 1.0 # Rod length
    g = 9.81
    I_cm = 1/12 * M * L**2
    I = I_cm + M * d**2 # Parallel axis theorem
    
    T = 2 * np.pi * np.sqrt(I / (M * g * d))
    
    plt.plot(d, T)
    plt.title('Period of a Physical Pendulum (Rod) vs Pivot Distance')
    plt.xlabel('Pivot Distance from CM d (m)')
    plt.ylabel('Period T (s)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
