# Compendio Subtemático: Mecanica Clasica Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Archimedes principle: Buoyant force = weight of displaced fluid
    v_submerged = np.linspace(0, 10, 100) # m^3
    rho_fluid = 1000 # water
    g = 9.81
    
    F_b = rho_fluid * v_submerged * g
    
    plt.plot(v_submerged, F_b)
    plt.title('Buoyant Force vs Submerged Volume')
    plt.xlabel('Submerged Volume (m^3)')
    plt.ylabel('Buoyant Force (N)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
