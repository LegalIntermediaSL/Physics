# Laboratorio Especializado: Mecanica Clasica Especializado 7

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def henon_heiles(t, y):
    x, y_pos, px, py = y
    dx = px
    dy = py
    dpx = -x - 2*x*y_pos
    dpy = -y_pos - x**2 + y_pos**2
    return [dx, dy, dpx, dpy]

E = 1/8
y0 = [0, 0.1, np.sqrt(2*E - 0.1**2 - 2/3*0.1**3), 0]
t_span = (0, 1000)
t_eval = np.linspace(*t_span, 50000)

sol = solve_ivp(henon_heiles, t_span, y0, t_eval=t_eval, rtol=1e-9, atol=1e-9)

crossings = np.where(np.diff(np.sign(sol.y[0])) > 0)[0]
y_sec = sol.y[1][crossings]
py_sec = sol.y[3][crossings]

plt.figure(figsize=(6,6))
plt.plot(y_sec, py_sec, '.', markersize=2)
plt.title("Poincaré Section of Hénon-Heiles System ($x=0, p_x>0$)")
plt.xlabel("$y$")
plt.ylabel("$p_y$")
plt.show()
```
