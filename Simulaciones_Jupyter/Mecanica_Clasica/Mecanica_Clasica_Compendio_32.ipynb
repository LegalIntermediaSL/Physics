# Compendio Subtemático: Mecanica Clasica Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Doppler effect: f' = f * (v + v_o) / (v - v_s)
    v_sound = 343.0
    f_source = 500.0
    v_source = np.linspace(-100, 100, 200) # Source moving towards (-) and away (+)
    
    # Observer stationary
    f_obs = f_source * v_sound / (v_sound + v_source)
    
    plt.plot(v_source, f_obs)
    plt.axvline(0, color='r', linestyle='--')
    plt.title('Doppler Effect (Stationary Observer, Moving Source)')
    plt.xlabel('Source Velocity (m/s) [negative = towards obs]')
    plt.ylabel('Observed Frequency (Hz)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
