# Compendio Subtemático: Mecanica Clasica Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Rocket equation: v(t) = v0 + ve * ln(m0 / m(t))
    t = np.linspace(0, 50, 100)
    m0 = 1000  # Initial mass
    dm_dt = 10 # Mass loss rate
    ve = 2500  # Exhaust velocity
    m = m0 - dm_dt * t
    v = ve * np.log(m0 / m)
    plt.plot(t, v)
    plt.title('Rocket Velocity in Gravity-Free Space')
    plt.xlabel('Time (s)')
    plt.ylabel('Velocity (m/s)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
