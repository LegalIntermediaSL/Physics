# Compendio Subtemático: Mecanica Clasica Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Conservation of angular momentum: I1 * w1 = I2 * w2
    # Skater pulls arms in: I decreases, w increases
    I_initial = 10.0
    w_initial = 2.0
    L = I_initial * w_initial
    
    I_final = np.linspace(10.0, 2.0, 100) # I decreasing
    w_final = L / I_final
    
    plt.plot(I_final, w_final)
    plt.gca().invert_xaxis() # Show decreasing I
    plt.title('Conservation of Angular Momentum')
    plt.xlabel('Moment of Inertia (kg m^2)')
    plt.ylabel('Angular Velocity (rad/s)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
