# Compendio Subtemático: Mecanica Clasica Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 100)
    alpha = 2.0  # Constant angular acceleration
    omega0 = 0
    theta = omega0 * t + 0.5 * alpha * t**2
    omega = omega0 + alpha * t
    
    plt.plot(t, theta, label='Angular Position (rad)')
    plt.plot(t, omega, label='Angular Velocity (rad/s)')
    plt.title('Rotational Kinematics')
    plt.xlabel('Time (s)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
