# Compendio Subtemático: Mecanica Clasica Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x = np.linspace(0, 10, 100)
    F = 2 * x # Variable force
    W = np.cumsum(F * (x[1]-x[0])) # Work done
    plt.plot(x, F, label='Force (N)')
    plt.plot(x, W, label='Work Done (J)')
    plt.title('Work-Energy Theorem')
    plt.xlabel('Position (m)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
