# Laboratorio Especializado: Mecanica Clasica Especializado 5

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def euler_equations(t, w, I1, I2, I3):
    w1, w2, w3 = w
    dw1 = (I2 - I3) / I1 * w2 * w3
    dw2 = (I3 - I1) / I2 * w3 * w1
    dw3 = (I1 - I2) / I3 * w1 * w2
    return [dw1, dw2, dw3]

I1, I2, I3 = 1.0, 2.0, 3.0
w0 = [0.1, 10.0, 0.1]
t_span = (0, 20)
t_eval = np.linspace(0, 20, 1000)

sol = solve_ivp(euler_equations, t_span, w0, t_eval=t_eval, args=(I1, I2, I3))

plt.plot(sol.t, sol.y[0], label='$\omega_1$')
plt.plot(sol.t, sol.y[1], label='$\omega_2$')
plt.plot(sol.t, sol.y[2], label='$\omega_3$')
plt.title('Dzhanibekov Effect (Intermediate Axis Theorem)')
plt.xlabel('Time')
plt.ylabel('Angular Velocity')
plt.legend()
plt.show()
```
