# Compendio Subtemático: Mecanica Clasica Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Torque = r x F
    r = np.array([2, 0, 0])
    F = np.array([0, 5, 0])
    tau = np.cross(r, F)
    print(f"Torque vector: {tau}")
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.quiver(0, 0, 0, r[0], r[1], r[2], color='r', label='r vector')
    ax.quiver(r[0], r[1], r[2], F[0], F[1], F[2], color='b', label='F vector')
    ax.quiver(0, 0, 0, tau[0], tau[1], tau[2], color='g', label='Torque vector')
    ax.set_xlim([-1, 3])
    ax.set_ylim([-1, 6])
    ax.set_zlim([-1, 12])
    plt.legend()
    plt.title('Torque Vector Cross Product')
    plt.show()

if __name__ == '__main__':
    main()
```
