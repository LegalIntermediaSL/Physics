# Compendio Subtemático: Mecanica Clasica Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Angular Momentum L = I * omega
    I = 5.0
    omega = np.linspace(0, 20, 100)
    L = I * omega
    plt.plot(omega, L)
    plt.title('Angular Momentum vs Angular Velocity')
    plt.xlabel('Angular Velocity (rad/s)')
    plt.ylabel('Angular Momentum (kg m^2/s)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
