# Compendio Subtemático: Mecanica Clasica Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 5, 100)
    v0 = 20
    theta = np.radians(45)
    g = 9.81
    x = v0 * np.cos(theta) * t
    y = v0 * np.sin(theta) * t - 0.5 * g * t**2
    plt.plot(x[y>=0], y[y>=0])
    plt.title('Projectile Motion')
    plt.xlabel('Distance (m)')
    plt.ylabel('Height (m)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
