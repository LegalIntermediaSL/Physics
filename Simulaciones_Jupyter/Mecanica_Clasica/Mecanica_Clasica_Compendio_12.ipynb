# Compendio Subtemático: Mecanica Clasica Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Positions and masses
    m = np.array([1, 2, 3])
    x = np.array([0, 2, 4])
    y = np.array([0, 1, 0])
    x_cm = np.sum(m * x) / np.sum(m)
    y_cm = np.sum(m * y) / np.sum(m)
    
    plt.scatter(x, y, s=m*100, label='Masses')
    plt.scatter(x_cm, y_cm, color='red', marker='x', s=100, label='Center of Mass')
    plt.title('Center of Mass of a System of Particles')
    plt.xlabel('X position')
    plt.ylabel('Y position')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
