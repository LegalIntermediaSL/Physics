# Oscilaciones de Langmuir

Ondas electrostáticas debido al desplazamiento coherente de electrones.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 300)
E = np.exp(-0.1*t) * np.cos(5*t)

plt.figure(figsize=(8,4))
plt.plot(t, E, 'm-', lw=2)
plt.title('Amortiguamiento del Campo Eléctrico en Plasma')
plt.xlabel('Tiempo')
plt.grid()
plt.show()
```
