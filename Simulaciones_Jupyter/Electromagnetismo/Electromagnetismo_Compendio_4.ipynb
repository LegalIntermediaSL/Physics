# Compendio Subtemático: Electromagnetismo Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x, y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
    Ex = np.zeros_like(x)
    Ey = np.zeros_like(y)
    for qx in np.linspace(-4, 4, 30):
        r1 = np.sqrt((x-qx)**2 + (y-2)**2)
        r2 = np.sqrt((x-qx)**2 + (y+2)**2)
        r1[r1 == 0] = 1e-10
        r2[r2 == 0] = 1e-10
        Ex += (x-qx)/r1**3 - (x-qx)/r2**3
        Ey += (y-2)/r1**3 - (y+2)/r2**3
    plt.quiver(x, y, Ex, Ey, color='m')
    plt.title("4. Electric Field of a Parallel Plate Capacitor")
    plt.show()

if __name__ == "__main__":
    main()
```
