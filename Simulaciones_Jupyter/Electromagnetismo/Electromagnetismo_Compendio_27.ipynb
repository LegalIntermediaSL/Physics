# Compendio Subtemático: Electromagnetismo Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    N = np.linspace(10, 1000, 100)
    mu0 = 4*np.pi*1e-7
    A = 0.01
    l = 0.5
    L = (mu0 * N**2 * A) / l
    plt.plot(N, L)
    plt.xlabel('Number of Turns (N)')
    plt.ylabel('Inductance (H)')
    plt.title("27. Inductance of a Solenoid vs N")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
