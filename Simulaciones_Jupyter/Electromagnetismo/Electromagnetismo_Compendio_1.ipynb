# Compendio Subtemático: Electromagnetismo Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x, y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
    r = np.sqrt(x**2 + y**2)
    r[r == 0] = 1e-10
    Ex = x / r**3
    Ey = y / r**3
    plt.quiver(x, y, Ex, Ey, color='r')
    plt.title("1. Electric Field of a Point Charge")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

if __name__ == "__main__":
    main()
```
