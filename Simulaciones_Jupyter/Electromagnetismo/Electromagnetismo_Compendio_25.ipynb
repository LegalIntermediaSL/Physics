# Compendio Subtemático: Electromagnetismo Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 1000)
    q = 1.0
    m = 1.0
    B = 1.0
    E = 0.5
    omega = q*B/m
    v_drift = E/B
    x = v_drift*t - (v_drift/omega)*np.sin(omega*t)
    y = (v_drift/omega)*(1 - np.cos(omega*t))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("25. Lorentz Force - Crossed E and B Fields")
    plt.axis('equal')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
