# Compendio Subtemático: Electromagnetismo Compendio 31

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    theta1 = np.linspace(0, np.pi/2, 100)
    n1, n2 = 1.0, 1.5
    theta2 = np.arcsin(n1/n2 * np.sin(theta1))
    plt.plot(np.degrees(theta1), np.degrees(theta2))
    plt.xlabel('Incident Angle (deg)')
    plt.ylabel('Refracted Angle (deg)')
    plt.title("31. Snell's Law - Refraction")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
