# Compendio Subtemático: Electromagnetismo Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    z = np.linspace(0, 4*np.pi, 200)
    E = np.sin(z)
    B = np.sin(z)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(z, E, 0, color='r', label='E field')
    ax.plot(z, 0, B, color='b', label='B field')
    ax.set_xlabel('Propagation (z)')
    ax.set_ylabel('E (x)')
    ax.set_zlabel('B (y)')
    plt.title("28. Electromagnetic Wave Propagation")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
