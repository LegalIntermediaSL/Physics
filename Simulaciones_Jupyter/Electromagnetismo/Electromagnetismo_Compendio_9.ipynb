# Compendio Subtemático: Electromagnetismo Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 5, 100)
    R = 2.0
    L = 1.0
    tau = L/R
    V0 = 10.0
    I = (V0/R) * (1 - np.exp(-t/tau))
    plt.plot(t, I, label='Current')
    plt.xlabel('Time (s)')
    plt.ylabel('Current (A)')
    plt.title("9. RL Circuit Energizing")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
