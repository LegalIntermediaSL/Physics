# Compendio Subtemático: Electromagnetismo Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x, y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
    d = 1.0
    r1 = np.sqrt((x-d)**2 + y**2)
    r2 = np.sqrt((x+d)**2 + y**2)
    r1[r1 == 0] = 1e-10
    r2[r2 == 0] = 1e-10
    Ex = (x-d)/r1**3 - (x+d)/r2**3
    Ey = y/r1**3 - y/r2**3
    plt.quiver(x, y, Ex, Ey, color='b')
    plt.title("2. Electric Field of a Dipole")
    plt.show()

if __name__ == "__main__":
    main()
```
