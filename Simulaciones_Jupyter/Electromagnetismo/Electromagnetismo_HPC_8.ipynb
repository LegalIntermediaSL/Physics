# Supercomputación: Electromagnetismo HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def simulate():
    N = 100
    x = np.linspace(-5, 5, N)
    y = np.linspace(-5, 5, N)
    X, Y = np.meshgrid(x, y)
    
    r2 = X**2 + Y**2 + 1e-6
    Ax = -Y / r2
    Ay = X / r2
    
    plt.streamplot(X, Y, Ax, Ay)
    plt.savefig("aharonov_bohm.png")

if __name__ == '__main__': simulate()
```
