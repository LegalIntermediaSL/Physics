# Compendio Subtemático: Electromagnetismo Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 0.1, 500)
    f = 50.0
    omega = 2*np.pi*f
    V0 = 10.0
    C = 1e-3
    V = V0 * np.sin(omega*t)
    I = V0 * omega * C * np.cos(omega*t)
    plt.plot(t, V, label='Voltage (V)')
    plt.plot(t, I, label='Current (A)')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title("15. AC Circuit with Capacitor (I leads V by 90 deg)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
