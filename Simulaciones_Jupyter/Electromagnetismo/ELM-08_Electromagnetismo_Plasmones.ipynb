# Investigación Avanzada: Electromagnetismo Plasmones

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def drude_permittivity(omega, omega_p, gamma):
    return 1 - omega_p**2 / (omega**2 + 1j * gamma * omega)

def spp_dispersion(omega, eps_m, eps_d):
    c = 3e8
    k0 = omega / c
    return k0 * np.sqrt((eps_m * eps_d) / (eps_m + eps_d))

def main():
    omega = np.linspace(0.1e15, 6e15, 1000)
    omega_p = 13.8e15 # Gold plasma frequency
    gamma = 1.0e14 # Collision frequency
    eps_d = 1.0 # Air
    
    eps_m = drude_permittivity(omega, omega_p, gamma)
    k_spp = spp_dispersion(omega, eps_m, eps_d)
    
    plt.figure(figsize=(8,6))
    plt.plot(np.real(k_spp) / 1e7, omega / 1e15, label='SPP Dispersion', color='b')
    plt.plot(omega / 3e8 / 1e7, omega / 1e15, '--', color='r', label='Light Line')
    plt.xlabel(r'Wave vector Re($k_{spp}$) ($10^7 m^{-1}$)')
    plt.ylabel(r'Frequency $\omega$ ($10^{15} s^{-1}$)')
    plt.title('Surface Plasmon Polariton Dispersion (Drude Model)')
    plt.legend()
    plt.grid(True)
    plt.ylim(0, 6)
    plt.xlim(0, 5)
    plt.show()

if __name__ == "__main__":
    main()
```
