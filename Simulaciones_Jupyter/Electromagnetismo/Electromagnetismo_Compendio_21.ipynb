# Compendio Subtemático: Electromagnetismo Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    z = np.linspace(-5, 5, 100)
    R = 2.0
    mu0 = 4*np.pi*1e-7
    I = 1.0
    B = (mu0 * I * R**2) / (2 * (z**2 + R**2)**1.5)
    plt.plot(z, B)
    plt.xlabel('Distance along axis z')
    plt.ylabel('Magnetic Field B_z')
    plt.title("21. Magnetic Field on Axis of a Circular Loop")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
