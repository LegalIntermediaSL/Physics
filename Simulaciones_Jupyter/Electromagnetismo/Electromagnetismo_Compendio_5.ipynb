# Compendio Subtemático: Electromagnetismo Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)
    R = np.sqrt(X**2 + Y**2)
    R[R == 0] = 1e-10
    V = 1.0 / R
    plt.contourf(X, Y, V, levels=50, cmap='viridis')
    plt.colorbar(label='Potential (V)')
    plt.title("5. Electric Potential of a Point Charge")
    plt.show()

if __name__ == "__main__":
    main()
```
