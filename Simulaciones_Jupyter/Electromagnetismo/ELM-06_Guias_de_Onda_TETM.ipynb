# Modos TE y TM en Guías de Onda

Cálculo de los nodos y antinodos en cavidades rectangulares.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi, 100)
y = np.linspace(0, np.pi, 100)
X, Y = np.meshgrid(x, y)
E_z = np.sin(X) * np.sin(2*Y)

plt.figure(figsize=(6,5))
plt.contourf(X, Y, E_z, cmap='coolwarm')
plt.title('Modo TM_12 en una Guía de Ondas Rectangular')
plt.colorbar(label='Amplitud del Campo')
plt.show()
```
