# Laboratorio Especializado: Electromagnetismo Especializado 7

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def biot_savart(r_eval, R_coil, Z_coil, I=1.0):
    theta = np.linspace(0, 2*np.pi, 100)
    dl = np.array([-np.sin(theta), np.cos(theta), np.zeros_like(theta)]) * R_coil * (2*np.pi/100)
    r_source = np.array([R_coil*np.cos(theta), R_coil*np.sin(theta), np.ones_like(theta)*Z_coil])
    
    B = np.zeros(3)
    for i in range(100):
        r_vec = r_eval - r_source[:, i]
        r_mag = np.linalg.norm(r_vec)
        B += (1e-7 * I) * np.cross(dl[:, i], r_vec) / r_mag**3
    return B

y = np.linspace(-2, 2, 20)
z = np.linspace(-2, 2, 20)
Y, Z = np.meshgrid(y, z)
By = np.zeros_like(Y)
Bz = np.zeros_like(Z)

for i in range(20):
    for j in range(20):
        r_eval = np.array([0, Y[i,j], Z[i,j]])
        B1 = biot_savart(r_eval, R_coil=1.0, Z_coil=0.5)
        B2 = biot_savart(r_eval, R_coil=1.0, Z_coil=-0.5)
        B = B1 + B2
        By[i,j] = B[1]
        Bz[i,j] = B[2]

plt.streamplot(Y, Z, By, Bz)
plt.title('Magnetic Field of Helmholtz Coils')
plt.xlabel('y')
plt.ylabel('z')
plt.show()
```
