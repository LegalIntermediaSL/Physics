# Compendio Subtemático: Electromagnetismo Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    H = np.linspace(-10, 10, 500)
    B_up = np.tanh(H + 2)
    B_down = np.tanh(H - 2)
    plt.plot(H, B_up, color='b', label='Magnetization')
    plt.plot(H, B_down, color='b')
    # connect the ends
    plt.plot([H[-1], H[-1]], [B_up[-1], B_down[-1]], color='b')
    plt.plot([H[0], H[0]], [B_up[0], B_down[0]], color='b')
    plt.xlabel('Magnetic Field Strength (H)')
    plt.ylabel('Magnetic Flux Density (B)')
    plt.title("35. Magnetic Hysteresis Loop")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
