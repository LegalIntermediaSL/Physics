# D-Branas interactuando a través de Cuerdas Abiertas
Simulación interactiva avanzada.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Dibujar dos planos (D2-branas) y una cuerda abierta entre ellos
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
Z1 = np.ones_like(X)
Z2 = -np.ones_like(X)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, alpha=0.3, color='blue')
ax.plot_surface(X, Y, Z2, alpha=0.3, color='red')

# Cuerda abierta
t = np.linspace(-1, 1, 50)
cx = 0.5 * np.sin(np.pi * t)
cy = 0.5 * np.cos(np.pi * t)
cz = t
ax.plot(cx, cy, cz, 'k-', linewidth=3, label='Cuerda Abierta (Vector Gauge)')

ax.set_title('Dos D2-Branas intercambiando interacciones')
ax.legend()
ax.axis('off')
plt.show()
```
