# [QG-07] Evolución de Espuma de Espín

Simulación interactiva generada para el bloque de Gravedad Cuántica de Lazos (LQG).

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Simulador de Vértice de Espuma de Espín (Spin Foam Vertex)
# Visualizamos cómo una red de espín bidimensional evoluciona en el tiempo a través de un vértice cuántico.

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Coordenadas de los nodos iniciales y finales
nodes_init = np.array([[-1, -1, 0], [1, -1, 0], [0, 1, 0]])
nodes_final = np.array([[-1, 1, 2], [1, 1, 2], [0, -1, 2]])
vertex = np.array([0, 0, 1])  # El evento de transición

# Dibujar las redes (Spin Networks)
for i in range(3):
    ax.scatter(*nodes_init[i], color='blue', s=100, label='Nodos Iniciales' if i==0 else "")
    ax.scatter(*nodes_final[i], color='green', s=100, label='Nodos Finales' if i==0 else "")

ax.scatter(*vertex, color='red', s=150, marker='*', label='Vértice de Spin Foam')

# Dibujar las caras de evolución (Path Integral)
for i in range(3):
    ax.plot([nodes_init[i][0], vertex[0]], [nodes_init[i][1], vertex[1]], [nodes_init[i][2], vertex[2]], 'k--')
    ax.plot([vertex[0], nodes_final[i][0]], [vertex[1], nodes_final[i][1]], [vertex[2], nodes_final[i][2]], 'r-')

ax.set_title('Evolución Dinámica: Vértice de Espuma de Espín (Spin Foam)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Tiempo (T)')
ax.legend()
plt.show()

```
