# [QG-08] Espectro Cuantizado de Área

Simulación interactiva generada para el bloque de Gravedad Cuántica de Lazos (LQG).

```python
import numpy as np
import matplotlib.pyplot as plt

# Espectro del Operador de Área en Gravedad Cuántica de Lazos (Rovelli-Smolin)
# A = 8 * pi * gamma * l_p^2 * sum(sqrt(j * (j + 1)))

gamma = 0.2375  # Parámetro de Barbero-Immirzi
j_vals = np.arange(0.5, 10.5, 0.5)  # Espines (1/2, 1, 3/2, 2...)

def area_eigenvalue(j):
    return 8 * np.pi * gamma * np.sqrt(j * (j + 1))

areas = [area_eigenvalue(j) for j in j_vals]

plt.figure(figsize=(10, 6))
plt.step(j_vals, areas, where='mid', color='purple', linewidth=2)
plt.plot(j_vals, areas, 'o', color='black')

plt.title('Espectro Discreto del Operador de Área en LQG')
plt.xlabel('Espin (j)')
plt.ylabel('Autovalor del Área (Unidades de $l_P^2$)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.annotate('El espacio salta en cuantos discretos,\nno existe el área continua.', 
             xy=(2, 10), xytext=(4, 5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()

```
