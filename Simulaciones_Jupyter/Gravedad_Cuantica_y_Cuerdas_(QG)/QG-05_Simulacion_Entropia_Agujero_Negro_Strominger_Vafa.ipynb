# Entropía Microscópica de D-Branas (Strominger-Vafa)
Simulación interactiva avanzada.

```python
import numpy as np
import matplotlib.pyplot as plt

# La entropía en microestados de cuerdas crece S ~ sqrt(N) para estados BPS
N = np.linspace(1, 1000, 100)
S_micro = 2 * np.pi * np.sqrt(N)
S_macro = S_micro + np.random.normal(0, 0.5, size=N.shape) # Añadimos ruido para visualización

plt.figure(figsize=(8, 5))
plt.plot(N, S_micro, 'k-', linewidth=2, label='Predicción Cuerdas (Microestados)')
plt.plot(N, S_macro, 'r.', alpha=0.5, label='Entropía de Bekenstein-Hawking (Macroscópica)')
plt.title('Concordancia de Entropía Holográfica en Teoría de Cuerdas')
plt.xlabel('Carga (N)')
plt.ylabel('Entropía $S$')
plt.legend()
plt.grid(True)
plt.show()
```
