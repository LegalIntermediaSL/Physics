# Tensión de la Cuerda vs Escala de Energía
Simulación interactiva avanzada.

```python
import numpy as np
import matplotlib.pyplot as plt

# A bajas energías la cuerda parece una partícula, a altas se revela su extensión
E = np.logspace(0, 19, 100) # Energía de 1 GeV a la masa de Planck 10^19 GeV
M_p = 1e19
# Sección eficaz gravitacional de cuerdas decae exponencialmente a altas energías (Comportamiento Veneziano)

sigma_pt = 1 / E**2
sigma_string = np.exp(-E/M_p) / E**2

plt.figure(figsize=(8, 5))
plt.loglog(E, sigma_pt, 'b--', label='Partícula Puntual (Divergencia UV)')
plt.loglog(E, sigma_string, 'r-', linewidth=2, label='Supercuerda (Suavizamiento UV)')
plt.title('Comportamiento de Dispersión a Altas Energías (Amplitud de Veneziano)')
plt.xlabel('Energía (GeV)')
plt.ylabel('Sección Eficaz $\\sigma$')
plt.legend()
plt.grid(True)
plt.show()
```
