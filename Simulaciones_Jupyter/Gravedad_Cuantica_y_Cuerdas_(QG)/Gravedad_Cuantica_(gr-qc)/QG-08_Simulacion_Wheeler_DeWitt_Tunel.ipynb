# [QG-08] Gravedad Cuántica: Minisuperespacio de Wheeler-DeWitt

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags
from scipy.sparse.linalg import eigs

# Gravedad Cuántica: Ecuación Cosmólogica de Wheeler-DeWitt (Minisuperespacio)
# H * Psi(a) = 0
# Resolvemos el origen del Universo desde la 'Nada' Matemática mediante
# Efecto Túnel Cuántico sobre el factor de escala cósmica "a".

# Espacio discretizado para el factor de escala 'a' (El tamaño del Universo)
N = 1000
a_min, a_max = 0.001, 3.0
a = np.linspace(a_min, a_max, N)
da = a[1] - a[0]

# Potencial Cosmológico (Modelo simplificado de Vilenkin)
# La relatividad general crea una "barrera" gravitacional repulsiva 
# (Constante Cosmológica) más un término de curvatura.
# V(a) = a^2 - Lambda * a^4
Lambda = 0.5
V = a**2 - Lambda * a**4

# Operador Cinético de Wheeler-DeWitt (derivada segunda de 'a')
# d2/da2
diagonals = [np.ones(N-1), -2*np.ones(N), np.ones(N-1)]
D2 = diags(diagonals, offsets=[1, 0, -1]) / (da**2)

# Hamiltoniano del Universo H_WDW = -D2 + V(a)
# (Técnicamente, en unidades de Planck y simplificaciones canónicas)
H_WDW = -D2 + diags(V, 0)

# Calculamos la Función de Onda del Universo Psi(a) 
# Buscamos los estados con E=0 (Por constricción hamiltoniana de la Relatividad General)
eigenvalues, eigenvectors = eigs(H_WDW, k=5, sigma=0.0)

# Seleccionamos la solución de más baja 'energía' pseudo-E
idx = np.argmin(np.abs(eigenvalues))
Psi = np.abs(eigenvectors[:, idx])
Psi = Psi / np.max(Psi) # Normalizamos

plt.figure(figsize=(10, 6))

plt.plot(a, V, 'r--', lw=2, label="Potencial Cosmológico $V(a)$ (Curvatura + $\Lambda$)")
plt.plot(a, Psi, 'b-', lw=3, label="Función de Onda Cosmológica $|\Psi(a)|$")

plt.fill_between(a, 0, Psi, color='blue', alpha=0.2)
plt.axhline(0, color='black', lw=1)

plt.xlim(0, 2.5)
plt.ylim(-1, 1.2)
plt.title("Gravedad Cuántica: El Origen del Universo por Efecto Túnel", fontsize=14)
plt.xlabel("Factor de Escala del Universo $a$ (Tamaño Cósmico)")
plt.ylabel("Amplitud Cuántica / Potencial")
plt.legend()
plt.grid(True)
plt.show()

# GRAVEDAD CUÁNTICA: La barrera roja es infranqueable para la física clásica.
# Matemáticamente, un Universo clásico de tamaño cero (a=0, La Nada Absoluta) 
# no puede saltar la barrera ni existir.
# Pero la línea azul (Función de Onda de Wheeler-DeWitt) demuestra que el espacio-tiempo 
# puede sufrir "Efecto Túnel". El Universo "Nace de la nada" probabilísticamente 
# y, al cruzar la barrera, la expansión cósmica (Inflación) toma el control (a -> infinito).

```
