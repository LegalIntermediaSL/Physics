# [QG-11] Visualizador de Redes de Espín (Gravedad Cuántica de Lazos)

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

# Simulación visual de una Red de Espín (Spin Network) de la Gravedad Cuántica de Lazos (LQG)
# En LQG, el espacio no es un fondo continuo, sino una superposición cuántica de estas redes discretas.
# Nodos = "Átomos" de volumen (nudos de 3 o más enlaces).
# Enlaces (links) = "Átomos" de área que conectan los volúmenes, representados por números cuánticos de espín (j).

np.random.seed(42)
G = nx.random_geometric_graph(30, 0.3)

plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, dim=2, seed=42)

# Coloreamos los nodos y enlaces según sus "cuantos" de espín
node_sizes = [300 * np.random.uniform(0.5, 2.0) for _ in range(len(G.nodes))]
node_colors = [np.random.randint(1, 10) for _ in range(len(G.nodes))]

edge_colors = [np.random.randint(1, 10) for _ in range(len(G.edges))]
edge_weights = [np.random.uniform(1, 3) for _ in range(len(G.edges))]

nodes = nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, cmap=plt.cm.plasma, alpha=0.9)
edges = nx.draw_networkx_edges(G, pos, width=edge_weights, edge_color=edge_colors, edge_cmap=plt.cm.viridis, alpha=0.7)

# Etiquetas de Espín simuladas en los enlaces
edge_labels = {edge: f'j={np.random.choice([0.5, 1, 1.5, 2])}' for edge in G.edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=7)

plt.title("Gravedad Cuántica de Lazos: Estructura Discreta del Espacio (Spin Network)", fontsize=14)
plt.axis("off")

# Barra de color para representar el "volumen" de los nodos
cbar = plt.colorbar(nodes)
cbar.set_label('Quantum Volume Eigenvalues (Nodos)', rotation=270, labelpad=15)
plt.show()

```
