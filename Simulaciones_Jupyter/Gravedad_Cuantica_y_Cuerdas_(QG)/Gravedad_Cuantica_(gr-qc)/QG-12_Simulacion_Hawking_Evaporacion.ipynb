# [QG-12] Termodinámica: Evaporación de Agujero Negro

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Termodinámica de Agujeros Negros y Radiación de Hawking
# Simulación Numérica de la Evaporación Cuántica (Explosión final)

# La pérdida de masa por radiación térmica de cuerpo negro viene de la Ley de Stefan-Boltzmann.
# dM/dt = - K / M^2 (Donde K es una constante proporcional a hbar*c^4/G^2)
K = 1.0 # Para simplificar numéricamente

def hawking_evaporation(M, t):
    if M <= 1e-3: return 0.0 # El agujero negro se ha evaporado (detenemos derivadas)
    dM_dt = -K / (M**2)
    return dM_dt

# Masa Inicial del Agujero Negro Primordial
M_0 = 10.0
# El tiempo de evaporación total escala como t_evap = (M_0^3) / (3K)
t_evap_theory = (M_0**3) / (3 * K)
time = np.linspace(0, t_evap_theory * 1.05, 5000)

sol = odeint(hawking_evaporation, M_0, time)
mass = sol[:, 0]
mass[mass < 1e-3] = np.nan # Filtrar masa cero o negativa

# La Temperatura de Hawking es inversamente proporcional a la masa T ~ 1/M
temperature = 1.0 / (mass + 1e-6)

fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:blue'
ax1.set_xlabel('Tiempo $t$ (Milisegundos finales)', fontsize=12)
ax1.set_ylabel('Masa del Agujero Negro ($M$)', color=color, fontsize=12)
ax1.plot(time, mass, color=color, lw=3, label='Masa $M(t)$')
ax1.tick_params(axis='y', labelcolor=color)

# Instanciamos un segundo eje y
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Temperatura de Hawking ($T$)', color=color, fontsize=12)
ax2.plot(time, temperature, color=color, lw=3, linestyle='--', label='Temperatura $T(t) \propto 1/M$')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_yscale('log') # Eje logarítmico para ver el estallido térmico a infinito

plt.title("Gravedad Cuántica: Evaporación y Explosión de un Agujero Negro", fontsize=14)
fig.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()
# FÍSICA: En los milisegundos finales antes de desaparecer (M->0), 
# la temperatura tiende al infinito (T->inf), emitiendo un flash de rayos gamma brutal.

```
