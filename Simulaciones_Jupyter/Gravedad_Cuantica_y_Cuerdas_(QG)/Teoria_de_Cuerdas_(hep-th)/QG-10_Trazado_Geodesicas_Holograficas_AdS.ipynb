# [QG-10] Geodésicas en AdS y Principio Holográfico

```python
import numpy as np
import matplotlib.pyplot as plt

# Geometría Hiperbólica: Disco de Poincaré (Proyección 2D del Espacio AdS)
# Trazamos geodésicas que en AdS son arcos de circunferencia ortogonales al borde (R=1).

theta = np.linspace(0, 2*np.pi, 100)
boundary_x = np.cos(theta)
boundary_y = np.sin(theta)

plt.figure(figsize=(8,8))
plt.plot(boundary_x, boundary_y, 'k-', lw=3, label="Frontera Conforme (CFT)")

# Trazado de geodésicas (Arcos ortogonales a la frontera)
for i in range(5):
    # Centro del círculo de la geodésica
    r_c = 1.0 + np.random.rand() * 1.5
    th_c = np.random.rand() * 2 * np.pi
    cx, cy = r_c * np.cos(th_c), r_c * np.sin(th_c)
    
    # Radio geométrico para que sea ortogonal: R_g = sqrt(r_c^2 - R_ads^2)
    R_g = np.sqrt(r_c**2 - 1.0)
    
    # Pintamos la geodésica interceptada con el disco
    arc_th = np.linspace(0, 2*np.pi, 200)
    ax, ay = cx + R_g * np.cos(arc_th), cy + R_g * np.sin(arc_th)
    
    # Máscara interior del disco AdS
    mask = (ax**2 + ay**2) <= 1.0
    plt.plot(ax[mask], ay[mask], lw=2)

plt.title("Holografía: Geodésicas en el Espacio Anti-de Sitter (AdS)")
plt.axis("equal")
plt.axis("off")
plt.legend(loc='upper right')
plt.show()

```
