# Coordenadas de Kruskal-Szekeres

Extensión analítica máxima del agujero negro de Schwarzschild.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

u = np.linspace(-3, 3, 100)
v = np.sqrt(1 + u**2)

plt.figure(figsize=(6,6))
plt.plot(u, v, 'r', lw=2, label='Singularidad (r=0)')
plt.plot(u, -v, 'r', lw=2)
plt.plot(u, u, 'k--', label='Horizonte de Eventos (r=2M)')
plt.plot(u, -u, 'k--')
plt.title('Diagrama de Kruskal del Agujero Negro')
plt.legend()
plt.grid(True)
plt.show()
```
