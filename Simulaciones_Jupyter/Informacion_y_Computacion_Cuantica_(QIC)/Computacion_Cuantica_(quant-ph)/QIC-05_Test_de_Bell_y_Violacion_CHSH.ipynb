# [QIC-05] Información Cuántica: Test de Bell

```python
import numpy as np
import matplotlib.pyplot as plt

# Mecánica Cuántica: Test de las Desigualdades de Bell (Límite CHSH)
# Simulamos el experimento de Aspect (Nobel 2022) midiendo la correlación
# del espín de fotones entrelazados en distintos ángulos de polarización.

# Generamos un par de fotones entrelazados en el estado singlete de Bell:
# |Psi> = 1/sqrt(2) * (|01> - |10>)
# Para este estado, la correlación mecanocuántica teórica entre Alice (ángulo a) 
# y Bob (ángulo b) es: E(a,b) = -cos(a-b).

def quantum_correlation(a, b):
    # La mecánica cuántica predice esta correlación exacta:
    return -np.cos(a - b)

# El Límite CHSH: S = |E(a,b) - E(a,b') + E(a',b) + E(a',b')|
# El Realismo Local de Einstein exige matemáticamente S <= 2
# La Mecánica Cuántica permite que S alcance 2 * sqrt(2) ≈ 2.828 (Límite Tsirelson)

# Escogemos los ángulos óptimos para maximizar la violación cuántica:
# a = 0, a' = pi/2
# b = pi/4, b' = 3pi/4
a = 0.0
a_prime = np.pi / 2
b = np.pi / 4
b_prime = 3 * np.pi / 4

# Calculamos las correlaciones cuánticas exactas
E_ab = quantum_correlation(a, b)
E_abp = quantum_correlation(a, b_prime)
E_apb = quantum_correlation(a_prime, b)
E_apbp = quantum_correlation(a_prime, b_prime)

S_quantum = abs(E_ab - E_abp + E_apb + E_apbp)

print(f"Límite Matemático Clásico (Realismo Local EPR): S_max = 2.000")
print(f"Valor Obtenido Computacionalmente (Cuántica): S = {S_quantum:.3f}")

# Visualización del barrido de S según el ángulo para mostrar la violación
theta = np.linspace(0, np.pi, 100)
# Dejamos a=0, a'=pi/2, b=theta, b'=theta+pi/2
S_sweep = np.abs(quantum_correlation(0, theta) - 
                 quantum_correlation(0, theta + np.pi/2) + 
                 quantum_correlation(np.pi/2, theta) + 
                 quantum_correlation(np.pi/2, theta + np.pi/2))

plt.figure(figsize=(8, 5))
plt.plot(theta, S_sweep, 'purple', lw=3, label="Predicción Cuántica $|S(\theta)|$")
plt.axhline(2.0, color='red', linestyle='--', lw=2, label="Límite Clásico de Bell (S=2)")
plt.axhline(2.828, color='green', linestyle=':', lw=2, label="Límite Tsirelson (S=2.82)")
plt.title("Violación Cuántica de la Desigualdad de Bell (CHSH)", fontsize=14)
plt.xlabel("Ángulo de barrido $\theta$ (radianes)")
plt.ylabel("Valor Estadístico S")
plt.legend()
plt.grid(True)
plt.show()

# FÍSICA: Puesto que la curva púrpura supera la línea roja punteada (límite clásico de Einstein),
# la física demuestra que no existen "Variables Ocultas Locales". O bien la información
# viaja instantáneamente violando la relatividad (No-Localidad), o bien las propiedades 
# no están definidas hasta medirlas (No-Realismo).

```
