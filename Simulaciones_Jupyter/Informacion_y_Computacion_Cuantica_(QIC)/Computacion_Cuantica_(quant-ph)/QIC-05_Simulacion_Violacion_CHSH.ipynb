# [QIC-05] Simulador Cuántico: Desigualdad CHSH

```python
import numpy as np
import matplotlib.pyplot as plt

# Violación de la Desigualdad de Bell (CHSH) usando correlaciones cuánticas exactas
# Estado de Bell |Phi+> = 1/sqrt(2) (|00> + |11>)
# La expectativa cuántica de observables de espín proyectados en ángulos a y b es cos(2*(a-b))

theta_Alice = np.linspace(0, 2*np.pi, 100)
# Bob mantiene sus ángulos fijos en la configuración óptima para violar CHSH
a, a_prime = 0, np.pi/4
b, b_prime = np.pi/8, -np.pi/8

# Generamos la curva S_CHSH variando el ángulo principal de Alice (theta)
S_quantum = np.cos(2*(theta_Alice - b)) - np.cos(2*(theta_Alice - b_prime)) + np.cos(2*(a_prime - b)) + np.cos(2*(a_prime - b_prime))

plt.figure(figsize=(10,6))
plt.plot(theta_Alice, S_quantum, 'b-', lw=3, label="Predicción Mecánica Cuántica $S_{QM}(\theta)$")
plt.axhline(2, color='red', linestyle='--', lw=2, label="Límite del Realismo Local (Variables Ocultas)")
plt.axhline(-2, color='red', linestyle='--', lw=2)
plt.axhline(2*np.sqrt(2), color='green', linestyle='-.', lw=2, label="Límite Máximo de Tsirelson ($2\sqrt{2}$)")
plt.axhline(-2*np.sqrt(2), color='green', linestyle='-.', lw=2)

# Resaltar la zona de violación
plt.fill_between(theta_Alice, 2, np.clip(S_quantum, 2, 4), color='red', alpha=0.3, label="Violación Cuántica")
plt.fill_between(theta_Alice, -2, np.clip(S_quantum, -4, -2), color='red', alpha=0.3)

plt.title("Violación de la Desigualdad de CHSH")
plt.xlabel("Ángulo del polarizador de Alice $\theta_A$ (Radianes)")
plt.ylabel("Valor del Parámetro $S$")
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

```
