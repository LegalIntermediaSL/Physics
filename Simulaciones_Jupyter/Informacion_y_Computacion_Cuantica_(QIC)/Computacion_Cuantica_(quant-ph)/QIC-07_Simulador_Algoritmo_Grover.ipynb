# [QIC-07] Computación Cuántica: Simulador y Grover

```python
import numpy as np
import matplotlib.pyplot as plt

# Computación Cuántica: Construcción de un Simulador Cuántico desde cero.
# Implementaremos el álgebra lineal tensorial de puertas lógicas universales
# (La base del Algoritmo de Grover y el Algoritmo de Shor).

# Bases del Qubit |0> y |1>
ket_0 = np.array([1, 0])
ket_1 = np.array([0, 1])

# Puertas Cuánticas Universales (Matrices Unitarias 2x2)
# Puerta X (NOT)
X = np.array([[0, 1], [1, 0]])
# Puerta Z (Desfase pi)
Z = np.array([[1, 0], [0, -1]])
# Puerta de Hadamard (Superposición)
H = (1 / np.sqrt(2)) * np.array([[1, 1], [1, -1]])

# Construimos un sistema de N=3 Qubits (Tamaño del vector de estado = 2^3 = 8)
# Estado Inicial |000>
state = np.kron(ket_0, np.kron(ket_0, ket_0))

print("Estado Inicial |000>:", state)

# Aplicamos la Puerta Hadamard simultáneamente a todos los qubits
# H_total = H ⊗ H ⊗ H
H3 = np.kron(H, np.kron(H, H))
superposition_state = np.dot(H3, state)

print("\nEstado en Superposición Cuántica Total (8 estados simultáneos):")
print(np.round(superposition_state, 3))

# Algoritmo de Grover (Esquema conceptual):
# 1. Un "Oráculo" invierte la fase del estado correcto que buscamos (ej: |101>)
# 2. El operador de difusión de Grover amplifica esa amplitud negativa.

# Imaginemos que buscamos el estado |101> (índice 5 en binario)
oracle_matrix = np.eye(8)
oracle_matrix[5, 5] = -1 # Invierte la fase del estado ganador

# Aplicamos oráculo
state_after_oracle = np.dot(oracle_matrix, superposition_state)

# Operador de Difusión de Grover (D = 2|s><s| - I)
s = superposition_state.reshape(8, 1) # vector columna
D = 2 * np.dot(s, s.T) - np.eye(8)

# Aplicamos difusión
final_state = np.dot(D, state_after_oracle)

# Probabilidades finales de medida
probabilities = np.abs(final_state)**2

plt.figure(figsize=(10, 5))
plt.bar([bin(i)[2:].zfill(3) for i in range(8)], probabilities, color='cyan', edgecolor='black')
plt.title("Algoritmo de Grover: Amplificación de Amplitud (1 iteración)", fontsize=14)
plt.ylabel("Probabilidad de Medida")
plt.xlabel("Estado de los 3 Qubits")
plt.show()

# COMPUTACIÓN: Empezamos con la probabilidad plana (12.5% cada estado).
# Tras sólo UNA consulta al oráculo y un ciclo de difusión (Grover), 
# la probabilidad de medir la respuesta correcta |101> salta del 12.5% a casi el 80%.
# Si la lista tuviera 1 millón de elementos, un PC clásico miraría 500,000 veces.
# El PC cuántico lo encuentra en solo 1,000 pasos. Matemáticas puras.

```
