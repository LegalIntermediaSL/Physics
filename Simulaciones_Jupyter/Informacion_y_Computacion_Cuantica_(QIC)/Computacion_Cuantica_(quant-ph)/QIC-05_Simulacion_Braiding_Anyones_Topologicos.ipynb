# [QIC-05] Computación Cuántica: Braiding de Anyones

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Computación Cuántica Topológica: Trenzado (Braiding) de Anyones en 2+1D
# Simulamos cómo las Líneas de Mundo (Worldlines) de tres Anyones No-Abelianos 
# intercambian posiciones en un sustrato 2D a lo largo del Tiempo.

t = np.linspace(0, 10, 500)

# Trayectorias en 2D (x, y) a lo largo del tiempo (z = t)
# Anyon 1 y Anyon 2 se intercambian ("trenzan") entre sí. Anyon 3 se queda quieto.
# Este intercambio físico en el espacio-tiempo implementa una Puerta Lógica Cuántica topológica.

x1 = -2.0 + 4.0 / (1 + np.exp(-2*(t-5)))  # A1 cruza hacia la derecha
x2 =  2.0 - 4.0 / (1 + np.exp(-2*(t-5)))  # A2 cruza hacia la izquierda
x3 = np.ones_like(t) * 4.0                # A3 estático como referencia

# Añadimos un pequeño desvío en el eje Y para que las hebras se "trencen" en 3D
y1 =  np.exp(-(t-5)**2)
y2 = -np.exp(-(t-5)**2)
y3 = np.zeros_like(t)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Eje Z es el Tiempo
ax.plot(x1, y1, t, 'r-', lw=4, label='Línea de mundo: Anyon A')
ax.plot(x2, y2, t, 'b-', lw=4, label='Línea de mundo: Anyon B')
ax.plot(x3, y3, t, 'g--', lw=2, alpha=0.5, label='Anyon C (Espectador)')

ax.set_title("Braiding de Anyones (Computación Cuántica Topológica)", fontsize=14)
ax.set_xlabel('Espacio 2D: X')
ax.set_ylabel('Espacio 2D: Y')
ax.set_zlabel('Tiempo T')
ax.legend()
ax.view_init(elev=20., azim=-45)
plt.show()

# FÍSICA: Este trenzado (knot theory) modifica el estado cuántico global.
# Como la información depende puramente de la 'topología' (de que se hayan enrollado), 
# el qubit es matemáticamente inmune al ruido térmico local. Esta es la arquitectura 
# hiper-estable que persiguen los laboratorios cuánticos como Microsoft Station Q.

```
