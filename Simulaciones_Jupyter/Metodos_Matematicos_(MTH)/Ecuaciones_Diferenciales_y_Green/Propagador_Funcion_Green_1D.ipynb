# Función de Green en 1D

Respuesta a un impulso puntual en un oscilador forzado.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 10, 400)
# Función de Green para un oscilador amortiguado (impulso en t=0)
omega_0, gamma = 2.0, 0.2
omega = np.sqrt(omega_0**2 - gamma**2)
G = np.heaviside(t, 1) * np.exp(-gamma * t) * np.sin(omega * t) / omega

plt.figure(figsize=(10,4))
plt.plot(t, G, 'g-', lw=2)
plt.axvline(0, color='r', linestyle='--', label='Impulso Delta de Dirac')
plt.title('Función de Green Retardada (Oscilador Amortiguado)')
plt.xlabel('Tiempo')
plt.legend()
plt.grid(True)
plt.show()
```
