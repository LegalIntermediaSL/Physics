# Métrica de Schwarzschild y Geodésicas

Simulación de las geodésicas alrededor de una masa esférica no rotante en Relatividad General.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

r = np.linspace(2.1, 10, 500)
V_eff = (1 - 2/r) * (1 + 10/r**2) # Potencial efectivo con momento angular

plt.figure(figsize=(8,5))
plt.plot(r, V_eff, 'r-', lw=2)
plt.title('Potencial Efectivo en la Métrica de Schwarzschild')
plt.xlabel('Radio r (unidades de masa)')
plt.ylabel('V_eff(r)')
plt.grid(True)
plt.show()
```
