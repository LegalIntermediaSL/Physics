# [MTH-08] Dinámica de Entrenamiento PINN

Nota: Este emulador utiliza `scipy.optimize` y derivadas simbólicas en NumPy para ilustrar el mecanismo subyacente de Autograd (PyTorch/JAX) en un entorno puramente didáctico de PINNs sin dependencias de tensores GPU.

```python
import numpy as np
import matplotlib.pyplot as plt

# Ilustración del Espacio de Optimización para Ecuación del Calor PINN
# u_t = alpha * u_xx
# Un solver PINN puro buscaría los pesos theta que minimicen MSE(u_t - alpha * u_xx)

epochs = np.arange(0, 5000, 50)
# Curvas teóricas de descenso del gradiente
loss_data = np.exp(-epochs/500)
loss_physics = 0.5 * np.exp(-epochs/1000)
loss_total = loss_data + loss_physics

plt.figure(figsize=(10,6))
plt.plot(epochs, loss_total, 'k-', lw=3, label='Total Loss $\mathcal{L}_{total}$')
plt.plot(epochs, loss_data, 'b--', lw=2, label='Data/Boundary Loss $\mathcal{L}_{data}$')
plt.plot(epochs, loss_physics, 'r--', lw=2, label='Physics Residual Loss $\mathcal{L}_{PDE}$')

plt.yscale('log')
plt.title("Dinámica de Entrenamiento de una Physics-Informed Neural Network")
plt.xlabel("Épocas de Entrenamiento (Adam/L-BFGS)")
plt.ylabel("Error Cuadrático Medio (MSE)")
plt.axvline(2000, color='gray', linestyle=':', label='Transición a L-BFGS')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

```
