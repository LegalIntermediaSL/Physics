# [MTH-05] Neural Network Informada por la Física

```python
import numpy as np
import matplotlib.pyplot as plt

# EMULADOR CONCEPTUAL DE PINN (Physics-Informed Neural Network)
# Resolveremos la ecuación del oscilador armónico cuántico (estado fundamental)
# Ecuación: -0.5 * u''(x) + 0.5 * x^2 * u(x) = E * u(x)

x_vals = np.linspace(-4, 4, 100)

# 1. Supongamos que una Red Neuronal parametrizada (con pesos theta) ya fue entrenada.
# Minimizó la función de pérdida residual L = | H*u - E*u |^2

# Solución analítica exacta (Gaussiana) para E_0 = 0.5
def exact_solution(x):
    return (1.0 / np.pi)**0.25 * np.exp(-x**2 / 2.0)

# "Predicción" de la Red Neuronal tras 5000 epochs (con ligero ruido/imperfección)
def pinn_prediction(x):
    # Simulamos el output de una red de 3 capas
    return exact_solution(x) * (1.0 + 0.05 * np.sin(3*x))

u_exact = exact_solution(x_vals)
u_pinn = pinn_prediction(x_vals)

plt.figure(figsize=(10,6))
plt.plot(x_vals, u_exact, 'k--', lw=3, label="Solución Exacta Analítica")
plt.plot(x_vals, u_pinn, 'r-', lw=2, alpha=0.8, label="Predicción de la PINN")

# Rellenamos el error (Loss residual físico)
plt.fill_between(x_vals, u_exact, u_pinn, color='red', alpha=0.2, label='Error Residual Físico L(θ)')

plt.title("Physics-Informed Neural Network: Aprendiendo a resolver Schrödinger")
plt.xlabel("Posición (x)")
plt.ylabel("Función de Onda u(x)")
plt.legend()
plt.grid(True)
plt.show()

```
