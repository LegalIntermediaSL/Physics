# Flujo de Ricci en 2D

Evolución de la métrica de una variedad gobernada por su curvatura, análogo a la ecuación del calor.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Simulación 1D de un flujo de suavizado (análogo térmico al Flujo de Ricci)
x = np.linspace(-5, 5, 100)
g = 1 + 2*np.exp(-x**2) # Métrica inicial con un "bulto"

plt.figure(figsize=(8,5))
plt.plot(x, g, label='t = 0 (Métrica deformada)')
for t in range(1, 4):
    g = g + 0.1 * np.gradient(np.gradient(g)) # Evolución de la métrica
    plt.plot(x, g, label=f't = {t}')

plt.title('Evolución de la Métrica por Flujo de Ricci (Análogo)')
plt.legend()
plt.grid(True)
plt.show()
```
