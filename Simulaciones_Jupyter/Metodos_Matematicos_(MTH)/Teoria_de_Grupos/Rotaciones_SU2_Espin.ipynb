# Rotaciones en SU(2) y Espín

Demostración geométrica del requerimiento de 720 grados (4π) para que un espinor de rango 1/2 vuelva a su estado original.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 4*np.pi, 400)
spinor_phase = np.exp(1j * theta / 2) # Fases de SU(2)

plt.figure(figsize=(8,8))
plt.plot(theta, np.real(spinor_phase), label='Componente Real')
plt.plot(theta, np.imag(spinor_phase), label='Componente Imaginaria')
plt.axvline(2*np.pi, color='k', linestyle='--', label='Giro de 360 grados (Cambio de signo)')
plt.title('Evolución de la fase de un Espinor bajo rotaciones SU(2)')
plt.xlabel('Ángulo de rotación')
plt.legend()
plt.grid(True)
plt.show()
```
