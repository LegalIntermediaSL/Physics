# Generadores del Álgebra de Lie

Exponenciación de generadores infinitesimales para construir rotaciones macroscópicas en SO(3).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm

# Generador L_z de SO(3)
L_z = np.array([[0, -1, 0], [1, 0, 0], [0, 0, 0]])

angles = np.linspace(0, 2*np.pi, 100)
v0 = np.array([1, 0, 1])
path = np.array([expm(theta * L_z) @ v0 for theta in angles])

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(path[:,0], path[:,1], path[:,2], 'm-', lw=3)
ax.scatter([0], [0], [0], c='k', s=100)
plt.title('Órbita generada por exp(θ L_z) en SO(3)')
plt.axis('off')
plt.show()
```
