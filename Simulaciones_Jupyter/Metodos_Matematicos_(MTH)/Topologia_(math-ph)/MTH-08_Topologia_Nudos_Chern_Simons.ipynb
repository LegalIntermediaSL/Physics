# [MTH-08] Topología: Generador Matemático de Nudos

```python
import numpy as np
import matplotlib.pyplot as plt

# Topología Matemática y Nudos Topológicos
# Renderizador paramétrico de nudos fundamentales para el estudio del Invariante 
# de Jones y estados topológicos (Chern-Simons).

t = np.linspace(0, 2 * np.pi, 1000)

# Ecuaciones paramétricas del Nudo de Trébol (Trefoil Knot, el nudo más simple, tipo (2,3) del toro)
# x = sin(t) + 2sin(2t)
# y = cos(t) - 2cos(2t)
# z = -sin(3t)
x_trefoil = np.sin(t) + 2 * np.sin(2 * t)
y_trefoil = np.cos(t) - 2 * np.cos(2 * t)
z_trefoil = -np.sin(3 * t)

# Ecuación del Enlace de Hopf (Hopf Link, dos anillos entrelazados)
# Un anillo fijo en Z=0
x_hopf1 = np.cos(t) - 0.5
y_hopf1 = np.sin(t)
z_hopf1 = np.zeros_like(t)

# Otro anillo rotado y desplazado
x_hopf2 = np.zeros_like(t) + 0.5
y_hopf2 = np.cos(t)
z_hopf2 = np.sin(t)

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot(x_trefoil, y_trefoil, z_trefoil, lw=4, color='magenta')
ax1.set_title("Nudo de Trébol (Invariante de Jones V(t))", fontsize=14)
ax1.axis('off')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot(x_hopf1, y_hopf1, z_hopf1, lw=4, color='cyan', label="Bucle 1")
ax2.plot(x_hopf2, y_hopf2, z_hopf2, lw=4, color='orange', label="Bucle 2")
ax2.set_title("Enlace Topológico de Hopf", fontsize=14)
ax2.legend()
ax2.axis('off')

plt.tight_layout()
plt.show()

# MATEMÁTICAS: Estos nudos son los pilares de la Mecánica Estadística Exacta. 
# Si representas la línea de mundo en el tiempo de dos partículas en 2D, al 
# dar la vuelta una sobre la otra construyes trenzas y nudos matemáticos: los Anyones.

```
