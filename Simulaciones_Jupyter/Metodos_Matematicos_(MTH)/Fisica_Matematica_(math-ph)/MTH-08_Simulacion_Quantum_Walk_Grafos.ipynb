# [MTH-08] Grafos Cuánticos: Caminata Cuántica 1D

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm

# Teoría de Grafos Cuánticos y Computación: Caminata Cuántica (Quantum Walk 1D)
# Comparativa matemática de la dispersión de probabilidad entre una partícula clásica 
# (Random Walk estocástico) y una partícula cuántica en superposición que se interfiere.

N_steps = 100
N_pos = 2 * N_steps + 1
pos_center = N_steps

# Estado cuántico global: vector de amplitudes de (Posición) x (Espín/Moneda)
# Moneda: 0 = Izquierda, 1 = Derecha
state = np.zeros((N_pos, 2), dtype=complex)
# Inicializamos la partícula cuántica en el centro, con espín simétrico
state[pos_center, 0] = 1.0 / np.sqrt(2)
state[pos_center, 1] = 1.0j / np.sqrt(2)

# Operador Moneda Cuántica de Hadamard
H = (1.0 / np.sqrt(2)) * np.array([[1, 1], [1, -1]])

# Evolución del Quantum Walk
for t in range(N_steps):
    new_state = np.zeros_like(state)
    for p in range(1, N_pos - 1):
        # 1. Tirar la Moneda (Superposición de Espín)
        coin_out = H @ state[p]
        # 2. Desplazamiento Condicional (Shift)
        new_state[p-1, 0] += coin_out[0] # Va a la Izquierda
        new_state[p+1, 1] += coin_out[1] # Va a la Derecha
    state = new_state

# Distribución de Probabilidad Cuántica
prob_quantum = np.sum(np.abs(state)**2, axis=1)

# Caminata Clásica Equivalente (Distribución Binomial)
from scipy.stats import binom
x_positions = np.arange(0, N_pos) - pos_center
prob_classical = np.zeros(N_pos)
# Para N pasos, las posiciones van de -N a N saltando de a 2
for k in range(N_steps + 1):
    pos_idx = pos_center + (2*k - N_steps)
    prob_classical[pos_idx] = binom.pmf(k, N_steps, 0.5)

plt.figure(figsize=(10, 6))
plt.plot(x_positions, prob_classical, 'k--', lw=2, label="Random Walk (Clásico)")
plt.plot(x_positions, prob_quantum, 'r-', lw=2, label="Quantum Walk (Interferencia Cuántica)")
plt.fill_between(x_positions, 0, prob_quantum, color='red', alpha=0.3)

plt.title(f"Teoría de Grafos Cuánticos: Propagación tras {N_steps} pasos", fontsize=14)
plt.xlabel("Posición del Nodo ($x$)")
plt.ylabel("Probabilidad $P(x)$")
plt.legend()
plt.grid(True)
plt.show()
# FÍSICA: Mientras el caminante clásico (Gaussiana negra) se aglutina en el centro con una 
# propagación lenta y difusiva (Sigma ~ sqrt(t)), el caminante cuántico sufre interferencia 
# destructiva en el centro y es empujado balísticamente hacia los extremos (Sigma ~ t). 
# ¡Es cuadráticamente más rápido! Base de los Algoritmos de Búsqueda Cuántica.

```
