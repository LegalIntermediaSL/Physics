# Compendio Subtemático: Astrofisica Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def accretion_disk():
    r = np.linspace(1, 100, 500)
    # T(r) proportional to r^(-3/4)
    T = 1e7 * (r)**(-0.75) * (1 - np.sqrt(1/r))**0.25
    
    plt.figure(figsize=(8, 6))
    plt.plot(r, T, 'darkorange')
    plt.xscale('log')
    plt.yscale('log')
    plt.title("Accretion Disk Temperature Profile")
    plt.xlabel("Radius ($R_{in}$)")
    plt.ylabel("Temperature (K)")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    accretion_disk()
```
