# Radiación Sincrotrón en Jets Relativistas

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nu = np.logspace(8, 18, 200)
p = 2.5 # Indice de ley de potencia energetica de electrones
I_nu = nu**(-(p-1)/2) * (1 - np.exp(- (nu/1e10)**(-2.5)))

plt.loglog(nu, I_nu, color='magenta')
plt.title('Espectro Sincrotrón de un Jet Cuásar (Ópticamente grueso a fino)')
plt.xlabel('Frecuencia $\nu$ (Hz)')
plt.ylabel('Intensidad $I_\nu$')
plt.grid(True)
plt.show()
```
