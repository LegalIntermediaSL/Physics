# Compendio Subtemático: Astrofisica Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def jeans_mass():
    T = np.linspace(10, 100, 100) # Temperature (K)
    rho = 1e-19 # Density (kg/m^3)
    # Jeans mass roughly proportional to T^(1.5) / rho^(0.5)
    M_J = 10 * (T/10)**1.5
    
    plt.figure(figsize=(8, 6))
    plt.plot(T, M_J, 'g-')
    plt.title("Jeans Mass vs Temperature (Constant Density)")
    plt.xlabel("Temperature (K)")
    plt.ylabel("Jeans Mass ($M_\odot$)")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    jeans_mass()
```
