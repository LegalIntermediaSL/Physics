# Termodinámica Cuántica de Agujeros Negros

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.linspace(1, 10, 100)
T_H = 1 / M  # Temperatura inversamente proporcional a la masa
S = M**2  # Entropía proporcional al area (M^2)

fig, ax1 = plt.subplots()
ax1.plot(M, T_H, 'r-', label='Temp. Hawking')
ax1.set_xlabel('Masa $M$')
ax1.set_ylabel('Temperatura', color='r')
ax2 = ax1.twinx()
ax2.plot(M, S, 'b-', label='Entropía')
ax2.set_ylabel('Entropía $S$', color='b')
plt.title('Leyes Termodinámicas de un Agujero Negro')
plt.grid(True)
plt.show()
```
