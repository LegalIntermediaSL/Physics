# Correcciones Post-Newtonianas en Binarias

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.linspace(2, 20, 200)
V_N = -1/r
V_1PN = -1/r - 3/(2*r**2) # Correccion 1PN muy simplificada

plt.plot(r, V_N, label='Newtoniano', linestyle='--')
plt.plot(r, V_1PN, label='1PN (Post-Newtoniano)')
plt.title('Potencial Orbital Gravitacional')
plt.xlabel('Distancia $r$')
plt.ylabel('$V(r)$')
plt.legend(); plt.grid(True)
plt.show()
```
