# Compendio Subtemático: Astrofisica Compendio 30

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def asteroid_orbit():
    t = np.linspace(0, 2*np.pi, 500)
    # Earth
    plt.plot(np.cos(t), np.sin(t), 'b-', label='Earth')
    # Asteroid (eccentric)
    a, e = 2.5, 0.6
    r = a * (1 - e**2) / (1 + e*np.cos(t))
    plt.plot(r*np.cos(t), r*np.sin(t), 'r-', label='Asteroid')
    plt.scatter([0], [0], c='y', s=100, label='Sun')
    plt.axis('equal')
    plt.title("Asteroid vs Earth Orbit")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    asteroid_orbit()
```
