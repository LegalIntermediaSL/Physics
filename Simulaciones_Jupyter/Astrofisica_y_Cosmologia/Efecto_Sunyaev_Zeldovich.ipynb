# Efecto Sunyaev-Zel'dovich Térmico

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.1, 10, 200) # Frecuencia adimensional x = h nu / k T_CMB
# Cambio de intensidad espectral SZ termico
delta_I = (x**4 * np.exp(x) / (np.exp(x)-1)**2) * (x * (np.exp(x)+1)/(np.exp(x)-1) - 4)

plt.plot(x, delta_I, color='orange', lw=2)
plt.axhline(0, color='k', linestyle='--')
plt.axvline(3.83, color='gray', linestyle=':', label='Frecuencia nula (217 GHz)')
plt.title("Efecto Sunyaev-Zel'dovich Térmico (tSZ)")
plt.xlabel('$x = h \nu / k T_{CMB}$')
plt.ylabel('$\Delta I / I_0$')
plt.legend(); plt.grid(True)
plt.show()
```
