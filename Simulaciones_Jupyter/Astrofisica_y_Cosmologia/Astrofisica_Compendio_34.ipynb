# Compendio Subtemático: Astrofisica Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def cosmic_ray():
    E = np.logspace(9, 20, 100) # eV
    # Power law flux ~ E^-2.7, knee at 1e15, ankle at 1e18
    flux = E**(-2.7)
    flux[E > 1e15] = (1e15)**(0.3) * E[E > 1e15]**(-3.0)
    flux[E > 1e18] = (1e15)**(0.3) * (1e18)**(-0.4) * E[E > 1e18]**(-2.6)
    
    plt.figure(figsize=(8, 6))
    plt.plot(E, flux * E**2.5, 'g-') # scaled by E^2.5 for visibility
    plt.xscale('log')
    plt.yscale('log')
    plt.title("Cosmic Ray Energy Spectrum (Scaled)")
    plt.xlabel("Energy (eV)")
    plt.ylabel("Flux $\\times E^{2.5}$")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    cosmic_ray()
```
