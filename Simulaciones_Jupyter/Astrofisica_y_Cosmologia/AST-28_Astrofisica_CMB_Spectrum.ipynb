# Investigación Avanzada: Astrofisica CMB Spectrum

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def planck_cmb_spectrum():
    """Simulates a mock CMB angular power spectrum."""
    l = np.arange(2, 2500)
    # Mock spectrum using a combination of spherical bessel functions and damping
    dl = 5000 * np.exp(-(l - 200)**2 / (2 * 50**2)) / (l*(l+1)) + \
         2000 * np.exp(-(l - 500)**2 / (2 * 100**2)) / (l*(l+1)) + \
         1500 * np.exp(-(l - 800)**2 / (2 * 120**2)) / (l*(l+1))
    
    # Scale appropriately
    dl *= l * (l + 1) / (2 * np.pi)
    dl += 10 # Noise floor
    
    plt.figure(figsize=(10, 6))
    plt.plot(l, dl, label='Mock CMB Spectrum', color='blue')
    plt.xlabel('Multipole moment, $\ell$')
    plt.ylabel('$\\mathcal{D}_\ell = \ell(\ell+1)C_\ell / 2\\pi \, [\\mu K^2]$')
    plt.title('CMB Angular Power Spectrum Simulation')
    plt.xscale('log')
    plt.grid(True, which='both', linestyle='--', alpha=0.5)
    plt.legend()
    plt.savefig('CMB_Spectrum.png')
    plt.show()

if __name__ == '__main__':
    planck_cmb_spectrum()
```
