# Compendio Subtemático: Astrofisica Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def hr_diagram():
    # Simulate Main Sequence
    t_ms = np.random.normal(5700, 1500, 500)
    l_ms = (t_ms/5700)**4 * np.random.normal(1.0, 0.2, 500)
    
    # Simulate Red Giants
    t_rg = np.random.normal(3500, 500, 100)
    l_rg = np.random.normal(100, 20, 100)
    
    # Simulate White Dwarfs
    t_wd = np.random.normal(10000, 2000, 50)
    l_wd = np.random.normal(0.01, 0.005, 50)
    
    plt.figure(figsize=(8, 8))
    plt.scatter(t_ms, l_ms, s=10, c='blue', label='Main Sequence')
    plt.scatter(t_rg, l_rg, s=20, c='red', label='Red Giants')
    plt.scatter(t_wd, l_wd, s=10, c='grey', label='White Dwarfs')
    
    plt.gca().invert_xaxis()
    plt.yscale('log')
    plt.title("Hertzsprung-Russell (H-R) Diagram")
    plt.xlabel("Temperature (K)")
    plt.ylabel("Luminosity ($L_\odot$)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    hr_diagram()
```
