# Compendio Subtemático: Astrofisica Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def radial_velocity():
    t = np.linspace(0, 10, 500) # days
    v1 = 50 * np.sin(2 * np.pi * t / 4) # Star 1
    v2 = -30 * np.sin(2 * np.pi * t / 4) # Star 2 (different mass)
    
    plt.figure(figsize=(10, 5))
    plt.plot(t, v1, label='Star 1')
    plt.plot(t, v2, label='Star 2')
    plt.title("Binary System Radial Velocity")
    plt.xlabel("Time (days)")
    plt.ylabel("Radial Velocity (km/s)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    radial_velocity()
```
