# Laboratorio Especializado: Astrofisica Especializado 5

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gw_strain(t, tc, f0, tau):
    mask = t < tc
    strain = np.zeros_like(t)
    f = f0 * (1 - t[mask]/tc)**(-3.0/8.0)
    phase = 2 * np.pi * np.cumsum(f) * (t[1]-t[0])
    amp = (1 - t[mask]/tc)**(-1.0/4.0)
    strain[mask] = amp * np.cos(phase) * np.exp(-(t[mask]-tc)**2/tau**2)
    return strain

t = np.linspace(0, 1.5, 2000)
h = gw_strain(t, 1.0, 20, 0.5)

plt.figure(figsize=(10,4))
plt.plot(t, h, color='blue')
plt.title('Simulated Gravitational Wave Strain (BBH Merger)')
plt.xlabel('Time (s)')
plt.ylabel('Strain h(t)')
plt.grid(True)
plt.savefig('gw_strain.png')
```
