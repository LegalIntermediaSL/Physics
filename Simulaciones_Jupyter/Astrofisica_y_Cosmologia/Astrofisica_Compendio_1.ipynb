# Compendio Subtemático: Astrofisica Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def kepler_orbit():
    t = np.linspace(0, 2*np.pi, 1000)
    a = 1.0 # semi-major axis
    e = 0.5 # eccentricity
    r = a * (1 - e**2) / (1 + e * np.cos(t))
    x = r * np.cos(t)
    y = r * np.sin(t)
    
    plt.figure(figsize=(8, 8))
    plt.plot(x, y, label=f'Orbit (e={e})')
    plt.plot(0, 0, 'yo', markersize=10, label='Star')
    plt.title("Kepler's First Law: Elliptical Orbit")
    plt.xlabel("x (AU)")
    plt.ylabel("y (AU)")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

if __name__ == '__main__':
    kepler_orbit()
```
