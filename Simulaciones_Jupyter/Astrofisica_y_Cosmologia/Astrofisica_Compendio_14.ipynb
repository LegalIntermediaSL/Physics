# Compendio Subtemático: Astrofisica Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def supernova():
    days = np.linspace(0, 150, 200)
    # Fast rise, exponential decay
    flux = (days / 15)**2 * np.exp(-days / 15) * 100 + 20 * np.exp(-days / 50)
    
    plt.figure(figsize=(10, 5))
    plt.plot(days, flux, 'm-')
    plt.title("Type Ia Supernova Light Curve")
    plt.xlabel("Days since explosion")
    plt.ylabel("Luminosity")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    supernova()
```
