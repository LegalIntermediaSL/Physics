# Compendio Subtemático: Astrofisica Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def ism_extinction():
    wav = np.linspace(0.1, 3.0, 500) # um
    # Cardelli-Clayton-Mathis approx shape
    A_lambda = 1 / wav + 0.2 / (wav**2 + 0.1)
    
    plt.figure(figsize=(8, 6))
    plt.plot(1/wav, A_lambda, 'b-')
    plt.title("Interstellar Extinction Curve")
    plt.xlabel("$1/\lambda$ ($\mu m^{-1}$)")
    plt.ylabel("$A_\lambda / A_V$")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    ism_extinction()
```
