# Supercomputación: Astrofisica HPC 2

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

N = 500
pos = np.random.randn(N, 2)
vel = np.zeros_like(pos)
mass = 1.0 / N
h = 0.5
k = 0.1

def W(r, h):
    return (1.0 / (h**2 * np.pi)) * np.exp(- (r/h)**2)

def gradW(r, dx, h):
    return W(r, h) * (-2.0 * dx / h**2)

for _ in range(50):
    r_mat = cdist(pos, pos)
    rho = np.sum(mass * W(r_mat, h), axis=1)
    pressure = k * rho**2
    
    force = np.zeros_like(pos)
    for i in range(N):
        dx = pos[i] - pos
        gw = np.array([gradW(r_mat[i], dx[:,0], h), gradW(r_mat[i], dx[:,1], h)]).T
        term = mass * (pressure[i]/rho[i]**2 + pressure/rho**2)[:, np.newaxis] * gw
        force[i] = -np.sum(term, axis=0)
    
    force -= 0.1 * pos / (np.linalg.norm(pos, axis=1, keepdims=True)**3 + 0.1)
    vel += force * 0.05
    pos += vel * 0.05

plt.scatter(pos[:,0], pos[:,1], c=rho, cmap='plasma', s=5)
plt.title('SPH Protostellar Cloud Collapse')
plt.show()
```
