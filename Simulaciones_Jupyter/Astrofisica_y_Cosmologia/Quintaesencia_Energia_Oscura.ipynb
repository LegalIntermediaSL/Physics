# Energía Oscura Dinámica: Parametrización CPL

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(0, 5, 100)
a = 1 / (1 + z)
w0, wa = -1.0, 0.1
w_z = w0 + wa * (1 - a)

plt.plot(z, w_z, color='purple', lw=2)
plt.title('Ecuación de Estado de la Energía Oscura (CPL)')
plt.xlabel('Redshift $z$')
plt.ylabel('Parámetro $w(z)$')
plt.axhline(-1, color='k', linestyle='--')
plt.grid(True)
plt.show()
```
