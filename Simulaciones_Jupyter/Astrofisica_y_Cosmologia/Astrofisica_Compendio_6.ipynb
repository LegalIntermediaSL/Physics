# Compendio Subtemático: Astrofisica Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def transit():
    t = np.linspace(-5, 5, 500)
    flux = np.ones_like(t)
    transit_depth = 0.02
    transit_duration = 2.0
    
    in_transit = np.abs(t) < (transit_duration/2)
    flux[in_transit] -= transit_depth
    flux += np.random.normal(0, 0.002, len(t))
    
    plt.figure(figsize=(10, 5))
    plt.plot(t, flux, 'ko-', markersize=2)
    plt.title("Exoplanet Transit Light Curve")
    plt.xlabel("Time (days)")
    plt.ylabel("Relative Flux")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    transit()
```
