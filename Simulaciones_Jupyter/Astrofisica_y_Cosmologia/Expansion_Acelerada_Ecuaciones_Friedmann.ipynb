# Ecuaciones de Friedmann y Energía Oscura

Integración de las ecuaciones de Friedmann mostrando cómo la constante cosmológica ($\Lambda$) provoca la expansión acelerada del universo en épocas tardías.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parámetros cosmológicos (Materia vs Energía Oscura)
Omega_m = 0.3
Omega_L = 0.7
H0 = 70.0 # km/s/Mpc

def friedmann(a, t):
    # da/dt = H0 * sqrt(Omega_m/a + Omega_L*a^2)
    if a[0] <= 0: return [0]
    dadt = H0 * np.sqrt(Omega_m / a[0] + Omega_L * a[0]**2)
    return [dadt]

t = np.linspace(0, 1.5, 500) # Tiempo en unidades cosmológicas
a0 = [1e-3]

sol = odeint(friedmann, a0, t)

plt.figure(figsize=(8,5))
plt.plot(t, sol[:,0], 'b-', lw=2)
plt.title('Factor de Escala a(t) en Universo dominado por Energía Oscura')
plt.xlabel('Tiempo Cósmico')
plt.ylabel('Factor de Escala a(t)')
plt.grid()
plt.show()
```
