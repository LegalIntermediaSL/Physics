# Nucleosíntesis Primordial (BBN)

Evolución de las abundancias de elementos ligeros (H, He, Li) conforme el universo primordial se enfría.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.logspace(0, -2, 200) # Temperatura bajando (MeV)
# Modelado simplificado de abundancias masivas (Fracciones de Masa)
H = 0.75 * np.ones_like(T) - 0.1 * np.exp(-1/T)
He4 = 0.25 * np.ones_like(T) + 0.1 * np.exp(-1/T)
Li7 = 1e-9 * np.exp(-2/T)

plt.figure(figsize=(8,5))
plt.plot(T, H, label='Hidrógeno ($^{1}$H)')
plt.plot(T, He4, label='Helio ($^{4}$He)')
plt.plot(T, Li7*1e8, label='Litio ($^{7}$Li) x $10^8$')
plt.xscale('log')
plt.gca().invert_xaxis() # El tiempo avanza hacia la izquierda al bajar T
plt.title('Abundancias Primordiales (BBN Simulado)')
plt.xlabel('Temperatura (MeV)')
plt.ylabel('Fracción de Masa')
plt.legend()
plt.grid()
plt.show()
```
