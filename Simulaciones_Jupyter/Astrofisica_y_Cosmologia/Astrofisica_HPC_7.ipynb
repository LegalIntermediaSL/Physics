# Supercomputación: Astrofisica HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def hyperchaos(state, t):
    x, y, z, u, w = state
    a, b, c, d, e = 10, 28, 8/3, 1.5, 0.2
    return [
        a*(y - x) + u,
        x*(b - z) - y + w,
        x*y - c*z,
        -d*x - u,
        -e*y - w
    ]

t = np.linspace(0, 50, 10000)
state0 = [1, 1, 1, 1, 1]
sol = odeint(hyperchaos, state0, t)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol[:,0], sol[:,1], sol[:,2], lw=0.5, color='darkorange')
ax.set_title('Astrophysical Plasma Hyperchaos (>4D)')
plt.show()
```
