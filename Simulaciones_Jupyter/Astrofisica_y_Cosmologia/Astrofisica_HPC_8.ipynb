# Supercomputación: Astrofisica HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def lorentz_force(state, t):
    r = state[:3]
    v = state[3:]
    R = np.linalg.norm(r)
    B = 1.0 / R**3 * np.array([3*r[0]*r[2]/R**2, 3*r[1]*r[2]/R**2, (3*r[2]**2/R**2 - 1)])
    E = np.cross(np.array([0, 0, 1.0]), r) * 0.1
    dv = np.cross(v, B) + E
    return np.concatenate((v, dv))

t = np.linspace(0, 10, 5000)
state0 = [1.5, 0, 1.5, 0.1, 0.1, 0.0]
sol = odeint(lorentz_force, state0, t)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol[:,0], sol[:,1], sol[:,2], lw=1.0)
ax.set_title('Particle in Pulsar Magnetosphere')
plt.show()
```
