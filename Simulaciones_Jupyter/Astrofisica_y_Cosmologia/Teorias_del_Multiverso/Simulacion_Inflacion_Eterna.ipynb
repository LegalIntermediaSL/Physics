# Inflación Eterna: Nucleación de Universos Burbuja

Modelo estocástico espacial simple (Autómata Celular) de cómo un 'falso vacío' en expansión exponencial crea burbujas de 'vacío verdadero' (universos como el nuestro) de forma incesante.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Autómata Celular Estocástico
size = 200
steps = 20
grid = np.zeros((size, size)) # 0 = Falso Vacío (Inflacionando)

plt.figure(figsize=(10, 8))
for t in range(1, steps + 1):
    # Nucleación aleatoria de burbujas (probabilidad muy baja)
    nucleation = np.random.rand(size, size) < 0.005
    grid[nucleation] = t # Marcamos el tiempo de creación del universo
    
    # Crecimiento de burbujas existentes (expansión a la velocidad de la luz)
    # Una simulación real requiere inflación hiper-rápida del falso vacío de fondo.
    # Aquí modelamos solo el crecimiento para visualización.
    new_grid = grid.copy()
    for i in range(1, size-1):
        for j in range(1, size-1):
            if grid[i, j] > 0:
                if grid[i+1, j] == 0: new_grid[i+1, j] = grid[i, j]
                if grid[i-1, j] == 0: new_grid[i-1, j] = grid[i, j]
                if grid[i, j+1] == 0: new_grid[i, j+1] = grid[i, j]
                if grid[i, j-1] == 0: new_grid[i, j-1] = grid[i, j]
    grid = new_grid

plt.imshow(grid, cmap='magma')
plt.title('Nucleación de Universos Burbuja (Multiverso de Nivel II)')
plt.colorbar(label='Generación Temporal del Universo')
plt.axis('off')
plt.show()
```
