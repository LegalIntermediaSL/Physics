# Interpretación de los Muchos Mundos (Árbol de Everett)

Visualización probabilística de la ramificación del estado cuántico universal por decoherencia cuántica tras una secuencia de mediciones binarias de espín.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def plot_tree(x, y, angle, length, depth, max_depth, ax):
    if depth > max_depth:
        return
    x_end = x + length * np.cos(angle)
    y_end = y + length * np.sin(angle)
    
    # El grosor de la línea representa la "amplitud" relativa de esa rama
    thickness = 5 / (2**depth)
    ax.plot([x, x_end], [y, y_end], 'k-', lw=thickness, alpha=0.7)
    
    # Cada evento de "medición" divide el universo en dos superposiciones no interactuantes
    plot_tree(x_end, y_end, angle - np.pi/6, length * 0.7, depth + 1, max_depth, ax)
    plot_tree(x_end, y_end, angle + np.pi/6, length * 0.7, depth + 1, max_depth, ax)

fig, ax = plt.subplots(figsize=(8, 8))
# 6 eventos de medición cuántica (2^6 = 64 universos observables resultantes)
plot_tree(0, 0, np.pi/2, 10, 0, 6, ax)

plt.title('Decoherencia y Ramificación: Interpretación de Everett (Multiverso Nivel III)')
plt.axis('off')
plt.show()
```
