# Compendio Subtemático: Astrofisica Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def galaxy_collision():
    # Simplistic visualization of 2 colliding disks
    theta = np.random.uniform(0, 2*np.pi, 500)
    r1 = np.random.normal(0, 1, 500)
    x1 = r1 * np.cos(theta)
    y1 = r1 * np.sin(theta)
    
    r2 = np.random.normal(0, 0.8, 500)
    x2 = r2 * np.cos(theta) + 1.5 # Offset
    y2 = r2 * np.sin(theta) + 1.5
    
    plt.figure(figsize=(8, 8))
    plt.scatter(x1, y1, s=5, c='blue', alpha=0.5, label='Galaxy 1')
    plt.scatter(x2, y2, s=5, c='red', alpha=0.5, label='Galaxy 2')
    plt.title("Simplistic Galaxy Collision")
    plt.axis('equal')
    plt.legend()
    plt.show()

if __name__ == '__main__':
    galaxy_collision()
```
