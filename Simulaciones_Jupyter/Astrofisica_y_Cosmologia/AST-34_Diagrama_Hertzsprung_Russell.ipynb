# Diagrama Hertzsprung-Russell

Distribución térmica y lumínica en la evolución estelar.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.random.normal(6000, 2000, 300)
L = (T/5778)**4 * np.random.normal(1, 0.2, 300) # Aproximación secuencia principal

plt.figure(figsize=(6,6))
plt.scatter(T, L, c=T, cmap='coolwarm_r', s=10)
plt.gca().invert_xaxis()
plt.yscale('log')
plt.title('Simulación Simple: Secuencia Principal H-R')
plt.xlabel('Temperatura (K)')
plt.ylabel('Luminosidad ($L_{\odot}$)')
plt.grid()
plt.show()
```
