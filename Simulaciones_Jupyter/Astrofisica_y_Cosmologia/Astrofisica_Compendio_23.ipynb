# Compendio Subtemático: Astrofisica Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def alt_az():
    az = np.linspace(0, 2*np.pi, 360)
    alt = np.abs(np.sin(az)) * 90 # Mock trajectory
    
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(8, 8))
    ax.plot(az, 90 - alt)
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    ax.set_yticks(range(0, 91, 15))
    ax.set_yticklabels(map(str, range(90, -1, -15)))
    plt.title("Star Trajectory in Alt-Az")
    plt.show()

if __name__ == '__main__':
    alt_az()
```
