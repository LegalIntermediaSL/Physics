# Laboratorio Especializado: Astrofisica Especializado 4

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N_stars = 10000
bp_rp = np.random.normal(1.0, 0.5, N_stars)
abs_mag = 5 * bp_rp + np.random.normal(0, 0.5, N_stars)

wd_bp_rp = np.random.normal(0.0, 0.2, 500)
wd_abs_mag = 5 * wd_bp_rp + 10 + np.random.normal(0, 0.5, 500)

bp_rp = np.concatenate([bp_rp, wd_bp_rp])
abs_mag = np.concatenate([abs_mag, wd_abs_mag])

plt.figure(figsize=(8,10))
plt.scatter(bp_rp, abs_mag, s=1, color='black', alpha=0.5)
plt.gca().invert_yaxis()
plt.title('Simulated Hertzsprung-Russell Diagram')
plt.xlabel('Color Index (BP-RP)')
plt.ylabel('Absolute Magnitude (M_G)')
plt.grid(True)
plt.savefig('hr_diagram.png')
```
