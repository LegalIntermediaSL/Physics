# Curvas de Rotación Galáctica

Diferencia entre perfiles keplerianos y perfiles isotermos (Materia Oscura).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

r = np.linspace(0.1, 10, 200)
v_kepler = 1/np.sqrt(r)
v_dark = np.sqrt(1 - np.exp(-r)) * 1.5

plt.figure(figsize=(8,5))
plt.plot(r, v_kepler, 'r--', label='Esperado (Kepleriano)')
plt.plot(r, v_dark, 'b-', label='Observado (Halo de Materia Oscura)')
plt.title('Anomalía en Curvas de Rotación Galáctica')
plt.xlabel('Radio (kpc)')
plt.ylabel('Velocidad Orbital')
plt.legend()
plt.grid()
plt.show()
```
