# Compendio Subtemático: Astrofisica Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rad_transfer():
    tau = np.linspace(0, 5, 100) # Optical depth
    I0 = 1.0 # Initial intensity
    S = 0.5 # Source function
    
    I_tau = I0 * np.exp(-tau) + S * (1 - np.exp(-tau))
    
    plt.figure(figsize=(8, 6))
    plt.plot(tau, I_tau, 'r-')
    plt.title("Radiative Transfer: Intensity vs Optical Depth")
    plt.xlabel("Optical Depth ($\\tau$)")
    plt.ylabel("Intensity $I(\\tau)$")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    rad_transfer()
```
