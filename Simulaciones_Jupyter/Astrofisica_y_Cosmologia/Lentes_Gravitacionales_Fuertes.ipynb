# Lentes Gravitacionales y Anillos de Einstein

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta_S = np.linspace(-3, 3, 500)
theta_E = 1.0  # Radio de Einstein
# Ecuación de la lente: theta_S = theta - theta_E^2 / theta
theta_I_plus = 0.5 * (theta_S + np.sqrt(theta_S**2 + 4*theta_E**2))
theta_I_minus = 0.5 * (theta_S - np.sqrt(theta_S**2 + 4*theta_E**2))

plt.plot(theta_S, theta_I_plus, label='Imagen Primaria', color='blue')
plt.plot(theta_S, theta_I_minus, label='Imagen Secundaria', color='red')
plt.plot(theta_S, theta_S, 'k--', label='Fuente sin lente')
plt.title('Desplazamiento angular por Lente Gravitacional')
plt.xlabel(r'Posición de la fuente $\theta_S$')
plt.ylabel(r'Posición aparente de la imagen $\\theta$')
plt.legend(); plt.grid(True)
plt.show()
```
