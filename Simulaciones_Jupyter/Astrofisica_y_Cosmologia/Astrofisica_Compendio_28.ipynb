# Compendio Subtemático: Astrofisica Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gw_inspiral():
    t = np.linspace(-1, 0, 1000)
    # Frequency increases as t->0
    f = 10 * (-t + 0.01)**(-3/8)
    # Amplitude increases as t->0
    A = (-t + 0.01)**(-1/4)
    h = A * np.cos(2 * np.pi * f * t)
    
    plt.figure(figsize=(10, 5))
    plt.plot(t, h, 'k-')
    plt.title("Gravitational Wave Inspiral (Chirp)")
    plt.xlabel("Time to merger (s)")
    plt.ylabel("Strain h")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    gw_inspiral()
```
