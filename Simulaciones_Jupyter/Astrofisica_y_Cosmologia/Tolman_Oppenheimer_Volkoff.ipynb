# Estrellas de Neutrones: Ecuación TOV

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Integración de la ecuación de Tolman-Oppenheimer-Volkoff
# dp/dr = - (G*m(r)*rho/r^2) * [correcciones relativistas]
def tov(r, y):
    p, m = y
    rho = p**(3/5)  # Ecuacion de estado politrópica simplificada
    if p <= 0: return [0, 0]
    dpdr = - (m * rho / r**2) * (1 + p/rho) * (1 + 4*np.pi*r**3*p/m) / (1 - 2*m/r)
    dmdr = 4 * np.pi * r**2 * rho
    return [dpdr, dmdr]

r_span = (0.01, 10.0)
y0 = [1.0, 0.001] # Presión central y masa inicial nula
sol = solve_ivp(tov, r_span, y0, max_step=0.1)
plt.plot(sol.t, sol.y[0])
plt.title('Perfil de Presión en una Estrella de Neutrones (TOV)')
plt.xlabel('Radio r'); plt.ylabel('Presión P(r)'); plt.grid(True)
plt.show()
```
