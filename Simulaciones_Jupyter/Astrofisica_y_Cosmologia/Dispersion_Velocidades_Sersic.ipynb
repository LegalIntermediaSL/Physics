# Cinemática Estelar y Perfil de Sérsic

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = np.linspace(0.1, 10, 100)
n = 4 # De Vaucouleurs
I_R = np.exp(-R**(1/n))

plt.plot(R, I_R)
plt.title('Brillo Superficial - Perfil de Sérsic (n=4)')
plt.xlabel('Radio')
plt.ylabel('$I(R)$')
plt.yscale('log')
plt.grid(True)
plt.show()
```
