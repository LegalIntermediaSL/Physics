# Compendio Subtemático: Astrofisica Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def cmb_spectrum():
    l = np.linspace(2, 2500, 500)
    # Simple approx of acoustic peaks
    power = 1000 * np.exp(-(l - 200)**2 / (2*50**2)) + \
            500 * np.exp(-(l - 500)**2 / (2*70**2)) + \
            300 * np.exp(-(l - 800)**2 / (2*100**2)) + \
            5000 / l
    
    plt.figure(figsize=(10, 6))
    plt.plot(l, power, 'b-')
    plt.title("Cosmic Microwave Background Power Spectrum (Mock)")
    plt.xlabel("Multipole moment $l$")
    plt.ylabel("Temperature variance $\Delta T^2$")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    cmb_spectrum()
```
