# Función de Masa Inicial (IMF) Estelar

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.logspace(-1, 2, 200)
imf_salpeter = M**(-2.35)
imf_kroupa = np.where(M<0.5, M**(-1.3), M**(-2.3))

plt.loglog(M, imf_salpeter, label='Salpeter', linestyle='--')
plt.loglog(M, imf_kroupa, label='Kroupa', lw=2)
plt.title('IMF - Formación Estelar Primordial')
plt.xlabel('Masa Estelar $M/M_\odot$')
plt.ylabel('$\phi(M)$')
plt.legend(); plt.grid(True)
plt.show()
```
