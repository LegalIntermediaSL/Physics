# Espectro de Potencia del CMB

Simulación paramétrica de los picos acústicos en el fondo cósmico de microondas debido a oscilaciones del plasma primordial.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

l = np.linspace(2, 2500, 1000)
# Aproximación fenomenológica de los picos acústicos del CMB
Dl = (1.0 / (1 + (l/1000)**2)) * (5000 * np.exp(-((l-220)/100)**2) +
                                  1500 * np.exp(-((l-540)/120)**2) +
                                  1800 * np.exp(-((l-800)/150)**2))

plt.figure(figsize=(8,5))
plt.plot(l, Dl, 'r-', lw=2)
plt.title('Espectro de Potencia Angular CMB (Aproximación)')
plt.xlabel('Momento Multipolo $\ell$')
plt.ylabel('$\mathcal{D}_\ell$ [$\mu$K$^2$]')
plt.grid()
plt.show()
```
