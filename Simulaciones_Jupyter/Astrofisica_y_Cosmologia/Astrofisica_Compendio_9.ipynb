# Compendio Subtemático: Astrofisica Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def sunspot_cycle():
    years = np.linspace(1900, 2020, 1200)
    cycle = 11.0 # years
    sunspots = 100 + 80 * np.sin(2 * np.pi * years / cycle) + 20 * np.random.randn(len(years))
    sunspots = np.clip(sunspots, 0, None)
    
    plt.figure(figsize=(12, 5))
    plt.plot(years, sunspots, color='orange')
    plt.title("Solar Cycle (Sunspot Number)")
    plt.xlabel("Year")
    plt.ylabel("Number of Sunspots")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    sunspot_cycle()
```
