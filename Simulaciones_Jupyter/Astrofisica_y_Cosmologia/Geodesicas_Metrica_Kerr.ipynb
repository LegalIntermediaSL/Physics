# Relatividad Numérica: Geodésicas en Espacio de Kerr

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Representación simbólica 2D del potencial efectivo radial
r = np.linspace(1.5, 10, 500)
a = 0.9 # Spin del Agujero Negro
L = 3.0 # Momento angular orbital
V_eff = -1/r + L**2/(2*r**2) - L**2/(r**3) - a**2/(2*r**3)

plt.plot(r, V_eff, color='k')
plt.title('Potencial Efectivo Radial en Métrica de Kerr (Plano Ecuatorial)')
plt.xlabel('Radio $r/M$')
plt.ylabel('$V_{eff}(r)$')
plt.axhline(0, color='gray', linestyle=':')
plt.grid(True)
plt.show()
```
