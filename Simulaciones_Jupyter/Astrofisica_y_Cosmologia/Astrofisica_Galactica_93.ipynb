# Astrofisica Galactica 93

Simulación matemática generada paramétricamente para el compendio galáctico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Simulation 93')
plt.show()
```
