# Compendio Subtemático: Astrofisica Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def nucleosynthesis():
    t = np.linspace(0, 10, 100)
    H = np.exp(-0.5 * t)
    He = 1 - np.exp(-0.5 * t) - 0.1 * t * np.exp(-0.2 * t)
    Metals = 1 - H - He
    
    plt.figure(figsize=(10, 6))
    plt.plot(t, H, label='Hydrogen')
    plt.plot(t, He, label='Helium')
    plt.plot(t, Metals, label='Metals')
    plt.title("Simplified Nucleosynthesis Evolution")
    plt.xlabel("Time (Arbitrary Units)")
    plt.ylabel("Mass Fraction")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    nucleosynthesis()
```
