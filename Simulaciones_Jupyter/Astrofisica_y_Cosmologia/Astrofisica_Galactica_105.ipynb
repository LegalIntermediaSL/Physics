# Astrofisica Galactica 105

Simulación matemática generada paramétricamente para el compendio galáctico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 100)
plt.plot(t, np.sin(t))
plt.title('Simulation')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```
