# Compendio Subtemático: Astrofisica Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rotation_curve():
    r = np.linspace(0.1, 30, 100) # kpc
    
    # Baryonic matter (disk + bulge)
    v_baryon = 150 * np.exp(-r/5) + 100 * (r / (r**2 + 1))
    
    # Dark matter halo (Navarro-Frenk-White profile approx)
    v_dm = 200 * np.sqrt(1 - (1/(1+r/10)))
    
    # Total
    v_total = np.sqrt(v_baryon**2 + v_dm**2)
    
    plt.figure(figsize=(10, 6))
    plt.plot(r, v_baryon, '--', label='Visible Matter')
    plt.plot(r, v_dm, '-.', label='Dark Matter Halo')
    plt.plot(r, v_total, '-', linewidth=2, label='Total Expected')
    plt.title("Galaxy Rotation Curve")
    plt.xlabel("Radius (kpc)")
    plt.ylabel("Orbital Velocity (km/s)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    rotation_curve()
```
