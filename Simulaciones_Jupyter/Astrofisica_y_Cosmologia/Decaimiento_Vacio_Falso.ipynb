# Colapso Cuántico: Decaimiento del Vacío Falso

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
phi = np.linspace(-2, 2, 200)
V = 0.5*(phi**2 - 1)**2 - 0.1*phi # Potencial asimétrico

plt.plot(phi, V, lw=2, color='red')
plt.title('Potencial de Vacío Falso y Verdadero')
plt.xlabel('Campo Escalar $\phi$')
plt.ylabel('Potencial $V(\phi)$')
plt.axvline(-1.05, color='gray', linestyle='--', label='Vacío Verdadero')
plt.axvline(0.95, color='black', linestyle='--', label='Vacío Falso')
plt.legend(); plt.grid(True)
plt.show()
```
