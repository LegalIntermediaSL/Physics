# Supercomputación: Astrofisica HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

K = 10.0
gamma = 5.0/3.0

def tov(y, r):
    P, M = y
    if P <= 0: return [0, 0]
    rho = (P / K)**(1/gamma)
    dP = - (M + 4*np.pi*r**3 * P) * (rho + P) / (r * (r - 2*M) + 1e-10)
    dM = 4 * np.pi * r**2 * rho
    return [dP, dM]

r = np.linspace(1e-5, 20.0, 1000)
P_c = 1e-3
sol = odeint(tov, [P_c, 0.0], r)

plt.plot(r, sol[:, 0], label='Pressure')
plt.plot(r, sol[:, 1], label='Mass')
plt.title('Tolman-Oppenheimer-Volkoff Solution (NS Structure)')
plt.xlabel('Radius'); plt.legend()
plt.show()
```
