# Compendio Subtemático: Astrofisica Compendio 26

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def agn_jet():
    z = np.linspace(0, 10, 500)
    width = 0.1 * z
    jet_x = width * np.sin(5*z)
    
    plt.figure(figsize=(6, 8))
    plt.fill_betweenx(z, jet_x - width, jet_x + width, color='purple', alpha=0.6)
    plt.scatter(0, 0, s=200, c='k', label='Black Hole')
    plt.title("AGN Jet Structure (Simplified)")
    plt.xlabel("Radial distance")
    plt.ylabel("Jet axis (kpc)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    agn_jet()
```
