# Acreción Esférica de Bondi-Hoyle

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v_inf = np.linspace(10, 100, 100)
M_dot = 1 / v_inf**3  # Tasa de acrecion proporcional a v^-3

plt.plot(v_inf, M_dot)
plt.title('Tasa de Acreción de Bondi vs Velocidad del gas')
plt.xlabel('Velocidad asíntotica $v_\infty$')
plt.ylabel('Tasa $\dot{M}$')
plt.grid(True)
plt.show()
```
