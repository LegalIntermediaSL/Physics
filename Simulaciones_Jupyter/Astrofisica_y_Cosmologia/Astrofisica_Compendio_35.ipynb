# Compendio Subtemático: Astrofisica Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def jeans_escape():
    v = np.linspace(0, 15000, 500) # m/s
    T = 1000 # K
    m_H = 1.67e-27 # kg
    k = 1.38e-23
    
    # Maxwell-Boltzmann distribution
    f_v = 4 * np.pi * (m_H / (2 * np.pi * k * T))**1.5 * v**2 * np.exp(-m_H * v**2 / (2 * k * T))
    
    v_esc_earth = 11200 # m/s
    
    plt.figure(figsize=(10, 6))
    plt.plot(v, f_v, 'b-')
    plt.axvline(v_esc_earth, color='r', linestyle='--', label='Earth Escape Velocity')
    plt.fill_between(v, f_v, where=(v > v_esc_earth), color='r', alpha=0.3)
    plt.title("Jeans Escape (Hydrogen at 1000K)")
    plt.xlabel("Velocity (m/s)")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    jeans_escape()
```
