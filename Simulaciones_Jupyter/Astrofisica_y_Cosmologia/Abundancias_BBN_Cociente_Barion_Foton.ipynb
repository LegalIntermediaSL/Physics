# Nucleosíntesis Primordial (BBN) y eta

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eta = np.logspace(-10, -9, 100)
Y_He = 0.24 + 0.01 * np.log10(eta / 1e-10)

plt.semilogx(eta, Y_He, color='green')
plt.title('Fracción de masa primordial de Helio-4 vs Cociente barión/fotón')
plt.xlabel('$\eta = n_b / n_\gamma$')
plt.ylabel('$Y_P(^4He)$')
plt.axvline(6e-10, color='gray', linestyle='--', label='$\eta$ WMAP/Planck')
plt.legend(); plt.grid(True)
plt.show()
```
