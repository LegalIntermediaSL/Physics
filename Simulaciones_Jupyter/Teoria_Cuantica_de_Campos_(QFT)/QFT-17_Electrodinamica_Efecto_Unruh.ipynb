# Radiación Térmica del Vacío: Efecto Unruh

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(1e20, 1e22, 100) # Aceleracion propia extrema (m/s^2)
c, hbar, kB = 3e8, 1.05e-34, 1.38e-23
T_U = (hbar * a) / (2 * np.pi * c * kB)

plt.plot(a, T_U, color='orange')
plt.title('Temperatura de Unruh vs Aceleración')
plt.xlabel('Aceleración Constante $a$ (m/$s^2$)')
plt.ylabel('Temperatura Percibida $T$ (K)')
plt.grid(True)
plt.show()
```
