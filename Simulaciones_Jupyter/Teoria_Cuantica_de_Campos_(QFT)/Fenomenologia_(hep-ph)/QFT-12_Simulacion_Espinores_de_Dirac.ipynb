# [QFT-12] Espinores de Dirac y Rotaciones SU(2)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Mecánica Cuántica Relativista: La extraña geometría del Espinor de Dirac.
# Demostración topológica de por qué una partícula de espín 1/2 (Fermión) 
# debe rotar 720 grados (4 pi) para volver a su estado cuántico original.

def R_z(theta):
    # Matriz de rotación cuántica en SU(2) generada por las matrices de Pauli
    return np.array([[np.exp(-1j * theta / 2), 0],
                     [0, np.exp(1j * theta / 2)]])

# Espinor inicial (Espín puro 'up' a lo largo del eje Z)
psi_0 = np.array([1, 0])

thetas = np.linspace(0, 4 * np.pi, 200) # De 0 a 720 grados
phases = []
components_0 = []
components_1 = []

for theta in thetas:
    psi_rotated = R_z(theta) @ psi_0
    components_0.append(np.real(psi_rotated[0]))
    components_1.append(np.imag(psi_rotated[0]))

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficamos la evolución de la componente del espinor en el plano complejo
ax.plot(thetas, components_0, components_1, 'b-', lw=3)
ax.scatter([2*np.pi], [-1], [0], color='red', s=100, label="Rotación de 360° (Estado invertido)")
ax.scatter([4*np.pi], [1], [0], color='green', s=100, label="Rotación de 720° (Estado restaurado)")

ax.set_title("Espinor de Dirac: Rotación Cuántica de un Electrón en SU(2)", fontsize=14)
ax.set_xlabel('Ángulo de Rotación $\theta$ (Radianes)')
ax.set_ylabel('Parte Real de $\psi_1$')
ax.set_zlabel('Parte Imag de $\psi_1$')
ax.set_xticks([0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi])
ax.set_xticklabels(['$0$', '$\pi$', '$2\pi$ (360°)', '$3\pi$', '$4\pi$ (720°)'])
ax.legend()
plt.show()

# FÍSICA: A los 360 grados, un objeto macroscópico clásico (SO(3)) vuelve a su estado inicial. 
# Pero el Espinor Cuántico de un electrón adquiere un signo negativo (fase de -1). 
# ¡Se necesita una rotación completa adicional (720 grados) para que el electrón vuelva a la normalidad!

```
