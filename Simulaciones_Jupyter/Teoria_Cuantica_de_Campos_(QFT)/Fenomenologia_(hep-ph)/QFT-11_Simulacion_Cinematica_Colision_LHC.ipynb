# [QFT-11] Física de Aceleradores: Cinemática y Dispersión (Jets)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Cinemática Invariante de Dispersión en Colisionadores de Hadrones (LHC)
# Simulador de un evento de colisión protón-protón produciendo "Jets" transversales.

# Centro de Masas (CM). Protón 1 va hacia +Z, Protón 2 hacia -Z.
# Energía Altísima (p >> m, asumimos masa despreciable c=1)
E_beam = 6800.0 # GeV (Mitad de los 13.6 TeV del LHC)
P1 = np.array([E_beam, 0, 0, E_beam])
P2 = np.array([E_beam, 0, 0, -E_beam])
s = (P1[0]+P2[0])**2 - np.sum((P1[1:]+P2[1:])**2) # Variable de Mandelstam 's' (Invariante)

# Simulación probabilística de un evento de aniquilación y creación de 2 Jets
# Jet 1 y Jet 2 (Back-to-back por conservación de momento transversal pT)
theta_jet = np.random.uniform(np.pi/4, 3*np.pi/4) # Ángulo polar
phi_jet = np.random.uniform(0, 2*np.pi)          # Ángulo azimutal
p_T = E_beam * np.sin(theta_jet) # Momento transversal

# Función para generar un cono de lluvia de partículas (Shower/Jet)
def generate_jet(theta_c, phi_c, E_total, n_particles=50):
    particles = []
    for _ in range(n_particles):
        t = theta_c + np.random.normal(0, 0.1)
        p = phi_c + np.random.normal(0, 0.1)
        E = E_total / n_particles * np.random.uniform(0.5, 1.5)
        px = E * np.sin(t) * np.cos(p)
        py = E * np.sin(t) * np.sin(p)
        pz = E * np.cos(t)
        particles.append((px, py, pz))
    return np.array(particles)

jet1 = generate_jet(theta_jet, phi_jet, E_beam)
jet2 = generate_jet(np.pi - theta_jet, phi_jet + np.pi, E_beam) # Jet opuesto

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Haces incidentes (Z-axis)
ax.plot([0, 0], [0, 0], [-E_beam, 0], color='blue', lw=4, label='Haces de Protones LHC')
ax.plot([0, 0], [0, 0], [0, E_beam], color='blue', lw=4)

# Dispersión (Jets)
for p in jet1: ax.plot([0, p[0]], [0, p[1]], [0, p[2]], color='red', lw=1, alpha=0.5)
for p in jet2: ax.plot([0, p[0]], [0, p[1]], [0, p[2]], color='orange', lw=1, alpha=0.5)

ax.set_title(f"Cinemática Colisionador: $\sqrt{{s}}$ = {np.sqrt(s):.0f} GeV", fontsize=14)
ax.set_xlabel('Px (GeV)')
ax.set_ylabel('Py (GeV)')
ax.set_zlabel('Pz (GeV)')
ax.legend()
plt.show()
# FÍSICA: En el LHC, la conservación del momento transversal (pT) exige que si un 
# partón colisiona de forma inelástica dura (hard scattering), salgan chorros (jets) de 
# quarks y gluones en direcciones opuestas en el plano transversal a los haces incidentes.

```
