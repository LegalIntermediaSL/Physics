# [QFT-06] Unificación de Fuerzas: Running Couplings

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación Conceptual: Running of Coupling Constants (Beta Functions)
# Alpha_1 (U(1) Electromagnética), Alpha_2 (SU(2) Débil), Alpha_3 (SU(3) Fuerte)

# Escala de Energía en GeV (log10)
log_Q = np.linspace(2, 18, 100)

# Inversos de las constantes (1/alpha_i) evolucionan casi linealmente con el logaritmo de la energía
# Valores iniciales a la masa del Z_boson (Q=10^2 GeV) aprox.
alpha_inv_1_SM = 59.0
alpha_inv_2_SM = 29.6
alpha_inv_3_SM = 8.5

# Pendientes de las funciones Beta del Modelo Estándar (SM)
b_SM = np.array([-41/10, 19/6, 7])
# Pendientes de las funciones Beta del MSSM (SUSY) - cambian debido a lazos de sparticles
b_SUSY = np.array([-33/5, -1, 3])

# Evolución
alpha_1_sm = alpha_inv_1_SM + b_SM[0]/(2*np.pi) * (log_Q - 2)
alpha_2_sm = alpha_inv_2_SM + b_SM[1]/(2*np.pi) * (log_Q - 2)
alpha_3_sm = alpha_inv_3_SM + b_SM[2]/(2*np.pi) * (log_Q - 2)

alpha_1_susy = alpha_inv_1_SM + b_SUSY[0]/(2*np.pi) * (log_Q - 2)
alpha_2_susy = alpha_inv_2_SM + b_SUSY[1]/(2*np.pi) * (log_Q - 2)
alpha_3_susy = alpha_inv_3_SM + b_SUSY[2]/(2*np.pi) * (log_Q - 2)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Modelo Estándar (Sin Convergencia Perfecta)
ax1.plot(log_Q, alpha_1_sm, 'b-', label='$\alpha_1^{-1}$ (U(1))')
ax1.plot(log_Q, alpha_2_sm, 'g-', label='$\alpha_2^{-1}$ (SU(2))')
ax1.plot(log_Q, alpha_3_sm, 'r-', label='$\alpha_3^{-1}$ (SU(3))')
ax1.set_title("Evolución en el Modelo Estándar (Sin SUSY)")
ax1.set_xlabel("Escala de Energía $\log_{10}(Q/GeV)$")
ax1.set_ylabel("$\alpha^{-1}$")
ax1.grid()
ax1.legend()

# Modelo Supersimétrico (Convergencia en 10^16 GeV)
ax2.plot(log_Q, alpha_1_susy, 'b--', lw=2, label='$\alpha_1^{-1}$ (U(1)) MSSM')
ax2.plot(log_Q, alpha_2_susy, 'g--', lw=2, label='$\alpha_2^{-1}$ (SU(2)) MSSM')
ax2.plot(log_Q, alpha_3_susy, 'r--', lw=2, label='$\alpha_3^{-1}$ (SU(3)) MSSM')
ax2.axvline(16.2, color='purple', linestyle=':', lw=3, label='Escala GUT ($10^{16}$ GeV)')
ax2.set_title("Evolución con Supersimetría (MSSM)")
ax2.set_xlabel("Escala de Energía $\log_{10}(Q/GeV)$")
ax2.grid()
ax2.legend()

plt.suptitle("Renormalización y la Gran Unificación de Fuerzas (GUT)", fontsize=16)
plt.show()

```
