# Espectro del Quarkonium (Potencial de Cornell)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.linspace(0.1, 2.0, 200) # Distancia inter-quark fm
alpha_s = 0.39
sigma = 0.89 # GeV/fm String tension
V = - (4/3) * alpha_s / r + sigma * r

plt.plot(r, V, color='navy')
plt.title('Potencial de Cornell para el Confinamiento de Quarks')
plt.xlabel('Distancia $r$ (fm)')
plt.ylabel('$V(r)$ (GeV)')
plt.grid(True)
plt.show()
```
