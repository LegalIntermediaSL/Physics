# Ruptura Espontánea Radiativa (Coleman-Weinberg)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
phi = np.linspace(-5, 5, 200)
# Potencial escalar puro (sin masa tachyónica, symetria estricta) + Corrección loop 1
v_tree = 0.05 * phi**4
v_loop = 0.01 * phi**4 * (np.log(phi**2 + 1e-10) - 0.5)
V_eff = v_tree + v_loop

plt.plot(phi, V_eff, color='darkgreen')
plt.title('Potencial Efectivo Cuántico de Coleman-Weinberg')
plt.xlabel('Campo Escalar $\\phi$')
plt.ylabel('$V_{eff}(\\phi)$')
plt.grid(True)
plt.show()
```
