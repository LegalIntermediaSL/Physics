# Potencial de Uehling (Polarización del Vacío QED)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
def uehling_integrand(t, x):
    return np.exp(-2*t*x) * (1 + 1/(2*t**2)) * np.sqrt(t**2 - 1) / t**2

def uehling_potential(r_m):
    # r_m = r / lambda_Compton
    alpha = 1/137.036
    coulomb = 1/r_m
    integral = quad(uehling_integrand, 1, np.inf, args=(r_m,))[0]
    correction = (2*alpha/(3*np.pi)) * integral
    return coulomb * (1 + correction)

rm_vals = np.logspace(-2, 0, 50)
v_eff = [uehling_potential(r) for r in rm_vals]

plt.loglog(rm_vals, v_eff, label='Potencial Efectivo (Uehling)')
plt.loglog(rm_vals, 1/rm_vals, '--', label='Coulomb Desnudo')
plt.title('Polarización del Vacío en QED')
plt.xlabel('Distancia $r / \lambda_c$')
plt.ylabel('Potencial normalizado')
plt.legend(); plt.grid(True)
plt.show()
```
