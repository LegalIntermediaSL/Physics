# Propagador de Feynman de Klein-Gordon en 2D
Simulación interactiva avanzada.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de la amplitud de propagación estacionaria G(r) ~ K_0(m r) en 2D Euclídeo
from scipy.special import k0

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
R[R < 0.1] = 0.1 # Evitar singularidad

m_mass = 1.0
G = k0(m_mass * R) # Función modificada de Bessel de segunda especie

plt.figure(figsize=(8, 6))
cp = plt.contourf(X, Y, G, levels=50, cmap='plasma')
plt.colorbar(cp, label='Amplitud del Propagador')
plt.title('Propagador de Feynman (Campo Escalar Libre 2D Euclídeo)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
