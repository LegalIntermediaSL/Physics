# [QFT-10] Cromodinámica Reticular: Markov Chain Monte Carlo

```python
import numpy as np
import matplotlib.pyplot as plt

# Teoría de Campos en un Retículo (Lattice Field Theory)
# Emularemos el Algoritmo de Monte Carlo Metropolis-Hastings en un Modelo de Ising 2D.
# Este modelo es el "Hola Mundo" para entender cómo los supercomputadores resuelven
# el confinamiento de quarks y gluones en Cromodinámica Cuántica (Lattice QCD).

N = 50  # Tamaño del Lattice (Grid 50x50)
T = 2.2 # Temperatura simulada (La T crítica de Onsager es ~2.269)
beta = 1.0 / T

# Inicializamos el retículo (espines aleatorios +1 o -1)
np.random.seed(42)
lattice = 2 * np.random.randint(2, size=(N, N)) - 1

def metropolis_step(lattice):
    for _ in range(N * N):
        x = np.random.randint(0, N)
        y = np.random.randint(0, N)
        
        S_i = lattice[x, y]
        # Suma de vecinos (Condiciones de contorno periódicas)
        S_vecinos = lattice[(x+1)%N, y] + lattice[(x-1)%N, y] + lattice[x, (y+1)%N] + lattice[x, (y-1)%N]
        
        # Variación de Energía si volteamos el espín S_i
        dE = 2 * S_i * S_vecinos
        
        # Aceptación de Metropolis-Hastings
        if dE < 0 or np.random.rand() < np.exp(-beta * dE):
            lattice[x, y] *= -1
            
    return lattice

# Calentamiento (Burn-in) del Monte Carlo
for _ in range(100):
    lattice = metropolis_step(lattice)

plt.figure(figsize=(8, 8))
plt.imshow(lattice, cmap='coolwarm', interpolation='nearest')
plt.title(f"Monte Carlo en el Retículo (Lattice Model, T={T})", fontsize=14)
plt.axis('off')
plt.show()
# FÍSICA: Observamos clústeres magnéticos macroscópicos. A través de este método,
# integrando sobre todas las configuraciones del espacio de gauge, calculamos
# observables no perturbativos como la masa exacta del protón en Lattice QCD.

```
