# Sección Eficaz Diferencial (Dispersión de Rutherford vía QED)
Simulación interactiva avanzada.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sección eficaz diferencial d_sigma/d_Omega ~ 1 / sin^4(theta/2)
theta = np.linspace(0.1, np.pi, 200)
d_sigma = 1 / (np.sin(theta/2)**4)

plt.figure(figsize=(8, 5))
plt.plot(theta, np.log10(d_sigma), 'g-')
plt.title('Sección Eficaz Diferencial de Dispersión e-e- (Límite no relativista)')
plt.xlabel('Ángulo de Dispersión $\\theta$ (Radianes)')
plt.ylabel('log($\\frac{d\\sigma}{d\\Omega}$)')
plt.grid(True)
plt.show()
```
