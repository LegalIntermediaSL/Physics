# Mecanismo de Higgs y Ruptura Espontánea de Simetría
Simulación interactiva avanzada.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Potencial del sombrero mexicano V(phi) = -mu^2 |phi|^2 + lambda |phi|^4
phi1 = np.linspace(-2, 2, 100)
phi2 = np.linspace(-2, 2, 100)
Phi1, Phi2 = np.meshgrid(phi1, phi2)
R2 = Phi1**2 + Phi2**2

mu2 = 1.0
lam = 0.5
V = -mu2 * R2 + lam * R2**2

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(Phi1, Phi2, V, cmap='viridis', alpha=0.8)

# El mínimo no está en el origen (0,0), sino en un anillo de radio v = sqrt(mu^2/lambda)
v = np.sqrt(mu2 / lam)
theta = np.linspace(0, 2*np.pi, 100)
ax.plot(v*np.cos(theta), v*np.sin(theta), -mu2**2/(4*lam), color='r', linewidth=3, label='Estado de Vacío (Ruptura)')

ax.set_title('Potencial de Higgs y Ruptura Espontánea de Simetría O(2)')
ax.set_xlabel('$\\phi_1$')
ax.set_ylabel('$\\phi_2$')
ax.set_zlabel('V($\\phi$)')
plt.legend()
plt.show()
```
