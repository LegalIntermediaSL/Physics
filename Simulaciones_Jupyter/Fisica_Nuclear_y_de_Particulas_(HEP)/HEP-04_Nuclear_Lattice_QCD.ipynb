# Investigación Avanzada: Nuclear Lattice QCD

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Toy Lattice QCD: 2D Ising Model as a placeholder for Lattice Gauge Theory
# Demonstrates Monte Carlo methods used in Lattice QCD.

def initial_lattice(N):
    return np.random.choice([-1, 1], size=(N, N))

def metropolis_step(lattice, beta):
    N = lattice.shape[0]
    x, y = np.random.randint(0, N, 2)
    S = lattice[x, y]
    
    # Periodic boundary conditions
    neighbors = lattice[(x+1)%N, y] + lattice[(x-1)%N, y] + \
                lattice[x, (y+1)%N] + lattice[x, (y-1)%N]
                
    delta_E = 2 * S * neighbors
    
    if delta_E < 0 or np.random.rand() < np.exp(-beta * delta_E):
        lattice[x, y] *= -1

if __name__ == "__main__":
    N = 50
    lattice = initial_lattice(N)
    steps = 100000
    beta = 0.44  # Critical beta for 2D Ising model ~ 0.4406
    
    # Thermalization
    for _ in range(steps):
        metropolis_step(lattice, beta)
        
    plt.imshow(lattice, cmap='gray', interpolation='nearest')
    plt.title(f'Toy Lattice Field Theory (Monte Carlo) at beta={beta}')
    plt.axis('off')
    plt.show()
```
