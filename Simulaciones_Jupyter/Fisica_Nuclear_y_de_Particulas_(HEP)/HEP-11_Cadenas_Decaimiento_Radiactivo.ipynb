# Decaimiento Radiactivo (Bateman)

Ecuaciones diferenciales acopladas para series radiactivas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 100)
N1 = np.exp(-t)
N2 = 2 * (np.exp(-0.5*t) - np.exp(-t))
N3 = 1 - N1 - N2

plt.figure(figsize=(8,5))
plt.plot(t, N1, label='Isótopo Padre')
plt.plot(t, N2, label='Isótopo Intermedio')
plt.plot(t, N3, label='Producto Estable')
plt.title('Cadena de Decaimiento Radiactivo')
plt.legend()
plt.grid()
plt.show()
```
