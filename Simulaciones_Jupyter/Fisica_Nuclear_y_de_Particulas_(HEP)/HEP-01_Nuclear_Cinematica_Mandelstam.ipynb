# Investigación Avanzada: Nuclear Cinematica Mandelstam

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Relativistic kinematics: Mandelstam variables
# For a process 1 + 2 -> 3 + 4
# s = (p1 + p2)^2
# t = (p1 - p3)^2
# u = (p1 - p4)^2

def compute_mandelstam(E_cm, m1, m2, m3, m4, theta):
    """
    Compute Mandelstam variables s, t, u given CM energy and scattering angle.
    Assumes natural units c=1.
    """
    s = E_cm**2
    
    # Momenta in CM frame
    p_initial = np.sqrt(max(0, (s - (m1 + m2)**2) * (s - (m1 - m2)**2))) / (2 * E_cm)
    p_final = np.sqrt(max(0, (s - (m3 + m4)**2) * (s - (m3 - m4)**2))) / (2 * E_cm)
    
    E1 = np.sqrt(m1**2 + p_initial**2)
    E3 = np.sqrt(m3**2 + p_final**2)
    
    t = m1**2 + m3**2 - 2*E1*E3 + 2*p_initial*p_final*np.cos(theta)
    u = m1**2 + m2**2 + m3**2 + m4**2 - s - t
    
    return s, t, u

if __name__ == "__main__":
    # E.g., elastic scattering of particles of mass m
    m = 1.0
    E_cm = 5.0
    
    theta_vals = np.linspace(0, np.pi, 100)
    t_vals = []
    u_vals = []
    
    for theta in theta_vals:
        s, t, u = compute_mandelstam(E_cm, m, m, m, m, theta)
        t_vals.append(t)
        u_vals.append(u)
        
    plt.figure(figsize=(8, 5))
    plt.plot(theta_vals, t_vals, label='t (Momentum transfer squared)', color='red')
    plt.plot(theta_vals, u_vals, label='u', color='blue')
    
    plt.title(f'Mandelstam Variables vs Scattering Angle (CM Energy = {E_cm})')
    plt.xlabel('Scattering Angle (radians)')
    plt.ylabel('Mandelstam Variables (GeV^2)')
    plt.legend()
    plt.grid(True)
    plt.show()
```
