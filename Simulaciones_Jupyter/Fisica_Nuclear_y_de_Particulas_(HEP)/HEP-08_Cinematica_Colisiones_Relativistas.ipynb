# Colisiones Relativistas (Mandelstam)

Dispersión invariante y conservación del cuadrimomento.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, np.pi, 100)
cross_section = 1 / (1 - np.cos(theta) + 0.1)**2

plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, cross_section, color='crimson')
plt.title('Sección Eficaz Diferencial (Dispersión de Rutherford/Mott)')
plt.show()
```
