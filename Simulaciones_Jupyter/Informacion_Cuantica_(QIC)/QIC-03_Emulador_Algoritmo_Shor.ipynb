# [QIC-03] Algoritmo de Shor

Simulación cuántica pura en Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Emulador Clásico del Algoritmo de Shor (Quantum Period Finding)
# Factorización de N usando la función modular f(x) = a^x mod N

N = 15  # Número a factorizar
a = 7   # Coprimo aleatorio (mcd(7, 15) = 1)

print(f"Objetivo: Factorizar N = {N}")
print(f"Base aleatoria elegida: a = {a}")

# 1. Búsqueda de período (Emulando el registro cuántico)
x_vals = np.arange(20)
f_x = [pow(a, int(x), N) for x in x_vals]

plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_x, 'ro--', markersize=8)
plt.title(f'Función Modular: f(x) = {a}^x mod {N}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xticks(x_vals)
plt.grid(True)
plt.show()

# Determinar el período r visualmente/clásicamente
r = 4  # Observando la gráfica, el patrón 1, 7, 4, 13 se repite cada 4 pasos
print(f"Período cuántico hallado (r): {r}")

if r % 2 == 0:
    factor1 = np.gcd(a**(r//2) - 1, N)
    factor2 = np.gcd(a**(r//2) + 1, N)
    print(f"Factores de {N}: {factor1} y {factor2}")
else:
    print("El período es impar, el algoritmo falla para esta base 'a'. Intente otra.")

```
