# [QIC-04] Algoritmo de Grover

Simulación cuántica pura en Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Emulador del Algoritmo de Grover (Búsqueda no estructurada)
n_qubits = 4
N = 2**n_qubits
target_x = 11  # El elemento secreto que estamos buscando

# 1. Estado inicial de superposición uniforme |s>
state = np.ones(N) / np.sqrt(N)

# 2. Oráculo O_f
def oracle(state):
    new_state = state.copy()
    new_state[target_x] *= -1  # Invertir fase de la solución
    return new_state

# 3. Operador de Difusión D (Inversión sobre la media)
def diffusion(state):
    mean = np.mean(state)
    return 2 * mean - state

# Número óptimo de iteraciones de Grover: pi/4 * sqrt(N)
iterations = int(np.pi / 4 * np.sqrt(N))

prob_history = []

for i in range(iterations + 1):
    prob_history.append(state**2)
    # Aplicar Grover
    state = oracle(state)
    state = diffusion(state)

# Visualizar la amplificación de probabilidad del elemento target
prob_history = np.array(prob_history)

plt.figure(figsize=(10, 6))
plt.plot(range(iterations + 1), prob_history[:, target_x], 'g-o', linewidth=2, label=f'Prob(Elemento {target_x})')
plt.plot(range(iterations + 1), prob_history[:, 0], 'r--', alpha=0.5, label='Prob(Otros Elementos)')
plt.title(f"Amplificación de Amplitud de Grover (N={N})")
plt.xlabel("Iteraciones de Grover")
plt.ylabel("Probabilidad de Medición")
plt.legend()
plt.grid(True)
plt.show()

```
