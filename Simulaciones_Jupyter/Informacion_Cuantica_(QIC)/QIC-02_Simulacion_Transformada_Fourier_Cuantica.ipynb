# [QIC-02] QFT

Simulación cuántica pura en Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de la Transformada de Fourier Cuántica (QFT)
# Calculamos la matriz unitaria de la QFT para n qubits.

n_qubits = 3
N = 2**n_qubits
omega = np.exp(2.0 * np.pi * 1j / N)

# Matriz de QFT
QFT_matrix = np.zeros((N, N), dtype=complex)
for j in range(N):
    for k in range(N):
        QFT_matrix[j, k] = omega ** (j * k) / np.sqrt(N)

# Visualizamos la parte real de la matriz unitaria QFT
plt.figure(figsize=(8, 8))
plt.imshow(np.real(QFT_matrix), cmap='coolwarm')
plt.title(f"Parte Real de la Matriz QFT para {n_qubits} Qubits")
plt.colorbar(label='Amplitud Real')
plt.xlabel("Estado j")
plt.ylabel("Estado k")
plt.show()

# Verificar unitariedad: QFT * QFT_dagger = Identidad
identity_check = np.dot(QFT_matrix, np.conj(QFT_matrix).T)
print("Verificación de unitariedad (debería ser la identidad):")
print(np.round(np.real(identity_check), 2))

```
