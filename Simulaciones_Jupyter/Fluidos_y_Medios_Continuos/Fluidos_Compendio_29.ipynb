# Compendio Subtemático: Fluidos Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def bulk_modulus():
    dP = np.linspace(0, 1e7, 100) # Pa
    K = 2.2e9 # Pa (water)
    dV_V = -dP / K
    plt.plot(dP/1e6, dV_V*100)
    plt.xlabel('Pressure Increase (MPa)')
    plt.ylabel('Volume Change (%)')
    plt.title('Fluid Compressibility')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    bulk_modulus()
```
