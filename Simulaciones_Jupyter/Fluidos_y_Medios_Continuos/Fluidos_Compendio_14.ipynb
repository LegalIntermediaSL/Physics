# Compendio Subtemático: Fluidos Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def siphon():
    h = np.linspace(0.1, 10, 100) # m
    g = 9.81
    v = np.sqrt(2 * g * h)
    plt.plot(h, v)
    plt.xlabel('Height Difference (m)')
    plt.ylabel('Velocity (m/s)')
    plt.title('Siphon Velocity')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    siphon()
```
