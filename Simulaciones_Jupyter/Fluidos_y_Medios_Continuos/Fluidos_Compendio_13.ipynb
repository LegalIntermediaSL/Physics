# Compendio Subtemático: Fluidos Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def pitot():
    delta_P = np.linspace(0, 5000, 100) # Pa
    rho = 1.225 # air
    v = np.sqrt((2 * delta_P) / rho)
    plt.plot(delta_P, v)
    plt.xlabel('Pressure Difference (Pa)')
    plt.ylabel('Velocity (m/s)')
    plt.title('Pitot Tube Measurement')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    pitot()
```
