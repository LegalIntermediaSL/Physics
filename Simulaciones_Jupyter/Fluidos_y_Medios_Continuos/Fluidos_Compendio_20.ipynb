# Compendio Subtemático: Fluidos Compendio 20

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def boundary_layer():
    x = np.linspace(0.1, 5, 100) # m
    Re_x = 1e5 * x
    delta = 5.0 * x / np.sqrt(Re_x)
    plt.plot(x, delta*1000)
    plt.xlabel('Distance along plate x (m)')
    plt.ylabel('Boundary Layer Thickness (mm)')
    plt.title('Blasius Boundary Layer')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    boundary_layer()
```
