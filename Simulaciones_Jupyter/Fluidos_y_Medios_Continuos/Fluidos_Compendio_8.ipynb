# Compendio Subtemático: Fluidos Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def capillary():
    r = np.linspace(0.0001, 0.005, 100) # m
    gamma = 0.0728 # N/m (water)
    theta = 0 # contact angle
    rho = 1000
    g = 9.81
    h = (2 * gamma * np.cos(theta)) / (rho * g * r)
    plt.plot(r*1000, h*100)
    plt.xlabel('Tube radius (mm)')
    plt.ylabel('Capillary height (cm)')
    plt.title('Capillary Action')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    capillary()
```
