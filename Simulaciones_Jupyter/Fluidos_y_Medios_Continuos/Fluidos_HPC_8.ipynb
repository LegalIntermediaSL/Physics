# Supercomputación: Fluidos HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def taylor_couette():
    r = np.linspace(1, 2, 50)
    theta = np.linspace(0, 2*np.pi, 100)
    R, Theta = np.meshgrid(r, theta)
    
    omega_inner = 1.0
    omega_outer = 0.0
    
    A = (omega_outer * 4 - omega_inner * 1) / (4 - 1)
    B = (omega_inner - omega_outer) * 4 / (4 - 1)
    v_theta = A * r + B / r
    
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.contourf(Theta, R, np.tile(v_theta, (100, 1)), cmap='cividis')
    plt.title('Taylor-Couette Flow Velocity Profile')
    plt.show()

if __name__ == '__main__':
    taylor_couette()
```
