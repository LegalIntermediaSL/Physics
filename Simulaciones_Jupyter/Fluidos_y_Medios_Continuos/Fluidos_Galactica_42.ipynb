# Fluidos Galactica 42

Simulación matemática generada paramétricamente para el compendio galáctico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def sim_42():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.title('Simulation 42')
    plt.show()

if __name__ == '__main__':
    sim_42()
```
