# Compendio Subtemático: Fluidos Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def orifice_flow():
    h = np.linspace(0.1, 10, 100) # m
    Cd = 0.6
    A = 0.01 # m^2
    g = 9.81
    Q = Cd * A * np.sqrt(2 * g * h)
    plt.plot(h, Q)
    plt.xlabel('Head h (m)')
    plt.ylabel('Flow Rate Q (m^3/s)')
    plt.title('Flow through Orifice')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    orifice_flow()
```
