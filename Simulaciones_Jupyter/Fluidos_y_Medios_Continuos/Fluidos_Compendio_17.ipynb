# Compendio Subtemático: Fluidos Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gravity_waves():
    lam = np.linspace(1, 100, 100) # m
    g = 9.81
    k = 2 * np.pi / lam
    omega = np.sqrt(g * k)
    c = omega / k
    plt.plot(lam, c)
    plt.xlabel('Wavelength (m)')
    plt.ylabel('Phase Velocity (m/s)')
    plt.title('Deep Water Gravity Waves')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    gravity_waves()
```
