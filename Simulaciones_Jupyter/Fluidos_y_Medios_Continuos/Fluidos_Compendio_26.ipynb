# Compendio Subtemático: Fluidos Compendio 26

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def froude():
    v = np.linspace(0, 10, 100)
    y = 2 # depth
    g = 9.81
    Fr = v / np.sqrt(g * y)
    plt.plot(v, Fr)
    plt.axhline(1, color='r', linestyle='--', label='Critical Flow (Fr=1)')
    plt.xlabel('Velocity (m/s)')
    plt.ylabel('Froude Number')
    plt.title('Froude Number vs Velocity')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    froude()
```
