# Compendio Subtemático: Fluidos Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rayleigh_taylor():
    k = np.linspace(0.1, 10, 100) # wavenumber
    rho1 = 1000 # heavy fluid top
    rho2 = 1.2 # light fluid bottom
    g = 9.81
    A = (rho1 - rho2) / (rho1 + rho2) # Atwood number
    omega_sq = g * k * A
    plt.plot(k, np.sqrt(omega_sq))
    plt.xlabel('Wavenumber k (1/m)')
    plt.ylabel('Growth Rate (1/s)')
    plt.title('Rayleigh-Taylor Instability Growth Rate')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    rayleigh_taylor()
```
