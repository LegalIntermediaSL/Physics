# Compendio Subtemático: Fluidos Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def drag():
    v = np.linspace(0, 30, 100) # m/s
    Cd = 0.47 # sphere
    rho = 1.225 # air
    A = np.pi * (0.1)**2
    Fd = 0.5 * rho * v**2 * Cd * A
    plt.plot(v, Fd)
    plt.xlabel('Velocity (m/s)')
    plt.ylabel('Drag Force (N)')
    plt.title('Drag Force vs Velocity')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    drag()
```
