# Compendio Subtemático: Fluidos Compendio 31

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def drop_evaporation():
    t = np.linspace(0, 100, 100) # time
    R0 = 1.0 # initial radius
    k = 0.01 # evaporation constant
    R = np.sqrt(np.maximum(R0**2 - k * t, 0))
    plt.plot(t, R)
    plt.xlabel('Time (s)')
    plt.ylabel('Drop Radius (mm)')
    plt.title('Droplet Evaporation (D^2 law)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    drop_evaporation()
```
