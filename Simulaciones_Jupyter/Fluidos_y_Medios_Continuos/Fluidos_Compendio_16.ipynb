# Compendio Subtemático: Fluidos Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def capillary_waves():
    lam = np.linspace(0.001, 0.05, 100) # m
    k = 2 * np.pi / lam
    gamma = 0.0728
    rho = 1000
    omega = np.sqrt((gamma * k**3) / rho)
    c = omega / k
    plt.plot(lam*1000, c)
    plt.xlabel('Wavelength (mm)')
    plt.ylabel('Phase Velocity (m/s)')
    plt.title('Capillary Waves')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    capillary_waves()
```
