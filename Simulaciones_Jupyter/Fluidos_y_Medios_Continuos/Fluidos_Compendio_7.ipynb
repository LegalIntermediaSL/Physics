# Compendio Subtemático: Fluidos Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def poiseuille():
    r = np.linspace(0, 0.05, 100) # m
    R = 0.05 # m
    delta_P = 1000 # Pa
    L = 10 # m
    mu = 0.001 # Pa.s
    v = (delta_P / (4 * mu * L)) * (R**2 - r**2)
    plt.plot(r, v)
    plt.xlabel('Radial distance (m)')
    plt.ylabel('Velocity (m/s)')
    plt.title('Poiseuille Flow Profile')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    poiseuille()
```
