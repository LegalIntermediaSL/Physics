# Laboratorio Especializado: Fluidos Especializado 1

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def solve_burgers(nx, nt, dt, nu):
    L = 2.0
    dx = L / (nx - 1)
    x = np.linspace(0, L, nx)
    u = np.sin(np.pi * x)
    un = np.empty_like(u)

    plt.figure(figsize=(10, 6))
    plt.plot(x, u, label='t = 0', lw=2)

    for n in range(nt):
        un = u.copy()
        for i in range(1, nx - 1):
            u[i] = un[i] - un[i] * dt / dx * (un[i] - un[i - 1]) + nu * dt / dx**2 * (un[i + 1] - 2 * un[i] + un[i - 1])
        u[0] = 0
        u[-1] = 0
        
        if n % (nt // 5) == 0 and n > 0:
            plt.plot(x, u, label=f't = {n*dt:.2f}')

    plt.title("1D Burgers' Equation: Shock Wave Formation")
    plt.xlabel('x')
    plt.ylabel('u(x,t)')
    plt.legend()
    plt.grid(True)
    plt.savefig('burgers_shock_wave.png')

if __name__ == "__main__":
    solve_burgers(nx=201, nt=400, dt=0.001, nu=0.01)
```
