# Vórtices (Navier-Stokes)

Visualización del campo vectorial de la vorticidad de un fluido.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.linspace(-2,2,20), np.linspace(-2,2,20))
u = -y / (x**2 + y**2 + 0.1)
v = x / (x**2 + y**2 + 0.1)

plt.figure(figsize=(6,6))
plt.quiver(x, y, u, v, color='navy')
plt.title('Campo Vectorial de un Vórtice 2D')
plt.axis('equal')
plt.grid(True)
plt.show()
```
