# Compendio Subtemático: Fluidos Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def pascal():
    A1 = 0.01 # m^2
    A2 = np.linspace(0.01, 0.5, 100) # m^2
    F1 = 100 # N
    F2 = F1 * (A2 / A1)
    plt.plot(A2, F2)
    plt.xlabel('Area 2 (m^2)')
    plt.ylabel('Force 2 (N)')
    plt.title('Pascal\\'s Principle (F1 = 100 N, A1 = 0.01 m^2)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    pascal()
```
