# Compendio Subtemático: Optica Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def newton_rings():
    x = np.linspace(-5e-3, 5e-3, 500)
    X, Y = np.meshgrid(x, x)
    R = np.sqrt(X**2 + Y**2)
    R_lens = 1.0
    lam = 500e-9
    d = R**2 / (2*R_lens)
    I = np.sin(2*np.pi*d/lam)**2
    
    plt.imshow(I, cmap='gray')
    plt.title("Newton's Rings")
    plt.axis('off')
    plt.show()

if __name__ == '__main__': 
    newton_rings()
```
