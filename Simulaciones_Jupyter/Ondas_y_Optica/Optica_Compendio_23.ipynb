# Compendio Subtemático: Optica Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fiber():
    n_core = 1.48
    n_clad = np.linspace(1.4, 1.479, 100)
    NA = np.sqrt(n_core**2 - n_clad**2)
    
    plt.plot(n_clad, NA)
    plt.title("Fiber Optic Numerical Aperture")
    plt.xlabel('Cladding Refractive Index')
    plt.ylabel('Numerical Aperture (NA)')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    fiber()
```
