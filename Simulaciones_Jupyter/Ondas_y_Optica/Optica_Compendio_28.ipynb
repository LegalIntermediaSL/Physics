# Compendio Subtemático: Optica Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def birefringence():
    d = np.linspace(0, 1e-4, 500)
    dn = 0.01
    lam = 500e-9
    phase = 2*np.pi*d*dn/lam
    
    plt.plot(d*1e6, np.sin(phase/2)**2)
    plt.title("Birefringence Intensity (Crossed Polarizers)")
    plt.xlabel('Thickness (um)')
    plt.ylabel('Transmitted Intensity')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    birefringence()
```
