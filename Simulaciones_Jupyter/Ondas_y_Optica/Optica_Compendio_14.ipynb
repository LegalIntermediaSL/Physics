# Compendio Subtemático: Optica Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def thin_film():
    d = np.linspace(0, 1e-6, 500)
    lam = 500e-9
    n = 1.5
    I = np.cos(2*np.pi*n*d/lam)**2
    plt.plot(d*1e9, I)
    plt.title("Thin Film Interference")
    plt.xlabel('Thickness (nm)')
    plt.ylabel('Reflected Intensity')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    thin_film()
```
