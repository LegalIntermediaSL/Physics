# Compendio Subtemático: Optica Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fabry_perot():
    phase = np.linspace(-np.pi, np.pi, 1000)
    R = 0.9
    F = 4*R/(1-R)**2
    T = 1 / (1 + F*np.sin(phase/2)**2)
    plt.plot(phase, T)
    plt.title("Fabry-Perot Transmittance")
    plt.xlabel('Phase Difference (rad)')
    plt.ylabel('Transmittance')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    fabry_perot()
```
