# Compendio Subtemático: Optica Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def pockels():
    V = np.linspace(0, 1000, 100)
    r = 1e-11
    n = 1.5
    dn = 0.5 * n**3 * r * V
    plt.plot(V, dn)
    plt.title("Pockels Effect (Linear Electro-optic)")
    plt.xlabel('Voltage (V)')
    plt.ylabel('Change in Refractive Index (dn)')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    pockels()
```
