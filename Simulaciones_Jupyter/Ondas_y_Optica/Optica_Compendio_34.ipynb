# Compendio Subtemático: Optica Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def laser():
    t = np.linspace(0, 1e-6, 1000)
    N2 = 1 - np.exp(-t/1e-7)
    plt.plot(t*1e9, N2)
    plt.title("Laser Population Inversion Growth")
    plt.xlabel('Time (ns)')
    plt.ylabel('Population of State 2 (N2)')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    laser()
```
