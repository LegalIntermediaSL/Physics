# Compendio Subtemático: Optica Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def chrom_ab():
    lam = np.linspace(400, 700, 100)
    n = 1.5 + 4e4 / lam**2
    f = 100 / (n - 1)
    
    plt.plot(lam, f)
    plt.title("Chromatic Aberration")
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Focal Length')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    chrom_ab()
```
