# Compendio Subtemático: Optica Compendio 22

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gauss_beam():
    z = np.linspace(-5, 5, 100)
    z0 = 1.0
    w0 = 1.0
    w = w0 * np.sqrt(1 + (z/z0)**2)
    
    plt.plot(z, w, 'b')
    plt.plot(z, -w, 'b')
    plt.fill_between(z, -w, w, color='blue', alpha=0.2)
    plt.title("Gaussian Beam Propagation")
    plt.xlabel('Propagation Distance (z)')
    plt.ylabel('Beam Width (w)')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    gauss_beam()
```
