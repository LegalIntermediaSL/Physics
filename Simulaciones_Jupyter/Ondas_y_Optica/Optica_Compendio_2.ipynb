# Compendio Subtemático: Optica Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def tir():
    n1, n2 = 1.5, 1.0
    theta1 = np.linspace(0, np.pi/2, 500)
    sin_theta2 = (n1/n2)*np.sin(theta1)
    
    arcsin_val = np.arcsin(np.clip(sin_theta2, -1, 1))
    R_s = np.where(sin_theta2 <= 1, np.abs((n1*np.cos(theta1) - n2*np.cos(arcsin_val)) / (n1*np.cos(theta1) + n2*np.cos(arcsin_val)))**2, 1)
    
    plt.plot(np.degrees(theta1), R_s)
    plt.xlabel('Angle of Incidence (degrees)')
    plt.ylabel('Reflectance')
    plt.title('Total Internal Reflection')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    tir()
```
