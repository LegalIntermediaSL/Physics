# Compendio Subtemático: Optica Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def concave_mirror():
    f, do = 10, 30
    di = 1/(1/f - 1/do)
    plt.plot([do, di], [1/do, 1/di], marker='o')
    plt.title("Concave Mirror Parameters")
    plt.xlabel('Distance')
    plt.ylabel('Inverse Distance')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    concave_mirror()
```
