# Laboratorio Especializado: Optica Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def michelson_interferometer():
    d = np.linspace(0, 2e-6, 1000) # Mirror displacement
    lambda_1 = 600e-9 # Wavelength 1
    lambda_2 = 605e-9 # Wavelength 2 (closely spaced)
    
    intensity_1 = 1 + np.cos(4 * np.pi * d / lambda_1)
    intensity_2 = 1 + np.cos(4 * np.pi * d / lambda_2)
    
    total_intensity = intensity_1 + intensity_2
    
    plt.figure(figsize=(8, 5))
    plt.plot(d * 1e9, total_intensity)
    plt.title("Michelson Interferometer (Beats from 2 closely spaced wavelengths)")
    plt.xlabel("Mirror Displacement (nm)")
    plt.ylabel("Intensity")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    michelson_interferometer()
```
