# Compendio Subtemático: Optica Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rect_diff():
    n = 256
    x = np.linspace(-1, 1, n)
    X, Y = np.meshgrid(x, x)
    aperture = (np.abs(X) <= 0.1) & (np.abs(Y) <= 0.2)
    I = np.abs(np.fft.fftshift(np.fft.fft2(aperture)))**2
    
    plt.imshow(I[n//2-30:n//2+30, n//2-30:n//2+30], cmap='hot')
    plt.title("Rectangular Aperture Diffraction")
    plt.axis('off')
    plt.show()

if __name__ == '__main__': 
    rect_diff()
```
