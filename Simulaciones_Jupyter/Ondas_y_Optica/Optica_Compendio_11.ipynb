# Compendio Subtemático: Optica Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def grating():
    theta = np.linspace(-0.2, 0.2, 1000)
    lam, d, N = 500e-9, 2e-6, 10
    delta = 2*np.pi*d*np.sin(theta)/lam
    I = (np.sin(N*delta/2) / (np.sin(delta/2) + 1e-9))**2
    plt.plot(theta, I)
    plt.title("Diffraction Grating")
    plt.xlabel('Angle (rad)')
    plt.ylabel('Intensity')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    grating()
```
