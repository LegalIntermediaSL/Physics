# Compendio Subtemático: Optica Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def malus():
    theta = np.linspace(0, 2*np.pi, 100)
    I = np.cos(theta)**2
    plt.plot(np.degrees(theta), I)
    plt.title("Malus's Law")
    plt.xlabel('Angle (degrees)')
    plt.ylabel('Normalized Intensity')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    malus()
```
