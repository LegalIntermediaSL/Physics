# Polarización y Parámetros de Stokes

Trazado del elipsoide de polarización del campo electromagnético.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 200)
Ex = np.cos(t)
Ey = np.cos(t + np.pi/4) # Desfase

plt.figure(figsize=(5,5))
plt.plot(Ex, Ey, color='magenta', lw=2)
plt.title('Elipse de Polarización (Luz Elíptica)')
plt.axis('equal')
plt.grid()
plt.show()
```
