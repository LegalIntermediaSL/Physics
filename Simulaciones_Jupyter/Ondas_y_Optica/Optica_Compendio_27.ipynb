# Compendio Subtemático: Optica Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def holography():
    n = 256
    x = np.linspace(-1, 1, n)
    X, Y = np.meshgrid(x, x)
    ref = np.exp(1j * 20 * X)
    obj = np.exp(1j * 50 * (X**2 + Y**2))
    
    intensity = np.abs(ref + obj)**2
    plt.imshow(intensity, cmap='gray')
    plt.title("Hologram Interference Pattern")
    plt.axis('off')
    plt.show()

if __name__ == '__main__': 
    holography()
```
