# Compendio Subtemático: Optica Compendio 26

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fourier_filter():
    n = 256
    x = np.linspace(-1, 1, n)
    X, Y = np.meshgrid(x, x)
    img = np.sin(20*X) + np.sin(50*Y)
    
    F = np.fft.fftshift(np.fft.fft2(img))
    filter_mask = (X**2 + Y**2) < 0.2**2
    filtered = np.fft.ifft2(np.fft.ifftshift(F * filter_mask))
    
    plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title("Original")
    plt.subplot(122), plt.imshow(np.abs(filtered), cmap='gray'), plt.title("Filtered")
    plt.show()

if __name__ == '__main__': 
    fourier_filter()
```
