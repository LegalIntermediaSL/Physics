# Laboratorio Especializado: Termodinamica Especializado 7

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def lennard_jones_force(r, epsilon=1.0, sigma=1.0):
    r_mag = np.linalg.norm(r, axis=1, keepdims=True)
    r_hat = r / r_mag
    f_mag = 48 * epsilon * ((sigma**12 / r_mag**13) - 0.5 * (sigma**6 / r_mag**7))
    return f_mag * r_hat

N = 10
pos = np.random.rand(N, 2) * 10
vel = np.random.randn(N, 2) * 0.1
dt = 0.01

trajectory = [pos.copy()]
for step in range(500):
    forces = np.zeros_like(pos)
    for i in range(N):
        r = pos - pos[i]
        r[i] = np.inf
        forces[i] = np.sum(lennard_jones_force(r), axis=0)
    vel += forces * dt
    pos += vel * dt
    trajectory.append(pos.copy())

trajectory = np.array(trajectory)
plt.figure(figsize=(8,8))
for i in range(N):
    plt.plot(trajectory[:, i, 0], trajectory[:, i, 1])
plt.title('Lennard-Jones 2D Gas Simulation')
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('lennard_jones.png')
```
