# Laboratorio Especializado: Relatividad Especializado 7

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def relativistic_rocket():
    tau = np.linspace(0, 5, 100)
    g = 1.0
    t = np.sinh(g * tau) / g
    x = (np.cosh(g * tau) - 1) / g
    plt.plot(t, x, label='Rocket Trajectory')
    plt.plot(t, t, 'k--', label='Light cone')
    plt.xlabel("Coordinate Time (t)")
    plt.ylabel("Coordinate Distance (x)")
    plt.legend()
    plt.title("Relativistic Rocket with Constant Proper Acceleration")
    plt.show()

if __name__ == "__main__":
    relativistic_rocket()
```
