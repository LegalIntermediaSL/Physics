# Laboratorio Especializado: Mecanica Clasica Especializado 2

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def standard_map(p, theta, K):
    p_next = p + K * np.sin(theta)
    theta_next = (theta + p_next) % (2 * np.pi)
    return p_next, theta_next

K = 0.9716
N = 1000
iters = 500

plt.figure(figsize=(8, 8))
for _ in range(N):
    theta = np.random.uniform(0, 2 * np.pi)
    p = np.random.uniform(-np.pi, np.pi)
    thetas, ps = [theta], [p]
    for _ in range(iters):
        p, theta = standard_map(p, theta, K)
        p = (p + np.pi) % (2 * np.pi) - np.pi
        thetas.append(theta)
        ps.append(p)
    plt.plot(thetas, ps, '.', markersize=0.5, alpha=0.5)

plt.title(f'Chirikov Standard Map (K={K}) - KAM Theorem Demonstration')
plt.xlabel('$\\theta$')
plt.ylabel('$p$')
plt.xlim(0, 2*np.pi)
plt.ylim(-np.pi, np.pi)
plt.show()
```
