# Laboratorio Especializado: Astrofisica Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def transit_light_curve(t, t0, duration, depth):
    flux = np.ones_like(t)
    transit_mask = np.abs(t - t0) < duration / 2
    flux[transit_mask] = 1.0 - depth
    return flux

t = np.linspace(0, 10, 1000)
flux = transit_light_curve(t, 5.0, 1.0, 0.01) + np.random.normal(0, 0.001, len(t))

plt.figure(figsize=(10,5))
plt.plot(t, flux, '.', color='black', alpha=0.5)
plt.title('Exoplanet Transit Light Curve')
plt.xlabel('Time (days)')
plt.ylabel('Relative Flux')
plt.grid(True)
plt.savefig('transit_lightcurve.png')
```
