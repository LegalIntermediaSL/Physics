# Laboratorio Especializado: Plasmas Especializado 4

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def grad_B_field(x, y, z):
    # Simple magnetic mirror field
    B0 = 1.0
    L = 10.0
    Bz = B0 * (1 + (z/L)**2)
    Br = -B0 * z * np.sqrt(x**2 + y**2) / L**2
    return np.array([Br * x, Br * y, Bz])

if __name__ == "__main__":
    z = np.linspace(-20, 20, 100)
    Bz = [grad_B_field(0, 0, zi)[2] for zi in z]
    plt.plot(z, Bz)
    plt.title('Magnetic Mirror Field')
    plt.xlabel('z')
    plt.ylabel('Bz')
    plt.savefig('magnetic_mirror.png')
```
