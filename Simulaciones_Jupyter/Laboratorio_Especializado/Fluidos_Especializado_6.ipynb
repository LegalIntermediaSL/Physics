# Laboratorio Especializado: Fluidos Especializado 6

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def shallow_water():
    nx, ny = 100, 100
    g = 9.81
    dt = 0.001
    dx = dy = 1.0
    steps = 500

    h = np.ones((nx, ny))
    h[40:60, 40:60] = 5.0  # initial water column
    u = np.zeros((nx, ny))
    v = np.zeros((nx, ny))

    for step in range(steps):
        h_old = h.copy()
        u_old = u.copy()
        v_old = v.copy()

        h[1:-1, 1:-1] = h_old[1:-1, 1:-1] - dt * (
            h_old[1:-1, 1:-1] * (u_old[1:-1, 2:] - u_old[1:-1, :-2]) / (2 * dx) +
            u_old[1:-1, 1:-1] * (h_old[1:-1, 2:] - h_old[1:-1, :-2]) / (2 * dx) +
            h_old[1:-1, 1:-1] * (v_old[2:, 1:-1] - v_old[:-2, 1:-1]) / (2 * dy) +
            v_old[1:-1, 1:-1] * (h_old[2:, 1:-1] - h_old[:-2, 1:-1]) / (2 * dy)
        )

        u[1:-1, 1:-1] = u_old[1:-1, 1:-1] - dt * (
            u_old[1:-1, 1:-1] * (u_old[1:-1, 2:] - u_old[1:-1, :-2]) / (2 * dx) +
            v_old[1:-1, 1:-1] * (u_old[2:, 1:-1] - u_old[:-2, 1:-1]) / (2 * dy) +
            g * (h_old[1:-1, 2:] - h_old[1:-1, :-2]) / (2 * dx)
        )

        v[1:-1, 1:-1] = v_old[1:-1, 1:-1] - dt * (
            u_old[1:-1, 1:-1] * (v_old[1:-1, 2:] - v_old[1:-1, :-2]) / (2 * dx) +
            v_old[1:-1, 1:-1] * (v_old[2:, 1:-1] - v_old[:-2, 1:-1]) / (2 * dy) +
            g * (h_old[2:, 1:-1] - h_old[:-2, 1:-1]) / (2 * dy)
        )

        # Reflective boundaries
        u[:, 0] = 0; u[:, -1] = 0; u[0, :] = 0; u[-1, :] = 0
        v[:, 0] = 0; v[:, -1] = 0; v[0, :] = 0; v[-1, :] = 0
        h[0, :] = h[1, :]; h[-1, :] = h[-2, :]
        h[:, 0] = h[:, 1]; h[:, -1] = h[:, -2]

    plt.figure(figsize=(8, 6))
    plt.imshow(h, cmap='Blues')
    plt.title('Shallow Water Equations: 2D Dam Break')
    plt.colorbar(label='Water height')
    plt.savefig('shallow_water.png')

if __name__ == "__main__":
    shallow_water()
```
