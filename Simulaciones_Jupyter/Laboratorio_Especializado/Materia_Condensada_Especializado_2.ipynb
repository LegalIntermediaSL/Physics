# Laboratorio Especializado: Materia Condensada Especializado 2

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar

def bcs_gap_equation():
    N_0 = 1.0   # Density of states at Fermi level
    V = 0.3     # Attractive interaction potential
    omega_D = 10.0 # Debye frequency
    k_B = 1.0   # Boltzmann constant
    
    def integrand(E, Delta, T):
        return np.tanh(np.sqrt(E**2 + Delta**2) / (2 * k_B * T)) / np.sqrt(E**2 + Delta**2)
        
    def gap_equation(Delta, T):
        # Numeric integration using Simpson's rule over energy
        E = np.linspace(0, omega_D, 1000)
        dE = E[1] - E[0]
        integral = np.sum(integrand(E, Delta, T)) * dE
        return 1.0 - N_0 * V * integral

    temperatures = np.linspace(0.01, 1.8, 50)
    gaps = []

    for T in temperatures:
        try:
            sol = root_scalar(gap_equation, args=(T,), bracket=[1e-5, 2.0], method='brentq')
            gaps.append(sol.root)
        except ValueError:
            gaps.append(0.0) # Gap closes at T_c
            
    plt.figure(figsize=(8, 5))
    plt.plot(temperatures, gaps, 'b-', lw=2, label='BCS Gap $\\Delta(T)$')
    plt.title('BCS Theory: Superconducting Gap vs Temperature')
    plt.xlabel('Temperature $T$')
    plt.ylabel('Gap $\\Delta(T)$')
    plt.grid(True)
    plt.legend()
    plt.savefig('bcs_gap.png')

if __name__ == "__main__":
    bcs_gap_equation()
```
