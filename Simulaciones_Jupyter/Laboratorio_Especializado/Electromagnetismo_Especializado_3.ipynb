# Laboratorio Especializado: Electromagnetismo Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

a, b = 2.0, 1.0 # Waveguide dimensions
m, n = 1, 1 # TE_11 mode
x = np.linspace(0, a, 100)
y = np.linspace(0, b, 100)
X, Y = np.meshgrid(x, y)

Ex = np.cos(m*np.pi*X/a) * np.sin(n*np.pi*Y/b)
Ey = -np.sin(m*np.pi*X/a) * np.cos(n*np.pi*Y/b)
Hz = np.cos(m*np.pi*X/a) * np.cos(n*np.pi*Y/b)

plt.figure(figsize=(10,4))
plt.subplot(121)
plt.streamplot(X, Y, Ex, Ey, color=Hz, cmap='coolwarm')
plt.title(f'TE_{m}{n} Electric Field')
plt.subplot(122)
plt.contourf(X, Y, Hz, 20, cmap='coolwarm')
plt.title(f'TE_{m}{n} Magnetic Field ($H_z$)')
plt.show()
```
