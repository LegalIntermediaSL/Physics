# Laboratorio Especializado: Plasmas Especializado 5

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def debye_shielding():
    r = np.linspace(0.1, 5, 100)
    lambda_D = 1.0
    phi = (1/r) * np.exp(-r/lambda_D)
    plt.plot(r, phi)
    plt.title('Debye Shielding Potential')
    plt.xlabel('r')
    plt.ylabel('Phi')
    plt.savefig('debye.png')

if __name__ == "__main__":
    debye_shielding()
```
