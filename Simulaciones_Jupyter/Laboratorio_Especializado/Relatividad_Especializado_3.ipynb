# Laboratorio Especializado: Relatividad Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def kerr_horizons():
    a = np.linspace(0, 1, 100)
    M = 1.0
    r_plus = M + np.sqrt(M**2 - a**2)
    r_minus = M - np.sqrt(M**2 - a**2)
    plt.plot(a, r_plus, label='Outer Event Horizon $r_+$')
    plt.plot(a, r_minus, label='Inner Cauchy Horizon $r_-$')
    plt.axhline(2*M, color='k', linestyle='--', label='Ergosurface at equator')
    plt.legend()
    plt.title("Kerr Black Hole Horizons")
    plt.xlabel("Spin Parameter a/M")
    plt.ylabel("Radius r/M")
    plt.show()

if __name__ == "__main__":
    kerr_horizons()
```
