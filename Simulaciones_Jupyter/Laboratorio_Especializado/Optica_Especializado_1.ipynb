# Laboratorio Especializado: Optica Especializado 1

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def second_harmonic_generation():
    delta_k = np.linspace(-10, 10, 500) # Phase mismatch
    L = 1.0 # Length of non-linear crystal
    
    # SHG intensity follows a sinc^2 function due to phase matching
    intensity = (np.sin(delta_k * L / 2) / (delta_k * L / 2))**2
    # Fix NaN at delta_k = 0
    intensity[delta_k == 0] = 1.0
    
    plt.figure(figsize=(8, 5))
    plt.plot(delta_k, intensity)
    plt.title("Second Harmonic Generation Intensity vs Phase Mismatch")
    plt.xlabel("Phase Mismatch $\\Delta k$")
    plt.ylabel("SHG Intensity (Normalized)")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    second_harmonic_generation()
```
