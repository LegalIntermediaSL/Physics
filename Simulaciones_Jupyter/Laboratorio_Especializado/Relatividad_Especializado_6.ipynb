# Laboratorio Especializado: Relatividad Especializado 6

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def tov():
    def deriv(y, r):
        P, m = y
        rho = P**(3/5) if P > 0 else 0
        dP = - (rho + P) * (m + 4*np.pi*r**3*P) / (r * (r - 2*m)) if r > 1e-5 else 0
        dm = 4*np.pi*r**2*rho if r > 1e-5 else 0
        return [dP, dm]
    r = np.linspace(1e-5, 10, 1000)
    y0 = [0.1, 0.0]
    sol = odeint(deriv, y0, r)
    plt.plot(r, sol[:, 0], label='Pressure P(r)')
    plt.plot(r, sol[:, 1], label='Mass m(r)')
    plt.legend()
    plt.title("TOV Equation Solution (Polytropic EOS)")
    plt.xlabel("Radius r")
    plt.ylabel("P(r), m(r)")
    plt.show()

if __name__ == "__main__":
    tov()
```
