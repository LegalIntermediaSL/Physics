# Laboratorio Especializado: Termodinamica Especializado 2

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fermi_dirac(E, mu, T):
    return 1.0 / (np.exp((E - mu) / T) + 1.0)

E = np.linspace(0, 10, 500)
mu = 5.0
temperatures = [0.01, 0.5, 1.0, 2.0]

plt.figure(figsize=(8,6))
for T in temperatures:
    plt.plot(E, fermi_dirac(E, mu, T), label=f'T={T}')

plt.axvline(mu, color='k', linestyle='--', label=r'$\mu$ (Fermi Energy)')
plt.title('Fermi-Dirac Distribution')
plt.xlabel('Energy (E)')
plt.ylabel('Probability f(E)')
plt.legend()
plt.grid(True)
plt.savefig('fermi_dirac.png')
```
