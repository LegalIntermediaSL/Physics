Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 8
d = 0.5 # Distance in units of lambda
theta = np.linspace(0, 2*np.pi, 500)
delta = np.pi/4 # Phase shift

AF = np.sin(N/2 * (2*np.pi*d*np.cos(theta) + delta)) / np.sin(1/2 * (2*np.pi*d*np.cos(theta) + delta))
AF = np.abs(AF) / N

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, AF)
ax.set_title(f'Antenna Array Factor (N={N}, $d=\lambda/2$, $\delta=\pi/4$)')
plt.show()
```
