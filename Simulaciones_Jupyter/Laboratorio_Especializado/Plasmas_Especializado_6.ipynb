Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def mhd_alfven():
    x = np.linspace(0, 4*np.pi, 100)
    t = 1.0
    vA = 2.0
    B_pert = np.sin(x - vA * t)
    plt.plot(x, B_pert)
    plt.title('MHD Alfven Wave')
    plt.xlabel('x')
    plt.ylabel('B_pert')
    plt.savefig('alfven.png')

if __name__ == "__main__":
    mhd_alfven()
```
