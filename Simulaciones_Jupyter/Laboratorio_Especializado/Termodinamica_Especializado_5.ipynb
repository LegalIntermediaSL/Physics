# Laboratorio Especializado: Termodinamica Especializado 5

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def planck_law(wav, T):
    h = 6.626e-34
    c = 3e8
    k = 1.38e-23
    a = 2.0 * h * c**2
    b = h * c / (wav * k * T)
    intensity = a / ( (wav**5) * (np.exp(b) - 1.0) )
    return intensity

wavelengths = np.linspace(1e-9, 3e-6, 500)
temperatures = [3000, 4000, 5000, 6000]

plt.figure(figsize=(8,6))
for T in temperatures:
    plt.plot(wavelengths * 1e9, planck_law(wavelengths, T), label=f'{T} K')

plt.title("Planck's Law of Black Body Radiation")
plt.xlabel('Wavelength (nm)')
plt.ylabel('Spectral Radiance (W/sr/m^3)')
plt.legend()
plt.grid(True)
plt.savefig('plancks_law.png')
```
