# Laboratorio Especializado: Termodinamica Especializado 4

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def maxwell_boltzmann(v, m, T, k=1.0):
    return (m / (2 * np.pi * k * T))**1.5 * 4 * np.pi * v**2 * np.exp(-m * v**2 / (2 * k * T))

v = np.linspace(0, 20, 500)
m = 1.0

plt.figure(figsize=(8,6))
for T in [10, 20, 50]:
    plt.plot(v, maxwell_boltzmann(v, m, T), label=f'T={T}')

plt.title('Maxwell-Boltzmann Speed Distribution')
plt.xlabel('Speed (v)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.savefig('maxwell_boltzmann.png')
```
