# Laboratorio Especializado: Materia Condensada Especializado 7

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh

def anderson_localization():
    N = 100 # Chain length
    W = 2.0 # Disorder strength
    t = 1.0 # Hopping
    
    # Construct tight-binding Hamiltonian with random on-site energies
    on_site = np.random.uniform(-W/2, W/2, N)
    
    H = np.zeros((N, N))
    np.fill_diagonal(H, on_site)
    np.fill_diagonal(H[1:], -t)
    np.fill_diagonal(H[:, 1:], -t)
    
    # Periodic boundary conditions
    H[0, -1] = -t
    H[-1, 0] = -t
    
    evals, evecs = eigh(H)
    
    # Calculate Inverse Participation Ratio (IPR)
    ipr = np.sum(np.abs(evecs)**4, axis=0)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Plot IPR vs Energy
    ax1.scatter(evals, ipr, color='red', alpha=0.6)
    ax1.set_xlabel('Energy')
    ax1.set_ylabel('Inverse Participation Ratio (IPR)')
    ax1.set_title('Anderson Localization: IPR')
    ax1.grid(True)
    
    # Plot a localized state
    mid_idx = np.argmin(np.abs(evals))
    ax2.plot(np.abs(evecs[:, mid_idx])**2, marker='o', linestyle='-')
    ax2.set_xlabel('Site index')
    ax2.set_ylabel('Probability density $|\psi|^2$')
    ax2.set_title('A Localized State near $E=0$')
    ax2.grid(True)
    
    plt.tight_layout()
    plt.savefig('anderson_localization.png')

if __name__ == "__main__":
    anderson_localization()
```
