# Laboratorio Especializado: Electromagnetismo Especializado 4

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def lorentz_force(t, Y, q, m, E, B):
    r = Y[:3]
    v = Y[3:]
    a = (q/m) * (E + np.cross(v, B))
    return np.concatenate((v, a))

q, m = 1.0, 1.0
E = np.array([0.0, 0.1, 0.0])
B = np.array([0.0, 0.0, 1.0])
Y0 = [0, 0, 0, 1.0, 0.0, 0.5]

sol = solve_ivp(lorentz_force, [0, 50], Y0, t_eval=np.linspace(0, 50, 1000), args=(q, m, E, B))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol.y[0], sol.y[1], sol.y[2])
ax.set_title('ExB Drift of a Charged Particle')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
