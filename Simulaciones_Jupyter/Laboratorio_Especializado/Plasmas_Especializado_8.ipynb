Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def tokamak_grad_shafranov():
    # Simplified Grad-Shafranov flux surfaces
    R = np.linspace(0.5, 1.5, 50)
    Z = np.linspace(-0.5, 0.5, 50)
    RR, ZZ = np.meshgrid(R, Z)
    psi = (RR**2 - 1)**2 + ZZ**2
    plt.contour(RR, ZZ, psi, levels=15)
    plt.title('Tokamak Flux Surfaces')
    plt.xlabel('R')
    plt.ylabel('Z')
    plt.savefig('tokamak.png')

if __name__ == "__main__":
    tokamak_grad_shafranov()
```
