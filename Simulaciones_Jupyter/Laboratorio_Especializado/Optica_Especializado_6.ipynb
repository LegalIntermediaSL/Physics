# Laboratorio Especializado: Optica Especializado 6

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import fresnel

def fresnel_diffraction():
    # Straight edge Fresnel diffraction
    v = np.linspace(-5, 5, 1000)
    S, C = fresnel(v)
    
    # Intensity of diffraction pattern
    intensity = 0.5 * ((C + 0.5)**2 + (S + 0.5)**2)
    
    plt.figure(figsize=(8, 5))
    plt.plot(v, intensity)
    plt.title("Fresnel Diffraction from a Straight Edge")
    plt.xlabel("Position Parameter v")
    plt.ylabel("Relative Intensity")
    plt.axvline(0, color='r', linestyle='--', label='Geometrical Shadow Edge')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    fresnel_diffraction()
```
