Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def semi_empirical_mass(A, Z):
    av, asurf, ac, asym, ap = 15.67, 17.23, 0.714, 93.15, 11.2
    
    vol = av * A
    surf = -asurf * A**(2/3)
    coul = -ac * Z**2 / A**(1/3)
    sym = -asym * (Z - A/2)**2 / A
    
    if A % 2 == 1:
        pair = 0
    elif Z % 2 == 0:
        pair = ap / np.sqrt(A)
    else:
        pair = -ap / np.sqrt(A)
        
    return vol + surf + coul + sym + pair

if __name__ == "__main__":
    A_vals = np.arange(10, 250, 2)
    BE_A = [semi_empirical_mass(A, A//2.1) / A for A in A_vals]
    
    plt.plot(A_vals, BE_A)
    plt.title('Binding Energy per Nucleon')
    plt.xlabel('A')
    plt.ylabel('BE / A (MeV)')
    plt.savefig('binding_energy.png')
```
