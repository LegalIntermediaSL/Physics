# Laboratorio Especializado: Relatividad Especializado 4

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def frame_dragging():
    r = np.linspace(2.1, 10, 100)
    a = 0.9
    omega = 2 * a / (r**3 + a**2 * r + 2 * a**2)
    plt.plot(r, omega)
    plt.title("Frame Dragging Angular Velocity (Equatorial)")
    plt.xlabel("r/M")
    plt.ylabel(r"$\omega$")
    plt.show()

if __name__ == "__main__":
    frame_dragging()
```
