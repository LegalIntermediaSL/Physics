# Laboratorio Especializado: Optica Especializado 5

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fraunhofer_diffraction():
    theta = np.linspace(-0.1, 0.1, 1000)
    lambda_wave = 500e-9
    k = 2 * np.pi / lambda_wave
    a = 20e-6 # Slit width
    d = 100e-6 # Slit separation
    N = 4 # Number of slits
    
    alpha = k * a * np.sin(theta) / 2
    beta = k * d * np.sin(theta) / 2
    
    # Avoid division by zero
    alpha[alpha == 0] = 1e-10
    beta[beta == 0] = 1e-10
    
    I_single = (np.sin(alpha) / alpha)**2
    I_multi = (np.sin(N * beta) / np.sin(beta))**2
    I_total = I_single * I_multi
    
    plt.figure(figsize=(8, 5))
    plt.plot(theta, I_total)
    plt.title(f"Fraunhofer Diffraction ({N} Slits)")
    plt.xlabel("Angle (rad)")
    plt.ylabel("Intensity")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    fraunhofer_diffraction()
```
