Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def accretion_disk_temp(R, M, M_dot):
    G = 6.674e-11
    sigma = 5.67e-8
    R_in = 6 * G * M / (3e8)**2
    T = (3 * G * M * M_dot / (8 * np.pi * sigma * R**3))**0.25 * (1 - np.sqrt(R_in / R))**0.25
    return np.where(R > R_in, T, 0)

M_bh = 10 * 1.989e30
M_dot = 1e-8 * 1.989e30 / (365.25 * 24 * 3600)
R = np.logspace(4, 7, 500)

T_disk = accretion_disk_temp(R, M_bh, M_dot)

plt.figure(figsize=(8,6))
plt.loglog(R, T_disk, lw=2, color='purple')
plt.title('Black Hole Accretion Disk Temperature Profile')
plt.xlabel('Radius (m)')
plt.ylabel('Temperature (K)')
plt.grid(True)
plt.savefig('accretion_disk.png')
```
