# Laboratorio Especializado: Mecanica Clasica Especializado 6

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def runge_kutta(f, y0, t, args=()):
    dt = t[1] - t[0]
    y = np.zeros((len(t), len(y0)))
    y[0] = y0
    for i in range(len(t) - 1):
        k1 = f(y[i], t[i], *args)
        k2 = f(y[i] + dt*k1/2, t[i] + dt/2, *args)
        k3 = f(y[i] + dt*k2/2, t[i] + dt/2, *args)
        k4 = f(y[i] + dt*k3, t[i] + dt, *args)
        y[i+1] = y[i] + dt*(k1 + 2*k2 + 2*k3 + k4)/6
        y[i+1, 0] = (y[i+1, 0] + np.pi) % (2*np.pi) - np.pi
    return y

def pendulum(y, t, q, F, omega_d):
    theta, omega = y
    return np.array([omega, -q*omega - np.sin(theta) + F*np.cos(omega_d*t)])

q = 0.5
omega_d = 2.0/3.0
F_vals = np.linspace(1.35, 1.5, 200)
t = np.linspace(0, 1000, 5000)
theta_points = []
f_points = []

for F in F_vals:
    sol = runge_kutta(pendulum, [0.2, 0], t, args=(q, F, omega_d))
    theta_points.extend(sol[4000::50, 0])
    f_points.extend([F]*len(sol[4000::50, 0]))

plt.figure(figsize=(10, 6))
plt.plot(f_points, theta_points, ',k', alpha=0.5)
plt.title('Bifurcation Diagram of Damped Driven Pendulum')
plt.xlabel('Drive Amplitude (F)')
plt.ylabel('$\\theta$')
plt.show()
```
