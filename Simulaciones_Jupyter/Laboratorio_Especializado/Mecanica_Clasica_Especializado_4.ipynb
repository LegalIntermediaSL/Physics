Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def fpu(y, t, N, alpha):
    q = y[:N]
    p = y[N:]
    dqdt = p
    dpdt = np.zeros(N)
    for i in range(1, N-1):
        dpdt[i] = (q[i+1] - 2*q[i] + q[i-1]) + alpha * ((q[i+1]-q[i])**2 - (q[i]-q[i-1])**2)
    return np.concatenate((dqdt, dpdt))

N = 32
alpha = 0.25
q0 = np.sin(np.pi * np.arange(N) / (N - 1))
p0 = np.zeros(N)
y0 = np.concatenate((q0, p0))
t = np.linspace(0, 1000, 2000)

sol = odeint(fpu, y0, t, args=(N, alpha))
energies = np.zeros((len(t), N//2))
for k in range(1, 5):
    mode = np.sin(k * np.pi * np.arange(N) / (N - 1))
    energies[:, k-1] = 0.5 * (np.dot(sol[:, :N], mode)**2 + np.dot(sol[:, N:], mode)**2)

plt.plot(t, energies[:, 0], label='Mode 1')
plt.plot(t, energies[:, 1], label='Mode 2')
plt.plot(t, energies[:, 2], label='Mode 3')
plt.plot(t, energies[:, 3], label='Mode 4')
plt.title('Fermi-Pasta-Ulam-Tsingou Recurrence')
plt.xlabel('Time')
plt.ylabel('Energy')
plt.legend()
plt.show()
```
