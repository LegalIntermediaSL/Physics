# Laboratorio Especializado: Plasmas Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def pic_step(pos, vel, dt, L):
    pos = (pos + vel * dt) % L
    return pos, vel

if __name__ == "__main__":
    Np = 1000
    L = 10.0
    pos = np.random.rand(Np) * L
    vel = np.random.randn(Np)
    # two stream
    vel[:Np//2] += 5.0
    vel[Np//2:] -= 5.0
    
    for _ in range(10):
        pos, vel = pic_step(pos, vel, 0.1, L)
        
    plt.hist(vel, bins=30)
    plt.title('Two Stream Instability - Velocity Distribution')
    plt.savefig('two_stream.png')
```
