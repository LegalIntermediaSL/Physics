# Laboratorio Especializado: Electromagnetismo Especializado 1

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 200)
phi = np.linspace(0, np.pi, 200)
THETA, PHI = np.meshgrid(theta, phi)
beta = 0.5

dP_dOmega = np.sin(PHI)**2 / (1 - beta*np.cos(PHI))**5

X = dP_dOmega * np.sin(PHI) * np.cos(THETA)
Y = dP_dOmega * np.sin(PHI) * np.sin(THETA)
Z = dP_dOmega * np.cos(PHI)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='inferno')
plt.title(f'Relativistic Larmor Radiation Pattern ($\\beta$={beta})')
plt.show()
```
