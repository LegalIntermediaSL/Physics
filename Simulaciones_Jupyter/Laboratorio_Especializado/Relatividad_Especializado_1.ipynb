# Laboratorio Especializado: Relatividad Especializado 1

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def geodesics():
    def deriv(y, phi):
        u, u_prime = y
        return [u_prime, 1.5 * u**2 - u]
    phi = np.linspace(0, 20*np.pi, 1000)
    y0 = [0.1, 0.01]
    sol = odeint(deriv, y0, phi)
    u = sol[:, 0]
    r = 1 / u
    x = r * np.cos(phi)
    y_coord = r * np.sin(phi)
    plt.plot(x, y_coord)
    plt.title("Photon Geodesic in Schwarzschild Spacetime")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.axis('equal')
    plt.show()

if __name__ == "__main__":
    geodesics()
```
