```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def stark_effect():
    electric_field = np.linspace(0, 100, 500)
    # Simplified Stark effect for n=2 states in Hydrogen
    # Degenerate perturbation theory splits the states linearly
    E_0 = -3.4 # eV for n=2
    k = 0.05 # proportional constant
    
    E_plus = E_0 + k * electric_field
    E_minus = E_0 - k * electric_field
    E_zero = np.full_like(electric_field, E_0)
    
    plt.figure(figsize=(8, 5))
    plt.plot(electric_field, E_plus, label='m=0, perturbed (+)')
    plt.plot(electric_field, E_minus, label='m=0, perturbed (-)')
    plt.plot(electric_field, E_zero, label='m=+-1, unperturbed', linestyle='--')
    plt.title("Linear Stark Effect in Hydrogen (n=2)")
    plt.xlabel("Electric Field Strength (kV/cm)")
    plt.ylabel("Energy (eV)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    stark_effect()
```
