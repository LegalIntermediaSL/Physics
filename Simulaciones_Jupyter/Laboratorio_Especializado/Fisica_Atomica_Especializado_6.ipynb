# Laboratorio Especializado: Fisica Atomica Especializado 6

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fine_structure():
    n = np.arange(1, 6)
    alpha = 1 / 137.036
    E_n = -13.6 / n**2
    
    # Fine structure correction (j = l + 1/2 for simplicity here)
    def fine_correction(n, j):
        return E_n * (alpha**2 / n**2) * (n / (j + 0.5) - 0.75)
    
    dE = fine_correction(n, 0.5)
    
    plt.figure(figsize=(8, 5))
    plt.stem(n, np.abs(dE) * 1e4, basefmt=" ") # Multiply by 10^4 for visibility
    plt.title("Fine Structure Energy Shift (j=1/2) vs Principal Quantum Number n")
    plt.xlabel("Principal Quantum Number (n)")
    plt.ylabel("Absolute Energy Shift x $10^4$ (eV)")
    plt.xticks(n)
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    fine_structure()
```
