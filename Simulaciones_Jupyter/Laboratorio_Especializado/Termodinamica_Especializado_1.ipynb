# Laboratorio Especializado: Termodinamica Especializado 1

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def density_of_states(E):
    return np.sqrt(E)

def bose_einstein(E, T, mu=0):
    return 1.0 / (np.exp((E - mu) / T) - 1.0)

def energy_integrand(E, T):
    if E == 0: return 0
    return E * density_of_states(E) * bose_einstein(E, T)

Temperatures = np.linspace(0.1, 5.0, 50)
U = []
for T in Temperatures:
    res, _ = quad(energy_integrand, 1e-5, 20*T, args=(T,))
    U.append(res)
U = np.array(U)
Cv = np.gradient(U, Temperatures)

plt.figure(figsize=(8,6))
plt.plot(Temperatures, Cv, lw=2, color='blue')
plt.title('Heat Capacity of an Ideal Bose Gas')
plt.xlabel('Temperature (T)')
plt.ylabel('Heat Capacity (Cv)')
plt.grid(True)
plt.savefig('bose_gas_cv.png')
```
