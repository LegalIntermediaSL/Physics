# Laboratorio Especializado: Astrofisica Especializado 8

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def star_orbit(M, e, a, num_points=1000):
    theta = np.linspace(0, 2*np.pi, num_points)
    r = a * (1 - e**2) / (1 + e * np.cos(theta))
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return x, y

a = 1000  # AU
e = 0.88  # Similar to S2 orbit

x, y = star_orbit(4e6, e, a)

plt.figure(figsize=(8,8))
plt.plot(x, y, label='S2-like Orbit')
plt.plot(0, 0, 'k*', markersize=15, label='Sgr A* (SMBH)')
plt.title('Orbit of a Star around a Supermassive Black Hole')
plt.xlabel('Distance (AU)')
plt.ylabel('Distance (AU)')
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.savefig('sgr_A_orbit.png')
```
