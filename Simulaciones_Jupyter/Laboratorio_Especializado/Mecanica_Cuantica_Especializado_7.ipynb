# Laboratorio Especializado: Mecanica Cuantica Especializado 7

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rabi():
    t = np.linspace(0, 20, 1000)
    Omega = 1.0
    delta = 0.2
    Omega_R = np.sqrt(Omega**2 + delta**2)
    P_up = (Omega**2 / Omega_R**2) * np.sin(Omega_R * t / 2)**2
    plt.plot(t, P_up)
    plt.title("Rabi Oscillations with Detuning")
    plt.xlabel("Time t")
    plt.ylabel("Spin-up Probability")
    plt.show()

if __name__ == "__main__":
    rabi()
```
