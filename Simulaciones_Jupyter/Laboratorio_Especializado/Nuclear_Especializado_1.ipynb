# Laboratorio Especializado: Nuclear Especializado 1

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def rutherford_force(t, y):
    r = y[:2]
    v = y[2:]
    dist = np.linalg.norm(r)
    # F = k / r^2
    a = r / dist**3
    return np.concatenate((v, a))

if __name__ == "__main__":
    y0 = np.array([-10.0, 1.0, 2.0, 0.0]) # b=1.0 impact parameter
    sol = solve_ivp(rutherford_force, [0, 10], y0, max_step=0.01)
    plt.plot(sol.y[0], sol.y[1])
    plt.scatter([0], [0], color='red') # Nucleus
    plt.title('Rutherford Scattering')
    plt.savefig('rutherford.png')
```
