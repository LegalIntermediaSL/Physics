# Laboratorio Especializado: Fluidos Especializado 2

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft, fftfreq

def simulate_kdv():
    N = 256
    L = 50.0
    dx = L / N
    x = np.arange(-L/2, L/2, dx)
    k = 2 * np.pi * fftfreq(N, d=dx)
    
    # Two solitons initial condition
    u = 3 * 2**2 / np.cosh(np.sqrt(2)*(x+10)/2)**2 + 3 * 1**2 / np.cosh(np.sqrt(1)*(x+2)/2)**2
    
    dt = 0.001
    t_max = 5.0
    steps = int(t_max / dt)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, u, label='Initial state', color='blue')
    
    u_hat = fft(u)
    
    for i in range(steps):
        # Pseudo-spectral method using integrating factor for linear part
        u_hat = u_hat * np.exp(1j * k**3 * dt / 2)
        u = np.real(ifft(u_hat))
        u_hat = u_hat - dt * 1j * k * fft(3 * u**2)
        u_hat = u_hat * np.exp(1j * k**3 * dt / 2)
        
    u_final = np.real(ifft(u_hat))
    
    plt.plot(x, u_final, label=f'State at t={t_max}', color='red')
    plt.title("KdV Equation: Soliton Interaction")
    plt.xlabel('x')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid(True)
    plt.savefig('kdv_solitons.png')

if __name__ == "__main__":
    simulate_kdv()
```
