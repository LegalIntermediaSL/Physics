```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def optical_bloch_steady_state():
    Delta = np.linspace(-5, 5, 400) # Detuning
    Gamma = 1.0 # Decay rate
    Omega_values = [0.1, 1.0, 5.0] # Rabi frequencies
    
    plt.figure(figsize=(8, 5))
    
    for Omega in Omega_values:
        # Steady state excited population
        rho_ee = (Omega**2 / 4) / (Delta**2 + (Gamma**2 / 4) + (Omega**2 / 2))
        plt.plot(Delta, rho_ee, label=f'$\\Omega$ = {Omega}')
        
    plt.title("Steady State Excited Population vs Detuning (Power Broadening)")
    plt.xlabel("Detuning $\\Delta$ (in units of $\\Gamma$)")
    plt.ylabel("Population $\\rho_{ee}$")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    optical_bloch_steady_state()
```
