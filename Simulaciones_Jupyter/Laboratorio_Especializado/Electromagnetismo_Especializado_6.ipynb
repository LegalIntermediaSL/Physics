# Laboratorio Especializado: Electromagnetismo Especializado 6

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

k = 2 * np.pi
r = np.linspace(0.1, 5, 200)
theta = np.linspace(0, 2*np.pi, 200)
R, THETA = np.meshgrid(r, theta)

E_theta = np.sin(THETA) * (1/R - 1j/(k*R**2) - 1/(k**2 * R**3)) * np.exp(1j*k*R)
S_r = 0.5 * np.real(E_theta * np.conj(E_theta))

plt.figure(figsize=(8,6))
ax = plt.subplot(111, polar=True)
c = ax.pcolormesh(THETA, R, S_r, shading='auto', cmap='plasma')
plt.colorbar(c)
plt.title('Hertzian Dipole Far-Field Radiation Density')
plt.show()
```
