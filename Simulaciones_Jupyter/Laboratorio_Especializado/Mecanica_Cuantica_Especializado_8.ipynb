Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def variational_helium():
    Z_eff = np.linspace(1, 3, 100)
    Z = 2
    E = Z_eff**2 - 2*Z*Z_eff + (5/8)*Z_eff
    plt.plot(Z_eff, E)
    plt.axvline(Z - 5/16, color='r', linestyle='--', label='Minimum energy Z_eff')
    plt.title("Variational Energy for Helium Ground State")
    plt.xlabel("Effective Charge $Z_{eff}$")
    plt.ylabel("Energy (Hartrees)")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    variational_helium()
```
