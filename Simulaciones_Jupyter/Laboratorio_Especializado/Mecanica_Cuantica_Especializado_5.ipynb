```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def scattering():
    E = np.linspace(0.1, 5, 500)
    V0 = 2.0
    a = 1.0
    kappa = np.sqrt(np.complex128(V0 - E))
    T = 1 / (1 + (V0**2 * np.sinh(kappa*a)**2) / (4 * E * (V0 - E)))
    plt.plot(E, np.abs(T))
    plt.axvline(V0, color='r', linestyle='--', label='$V_0$')
    plt.title("1D Transmission Coefficient for Rectangular Barrier")
    plt.xlabel("Energy E")
    plt.ylabel("Transmission T")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    scattering()
```
