# Laboratorio Especializado: Materia Condensada Especializado 4

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def graphene_bands():
    t = -2.7 # Hopping parameter in eV
    a = 1.42 # Carbon-carbon distance
    
    # Reciprocal lattice vectors
    K = np.array([4*np.pi/(3*np.sqrt(3)*a), 0])
    Gamma = np.array([0, 0])
    M = np.array([np.pi/(np.sqrt(3)*a), np.pi/(3*a)])
    
    def energy_dispersion(kx, ky):
        f = np.exp(1j * kx * a) + 2 * np.exp(-1j * kx * a / 2) * np.cos(np.sqrt(3) * ky * a / 2)
        return np.sqrt(np.abs(f)**2)
        
    path_G_K = np.linspace(Gamma, K, 100)
    path_K_M = np.linspace(K, M, 50)
    path_M_G = np.linspace(M, Gamma, 100)
    
    k_path = np.vstack([path_G_K, path_K_M, path_M_G])
    
    E = []
    for kx, ky in k_path:
        E.append(t * energy_dispersion(kx, ky))
        
    E = np.array(E)
    
    plt.figure(figsize=(8, 6))
    plt.plot(E, 'b-', lw=2)
    plt.plot(-E, 'r-', lw=2)
    plt.axhline(0, color='k', linestyle='--')
    plt.xticks([0, 100, 150, 250], ['$\\Gamma$', '$K$', '$M$', '$\\Gamma$'])
    plt.ylabel('Energy (eV)')
    plt.title('Graphene Tight-Binding Band Structure')
    plt.grid(True)
    plt.savefig('graphene_bands.png')

if __name__ == "__main__":
    graphene_bands()
```
