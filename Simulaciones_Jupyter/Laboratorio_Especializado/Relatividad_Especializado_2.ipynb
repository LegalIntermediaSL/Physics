# Laboratorio Especializado: Relatividad Especializado 2

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def penrose():
    u = np.linspace(-1, 1, 100)
    v = np.linspace(-1, 1, 100)
    U, V = np.meshgrid(u, v)
    T = U + V
    R = U - V
    plt.contour(R, T, U, 20, colors='b', alpha=0.5)
    plt.contour(R, T, V, 20, colors='r', alpha=0.5)
    plt.title("Minkowski Penrose Diagram Lines of Constant U, V")
    plt.xlabel("Compactified Spatial Coordinate R")
    plt.ylabel("Compactified Temporal Coordinate T")
    plt.axis('equal')
    plt.show()

if __name__ == "__main__":
    penrose()
```
