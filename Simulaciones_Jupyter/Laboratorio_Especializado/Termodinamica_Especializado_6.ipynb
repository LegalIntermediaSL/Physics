# Laboratorio Especializado: Termodinamica Especializado 6

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def vdw_pressure(V, T, a=3.59, b=0.0427, R=0.08206):
    return (R * T) / (V - b) - a / (V**2)

V = np.linspace(0.06, 0.5, 500)
temperatures = [273, 300, 350, 400]

plt.figure(figsize=(8,6))
for T in temperatures:
    P = vdw_pressure(V, T)
    plt.plot(V, P, label=f'T={T} K')

plt.ylim(0, 150)
plt.title('Van der Waals Isotherms')
plt.xlabel('Molar Volume (L/mol)')
plt.ylabel('Pressure (atm)')
plt.legend()
plt.grid(True)
plt.savefig('vdw_isotherms.png')
```
