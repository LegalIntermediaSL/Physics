# Laboratorio Especializado: Mecanica Cuantica Especializado 1

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def path_integral_mc():
    N = 100
    beta = 1.0
    dtau = beta / N
    x = np.zeros(N)
    steps = 10000
    for _ in range(steps):
        i = np.random.randint(N)
        dx = np.random.normal(0, 0.5)
        x_new = x.copy()
        x_new[i] += dx
        S_old = np.sum((x[1:] - x[:-1])**2) / (2*dtau) + 0.5 * dtau * np.sum(x**2)
        S_new = np.sum((x_new[1:] - x_new[:-1])**2) / (2*dtau) + 0.5 * dtau * np.sum(x_new**2)
        if np.exp(-(S_new - S_old)) > np.random.rand():
            x = x_new
    plt.plot(x)
    plt.title("Path Integral Monte Carlo for Harmonic Oscillator")
    plt.xlabel("Imaginary Time Slice")
    plt.ylabel("Position")
    plt.show()

if __name__ == "__main__":
    path_integral_mc()
```
