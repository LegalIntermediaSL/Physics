# Laboratorio Especializado: Mecanica Clasica Especializado 1

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def derivs(t, state, L1, L2, m1, m2, g):
    theta1, z1, theta2, z2 = state
    c, s = np.cos(theta1-theta2), np.sin(theta1-theta2)
    theta1dot = z1
    z1dot = (m2*g*np.sin(theta2)*c - m2*s*(L1*z1**2*c + L2*z2**2) - (m1+m2)*g*np.sin(theta1)) / L1 / (m1 + m2*s**2)
    theta2dot = z2
    z2dot = ((m1+m2)*(L1*z1**2*s - g*np.sin(theta2) + g*np.sin(theta1)*c) + m2*L2*z2**2*s*c) / L2 / (m1 + m2*s**2)
    return [theta1dot, z1dot, theta2dot, z2dot]

L1, L2, m1, m2, g = 1.0, 1.0, 1.0, 1.0, 9.81
state0 = [np.pi/2, 0.0, np.pi/2, 0.0]
t = np.arange(0, 20, 0.05)
sol = solve_ivp(derivs, [0, 20], state0, t_eval=t, args=(L1, L2, m1, m2, g), method='Radau')

x1 = L1 * np.sin(sol.y[0])
y1 = -L1 * np.cos(sol.y[0])
x2 = x1 + L2 * np.sin(sol.y[2])
y2 = y1 - L2 * np.cos(sol.y[2])

fig, ax = plt.subplots()
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)
line, = ax.plot([], [], 'o-', lw=2)
trace, = ax.plot([], [], 'r-', lw=1, alpha=0.5)

def animate(i):
    line.set_data([0, x1[i], x2[i]], [0, y1[i], y2[i]])
    trace.set_data(x2[:i], y2[:i])
    return line, trace

ani = animation.FuncAnimation(fig, animate, frames=len(t), interval=50, blit=True)
plt.title('Double Pendulum Simulation')
plt.show()
```
