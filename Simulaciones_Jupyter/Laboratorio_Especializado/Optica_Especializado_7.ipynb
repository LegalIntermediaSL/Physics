# Laboratorio Especializado: Optica Especializado 7

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gaussian_beam():
    z = np.linspace(-5, 5, 500)
    w0 = 1.0 # Beam waist
    lambda_wave = 0.1
    z_R = np.pi * w0**2 / lambda_wave # Rayleigh length
    
    w_z = w0 * np.sqrt(1 + (z / z_R)**2)
    
    plt.figure(figsize=(8, 5))
    plt.plot(z, w_z, 'b', label='Beam Radius w(z)')
    plt.plot(z, -w_z, 'b')
    plt.fill_between(z, w_z, -w_z, color='b', alpha=0.2)
    plt.title("Gaussian Beam Propagation")
    plt.xlabel("Propagation Distance z")
    plt.ylabel("Transverse Distance")
    plt.axvline(0, color='k', linestyle='--', label='Beam Waist Location')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    gaussian_beam()
```
