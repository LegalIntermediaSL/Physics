Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def wkb_approximation():
    x = np.linspace(-5, 5, 1000)
    V = 0.5 * x**2
    E = 1.0
    p = np.sqrt(np.maximum(2*(E - V), 0))
    plt.plot(x, V, label='Potential $V(x)$')
    plt.plot(x, p, label='Momentum $p(x)$')
    plt.axhline(E, color='r', linestyle='--', label='Energy $E$')
    plt.legend()
    plt.title("WKB Approximation: Momentum in Harmonic Potential")
    plt.xlabel("Position x")
    plt.show()

if __name__ == "__main__":
    wkb_approximation()
```
