# Laboratorio Especializado: Materia Condensada Especializado 8

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse as sp
from scipy.sparse.linalg import eigsh

def heisenberg_chain():
    L = 12 # Number of spins
    J = 1.0 # Antiferromagnetic coupling
    
    # Pauli matrices
    sz = sp.csr_matrix([[1, 0], [0, -1]])
    sp_op = sp.csr_matrix([[0, 1], [0, 0]]) # S+
    sm_op = sp.csr_matrix([[0, 0], [1, 0]]) # S-
    eye = sp.eye(2)
    
    def tensor_prod(op, i):
        res = 1
        for j in range(L):
            if j == i:
                res = sp.kron(res, op) if j > 0 else op
            else:
                res = sp.kron(res, eye) if j > 0 else eye
        return res
        
    H = sp.csr_matrix((2**L, 2**L))
    
    # Construct Hamiltonian H = J \sum \vec{S}_i \cdot \vec{S}_{i+1}
    for i in range(L - 1):
        sz_i = tensor_prod(sz, i)
        sz_j = tensor_prod(sz, i+1)
        sp_i = tensor_prod(sp_op, i)
        sm_j = tensor_prod(sm_op, i+1)
        sm_i = tensor_prod(sm_op, i)
        sp_j = tensor_prod(sp_op, i+1)
        
        H += J * (sz_i @ sz_j + 0.5 * (sp_i @ sm_j + sm_i @ sp_j))
        
    # Find ground state and first excited state
    evals, evecs = eigsh(H, k=3, which='SA')
    
    print(f"Ground state energy: {evals[0]:.4f}")
    print(f"First excited state energy: {evals[1]:.4f}")
    print(f"Gap: {evals[1] - evals[0]:.4f}")
    
    plt.figure(figsize=(6, 4))
    plt.plot(evals, 'ro-', markersize=8)
    plt.title('Low-lying Energy Levels of Heisenberg Chain (L=12)')
    plt.xlabel('State index')
    plt.ylabel('Energy')
    plt.grid(True)
    plt.savefig('heisenberg_spectrum.png')

if __name__ == "__main__":
    heisenberg_chain()
```
