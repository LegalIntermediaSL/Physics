# Laboratorio Especializado: Astrofisica Especializado 7

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def nfw_profile(r, rho0, Rs):
    return rho0 / ((r / Rs) * (1 + r / Rs)**2)

r = np.logspace(0, 3, 500)
rho0 = 1e7
Rs_values = [10, 50, 100]

plt.figure(figsize=(8,6))
for Rs in Rs_values:
    plt.loglog(r, nfw_profile(r, rho0, Rs), label=f'Rs = {Rs} kpc')

plt.title('Dark Matter Halo Density Profiles (NFW)')
plt.xlabel('Radius (kpc)')
plt.ylabel(r'Density $\rho(r)$ ($M_\odot / kpc^3$)')
plt.legend()
plt.grid(True)
plt.savefig('nfw_profile.png')
```
