# Laboratorio Especializado: Astrofisica Especializado 6

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def friedmann(a, t, Omega_m, Omega_Lambda):
    H0 = 70.0
    dadt = H0 * np.sqrt(Omega_m / a + Omega_Lambda * a**2)
    return dadt

t = np.linspace(0, 14, 500)
a0 = 1e-3

universes = [(0.3, 0.7, 'Lambda-CDM'), (1.0, 0.0, 'Matter Only'), (0.0, 1.0, 'Dark Energy Only')]
plt.figure(figsize=(8,6))

for Om, Ol, label in universes:
    a = odeint(friedmann, a0, t, args=(Om, Ol))
    plt.plot(t, a, label=label)

plt.title('Cosmological Expansion (Scale Factor a(t))')
plt.xlabel('Time (Gyr)')
plt.ylabel('Scale Factor a')
plt.legend()
plt.grid(True)
plt.savefig('cosmic_expansion.png')
```
