# Laboratorio Especializado: Optica Especializado 4

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fabry_perot():
    phase = np.linspace(0, 4*np.pi, 1000)
    reflectivities = [0.5, 0.8, 0.95]
    
    plt.figure(figsize=(8, 5))
    
    for R in reflectivities:
        F = 4 * R / (1 - R)**2 # Coefficient of finesse
        transmission = 1 / (1 + F * np.sin(phase/2)**2)
        plt.plot(phase, transmission, label=f'R = {R}')
        
    plt.title("Fabry-Perot Interferometer Transmission")
    plt.xlabel("Phase difference $\\delta$ (rad)")
    plt.ylabel("Transmission")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    fabry_perot()
```
