Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def rabi_oscillations():
    # Optical Bloch Equations for a 2-level system
    def bloch_equations(state, t, Omega, Delta, Gamma):
        rho_ee, rho_gg, u, v = state
        drho_ee = -Gamma * rho_ee - 0.5 * Omega * v
        drho_gg = Gamma * rho_ee + 0.5 * Omega * v
        du = -Delta * v - 0.5 * Gamma * u
        dv = Delta * u - 0.5 * Gamma * v + Omega * (rho_ee - rho_gg)
        return [drho_ee, drho_gg, du, dv]

    t = np.linspace(0, 10, 1000)
    Omega = 5.0  # Rabi frequency
    Delta = 0.0  # Detuning
    Gamma = 0.5  # Spontaneous emission rate
    
    initial_state = [0.0, 1.0, 0.0, 0.0] # Starts in ground state
    
    result = odeint(bloch_equations, initial_state, t, args=(Omega, Delta, Gamma))
    rho_ee = result[:, 0]
    
    plt.figure(figsize=(8, 5))
    plt.plot(t, rho_ee, label='Excited State Population')
    plt.title("Rabi Oscillations with Damping")
    plt.xlabel("Time")
    plt.ylabel("Population $\\rho_{ee}$")
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    rabi_oscillations()
```
