# [CLF-09] Fluidos Extremos: Rompimiento de Onda de Choque

```python
import numpy as np
import matplotlib.pyplot as plt

# Acústica No-Lineal y Fluidos Extremos
# Integrador PDE de la Ecuación Viscosa de Burgers 1D: du/dt + u(du/dx) = nu(d2u/dx2)
# Modela cómo una onda acústica suave rompe aerodinámicamente formando una Onda de Choque.

N_x = 250
L = 2.0 * np.pi
dx = L / (N_x - 1)
x = np.linspace(0, L, N_x)

nu = 0.02 # Viscosidad Cinemática (Evita que el gradiente se haga infinito matemático)
dt = 0.005
N_t = 300

# Onda acústica inicial: un pulso senoidal suave
u = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, u, 'b--', lw=2, label="Onda Acústica Inicial $t=0$")

u_new = np.zeros(N_x)

# Integración Numérica en Derivadas Parciales (Forward-Time Central-Space + Upwind)
for n in range(N_t):
    for i in range(1, N_x - 1):
        # Término Difusivo (Viscosidad)
        diffusion = nu * (u[i+1] - 2*u[i] + u[i-1]) / dx**2
        
        # Término Advectivo No-Lineal (u * du/dx). Usamos Upwind para evitar inestabilidades
        if u[i] > 0: advection = u[i] * (u[i] - u[i-1]) / dx
        else:        advection = u[i] * (u[i+1] - u[i]) / dx
            
        u_new[i] = u[i] - dt * advection + dt * diffusion
    
    # Condiciones de frontera periódicas
    u_new[0] = u_new[-2]
    u_new[-1] = u_new[1]
    
    u = u_new.copy()
    
    if n == int(N_t / 3):
        plt.plot(x, u, 'g-', lw=2, alpha=0.5, label="Onda distorsionándose")

plt.plot(x, u, 'r-', lw=3, label=f"Onda de Choque Final ($t={N_t*dt}$)")

plt.title("Rompimiento Acústico: Ecuación No-Lineal de Burgers")
plt.xlabel("Posición Espacial $x$")
plt.ylabel("Amplitud de la Onda Acústica (Presión/Velocidad)")
plt.legend()
plt.grid(True)
plt.show()
# FÍSICA: Al propagarse, las partes de mayor amplitud viajan más rápido debido 
# al término no lineal. La cresta alcanza al valle y el gradiente acústico colapsa, 
# formando un 'muro' discontinuo casi vertical: es una Onda de Choque (Shockwave).

```
