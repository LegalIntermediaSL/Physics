# Algoritmo Cuántico de Grover

Amplificación de amplitud teórica para la búsqueda no estructurada.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 100 # Base states
prob = np.ones(N) / N
prob[42] = 0.5 # Amplificado

plt.figure(figsize=(8,4))
plt.bar(range(N), prob, color='green')
plt.title('Amplificación de Amplitud (Grover)')
plt.xlabel('Estado Base |x>')
plt.ylabel('Probabilidad')
plt.show()
```
