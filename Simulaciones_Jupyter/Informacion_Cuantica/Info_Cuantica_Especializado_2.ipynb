# Laboratorio Especializado: Info Cuantica Especializado 2

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import math

def shor_period_finding(a, N):
    # Classical simulation of quantum period finding
    for r in range(1, N):
        if pow(a, r, N) == 1:
            return r
    return None

if __name__ == "__main__":
    N = 15
    a = 7
    r = shor_period_finding(a, N)
    print(f"Period of {a} mod {N} is {r}")
    if r % 2 == 0:
        p = math.gcd(a**(r//2) - 1, N)
        q = math.gcd(a**(r//2) + 1, N)
        print(f"Factors: {p}, {q}")
```
