# Compendio Subtemático: Info Cuantica Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # 2-qubit Grover search for state |11>
    states = ['|00>', '|01>', '|10>', '|11>']
    # After 1 iteration, prob of |11> is 1.0 in ideal case
    probs = [0, 0, 0, 1]
    
    plt.bar(states, probs, color='magenta')
    plt.title("Grover's Search Algorithm (Target: |11>)")
    plt.ylabel('Probability Amplitude Squared')
    plt.show()

if __name__ == '__main__':
    main()
```
