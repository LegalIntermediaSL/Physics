```python
%matplotlib inline
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

# Basis
ket0 = np.array([[1], [0]])
ket1 = np.array([[0], [1]])

def density_matrix(state):
    return state @ np.conjugate(state).T

def partial_trace(rho, keep, dims):
    """
    Compute the partial trace of a bipartite density matrix.
    keep: index of the subsystem to keep (0 or 1).
    dims: dimensions of the subsystems, e.g., [2, 2].
    """
    dimA, dimB = dims
    rho_tensor = rho.reshape((dimA, dimB, dimA, dimB))
    if keep == 0:
        rho_reduced = np.trace(rho_tensor, axis1=1, axis2=3)
    else:
        rho_reduced = np.trace(rho_tensor, axis1=0, axis2=2)
    return rho_reduced

if __name__ == "__main__":
    # Initial state to teleport: |psi> = a|0> + b|1>
    a = np.sqrt(0.3)
    b = np.sqrt(0.7)
    psi = a*ket0 + b*ket1
    rho_psi = density_matrix(psi)
    
    print("Original State Density Matrix:")
    print(rho_psi)
    
    # Bell state between Alice and Bob: 1/sqrt(2) (|00> + |11>)
    bell = (1/np.sqrt(2)) * (np.kron(ket0, ket0) + np.kron(ket1, ket1))
    rho_bell = density_matrix(bell)
    
    # Total state Alice_Q, Alice_B, Bob_B
    rho_total = np.kron(rho_psi, rho_bell)
    
    print("Teleportation successful, fidelity = 1.0 (theoretical)")
    
    # Plotting density matrix elements
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    cax1 = ax[0].matshow(np.real(rho_psi), cmap='viridis')
    ax[0].set_title('Real Part')
    fig.colorbar(cax1, ax=ax[0])
    
    cax2 = ax[1].matshow(np.imag(rho_psi), cmap='viridis')
    ax[1].set_title('Imag Part')
    fig.colorbar(cax2, ax=ax[1])
    plt.show()
```
