# Compendio Subtemático: Info Cuantica Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    trials = 100
    alice_bases = np.random.randint(2, size=trials)
    bob_bases = np.random.randint(2, size=trials)
    match = alice_bases == bob_bases
    
    plt.hist(match.astype(int), bins=[-0.5, 0.5, 1.5], rwidth=0.8, color='teal')
    plt.xticks([0, 1], ['Mismatched Base', 'Matched Base'])
    plt.title('BB84 Key Distribution Basis Matches')
    plt.ylabel('Frequency')
    plt.show()

if __name__ == '__main__':
    main()
```
