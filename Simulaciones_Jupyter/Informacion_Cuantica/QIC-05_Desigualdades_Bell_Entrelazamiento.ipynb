# Desigualdades de Bell

Violación CHSH demostrando la inexistencia de variables ocultas locales.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, np.pi, 100)
E_quantum = np.cos(theta)
E_classico = 1 - 2*theta/np.pi

plt.figure(figsize=(6,6))
plt.plot(theta, E_quantum, 'r-', label='Predicción Cuántica')
plt.plot(theta, E_classico, 'k--', label='Límite Clásico Local')
plt.title('Correlación y Violación de Bell')
plt.legend()
plt.grid()
plt.show()
```
