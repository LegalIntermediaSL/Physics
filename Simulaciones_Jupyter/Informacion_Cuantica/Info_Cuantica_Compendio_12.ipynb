# Compendio Subtemático: Info Cuantica Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    steps = 50
    pos = np.zeros(2*steps + 1)
    pos[steps] = 1.0 # Initial delta distribution (simplified classical random walk for analogy)
    
    for _ in range(steps):
        new_pos = np.zeros_like(pos)
        for i in range(1, 2*steps):
            new_pos[i] = 0.5 * pos[i-1] + 0.5 * pos[i+1]
        pos = new_pos
        
    plt.plot(np.arange(-steps, steps+1), pos)
    plt.title('Random Walk (Classical Analogue to Quantum Walk)')
    plt.xlabel('Position')
    plt.ylabel('Probability')
    plt.show()

if __name__ == '__main__':
    main()
```
