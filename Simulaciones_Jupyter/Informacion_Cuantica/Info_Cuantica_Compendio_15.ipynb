# Compendio Subtemático: Info Cuantica Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    phases = np.linspace(0, 2*np.pi, 100)
    # Interference fringe pattern
    prob = np.cos(phases/2)**2
    
    plt.plot(phases, prob, color='navy')
    plt.title('Quantum Phase Estimation (Interference Fringe)')
    plt.xlabel('Phase shift')
    plt.ylabel('Probability of |0>')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
