# Compendio Subtemático: Info Cuantica Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Pure state |+>
    rho = 0.5 * np.array([[1, 1], [1, 1]])
    
    plt.matshow(rho.real, cmap='Blues')
    plt.colorbar()
    plt.title('Density Matrix of |+> state (Real Part)', pad=20)
    plt.show()

if __name__ == '__main__':
    main()
```
