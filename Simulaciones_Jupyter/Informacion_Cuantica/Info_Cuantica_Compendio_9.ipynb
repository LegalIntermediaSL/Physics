# Compendio Subtemático: Info Cuantica Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    N = 4
    qft_matrix = np.zeros((N, N), dtype=complex)
    omega = np.exp(2j * np.pi / N)
    for i in range(N):
        for j in range(N):
            qft_matrix[i, j] = omega**(i * j) / np.sqrt(N)
    
    plt.imshow(np.abs(qft_matrix), cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.title('Quantum Fourier Transform Matrix Magnitudes (N=4)')
    plt.show()

if __name__ == '__main__':
    main()
```
