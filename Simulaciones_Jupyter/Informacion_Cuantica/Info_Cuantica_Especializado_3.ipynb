# Laboratorio Especializado: Info Cuantica Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np

def grover_search():
    N = 8
    target = 5
    state = np.ones(N) / np.sqrt(N)
    
    iterations = int(np.pi / 4 * np.sqrt(N))
    for _ in range(iterations):
        # Oracle
        state[target] *= -1
        # Diffusion
        mean = np.mean(state)
        state = 2*mean - state
        
    print("State probabilities:", np.abs(state)**2)

if __name__ == "__main__":
    grover_search()
```
