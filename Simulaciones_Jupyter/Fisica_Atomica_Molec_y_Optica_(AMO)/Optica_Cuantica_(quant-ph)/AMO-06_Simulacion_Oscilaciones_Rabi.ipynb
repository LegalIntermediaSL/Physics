# [AMO-06] Óptica Cuántica: Emisión Estimulada Láser

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Ecuaciones Ópticas de Bloch (Sistema de Dos Niveles acoplado a un Láser)
# Simulamos un átomo iluminado por un láser continuo monocromático (Frecuencia w_L).
# Variables: u (Coherencia en fase), v (Coherencia en cuadratura), w (Inversión de Población).

Omega = 2.0 * np.pi * 1.0  # Frecuencia de Rabi (Intensidad del acoplamiento láser-átomo)
Delta = 0.0                # Detuning (Resonancia Perfecta w_L = w_atomo)
Gamma_sp = 0.5             # Decaimiento por Emisión Espontánea (Tasa de pérdida)

def optical_bloch(y, t):
    u, v, w = y
    du_dt = Delta * v - 0.5 * Gamma_sp * u
    dv_dt = -Delta * u + Omega * w - 0.5 * Gamma_sp * v
    dw_dt = -Omega * v - Gamma_sp * (w + 1) # Relajación hacia el estado base (w=-1)
    return [du_dt, dv_dt, dw_dt]

# Estado inicial: El electrón empieza en el estado fundamental (Inversión w = -1, Coherencias = 0)
y0 = [0.0, 0.0, -1.0]
time = np.linspace(0, 5, 1000)

sol = odeint(optical_bloch, y0, time)
w_sol = sol[:, 2]

# Probabilidad de encontrar el electrón en el estado excitado (P_e = (w+1)/2)
P_excitado = (w_sol + 1.0) / 2.0

plt.figure(figsize=(10, 5))
plt.plot(time, P_excitado, 'r-', lw=3, label='Probabilidad de Estado Excitado $P_e(t)$')
plt.title("Óptica Cuántica: Oscilaciones de Rabi Amortiguadas")
plt.xlabel("Tiempo $t$ (microsegundos)")
plt.ylabel("Probabilidad")
plt.ylim([0, 1.05])
plt.grid(True)
plt.legend()
plt.show()
# FÍSICA: El electrón absorbe un fotón y sube, es forzado por el mismo láser 
# mediante emisión estimulada a bajar (desorbitar). La decaimiento espontáneo amortigua el sistema a estado estacionario.

```
