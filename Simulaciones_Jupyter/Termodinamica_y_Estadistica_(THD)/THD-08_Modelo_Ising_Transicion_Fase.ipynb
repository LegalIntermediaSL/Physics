# Modelo de Ising

Representación de espines en redes ferromagnéticas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

spin_lattice = np.random.choice([-1, 1], size=(50,50))

plt.figure(figsize=(6,6))
plt.imshow(spin_lattice, cmap='gray')
plt.title('Microestado del Modelo de Ising 2D')
plt.axis('off')
plt.show()
```
