# [THD-09] Dinámica de Langevin de un Trinquete

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de un Trinquete Browniano (Flashing Ratchet)
# Dinámica de Langevin: dx = -V'(x,t)*dt/gamma + sqrt(2D*dt)*dW

N_particles = 1000
dt = 0.01
steps = 2000
D = 0.5  # Difusión

x = np.zeros(N_particles)
x_history = []

# Potencial asimétrico diente de sierra
def dV_dx(x):
    mod_x = x % 1.0
    # Asimetría: subida lenta (0 a 0.8), bajada rápida (0.8 a 1.0)
    force = np.where(mod_x < 0.8, 1.0/0.8, -1.0/0.2)
    return force

# El potencial se enciende/apaga periódicamente
tau_on = 10
tau_off = 10

for i in range(steps):
    x_history.append(np.mean(x))
    
    # Estado del potencial
    is_on = (i % (tau_on + tau_off)) < tau_on
    
    # Fuerza determinista (0 si está apagado)
    f_det = -dV_dx(x) if is_on else 0.0
    
    # Ruido térmico (Wiener process)
    noise = np.sqrt(2 * D * dt) * np.random.randn(N_particles)
    
    # Integración de Euler-Maruyama
    x += f_det * dt + noise

plt.figure(figsize=(10, 5))
plt.plot(np.arange(steps)*dt, x_history, 'b-', lw=2)
plt.title("Corriente Neta en un Trinquete Browniano (Flashing Ratchet)")
plt.xlabel("Tiempo")
plt.ylabel("Posición Promedio <x>")
plt.grid(True)
plt.show()

print(f"Velocidad neta de deriva: {x_history[-1]/(steps*dt):.4f}")

```
