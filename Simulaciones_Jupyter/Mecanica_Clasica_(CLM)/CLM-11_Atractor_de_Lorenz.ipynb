# Atractor de Lorenz

Representación tridimensional del atractor de Lorenz, un clásico de los sistemas dinámicos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

def lorenz(state, t):
    x, y, z = state
    return [10*(y-x), x*(28-z)-y, x*y-8/3*z]

t = np.linspace(0, 40, 10000)
state0 = [1.0, 1.0, 1.0]
sol = odeint(lorenz, state0, t)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol[:,0], sol[:,1], sol[:,2], lw=0.5, color='red')
plt.title('Atractor Caótico de Lorenz')
plt.axis('off')
plt.show()
```
