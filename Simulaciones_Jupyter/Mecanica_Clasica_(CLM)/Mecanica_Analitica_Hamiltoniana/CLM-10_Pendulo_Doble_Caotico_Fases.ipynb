# [CLM-10] Dinámica Analítica: Caos y Péndulo Doble

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Mecánica Analítica Lagrangiana: Péndulo Doble Caótico
# Las Ecuaciones de Euler-Lagrange para este sistema determinista generan 
# Caos en altas energías. Para desentrañar el desorden, trazamos "Mapas de Poincaré".

# Constantes del Péndulo Doble (masas M=1, longitudes L=1, gravedad G=9.8)
def deriv_double_pendulum(t, state):
    t1, w1, t2, w2 = state
    d = t1 - t2
    
    # Ecuaciones de Euler-Lagrange reducidas algebráicamente
    den1 = 1 + np.sin(d)**2
    dt1 = w1
    dt2 = w2
    
    dw1 = (w2**2 * np.sin(d) * np.cos(d) + w2**2 * np.sin(d) - 2 * 9.8 * np.sin(t1) + 9.8 * np.sin(t2) * np.cos(d)) / den1
    dw2 = (-2 * w1**2 * np.sin(d) - w2**2 * np.sin(d) * np.cos(d) + 2 * 9.8 * np.sin(t1) * np.cos(d) - 2 * 9.8 * np.sin(t2)) / den1
    
    return [dt1, dw1, dt2, dw2]

# Alta energía = Caos total
state0 = [np.pi/2, 0.0, np.pi/2, 0.0]
t_span = (0, 30)
t_eval = np.linspace(*t_span, 5000)

sol = solve_ivp(deriv_double_pendulum, t_span, state0, t_eval=t_eval)

# Coordenadas Cartesianas (Transformación Cinematica)
x1 = np.sin(sol.y[0])
y1 = -np.cos(sol.y[0])
x2 = x1 + np.sin(sol.y[2])
y2 = y1 - np.cos(sol.y[2])

fig = plt.figure(figsize=(12, 5))

# Espacio Cartesiano
ax1 = fig.add_subplot(121)
ax1.plot(x1, y1, 'blue', alpha=0.4, lw=1)
ax1.plot(x2, y2, 'red', alpha=0.8, lw=1)
ax1.set_title("Espacio Cartesiano (Trayectoria Caótica)", fontsize=14)
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.axis('equal')

# Espacio de Fases del brazo 1 (theta vs momento angular omega)
ax2 = fig.add_subplot(122)
ax2.plot(sol.y[0] % (2*np.pi), sol.y[1], 'k.', markersize=0.5, alpha=0.5)
ax2.set_title("Espacio de Fases $q_1 - p_1$ (Seudomapa de Poincaré)", fontsize=14)
ax2.set_xlabel("Ángulo $\theta_1$")
ax2.set_ylabel("Vel. Angular $\omega_1$")

plt.tight_layout()
plt.show()

# FÍSICA: En coordenadas cartesianas el movimiento es un infierno entrelazado imposible
# de predecir o resolver analíticamente (El Teorema KAM demuestra la ruptura de los toros 
# invariantes del espacio de fases de Hamilton). Las Secciones de Poincaré nos permiten
# detectar 'Islas de Estabilidad' dentro del mar caótico del diagrama de fases.

```
