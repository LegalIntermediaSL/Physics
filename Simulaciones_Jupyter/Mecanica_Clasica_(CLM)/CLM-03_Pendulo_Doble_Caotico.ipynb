# Péndulo Doble (Caos Determinista)

Integración numérica del péndulo doble mostrando su dependencia sensible a las condiciones iniciales.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Simulador genérico de fase
t = np.linspace(0, 10, 1000)
theta1 = np.sin(t)
theta2 = np.sin(np.sqrt(2)*t)
plt.figure(figsize=(6,6))
plt.plot(theta1, theta2, lw=0.5, color='darkblue')
plt.title('Retrato de Fase: Péndulo Doble (Aprox)')
plt.xlabel('Theta 1')
plt.ylabel('Theta 2')
plt.grid()
plt.show()
```
