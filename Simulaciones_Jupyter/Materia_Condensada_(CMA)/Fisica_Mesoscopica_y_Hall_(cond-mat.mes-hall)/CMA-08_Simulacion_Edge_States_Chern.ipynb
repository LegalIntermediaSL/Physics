# [CMA-08] Simulación de Estados de Borde

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de Modos de Borde en un Aislante Topológico (Modelo BHZ / Chern)
# Espectro de energía de una banda cilíndrica a lo largo de kx.

N_y = 40  # Sitios en la dirección y (finita)
k_x = np.linspace(-np.pi, np.pi, 100)

def hamiltonian_1d(kx, M, B, A):
    H = np.zeros((N_y, N_y), dtype=complex)
    for j in range(N_y):
        # Término diagonal (masa efectiva)
        H[j, j] = M - 2*B*(1 - np.cos(kx)) - 2*B
        # Salto a primeros vecinos en y
        if j < N_y - 1:
            H[j, j+1] = B + 1j * A / 2.0
            H[j+1, j] = B - 1j * A / 2.0
    return H

# Parámetros fase topológica (M/2B < 2)
M, B, A = -1.0, -1.0, 1.0

energies = []
for k in k_x:
    H = hamiltonian_1d(k, M, B, A)
    evals = np.linalg.eigvalsh(H)
    energies.append(evals)

energies = np.array(energies)

plt.figure(figsize=(8,6))
for i in range(N_y):
    # Pintamos en rojo los modos de borde si cruzan E=0
    color = 'r' if np.min(np.abs(energies[:, i])) < 0.1 else 'k'
    plt.plot(k_x, energies[:, i], color=color, alpha=0.6)

plt.title("Espectro de Energía (Cilindro finito): Aislante Topológico")
plt.xlabel("Momento k_x")
plt.ylabel("Energía E")
plt.axhline(0, color='b', linestyle='--')
plt.grid(True)
plt.show()

```
