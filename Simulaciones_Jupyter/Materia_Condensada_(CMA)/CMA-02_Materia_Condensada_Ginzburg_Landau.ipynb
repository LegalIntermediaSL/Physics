# Investigación Avanzada: Materia Condensada Ginzburg Landau

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

# 1D Ginzburg-Landau equations for a superconductor-normal interface
# Let f be the normalized order parameter (f = |psi|/psi_infty)
# Let a be the normalized vector potential (a = A / (sqrt(2)*H_c*lambda))
# Equations:
# (1/kappa^2) * f'' = f * (f^2 - 1 + a^2)
# a'' = f^2 * a

kappa = 0.5 # Type I superconductor

def gl_equations(x, y):
    f, df, a, da = y
    d2f = kappa**2 * f * (f**2 - 1 + a**2)
    d2a = f**2 * a
    return np.vstack((df, d2f, da, d2a))

def bc(ya, yb):
    # ya: at x = 0 (Normal state interface or deep inside normal region)
    # yb: at x = L (Deep inside superconducting region)
    return np.array([
        ya[0] - 0.0,  # f(0) = 0
        ya[2] - 1.0,  # a(0) = 1 (magnetic field penetration)
        yb[0] - 1.0,  # f(L) = 1 (superconducting order parameter)
        yb[3] - 0.0   # a'(L) = 0 (no magnetic field)
    ])

x = np.linspace(0, 10, 100)
y_guess = np.zeros((4, x.size))
y_guess[0] = x / 10.0 # Linear guess for f
y_guess[2] = 1.0 - x / 10.0 # Linear guess for a

res = solve_bvp(gl_equations, bc, x, y_guess, tol=1e-4, max_nodes=1000)

plt.figure(figsize=(8, 6))
plt.plot(res.x, res.y[0], label='Order Parameter $f$', color='blue', linewidth=2)
plt.plot(res.x, res.y[2], label='Vector Potential $a$', color='red', linewidth=2)
plt.plot(res.x, res.y[3], label='Magnetic Field $b = da/dx$', color='green', linewidth=2)
plt.title(f'Ginzburg-Landau 1D Solution ($\kappa$={kappa})')
plt.xlabel('Distance $x / \\lambda$')
plt.ylabel('Normalized Quantities')
plt.legend()
plt.grid(True)
plt.savefig('ginzburg_landau.png')
```
