# Efecto Meissner

Perfil de penetración del campo magnético en superconductores (Ecuación de London).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 100)
B = np.exp(-x) # Decaimiento de London

plt.figure(figsize=(8,4))
plt.plot(x, B, 'c-', lw=3, label='Campo Magnético B(x)')
plt.fill_between(x, B, color='cyan', alpha=0.3)
plt.title('Penetración Magnética en Superconductor (Meissner)')
plt.xlabel('Profundidad x')
plt.legend()
plt.grid()
plt.show()
```
