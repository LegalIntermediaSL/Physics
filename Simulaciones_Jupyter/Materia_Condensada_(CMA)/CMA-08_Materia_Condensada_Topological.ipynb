# Investigación Avanzada: Materia Condensada Topological

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def haldane_model(kx, ky, t1=1.0, t2=0.1, phi=np.pi/2, m=0.0):
    # Primitive lattice vectors
    a1 = np.array([np.sqrt(3)/2, 3/2])
    a2 = np.array([-np.sqrt(3)/2, 3/2])
    
    # Nearest neighbor vectors
    delta1 = np.array([0, 1])
    delta2 = np.array([np.sqrt(3)/2, -1/2])
    delta3 = np.array([-np.sqrt(3)/2, -1/2])
    
    # Next nearest neighbor vectors
    b1 = a1
    b2 = a2
    b3 = a2 - a1
    
    h0 = 2 * t2 * np.cos(phi) * (np.cos(np.dot([kx, ky], b1)) + 
                                 np.cos(np.dot([kx, ky], b2)) + 
                                 np.cos(np.dot([kx, ky], b3)))
    hx = t1 * (np.cos(np.dot([kx, ky], delta1)) + 
               np.cos(np.dot([kx, ky], delta2)) + 
               np.cos(np.dot([kx, ky], delta3)))
    hy = t1 * (np.sin(np.dot([kx, ky], delta1)) + 
               np.sin(np.dot([kx, ky], delta2)) + 
               np.sin(np.dot([kx, ky], delta3)))
    hz = m - 2 * t2 * np.sin(phi) * (np.sin(np.dot([kx, ky], b1)) + 
                                     np.sin(np.dot([kx, ky], b2)) + 
                                     np.sin(np.dot([kx, ky], b3)))
    
    E_plus = h0 + np.sqrt(hx**2 + hy**2 + hz**2)
    E_minus = h0 - np.sqrt(hx**2 + hy**2 + hz**2)
    
    return E_minus, E_plus

# Generate BZ grid
kx = np.linspace(-np.pi, np.pi, 100)
ky = np.linspace(-np.pi, np.pi, 100)
KX, KY = np.meshgrid(kx, ky)

E1, E2 = np.zeros_like(KX), np.zeros_like(KX)
for i in range(KX.shape[0]):
    for j in range(KX.shape[1]):
        E1[i, j], E2[i, j] = haldane_model(KX[i, j], KY[i, j])

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(KX, KY, E1, cmap='viridis', alpha=0.8)
ax.plot_surface(KX, KY, E2, cmap='plasma', alpha=0.8)
ax.set_title('Haldane Model Band Structure')
ax.set_xlabel('kx')
ax.set_ylabel('ky')
ax.set_zlabel('Energy')
plt.savefig('haldane_model.png')
```
