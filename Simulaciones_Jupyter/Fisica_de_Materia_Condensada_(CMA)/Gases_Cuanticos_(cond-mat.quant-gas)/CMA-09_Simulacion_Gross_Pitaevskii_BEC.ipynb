# [CMA-09] Fluidos Cuánticos: Condensado de Bose-Einstein

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal

# Hidrodinámica Cuántica: Condensado de Bose-Einstein (BEC)
# Solución de Estado Base de la Ecuación de Gross-Pitaevskii (GPE) en 1D
# H_gpe * Psi = (-hbar^2/2m d^2/dx^2 + V_trap(x) + g*|Psi|^2) * Psi = mu * Psi

# Parámetros (Unidades adimensionales)
N_points = 500
x = np.linspace(-10, 10, N_points)
dx = x[1] - x[0]
V_trap = 0.5 * x**2  # Trampa Armónica Magnética
g_list = [0, 10, 50] # Fuerzas de repulsión s-wave (0 = Gas ideal de Bose, >0 = BEC Repulsivo)

plt.figure(figsize=(10, 6))
plt.plot(x, V_trap, 'k--', label="Trampa Armónica $V(x)$", alpha=0.5)

# Iteración de Relajación en Tiempo Imaginario para resolver la GPE No-Lineal
for g in g_list:
    # Estado inicial: Gaussiana simple
    psi = np.exp(-x**2 / 2.0)
    psi /= np.sqrt(np.sum(np.abs(psi)**2) * dx) # Normalización inicial
    
    # Parámetros del tiempo imaginario (Evolución dt)
    dt = 0.01
    
    # 500 pasos de relajación para encontrar el estado base
    for _ in range(500):
        # Operador de Energía Cinética (Diferencias finitas, segunda derivada)
        laplacian = (np.roll(psi, -1) - 2*psi + np.roll(psi, 1)) / dx**2
        
        # Potencial total = Trampa + Interacción Mean-Field de Gross-Pitaevskii
        V_eff = V_trap + g * np.abs(psi)**2
        
        # Evolución en tiempo imaginario (psi -> psi - dt * H * psi)
        psi = psi - dt * (-0.5 * laplacian + V_eff * psi)
        
        # Renormalización forzada
        psi /= np.sqrt(np.sum(np.abs(psi)**2) * dx)
    
    density = np.abs(psi)**2
    
    label_txt = f"Gas Ideal Bose (g={g})" if g == 0 else f"Superfluido BEC (g={g})"
    plt.plot(x, density, lw=2.5, label=label_txt)

plt.title("Hidrodinámica Cuántica: Perfil de Densidad del Superfluido (Gross-Pitaevskii 1D)")
plt.xlabel("Posición Espacial $x$")
plt.ylabel("Densidad Probabilidad $|\Psi|^2$")
plt.xlim([-5, 5])
plt.ylim([0, 0.7])
plt.grid(True)
plt.legend()
plt.show()
# FÍSICA: Al incrementar la repulsión cuántica (g), el condensado se deforma,
# pasando de una Gaussiana de gas ideal a un perfil de parábola invertida 
# conocido como el límite de Thomas-Fermi (característico de los superfluidos atómicos reales).

```
