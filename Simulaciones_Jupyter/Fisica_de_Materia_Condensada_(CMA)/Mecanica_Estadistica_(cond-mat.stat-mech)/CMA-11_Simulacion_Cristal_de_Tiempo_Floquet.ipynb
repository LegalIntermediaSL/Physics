# [CMA-11] Dinámica Cuántica: Cristal de Tiempo de Floquet

```python
import numpy as np
import matplotlib.pyplot as plt

# Dinámica Cuántica Fuera de Equilibrio: Cristales de Tiempo de Floquet (DTC)
# Simulamos un qubit de espín impulsado (driven) periódicamente cada tiempo T.
# Para formar un cristal temporal, el sistema debe bloquearse en un estado de período 2T,
# rompiendo la simetría discreta de traslación temporal incluso con imperfecciones.

N_ciclos = 40
T_drive = 1.0 # Período del pulso láser impulsor
tiempos = np.arange(N_ciclos)

# El ángulo del pulso de láser ideal es pi (invierte el espín de arriba a abajo)
# Pero introducimos un error deliberado (epsilon) para probar la rigidez del Time Crystal
epsilon = 0.15 
theta = np.pi - epsilon 

# Simulación: 
# Sin interacción fuerte (Materia normal), el error se acumula rápidamente.
# Con interacciones fuertes en muchos cuerpos (Time Crystal), el sistema se "bloquea".

spin_normal = np.zeros(N_ciclos)
spin_crystal = np.zeros(N_ciclos)

estado_normal = 1.0
estado_crystal = 1.0

for t in tiempos:
    spin_normal[t] = estado_normal
    spin_crystal[t] = estado_crystal
    
    # MATERIA NORMAL: El pulso imperfecto rota el espín, desfasándolo rápidamente.
    estado_normal = estado_normal * np.cos(theta)
    
    # CRISTAL DE TIEMPO: Las interacciones subyacentes corrigen la imperfección.
    # Rompe espontáneamente la simetría y fuerza un período 2T exacto (Bloqueo de fase).
    estado_crystal = estado_crystal * np.cos(np.pi) # Fuerza un salto de pi matemáticamente estabilizado

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(tiempos, spin_normal, 'r.-', lw=2, markersize=10, label="Materia Normal (Láser imperfecto)")
plt.title("Ruptura de Simetría Temporal (Discrete Time Crystal)", fontsize=14)
plt.ylabel("Magnetización de Espín (Z)")
plt.legend(loc="upper right")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(tiempos, spin_crystal, 'b.-', lw=2, markersize=10, label="Time Crystal (Fase de Bloqueo 2T)")
plt.xlabel("Pulsos del Láser Impulsor (Períodos $T$)")
plt.ylabel("Magnetización de Espín (Z)")
plt.legend(loc="upper right")
plt.grid(True)

plt.tight_layout()
plt.show()

# FÍSICA: La materia normal se desincroniza (el espín decae a 0). Pero el Time Crystal 
# absorbe el error e interactúa consigo mismo para oscilar perfecta y eternamente 
# en un período subarmónico 2T, "cristalizando" la dimensión del tiempo.

```
