# [CMA-10] Aislantes Topológicos: Conos de Dirac (Grafeno)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Física del Grafeno: Estructura de Bandas y Conos de Dirac
# El Hamiltoniano de Tight-Binding para el grafeno (retículo hexagonal) resulta en 
# una relación de dispersión E(kx, ky) donde las bandas de valencia y conducción se tocan.

# Constantes del retículo y salto (hopping energy t)
t = 2.8 # eV
a = 1.0 # Constante del cristal escalada

kx = np.linspace(-np.pi, np.pi, 200)
ky = np.linspace(-np.pi, np.pi, 200)
KX, KY = np.meshgrid(kx, ky)

# Relación de Dispersión Cuántica del Grafeno
f_k = np.sqrt(1 + 4*np.cos(np.sqrt(3)*KX*a/2)*np.cos(KY*a/2) + 4*(np.cos(KY*a/2))**2)
E_plus = t * f_k    # Banda de Conducción
E_minus = -t * f_k  # Banda de Valencia

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Ploteamos ambas bandas de energía
surf1 = ax.plot_surface(KX, KY, E_plus, cmap='cool', alpha=0.8, edgecolor='none')
surf2 = ax.plot_surface(KX, KY, E_minus, cmap='autumn', alpha=0.8, edgecolor='none')

ax.set_title("Grafeno: Conos de Dirac y Fermiones Relativistas", fontsize=14)
ax.set_xlabel('Vector de onda $k_x$')
ax.set_ylabel('Vector de onda $k_y$')
ax.set_zlabel('Energía $E$ (eV)')

# Cámara y visualización
ax.view_init(elev=15., azim=45)
plt.show()
# FÍSICA: Observa los 6 puntos en el plano E=0 (Puntos K y K').
# En esos puntos cónicos, los electrones obedecen la ecuación relativista E = vp 
# como si no tuvieran masa, viajando a 1 millón de m/s. 

```
