# [CMA-08] Física de Transistores: Bandas Prohibidas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Solución Matemática del Modelo de Kronig-Penney (Estructura de Bandas)
# En un cristal 1D con periodicidad atómica 'a', el Teorema de Bloch fuerza una relación de dispersión.
# Para un potencial delta-Dirac de fuerza P, la ecuación trascendental es:
# cos(k*a) = P * sin(alpha*a)/(alpha*a) + cos(alpha*a)
# donde alpha proproporcional a sqrt(E).

# Fuerza del potencial de la red iónica (Impureza cristalina)
P = (3.0 * np.pi) / 2.0 
a = 1.0 # Constante de red

alpha_a = np.linspace(1e-3, 4 * np.pi, 2000)
# Lado derecho de la ecuación trascendental
f_alpha = P * np.sin(alpha_a) / alpha_a + np.cos(alpha_a)

plt.figure(figsize=(10, 6))
plt.plot(alpha_a / np.pi, f_alpha, 'b-', lw=2, label="$f(\alpha a)$")

# Límites de cos(ka) (Entre +1 y -1). 
# Las zonas donde f(alpha*a) supera 1 o baja de -1 son los BAND GAPS (Energías Prohibidas).
plt.axhline(1, color='r', linestyle='--', lw=2, label="Límite Físico del Teorema de Bloch")
plt.axhline(-1, color='r', linestyle='--', lw=2)

plt.fill_between(alpha_a / np.pi, -1, 1, color='lightgreen', alpha=0.3, label="Zonas Permitidas (Bandas)")

plt.title("Física de Semiconductores: Modelo de Kronig-Penney (Band Gaps)")
plt.xlabel("Energía Proporcional $\alpha a / \pi \sim \sqrt{E}$")
plt.ylabel("Amplitud de Función Bloch")
plt.ylim([-3, 3])
plt.grid(True)
plt.legend()
plt.show()
# FÍSICA: Las zonas fuera de la banda verde (-1 a 1) representan niveles de energía
# totalmente prohibidos por la difracción cuántica (Interferencia destructiva de Bragg).
# Esta es la base matemática sobre la cual operan todos los transistores (Semiconductores).

```
