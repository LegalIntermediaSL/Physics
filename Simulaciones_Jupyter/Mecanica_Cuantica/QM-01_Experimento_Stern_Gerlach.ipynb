# Experimento de Stern-Gerlach

Cuantización espacial del espín 1/2 en campos magnéticos inhomogéneos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 100)
z_up = 0.5 * x**2
z_down = -0.5 * x**2

plt.figure(figsize=(8,4))
plt.plot(x, z_up, 'b-', label='Espín Up (+1/2)')
plt.plot(x, z_down, 'r-', label='Espín Down (-1/2)')
plt.title('Desdoblamiento de Haces: Stern-Gerlach')
plt.xlabel('Eje x (Propagación)')
plt.ylabel('Eje z (Deflexión Magnética)')
plt.legend()
plt.grid()
plt.show()
```
