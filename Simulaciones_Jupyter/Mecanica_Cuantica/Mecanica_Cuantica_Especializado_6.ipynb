# Laboratorio Especializado: Mecanica Cuantica Especializado 6

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import genlaguerre

def hydrogen_radial():
    r = np.linspace(0, 20, 1000)
    n, l = 3, 1
    L = genlaguerre(n-l-1, 2*l+1)(2*r/n)
    R = (2*r/n)**l * np.exp(-r/n) * L
    prob_density = R**2 * r**2
    plt.plot(r, prob_density)
    plt.title(f"Hydrogen Radial Probability Density n={n}, l={l}")
    plt.xlabel("Radius r (Bohr radii)")
    plt.ylabel("$r^2 |R(r)|^2$")
    plt.show()

if __name__ == "__main__":
    hydrogen_radial()
```
