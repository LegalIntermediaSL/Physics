# Polinomios de Hermite (Oscilador Armónico)

Visualización de las funciones de onda estacionarias del oscilador cuántico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite
import math

x = np.linspace(-4, 4, 200)

plt.figure(figsize=(8,5))
for n in range(4):
    H_n = hermite(n)(x)
    psi = (1/math.sqrt(2**n * math.factorial(n))) * np.exp(-x**2/2) * H_n
    plt.plot(x, psi + n, label=f'n={n}')
plt.title('Funciones de Onda del Oscilador Armónico Cuántico')
plt.legend()
plt.grid()
plt.show()
```
