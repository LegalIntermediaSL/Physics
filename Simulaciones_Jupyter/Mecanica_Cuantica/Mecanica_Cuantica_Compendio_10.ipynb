# Compendio Subtemático: Mecanica Cuantica Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
alpha = 2.0
m = 1.0
hbar = 1.0
kappa = m * alpha / hbar**2

psi = np.sqrt(kappa) * np.exp(-kappa * np.abs(x))

plt.plot(x, psi)
plt.title('Bound state of attractive delta potential')
plt.xlabel('x')
plt.ylabel(r'$\psi(x)$')
plt.grid(True)
plt.show()
```
