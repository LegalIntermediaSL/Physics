# Compendio Subtemático: Mecanica Cuantica Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def psi_n(x, n, L):
    return np.sqrt(2/L) * np.sin(n * np.pi * x / L)

x = np.linspace(0, 1, 1000)
for n in range(1, 4):
    plt.plot(x, psi_n(x, n, 1), label=f'n={n}')
plt.title('Particle in a 1D Box: Wavefunctions')
plt.xlabel('x')
plt.ylabel(r'$\psi_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```
