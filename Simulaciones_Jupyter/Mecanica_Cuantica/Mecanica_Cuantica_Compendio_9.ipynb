# Compendio Subtemático: Mecanica Cuantica Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 20, 1000)
Omega_0 = 1.0 
Delta = 0.5   
Omega = np.sqrt(Omega_0**2 + Delta**2)

P_excited = (Omega_0**2 / Omega**2) * np.sin(Omega * t / 2)**2

plt.plot(t, P_excited)
plt.title(f'Rabi Oscillations (Detuning = {Delta})')
plt.xlabel('Time')
plt.ylabel('Probability of Excited State')
plt.grid(True)
plt.show()
```
