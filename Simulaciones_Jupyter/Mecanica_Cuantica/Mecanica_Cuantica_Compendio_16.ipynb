# Compendio Subtemático: Mecanica Cuantica Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, 500)
V = 5 * (x**2 - 2)**2
dx = x[1] - x[0]
H = -0.5 * (np.diag(-2*np.ones(500)) + np.diag(np.ones(499), 1) + np.diag(np.ones(499), -1)) / dx**2 + np.diag(V)
E, V_eig = np.linalg.eigh(H)
plt.plot(x, V_eig[:, 0]**2 + E[0], label='Ground state')
plt.plot(x, V_eig[:, 1]**2 + E[1], label='1st Excited state')
plt.plot(x, V * 0.01) 
plt.title('Double Well Potential Wavefunctions')
plt.legend()
plt.grid()
plt.show()
```
