# Laboratorio Especializado: Mecanica Cuantica Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite

def qho_2d():
    x = np.linspace(-3, 3, 100)
    y = np.linspace(-3, 3, 100)
    X, Y = np.meshgrid(x, y)
    nx, ny = 2, 2
    Hx = hermite(nx)(X)
    Hy = hermite(ny)(Y)
    psi = Hx * Hy * np.exp(-0.5*(X**2 + Y**2))
    plt.contourf(X, Y, psi**2, 50, cmap='viridis')
    plt.colorbar()
    plt.title("2D QHO Probability Density (nx=2, ny=2)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

if __name__ == "__main__":
    qho_2d()
```
