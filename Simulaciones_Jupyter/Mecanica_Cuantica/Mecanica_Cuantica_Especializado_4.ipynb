# Laboratorio Especializado: Mecanica Cuantica Especializado 4

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def time_dep_pert():
    t = np.linspace(0, 10, 1000)
    omega = 1.0
    omega_0 = 1.1
    delta_omega = omega - omega_0
    P = (np.sin(delta_omega * t / 2)**2) / (delta_omega**2)
    plt.plot(t, P)
    plt.title("Transition Probability (Time-dependent Perturbation)")
    plt.xlabel("Time t")
    plt.ylabel("Probability")
    plt.show()

if __name__ == "__main__":
    time_dep_pert()
```
