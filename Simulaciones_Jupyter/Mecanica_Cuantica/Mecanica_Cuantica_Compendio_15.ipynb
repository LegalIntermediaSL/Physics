# Compendio Subtemático: Mecanica Cuantica Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
theta = np.linspace(0, np.pi, 10)
phi = np.linspace(0, 2*np.pi, 10)
T, P = np.meshgrid(theta, phi)
X = np.sin(T)*np.cos(P)
Y = np.sin(T)*np.sin(P)
plt.plot(X.flatten(), Y.flatten(), 'o')
plt.title('Bloch Sphere (Equatorial & Polar projections)')
plt.axis('equal')
plt.grid()
plt.show()
```
