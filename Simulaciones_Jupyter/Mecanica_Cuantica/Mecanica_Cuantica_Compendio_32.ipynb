# Compendio Subtemático: Mecanica Cuantica Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
k = np.linspace(0.1, 5, 100)
a = 1.0
delta_0 = -k * a
sigma = (4 * np.pi / k**2) * np.sin(delta_0)**2
plt.plot(k, sigma)
plt.title('S-wave Scattering Cross Section (Hard Sphere)')
plt.xlabel('k')
plt.grid()
plt.show()
```
