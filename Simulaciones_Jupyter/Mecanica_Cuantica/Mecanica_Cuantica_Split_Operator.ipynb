# Investigación Avanzada: Mecanica Cuantica Split Operator

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def main():
    # 2D Split-Operator Method for Free Particle
    N = 128
    L = 10.0
    x = np.linspace(-L/2, L/2, N)
    y = np.linspace(-L/2, L/2, N)
    X, Y = np.meshgrid(x, y)
    
    dx = x[1] - x[0]
    kx = np.fft.fftfreq(N, d=dx) * 2 * np.pi
    ky = np.fft.fftfreq(N, d=dx) * 2 * np.pi
    KX, KY = np.meshgrid(kx, ky)
    
    dt = 0.05
    steps = 200
    
    # Initial wavepacket
    sigma = 0.5
    k0x = 5.0
    k0y = 5.0
    psi = np.exp(-(X**2 + Y**2)/(2*sigma**2)) * np.exp(1j * (k0x*X + k0y*Y))
    psi /= np.linalg.norm(psi)
    
    # Kinetic energy operator
    T = (KX**2 + KY**2) / 2.0
    U_k = np.exp(-1j * T * dt)
    
    history_prob = np.zeros((steps, N, N))
    
    for t in range(steps):
        history_prob[t] = np.abs(psi)**2
        # Move to momentum space
        psi_k = np.fft.fft2(psi)
        # Apply kinetic operator
        psi_k *= U_k
        # Move back to position space
        psi = np.fft.ifft2(psi_k)
        
    fig, ax = plt.subplots(figsize=(6, 6))
    im = ax.imshow(history_prob[0], extent=[-L/2, L/2, -L/2, L/2], origin='lower', cmap='viridis')
    ax.set_title("2D Wavepacket Evolution (Split-Operator)")
    
    def update(frame):
        im.set_array(history_prob[frame])
        return [im]
        
    ani = animation.FuncAnimation(fig, update, frames=steps, interval=30, blit=True)
    plt.show()

if __name__ == "__main__":
    main()
```
