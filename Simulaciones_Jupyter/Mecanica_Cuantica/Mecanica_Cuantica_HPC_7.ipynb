# Supercomputación: Mecanica Cuantica HPC 7

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

q_max = 50
energies = []
alphas = []

for q in range(1, q_max + 1):
    for p in range(0, q + 1):
        if np.gcd(p, q) == 1:
            alpha = p / q
            H = np.zeros((q, q), dtype=complex)
            for i in range(q):
                H[i, i] = 2 * np.cos(2 * np.pi * alpha * i)
                H[i, (i+1)%q] = 1
                H[(i+1)%q, i] = 1
            e = np.linalg.eigvalsh(H)
            energies.extend(e)
            alphas.extend([alpha]*q)

plt.scatter(alphas, energies, s=0.1, c='black')
plt.title('Hofstadter Butterfly (Harper Equation)')
plt.xlabel(r'$\alpha = \Phi / \Phi_0$')
plt.ylabel('Energy')
plt.savefig('Mecanica_Cuantica_HPC_7_output.png')
```
