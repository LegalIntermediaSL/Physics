# Compendio Subtemático: Mecanica Cuantica Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
alpha = np.linspace(0.1, 2, 100)
E_var = alpha / 2 + 1 / (8 * alpha)
plt.plot(alpha, E_var)
plt.axvline(0.5, color='r', linestyle='--', label='Min at alpha=0.5')
plt.title('Variational Energy for H.O.')
plt.legend()
plt.grid()
plt.show()
```
