# Supercomputación: Mecanica Cuantica HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import scipy.sparse as sparse
import scipy.sparse.linalg as spla
import matplotlib.pyplot as plt

N = 100
L = 1.0
dx = L / N
x = np.linspace(0, L, N)
y = np.linspace(0, L, N)
X, Y = np.meshgrid(x, y)

D1 = sparse.diags([-1, 2, -1], [-1, 0, 1], shape=(N, N)) / dx**2
I = sparse.eye(N)
H = sparse.kronsum(D1, D1)

V = np.zeros((N, N))
V[40:60, 40:60] = 1000 # potential barrier
H_pot = sparse.diags([V.flatten()], [0], shape=(N*N, N*N))
H = H + H_pot

evals, evecs = spla.eigsh(H, k=5, which='SM')

plt.imshow(evecs[:,0].reshape(N, N)**2, extent=[0,L,0,L], origin='lower')
plt.title('Ground state density in 2D box with barrier')
plt.colorbar()
plt.savefig('Mecanica_Cuantica_HPC_3_output.png')
```
