# Compendio Subtemático: Mecanica Cuantica Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
E = np.linspace(-2, 2, 200)
T_vals = [0.01, 0.2, 0.5]
for T in T_vals:
    f = 1 / (np.exp(E / T) + 1)
    plt.plot(E, f, label=f'T={T}')
plt.title('Fermi-Dirac Distribution')
plt.axvline(0, color='k', linestyle=':')
plt.legend()
plt.grid()
plt.show()
```
