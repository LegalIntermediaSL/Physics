# Efecto Túnel Cuántico

Evolución de un paquete de ondas atravesando una barrera de potencial.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 500)
barrera = np.where((x>4) & (x<5), 1.0, 0)
psi = np.exp(-(x-3)**2) + 0.2*np.exp(-(x-6)**2)

plt.figure(figsize=(8,4))
plt.plot(x, barrera, 'r-', label='Barrera de Potencial')
plt.plot(x, psi, 'b-', label='Densidad de Probabilidad |Ψ|^2')
plt.title('Representación Estática del Efecto Túnel')
plt.legend()
plt.grid()
plt.show()
```
