# Compendio Subtemático: Mecanica Cuantica Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 500)
omega = 1.0 
theta = np.pi / 4 
phi_0 = 0.0

Sz = 0.5 * np.cos(theta) * np.ones_like(t)
Sx = 0.5 * np.sin(theta) * np.cos(omega * t + phi_0)
Sy = 0.5 * np.sin(theta) * np.sin(omega * t + phi_0)

plt.plot(t, Sx, label='<Sx>')
plt.plot(t, Sy, label='<Sy>')
plt.plot(t, Sz, label='<Sz>')
plt.title('Spin 1/2 Precession')
plt.xlabel('Time')
plt.ylabel('Expectation Value')
plt.legend()
plt.grid(True)
plt.show()
```
