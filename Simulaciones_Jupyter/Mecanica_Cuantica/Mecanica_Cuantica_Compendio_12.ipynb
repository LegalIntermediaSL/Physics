# Compendio Subtemático: Mecanica Cuantica Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm
theta = np.linspace(0, 2*np.pi, 500)
Y = sph_harm(0, 1, theta, 0)
r = np.abs(Y)**2
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_title('|Y_1^0|^2 Angular Dependence')
plt.show()
```
