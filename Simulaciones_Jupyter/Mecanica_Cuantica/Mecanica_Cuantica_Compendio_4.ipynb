# Compendio Subtemático: Mecanica Cuantica Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
k0 = 5.0
sigma = 1.0
psi = (1/(sigma*np.sqrt(np.pi)))**0.5 * np.exp(-x**2 / (2*sigma**2)) * np.exp(1j*k0*x)

plt.plot(x, np.real(psi), label='Re(psi)')
plt.plot(x, np.abs(psi)**2, label='|psi|^2')
plt.title('Free Particle Gaussian Wave Packet (t=0)')
plt.xlabel('x')
plt.legend()
plt.grid(True)
plt.show()
```
