# Compendio Subtemático: Mecanica Cuantica Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
E = np.linspace(0.1, 10, 100)
V0 = 10
gamma = np.sqrt(V0 - E)
T = np.exp(-2 * gamma)
plt.plot(E, T)
plt.title('WKB Transmission vs Energy')
plt.yscale('log')
plt.grid()
plt.show()
```
