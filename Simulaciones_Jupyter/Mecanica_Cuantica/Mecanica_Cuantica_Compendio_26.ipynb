# Compendio Subtemático: Mecanica Cuantica Compendio 26

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 10, 100)
alpha = 2.0
exp_x = np.sqrt(2) * alpha * np.cos(t)
exp_p = -np.sqrt(2) * alpha * np.sin(t)
plt.plot(t, exp_x, label='<x>')
plt.plot(t, exp_p, label='<p>')
plt.title('Coherent State Time Evolution')
plt.legend()
plt.grid()
plt.show()
```
