# Compendio Subtemático: Mecanica Cuantica Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)
psi = 2 * np.sin(np.pi * X) * np.sin(2 * np.pi * Y)
plt.contourf(X, Y, psi**2, 50, cmap='viridis')
plt.colorbar()
plt.title('2D Box Probability Density (n_x=1, n_y=2)')
plt.show()
```
