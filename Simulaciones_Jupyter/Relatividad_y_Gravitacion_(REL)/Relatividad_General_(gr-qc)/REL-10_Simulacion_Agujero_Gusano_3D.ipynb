# [REL-10] Visualizador del Paraboloide de Flamm (Agujero de Gusano)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Visualización del Embedding 3D de un Agujero de Gusano Atravesable (Métrica Espacial de Morris-Thorne)
# Paraboloide de Flamm interconectando dos universos asintóticamente planos a través de una "Garganta".

# Ecuación del perfil z(r) para la incrustación geométrica espacial (t=const, theta=pi/2)
# dz/dr = +/- sqrt( b_0 / (r - b_0) ) -> z(r) = +/- 2 * sqrt(b_0 * (r - b_0))
b_0 = 1.0 # Radio de la garganta del agujero de gusano

r = np.linspace(b_0, 6, 50)
theta = np.linspace(0, 2 * np.pi, 50)
R, Theta = np.meshgrid(r, theta)

X = R * np.cos(Theta)
Y = R * np.sin(Theta)
# Universo Superior (z > 0) y Universo Inferior (z < 0)
Z_up = 2.0 * np.sqrt(b_0 * (R - b_0))
Z_down = -2.0 * np.sqrt(b_0 * (R - b_0))

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Superficie superior
surf1 = ax.plot_surface(X, Y, Z_up, cmap='viridis', alpha=0.7, edgecolor='none')
# Superficie inferior
surf2 = ax.plot_surface(X, Y, Z_down, cmap='plasma', alpha=0.7, edgecolor='none')

# Dibujamos la Garganta (Throat)
throat_theta = np.linspace(0, 2*np.pi, 100)
ax.plot(b_0*np.cos(throat_theta), b_0*np.sin(throat_theta), 0, color='red', lw=4, label='Garganta (Throat)')

ax.set_title("Agujero de Gusano (Puente Atravesable de Morris-Thorne)")
ax.set_zlabel("Coordenada de Incrustación Z")
ax.legend()
plt.axis("off")
plt.show()

```
