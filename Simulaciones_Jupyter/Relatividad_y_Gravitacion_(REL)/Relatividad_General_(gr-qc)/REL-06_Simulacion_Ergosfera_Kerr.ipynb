# [REL-06] Relatividad Extrema: Agujero Negro de Kerr

```python
import numpy as np
import matplotlib.pyplot as plt

# Relatividad General Avanzada: Agujero Negro Rotatorio (Métrica de Kerr)
# Visualizamos la región límite estacionaria exterior y el horizonte de eventos.
# La región entre ambos es la "Ergósfera", donde opera el "Proceso de Penrose".

G = 1.0
M = 1.0
c = 1.0

# Parámetro de spin del agujero negro de Kerr (a = J/M)
# Métrica extrema (máxima rotación posible a=M)
a = 0.99 * M 

# Coordenadas polares
theta = np.linspace(0, 2*np.pi, 200)

# 1. Horizonte de Eventos Exterior (Event Horizon)
# r_H = G*M/c^2 + sqrt((G*M/c^2)^2 - a^2)
r_horizon = M + np.sqrt(M**2 - a**2)

# 2. Límite Estático (Ergosuperficie)
# r_E(theta) = G*M/c^2 + sqrt((G*M/c^2)^2 - a^2 * cos^2(theta))
r_ergosphere = M + np.sqrt(M**2 - (a * np.cos(theta))**2)

# Conversión a coordenadas cartesianas para ploteo
x_H = r_horizon * np.cos(theta)
y_H = r_horizon * np.sin(theta)

x_E = r_ergosphere * np.cos(theta)
y_E = r_ergosphere * np.sin(theta)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)

# Pintamos el Agujero Negro
ax.fill(x_H, y_H, color='black', label="Horizonte de Eventos (Punto de no retorno)")

# Pintamos la Ergósfera
ax.fill(x_E, y_E, color='orange', alpha=0.3, label="Ergósfera (Proceso de Penrose)")
ax.plot(x_E, y_E, color='red', lw=2, linestyle='--')

ax.set_aspect('equal')
ax.set_title("Métrica de Kerr (Agujero Negro en Rotación Extrema $a \approx M$)", fontsize=14)
ax.set_xlabel("x (Coordenadas de Boyer-Lindquist)")
ax.set_ylabel("y")
ax.legend()
plt.show()

# FÍSICA: En la zona naranja (Ergósfera), el "Arrastre del Espacio-Tiempo" 
# es superior a la velocidad de la luz. Estás obligado a girar con el Agujero Negro.
# Penrose demostró matemáticamente que puedes entrar en esta zona con tu nave,
# soltar carga hacia adentro en contra del giro, y la física gravitacional "empujará"
# tu nave hacia afuera a una velocidad tremenda. Acabas de "robarle" masa-energía 
# al monstruo.

```
