# Simetría de Gauge U(1)

Invarianza de fase electromagnética y transformación local de campos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
psi_original = np.exp(-x**2) * np.exp(1j * 5 * x)
fase_gauge = np.exp(1j * np.sin(x)) # Transformación local U(1)
psi_transformada = psi_original * fase_gauge

plt.figure(figsize=(10,4))
plt.plot(x, np.real(psi_original), 'b-', label='Re(Ψ) Original')
plt.plot(x, np.real(psi_transformada), 'r--', label='Re(Ψ) Transformada bajo U(1)')
plt.title('Transformación de Gauge Local U(1)')
plt.legend()
plt.grid(True)
plt.show()
```
