# Variedad Topológica: El Toro

Construcción paramétrica de un toro $T^2$, variedad fundamental en teorías con condiciones de contorno periódicas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(0, 2*np.pi, 100)
U, V = np.meshgrid(u, v)

R = 3
r = 1
X = (R + r*np.cos(V)) * np.cos(U)
Y = (R + r*np.cos(V)) * np.sin(U)
Z = r * np.sin(V)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='twilight', edgecolor='none')
plt.title('Variedad Topológica $T^2$ (Toro)')
plt.axis('off')
plt.show()
```
