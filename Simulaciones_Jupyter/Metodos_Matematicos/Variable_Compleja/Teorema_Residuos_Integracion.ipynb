# Teorema de los Residuos

Visualización de polos en el plano complejo para evaluación de integrales.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

# Función con polos en i y -i: f(z) = 1 / (z^2 + 1)
W = 1 / (Z**2 + 1)

plt.figure(figsize=(8,8))
plt.contourf(X, Y, np.abs(W), levels=np.linspace(0, 5, 50), cmap='magma', extend='max')
plt.scatter([0, 0], [1, -1], color='cyan', marker='x', s=200, label='Polos (Residuos)')
plt.title('Módulo Analítico y Localización de Polos en el Plano Complejo')
plt.legend()
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```
