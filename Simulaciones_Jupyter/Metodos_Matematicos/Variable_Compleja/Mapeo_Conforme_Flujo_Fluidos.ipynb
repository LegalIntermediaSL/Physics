# Transformación de Joukowsky

Uso de variable compleja para mapear cilindros en perfiles alares aerodinámicos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Círculo desplazado en el plano complejo
theta = np.linspace(0, 2*np.pi, 200)
z = (1.1 * np.exp(1j * theta)) - 0.1 + 0.1j

# Transformación de Joukowsky w = z + 1/z
w = z + 1/z

plt.figure(figsize=(8,8))
plt.plot(np.real(z), np.imag(z), 'b--', label='Cilindro Original')
plt.plot(np.real(w), np.imag(w), 'r-', lw=2, label='Perfil Alar Transformado')
plt.title('Mapeo Conforme de Joukowsky')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```
