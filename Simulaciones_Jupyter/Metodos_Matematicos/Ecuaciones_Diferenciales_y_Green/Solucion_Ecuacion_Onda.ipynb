# Solución de D'Alembert

Propagación de una onda viajera en 1D resuelta mediante el formalismo diferencial.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
c = 1.0 # Velocidad de la onda

def pulso(x): return np.exp(-x**2)

plt.figure(figsize=(8,5))
for t in [0, 2, 4, 6]:
    # Solución f(x-ct) + g(x+ct)
    u = 0.5 * (pulso(x - c*t) + pulso(x + c*t))
    plt.plot(x, u, label=f't={t}')

plt.title("Propagación de Pulso (Ecuación de Onda 1D)")
plt.legend()
plt.grid(True)
plt.show()
```
