# Armónicos Esféricos 3D

Soluciones angulares de la ecuación de Laplace y orbitales del átomo de hidrógeno.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm
from mpl_toolkits.mplot3d import Axes3D

phi = np.linspace(0, np.pi, 100)
theta = np.linspace(0, 2*np.pi, 100)
phi, theta = np.meshgrid(phi, theta)

# Armónico Y_2^0
Y = sph_harm(0, 2, theta, phi)
r = np.abs(Y)

x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='orange', edgecolor='k', lw=0.2)
plt.title('Armónico Esférico $|Y_2^0|$ (Orbital d_{z^2})')
plt.axis('off')
plt.show()
```
