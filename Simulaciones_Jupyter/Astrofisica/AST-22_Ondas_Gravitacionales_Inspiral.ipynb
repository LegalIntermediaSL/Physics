# Radiación Gravitacional de Binaria (Fórmula Cuadrupolar)

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 1000)
t_c = 10.5 # Tiempo de coalescencia
f_gw = 10 * (t_c - t)**(-3/8)  # Chirp frequency evolution
h_plus = (f_gw**(2/3)) * np.cos(2*np.pi*np.cumsum(f_gw)*(t[1]-t[0]))

plt.figure(figsize=(10,4))
plt.plot(t, h_plus, color='indigo')
plt.title('Strain de Onda Gravitacional $h_+$ (Inspiral phase)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```
