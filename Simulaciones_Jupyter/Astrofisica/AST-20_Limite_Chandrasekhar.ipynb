# Límite de Chandrasekhar

Presión de degeneración electrónica frente al colapso gravitatorio.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

M = np.linspace(0.1, 1.44, 100)
R = (M**(-1/3)) * (1 - (M/1.44)**(4/3))**0.5

plt.figure(figsize=(8,5))
plt.plot(M, R, 'darkred', lw=2)
plt.axvline(1.44, color='k', linestyle='--', label='Masa de Chandrasekhar (1.44 $M_{\odot}$)')
plt.title('Radio de una Enana Blanca vs Masa')
plt.xlabel('Masa')
plt.ylabel('Radio')
plt.legend()
plt.grid()
plt.show()
```
