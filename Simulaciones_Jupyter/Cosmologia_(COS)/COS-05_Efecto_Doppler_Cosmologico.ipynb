# Doppler vs Corrimiento al Rojo Cosmológico

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v_c = np.linspace(0, 0.9, 100)
z_doppler = np.sqrt((1+v_c)/(1-v_c)) - 1
z_cosmo = v_c # Lineal aproximacion a bajo redshift de Hubble

plt.plot(v_c, z_doppler, label='Doppler Relativista Específico')
plt.plot(v_c, z_cosmo, label='Ley Hubble Lineal', linestyle='--')
plt.title('Corrimiento al rojo local vs Cosmológico')
plt.xlabel('Velocidad recesional $v/c$')
plt.ylabel('Redshift $z$')
plt.legend(); plt.grid(True)
plt.show()
```
