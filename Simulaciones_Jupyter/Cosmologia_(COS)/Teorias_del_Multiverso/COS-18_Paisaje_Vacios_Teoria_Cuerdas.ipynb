# El Paisaje (Landscape) de la Teoría de Cuerdas

Renderizado 3D de una función de potencial escalar compleja $V(\phi_1, \phi_2)$ para ilustrar los $10^{500}$ posibles estados de vacío en los que pueden 'caer' diferentes universos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

phi1 = np.linspace(-3, 3, 100)
phi2 = np.linspace(-3, 3, 100)
Phi1, Phi2 = np.meshgrid(phi1, phi2)

# Función topológica con múltiples mínimos (simulando los vacíos de Calabi-Yau)
V = np.sin(3*Phi1) * np.cos(3*Phi2) + 0.5*(Phi1**2 + Phi2**2) - np.cos(5*(Phi1+Phi2))*0.3

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(Phi1, Phi2, V, cmap='terrain', edgecolor='none', alpha=0.9)

# Simulación de un universo "cayendo" a un vacío local aleatorio
u_x, u_y = 1.0, -1.0 # Punto inicial
path_x, path_y, path_z = [u_x], [u_y], [np.sin(3*u_x)*np.cos(3*u_y) + 0.5*(u_x**2 + u_y**2)]
ax.scatter(path_x, path_y, path_z, color='red', s=100, label='Universo Local (Ajuste Fino)')

plt.title('El Paisaje de la Teoría de Cuerdas ($10^{500}$ vacíos posibles)')
plt.legend()
plt.axis('off')
plt.show()
```
