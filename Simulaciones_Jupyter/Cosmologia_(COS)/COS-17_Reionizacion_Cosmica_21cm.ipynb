# Cosmología de 21cm y Reionización

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(6, 15, 100)
x_HII = 0.5 * (1 - np.tanh((z - 9.0)/1.5)) # Fraccion de ionizacion
delta_T = 27 * (1 - x_HII) * np.sqrt((1+z)/10) # Señal 21cm

plt.plot(z, delta_T, color='red')
plt.title('Señal global a 21cm durante la Época de Reionización (EoR)')
plt.xlabel('Redshift $z$')
plt.ylabel('Temperatura de brillo $\delta T_b$ (mK)')
plt.grid(True)
plt.show()
```
