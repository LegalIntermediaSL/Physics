# Red Cósmica y Aproximación de Zel'dovich

Colapso gravitacional de perturbaciones de densidad iniciales para formar filamentos de materia oscura.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 500
q_x = np.random.uniform(0, 10, N)
q_y = np.random.uniform(0, 10, N)

# Perturbación inicial gaussiana (Campo de velocidades)
psi_x = np.sin(q_x) * np.cos(q_y)
psi_y = np.cos(q_x) * np.sin(q_y)

# Posición euleriana x(t) = q + D(t)*psi(q)
D = 1.5 # Factor de crecimiento
x = q_x + D * psi_x
y = q_y + D * psi_y

plt.figure(figsize=(6,6))
plt.scatter(x, y, s=2, color='k', alpha=0.6)
plt.title('Formación de Estructuras: Aproximación de Zel\'dovich')
plt.axis('equal')
plt.show()
```
