# [COS-07] Evolución en la CCC

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación Conceptual: Evolución de Escala en Cosmología Cíclica Conforme
# Unimos el a(t) -> infinito de un eón con el a(t) -> 0 del siguiente.

eons = 3
t_per_eon = np.linspace(0.01, 10, 500)

plt.figure(figsize=(12, 5))

for n in range(eons):
    # Expansión de Friedmann a(t) ~ t^(2/3) seguida de a(t) ~ e^(H t)
    # Factor de escala físico
    a_t = t_per_eon**(2/3) + 0.05 * np.exp(t_per_eon)
    
    # Rescalado conforme Omega (simplificado para que empate)
    # Omega decrece hacia 0 al final del eón
    Omega = 1.0 / np.sqrt(a_t)
    
    # Métrica conforme efectiva: g_hat = Omega^2 g_fisica
    a_conformal = a_t * Omega
    
    time_offset = n * 10
    
    plt.plot(t_per_eon + time_offset, a_t, 'b--', alpha=0.3)
    plt.plot(t_per_eon + time_offset, a_conformal, 'r-', lw=2)
    plt.axvline(time_offset, color='k', linestyle=':')

plt.title("Cosmología Cíclica Conforme (CCC): Isomorfismo entre Eones")
plt.xlabel("Tiempo Cósmico")
plt.ylabel("Factor de Escala")
plt.plot([], [], 'b--', label='Factor de Escala Físico a(t)')
plt.plot([], [], 'r-', label='Factor de Escala Conforme $\hat{a}(t)$')
plt.yscale('log')
plt.legend()
plt.grid(True)
plt.show()

```
