# Supercomputación: Fisica Atomica HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Dicke Superradiance toy model (Master equation for N two-level atoms)
# Simulating the decay of a fully inverted state of N atoms
N_atoms = 50
max_J = N_atoms / 2.0
M_states = np.arange(-max_J, max_J + 1)
dim = len(M_states)
rho = np.zeros(dim)
rho[-1] = 1.0  # Fully inverted state

gamma = 1.0
dt = 0.01
steps = 500

intensity = []
for t in range(steps):
    drho = np.zeros(dim)
    for i in range(dim - 1):
        m = M_states[i+1]
        rate = gamma * (max_J + m) * (max_J - m + 1)
        drho[i] += rate * rho[i+1]
        drho[i+1] -= rate * rho[i+1]
    
    rho += drho * dt
    
    I_t = 0
    for i in range(1, dim):
        m = M_states[i]
        rate = gamma * (max_J + m) * (max_J - m + 1)
        I_t += rate * rho[i]
    intensity.append(I_t)

plt.plot(intensity)
plt.title('Dicke Superradiance Pulse')
plt.savefig('superradiance.png')
```
