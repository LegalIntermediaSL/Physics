# Compendio Subtemático: Fisica Atomica Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rabi_oscillations(omega_r, detuning):
    # Generalized Rabi frequency
    Omega = np.sqrt(omega_r**2 + detuning**2)
    
    t = np.linspace(0, 5 * (2*np.pi/omega_r), 500)
    
    # Probability of being in excited state
    P_e = (omega_r**2 / Omega**2) * np.sin(Omega * t / 2)**2
    
    plt.figure(figsize=(8, 6))
    plt.plot(t, P_e, 'b-', label=f'$\Omega_R$={omega_r}, $\Delta$={detuning}')
    plt.xlabel('Time')
    plt.ylabel('Excited State Probability $P_e(t)$')
    plt.title('Rabi Oscillations in a Two-Level Atom')
    plt.ylim(0, 1.05)
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    rabi_oscillations(1.0, 0.0) # On resonance
    rabi_oscillations(1.0, 0.5) # Off resonance
```
