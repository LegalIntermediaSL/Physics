# Compendio Subtemático: Fisica Atomica Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def stark_effect_n2(E_field):
    # n=2 level splitting in Hydrogen
    # First order Stark effect splits n=2 into 3 levels
    # Energy shift Delta E = 3 e a0 E_field
    # in atomic units e=1, a0=1, so Delta E = 3 * E_field (in Hartrees)
    # let's just plot the proportional splitting
    
    E_fields = np.linspace(0, E_field, 100)
    
    E0 = 0 # unperturbed energy of n=2 relative to itself
    E_plus = E0 + 3 * E_fields
    E_minus = E0 - 3 * E_fields
    E_zero = np.zeros_like(E_fields)
    
    plt.figure(figsize=(8, 6))
    plt.plot(E_fields, E_plus, 'r', label='$|2, 0, 0\\rangle - |2, 1, 0\\rangle$')
    plt.plot(E_fields, E_minus, 'b', label='$|2, 0, 0\\rangle + |2, 1, 0\\rangle$')
    plt.plot(E_fields, E_zero, 'g', label='m = ±1 (No 1st order shift)')
    
    plt.xlabel('Electric Field E (atomic units)')
    plt.ylabel('Energy Shift $\Delta E$ (atomic units)')
    plt.title('Linear Stark Effect in Hydrogen (n=2)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    stark_effect_n2(0.1)
```
