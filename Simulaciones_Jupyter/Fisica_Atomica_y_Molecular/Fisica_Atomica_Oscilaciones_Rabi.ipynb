# Investigación Avanzada: Fisica Atomica Oscilaciones Rabi

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def rabi_oscillations():
    """Simulates Rabi oscillations in a two-level atom."""
    Omega = 2.0 * np.pi # Rabi frequency
    Delta = 0.5 * np.pi # Detuning
    
    def bloch_equations(t, y):
        # y = [u, v, w]
        u, v, w = y
        du = Delta * v
        dv = -Delta * u + Omega * w
        dw = -Omega * v
        return [du, dv, dw]
    
    t_span = (0, 5)
    t_eval = np.linspace(t_span[0], t_span[1], 1000)
    y0 = [0, 0, -1] # Start in ground state
    
    sol = solve_ivp(bloch_equations, t_span, y0, t_eval=t_eval)
    
    plt.figure(figsize=(10,6))
    plt.plot(sol.t, sol.y[2], label='Population Inversion ($w$)')
    plt.plot(sol.t, sol.y[0], label='Coherence ($u$)')
    plt.plot(sol.t, sol.y[1], label='Coherence ($v$)')
    plt.xlabel('Time $t$')
    plt.ylabel('Bloch Vector Components')
    plt.title('Rabi Oscillations (Two-Level System)')
    plt.legend()
    plt.grid()
    plt.savefig('Rabi_Oscillations.png')

if __name__ == '__main__':
    rabi_oscillations()
```
