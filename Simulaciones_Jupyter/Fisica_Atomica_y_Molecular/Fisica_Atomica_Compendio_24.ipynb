# Compendio Subtemático: Fisica Atomica Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def rate_equations(y, t, W, tau, gamma):
    N1, N2, N3 = y
    # N1: ground state, N2: lasing state, N3: pump state
    # W: pumping rate N1 -> N3
    # tau: lifetime of N3 -> N2 (fast non-radiative decay)
    # gamma: spontaneous emission N2 -> N1
    
    dN1_dt = -W * N1 + gamma * N2
    dN2_dt = (1/tau) * N3 - gamma * N2
    dN3_dt = W * N1 - (1/tau) * N3
    
    return [dN1_dt, dN2_dt, dN3_dt]

def simulate_laser():
    # parameters
    W = 10.0
    tau = 0.1
    gamma = 1.0
    
    y0 = [100.0, 0.0, 0.0] # start all in ground state
    t = np.linspace(0, 5, 200)
    
    sol = odeint(rate_equations, y0, t, args=(W, tau, gamma))
    
    plt.figure(figsize=(8, 6))
    plt.plot(t, sol[:, 0], label='N1 (Ground)')
    plt.plot(t, sol[:, 1], label='N2 (Lasing)')
    plt.plot(t, sol[:, 2], label='N3 (Pump)')
    
    plt.xlabel('Time')
    plt.ylabel('Population')
    plt.title('Three-Level Laser Rate Equations')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    simulate_laser()
```
