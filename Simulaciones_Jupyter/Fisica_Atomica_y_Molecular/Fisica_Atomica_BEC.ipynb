# Investigación Avanzada: Fisica Atomica BEC

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def gross_pitaevskii():
    """1D Gross-Pitaevskii Equation for Bose-Einstein Condensate."""
    # Split-step Fourier method or finite difference for ground state.
    # Using imaginary time evolution.
    N = 512
    L = 20.0
    x = np.linspace(-L/2, L/2, N)
    dx = x[1] - x[0]
    
    # Potential: Harmonic oscillator
    V = 0.5 * x**2
    
    # Initial guess: Gaussian
    psi = np.exp(-x**2 / 2) / np.pi**0.25
    
    # Kinetic energy operator in Fourier space
    k = 2 * np.pi * np.fft.fftfreq(N, d=dx)
    T = k**2 / 2
    
    dt = 0.01
    g = 10.0 # Interaction strength
    
    for _ in range(1000):
        # Half step potential
        psi *= np.exp(-dt * (V + g * np.abs(psi)**2) / 2)
        # Full step kinetic
        psi_k = np.fft.fft(psi)
        psi_k *= np.exp(-dt * T)
        psi = np.fft.ifft(psi_k)
        # Half step potential
        psi *= np.exp(-dt * (V + g * np.abs(psi)**2) / 2)
        # Normalize
        psi /= np.sqrt(np.sum(np.abs(psi)**2) * dx)
        
    plt.figure(figsize=(8,6))
    plt.plot(x, np.abs(psi)**2, label='BEC Density ($|\psi|^2$)')
    plt.plot(x, V, '--', label='Harmonic Trap $V(x)$')
    plt.ylim(0, 1)
    plt.xlim(-5, 5)
    plt.xlabel('Position $x$')
    plt.ylabel('Density / Potential')
    plt.title('Bose-Einstein Condensate Ground State (GPE)')
    plt.legend()
    plt.grid()
    plt.savefig('BEC_GroundState.png')

if __name__ == '__main__':
    gross_pitaevskii()
```
