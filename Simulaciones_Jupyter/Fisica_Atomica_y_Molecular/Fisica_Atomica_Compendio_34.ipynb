# Compendio Subtemático: Fisica Atomica Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def relativistic_correction(n, l):
    # Relativistic correction to kinetic energy for Hydrogen
    # E_rel = E_n * (alpha^2 / n^2) * (n / (l + 1/2) - 3/4)
    alpha = 1 / 137.036
    E_n = -13.6 / (n**2)
    return E_n * (alpha**2 / n**2) * (n / (l + 0.5) - 0.75)

def plot_corrections():
    n = 2
    l_values = [0, 1] # s and p
    
    plt.figure(figsize=(6, 6))
    plt.hlines(0, 0, 2, colors='k', linestyles='dashed', label='Non-relativistic n=2')
    
    for l in l_values:
        shift = relativistic_correction(n, l)
        plt.hlines(shift * 1e4, 0.5 + l*0.5, 1.0 + l*0.5, label=f'l={l} (x$10^4$)')
        plt.text(0.75 + l*0.5, shift * 1e4 + 0.05, f'l={l}', ha='center')
        
    plt.ylabel('Energy shift (x$10^{-4}$ eV)')
    plt.title('Relativistic Kinetic Energy Correction (n=2)')
    plt.xticks([])
    plt.legend()
    plt.show()

if __name__ == "__main__":
    plot_corrections()
```
