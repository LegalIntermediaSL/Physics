# Supercomputación: Fisica Atomica HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
import matplotlib.pyplot as plt

# Transverse Field Ising Model - Exact Diagonalization
N = 10
J = 1.0
h = 0.5

# Build Hamiltonian
sx = sp.csr_matrix([[0, 1], [1, 0]])
sz = sp.csr_matrix([[1, 0], [0, -1]])
I = sp.csr_matrix([[1, 0], [0, 1]])

def kron_list(op_list):
    res = op_list[0]
    for op in op_list[1:]:
        res = sp.kron(res, op)
    return res

H = sp.csr_matrix((2**N, 2**N))
for i in range(N):
    ops = [I]*N
    ops[i] = sx
    H -= h * kron_list(ops)
    
for i in range(N-1):
    ops = [I]*N
    ops[i] = sz
    ops[i+1] = sz
    H -= J * kron_list(ops)

evals, evecs = spla.eigsh(H, k=1, which='SA')
plt.bar([0], evals[0])
plt.title('Ground State Energy of TFIM')
plt.savefig('tfim_energy.png')
```
