Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import matplotlib.pyplot as plt

def lamb_shift_diagram():
    # Diagram showing Lamb shift splitting of 2S_1/2 and 2P_1/2
    # In Dirac theory, 2S_1/2 and 2P_1/2 are degenerate
    # QED shifts 2S_1/2 up by ~1057 MHz
    
    plt.figure(figsize=(6, 6))
    
    # Dirac levels
    plt.hlines(0, 0, 1, color='k', linestyles='dashed', label='Dirac Theory (Degenerate)')
    plt.text(0.5, 0.1, '$2S_{1/2}, 2P_{1/2}$', ha='center')
    
    # QED levels
    plt.hlines(1.057, 1.5, 2.5, color='r', label='QED (Lamb Shift)')
    plt.text(2.0, 1.15, '$2S_{1/2}$', ha='center')
    
    plt.hlines(0, 1.5, 2.5, color='b')
    plt.text(2.0, -0.1, '$2P_{1/2}$', ha='center')
    
    # Arrow for shift
    plt.annotate('', xy=(2.0, 1.057), xytext=(2.0, 0), arrowprops=dict(arrowstyle='<->'))
    plt.text(2.1, 0.5, '1057 MHz')
    
    plt.xlim(0, 3)
    plt.ylim(-0.5, 1.5)
    plt.ylabel('Energy shift (GHz)')
    plt.title('Lamb Shift in Hydrogen (n=2)')
    plt.xticks([])
    plt.legend()
    plt.show()

if __name__ == "__main__":
    lamb_shift_diagram()
```
