# Compendio Subtemático: Fisica Atomica Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def compton_scattering():
    lambda_c = 2.426e-12 # Compton wavelength of electron in m
    theta = np.linspace(0, np.pi, 100)
    
    delta_lambda = lambda_c * (1 - np.cos(theta))
    
    plt.figure(figsize=(8, 6))
    plt.plot(np.degrees(theta), delta_lambda * 1e12, 'b-', linewidth=2)
    plt.xlabel('Scattering Angle $\\theta$ (degrees)')
    plt.ylabel('Wavelength Shift $\Delta\lambda$ (pm)')
    plt.title('Compton Scattering Wavelength Shift')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    compton_scattering()
```
