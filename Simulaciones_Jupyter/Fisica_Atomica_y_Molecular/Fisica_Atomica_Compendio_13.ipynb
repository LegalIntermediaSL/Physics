# Compendio Subtemático: Fisica Atomica Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def stern_gerlach():
    # Simulate Ag atoms (L=0, S=1/2) passing through inhomogeneous B field
    N = 1000
    # initial positions
    z = np.random.normal(0, 0.1, N)
    y = np.random.normal(0, 0.1, N)
    
    # Spin 1/2 means two states
    spins = np.random.choice([-1, 1], N)
    
    # Deflection proportional to spin
    deflection = spins * 2.0
    
    z_final = z + deflection
    
    plt.figure(figsize=(8, 6))
    plt.scatter(y, z_final, c=spins, cmap='bwr', alpha=0.5)
    plt.xlabel('y position (mm)')
    plt.ylabel('z position (mm)')
    plt.title('Stern-Gerlach Experiment: Detection Screen')
    plt.yticks([-2, 0, 2], ['Spin Down', '0', 'Spin Up'])
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    stern_gerlach()
```
