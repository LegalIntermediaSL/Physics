# Compendio Subtemático: Fisica Atomica Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fine_structure(n, j):
    # Energy correction in eV for Hydrogen
    # E_fs = E_n * (alpha^2 / n^2) * (n / (j + 1/2) - 3/4)
    alpha = 1 / 137.036
    E_n = -13.6 / (n**2)
    return E_n * (alpha**2 / n**2) * (n / (j + 0.5) - 0.75)

def plot_fine_structure():
    n = 2
    j_values = [0.5, 1.5] # For l=0 (j=1/2), l=1 (j=1/2, 3/2)
    
    E_n = -13.6 / (n**2)
    
    plt.figure(figsize=(6, 8))
    plt.hlines(0, 0, 2, linestyles='dashed', colors='k', label='Unperturbed n=2')
    
    for j in j_values:
        shift = fine_structure(n, j)
        # multiply by 1e5 for visibility
        plt.hlines(shift * 1e5, 0.5, 1.5, label=f'j={j} (x$10^5$)')
        plt.text(1.55, shift * 1e5, f'j={j}')
        
    plt.ylabel('Energy Shift (x$10^{-5}$ eV)')
    plt.title('Fine Structure Splitting of Hydrogen (n=2)')
    plt.xticks([])
    plt.legend()
    plt.show()

if __name__ == "__main__":
    plot_fine_structure()
```
