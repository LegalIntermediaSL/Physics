Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def zeeman_effect():
    B = np.linspace(0, 5, 200) # Magnetic field in Tesla
    mu_B = 5.788e-5 # Bohr magneton in eV/T
    
    # Simulating normal Zeeman effect (e.g. spin zero transition)
    E_0 = 2.1 # Transition energy in eV
    
    delta_E = mu_B * B
    
    plt.figure(figsize=(8, 5))
    plt.plot(B, E_0 + delta_E, label='m = +1')
    plt.plot(B, np.full_like(B, E_0), label='m = 0')
    plt.plot(B, E_0 - delta_E, label='m = -1')
    
    plt.title("Normal Zeeman Effect")
    plt.xlabel("Magnetic Field B (T)")
    plt.ylabel("Energy Level (eV)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    zeeman_effect()
```
