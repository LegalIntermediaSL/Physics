# Compendio Subtemático: Fisica Atomica Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def quantum_defect(n, delta_l):
    # Energy levels for alkali atoms using quantum defect delta_l
    # E = - R_H / (n - delta_l)^2
    R_H = 13.6 # eV
    return -R_H / (n - delta_l)**2

def plot_quantum_defects():
    n_values = np.arange(3, 8)
    
    # Sodium quantum defects approximately:
    defects = {'s (l=0)': 1.37, 'p (l=1)': 0.88, 'd (l=2)': 0.01}
    colors = ['b', 'r', 'g']
    
    plt.figure(figsize=(8, 6))
    
    for (label, delta), c in zip(defects.items(), colors):
        E = quantum_defect(n_values, delta)
        plt.plot(n_values, E, marker='o', color=c, label=f'{label}, $\delta$={delta}')
        
    # Hydrogen reference (delta = 0)
    plt.plot(n_values, -13.6/n_values**2, 'k--', label='Hydrogen')
    
    plt.xlabel('Principal Quantum Number $n$')
    plt.ylabel('Energy (eV)')
    plt.title('Energy Levels of Sodium (Quantum Defect Model)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_quantum_defects()
```
