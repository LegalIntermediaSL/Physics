# Compendio Subtemático: Fisica Atomica Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def hyperfine_structure(F):
    # A is the hyperfine structure constant
    # Delta E = A/2 * (F(F+1) - I(I+1) - J(J+1))
    # For Hydrogen ground state, 1s1/2: J=1/2, I=1/2 (proton spin)
    # F can be 0 or 1
    J = 0.5
    I = 0.5
    A = 5.87e-6 # eV (approx 1420 MHz)
    return (A / 2) * (F*(F+1) - I*(I+1) - J*(J+1))

def plot_hyperfine():
    F_values = [0, 1]
    
    plt.figure(figsize=(5, 6))
    plt.hlines(0, 0, 1, linestyles='dashed', colors='k', label='Fine structure level')
    
    for F in F_values:
        shift = hyperfine_structure(F)
        plt.hlines(shift, 0.2, 0.8)
        plt.text(0.85, shift, f'F={F}')
        
    delta_E = hyperfine_structure(1) - hyperfine_structure(0)
    freq = delta_E * 2.418e14 # convert eV to Hz
    
    plt.annotate(f'21 cm line\n{freq*1e-6:.0f} MHz', 
                 xy=(0.5, hyperfine_structure(0)), xycoords='data',
                 xytext=(0.5, hyperfine_structure(1)), textcoords='data',
                 arrowprops=dict(arrowstyle="<->", connectionstyle="arc3"))
                 
    plt.title('Hyperfine Splitting of Hydrogen 1s')
    plt.ylabel('Energy Shift (eV)')
    plt.xticks([])
    plt.show()

if __name__ == "__main__":
    plot_hyperfine()
```
