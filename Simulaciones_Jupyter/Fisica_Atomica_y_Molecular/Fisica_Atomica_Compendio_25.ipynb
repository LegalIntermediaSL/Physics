# Compendio Subtemático: Fisica Atomica Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def helium_variational():
    # Variational method for Helium ground state
    # Trial wavefunction with effective nuclear charge Z'
    # E(Z') = Z'^2 - 2 Z Z' + (5/8) Z' (in atomic units / Hartrees)
    # For Helium Z = 2
    
    Z = 2
    Z_eff = np.linspace(1, 3, 100)
    
    E = Z_eff**2 - 2 * Z * Z_eff + (5/8) * Z_eff
    
    # Minimum occurs at Z' = Z - 5/16
    Z_min = Z - 5/16
    E_min = Z_min**2 - 2 * Z * Z_min + (5/8) * Z_min
    
    # Experimental energy approx -2.903 Hartrees
    
    plt.figure(figsize=(8, 6))
    plt.plot(Z_eff, E, 'b-', label='Variational Energy $E(Z\')$')
    plt.plot(Z_min, E_min, 'ro', label=f'Minimum at Z\'={Z_min:.3f}\nE={E_min:.3f} a.u.')
    
    plt.axhline(-2.903, color='g', linestyle='--', label='Experimental E = -2.903 a.u.')
    
    plt.xlabel("Effective Nuclear Charge Z'")
    plt.ylabel('Energy (Hartrees)')
    plt.title('Variational Method for Helium Ground State')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    helium_variational()
```
