# Compendio Subtemático: Fisica Atomica Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import matplotlib.pyplot as plt

def selection_rules():
    # Electric dipole selection rules: Delta l = +-1, Delta m = 0, +-1
    
    levels = {
        's': 0,
        'p': 1,
        'd': 2,
        'f': 3
    }
    
    plt.figure(figsize=(8, 6))
    for i, (name, l) in enumerate(levels.items()):
        plt.hlines(i, l - 0.2, l + 0.2, colors='k')
        plt.text(l, i + 0.1, f'{name}-state', ha='center')
        
    # Draw allowed transitions
    # s to p
    plt.annotate('', xy=(1, 1), xytext=(0, 0), arrowprops=dict(arrowstyle='<->', color='b'))
    # p to d
    plt.annotate('', xy=(2, 2), xytext=(1, 1), arrowprops=dict(arrowstyle='<->', color='g'))
    # d to f
    plt.annotate('', xy=(3, 3), xytext=(2, 2), arrowprops=dict(arrowstyle='<->', color='r'))
    # p to s (already drawn by s to p)
    
    plt.xlim(-1, 4)
    plt.ylim(-1, 4)
    plt.xticks(list(levels.values()), list(levels.keys()))
    plt.xlabel('Orbital angular momentum $l$')
    plt.ylabel('Energy Level (arbitrary)')
    plt.title('Allowed Electric Dipole Transitions ($\Delta l = \pm 1$)')
    plt.grid(axis='x')
    plt.show()

if __name__ == "__main__":
    selection_rules()
```
