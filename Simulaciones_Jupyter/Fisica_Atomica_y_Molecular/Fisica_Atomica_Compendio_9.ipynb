# Compendio Subtemático: Fisica Atomica Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import itertools
import matplotlib.pyplot as plt

def ls_coupling_terms(l1, l2, s1=0.5, s2=0.5):
    # Calculate all possible L and S values
    L_values = list(range(abs(l1 - l2), l1 + l2 + 1))
    S_values = list(range(int(abs(s1 - s2)), int(s1 + s2 + 1)))
    
    terms = []
    for L in L_values:
        for S in S_values:
            J_values = list(range(abs(L - S), L + S + 1))
            for J in J_values:
                terms.append((S, L, J))
    return terms

def L_to_letter(L):
    letters = "SPDFGHIKLMNOQRTU"
    return letters[L] if L < len(letters) else str(L)

def plot_terms(l1, l2):
    terms = ls_coupling_terms(l1, l2)
    
    plt.figure(figsize=(10, 6))
    
    # Just space them out arbitrarily for visualization
    x = 0
    for S, L, J in terms:
        multiplicity = int(2*S + 1)
        term_symbol = f"$^{{{multiplicity}}}{L_to_letter(L)}_{{{J}}}$"
        
        # approximate energy order (Hund's rules naive)
        energy = -S * 10 + L * 2 + J * 0.5
        
        plt.plot(x, energy, 'ro', markersize=10)
        plt.text(x+0.1, energy, term_symbol, fontsize=12)
        x += 1
        
    plt.title(f'LS Coupling Term Symbols for $l_1={l1}, l_2={l2}$')
    plt.ylabel('Relative Energy (arbitrary units)')
    plt.xticks([])
    plt.grid(True, alpha=0.3)
    plt.show()

if __name__ == "__main__":
    plot_terms(1, 1) # p^2 configuration (non-equivalent electrons)
```
