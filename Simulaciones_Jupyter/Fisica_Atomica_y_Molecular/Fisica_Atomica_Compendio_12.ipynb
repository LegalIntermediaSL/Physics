# Compendio Subtemático: Fisica Atomica Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def franck_hertz(V):
    # Simulated current for Franck-Hertz experiment (Hg atoms, E_exc ~ 4.9 eV)
    E_exc = 4.9
    
    # Base current
    I = V**1.5 * 0.05 
    
    # Dips due to inelastic collisions
    for i in range(1, int(max(V)/E_exc) + 1):
        center = i * E_exc
        width = 0.5
        depth = 0.3 * V
        dip = depth * np.exp(-((V - center)**2) / (2 * width**2))
        I = I - dip
        
    I = np.maximum(I, 0)
    return I

def plot_franck_hertz():
    V = np.linspace(0, 30, 500)
    I = franck_hertz(V)
    
    plt.figure(figsize=(8, 6))
    plt.plot(V, I, 'b-', linewidth=2)
    plt.xlabel('Accelerating Voltage (V)')
    plt.ylabel('Current (arbitrary units)')
    plt.title('Franck-Hertz Experiment Simulation (Hg)')
    
    for i in range(1, 7):
        plt.axvline(i * 4.9, color='r', linestyle='--', alpha=0.3)
        
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_franck_hertz()
```
