# Laboratorio Especializado: Fisica Atomica Especializado 5

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def laser_cooling():
    velocity = np.linspace(-10, 10, 500)
    Gamma = 1.0 # Linewidth
    k = 1.0 # Wavenumber
    Delta = -1.0 * Gamma # Detuning
    
    # Radiation force from two counter-propagating beams
    def scattering_rate(v, delta, k, Gamma):
        return (Gamma / 2) / (1 + 4 * ((delta - k * v) / Gamma)**2)
    
    F_plus = scattering_rate(velocity, Delta, k, Gamma)
    F_minus = -scattering_rate(velocity, Delta, -k, Gamma)
    F_total = F_plus + F_minus
    
    plt.figure(figsize=(8, 5))
    plt.plot(velocity, F_plus, '--', label='Force from Right beam')
    plt.plot(velocity, F_minus, '--', label='Force from Left beam')
    plt.plot(velocity, F_total, 'k-', linewidth=2, label='Total Force')
    plt.title("1D Optical Molasses (Laser Cooling Force)")
    plt.xlabel("Velocity (v)")
    plt.ylabel("Force (Arbitrary Units)")
    plt.axvline(0, color='gray', linestyle=':')
    plt.axhline(0, color='gray', linestyle=':')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    laser_cooling()
```
