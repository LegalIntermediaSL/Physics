# Compendio Subtemático: Fisica Atomica Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import matplotlib.pyplot as plt

def hunds_rules(config="p2"):
    # Simplified simulation for plotting Hund's rules concept
    # Just hardcoding a few equivalent electron configs for p orbital
    
    configs = {
        "p1": {"S": 1/2, "L": 1, "J": 1/2, "term": "^2P_{1/2}"},
        "p2": {"S": 1, "L": 1, "J": 0, "term": "^3P_0"},
        "p3": {"S": 3/2, "L": 0, "J": 3/2, "term": "^4S_{3/2}"},
        "p4": {"S": 1, "L": 1, "J": 2, "term": "^3P_2"},
        "p5": {"S": 1/2, "L": 1, "J": 3/2, "term": "^2P_{3/2}"},
        "p6": {"S": 0, "L": 0, "J": 0, "term": "^1S_0"}
    }
    
    if config not in configs:
        return
        
    term = configs[config]["term"]
    
    plt.figure(figsize=(6, 4))
    plt.text(0.5, 0.5, f"Ground State Term Symbol for {config}:\n\n${term}$", 
             fontsize=24, ha='center', va='center')
    plt.axis('off')
    plt.title(f"Hund's Rules Application")
    plt.show()

if __name__ == "__main__":
    hunds_rules("p2")
```
