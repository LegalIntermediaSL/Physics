# Investigación Avanzada: Fisica Atomica Lennard Jones

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def lennard_jones_md():
    """Molecular dynamics simulation of Lennard-Jones gas."""
    N = 50
    L = 10.0
    dt = 0.005
    steps = 1000
    
    # Init pos
    pos = np.random.rand(N, 2) * L
    # Init vel (zero total momentum)
    vel = np.random.randn(N, 2)
    vel -= np.mean(vel, axis=0)
    
    def compute_forces(p):
        forces = np.zeros_like(p)
        potential = 0
        for i in range(N):
            for j in range(i+1, N):
                dp = p[i] - p[j]
                # Periodic boundary conditions
                dp = dp - L * np.round(dp / L)
                r2 = np.sum(dp**2)
                if r2 < 9.0: # Cutoff
                    r6 = r2**3
                    r12 = r6**2
                    f_mag = 24 * (2/r12 - 1/r6) / r2
                    f = f_mag * dp
                    forces[i] += f
                    forces[j] -= f
                    potential += 4 * (1/r12 - 1/r6)
        return forces, potential
    
    f, _ = compute_forces(pos)
    
    energies = []
    
    for _ in range(steps):
        # Velocity Verlet
        pos += vel * dt + 0.5 * f * dt**2
        pos = pos % L
        
        f_new, ep = compute_forces(pos)
        vel += 0.5 * (f + f_new) * dt
        f = f_new
        
        ek = 0.5 * np.sum(vel**2)
        energies.append(ek + ep)
        
    plt.figure(figsize=(10,5))
    plt.subplot(1, 2, 1)
    plt.scatter(pos[:,0], pos[:,1], c='purple')
    plt.xlim(0, L)
    plt.ylim(0, L)
    plt.title('LJ Gas Final State')
    
    plt.subplot(1, 2, 2)
    plt.plot(energies)
    plt.xlabel('Step')
    plt.ylabel('Total Energy')
    plt.title('Energy Conservation')
    
    plt.tight_layout()
    plt.savefig('Lennard_Jones.png')

if __name__ == '__main__':
    lennard_jones_md()
```
