# Simulación Computacional: Elasticidad y Medios Continuos
**Área:** Fluidos y Medios Continuos

Este cuaderno interactivo de Jupyter ejecuta numéricamente los modelos físicos descritos en el repositorio. Puedes modificar los parámetros iniciales y ejecutar las celdas para visualizar gráficos o animaciones interactivos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del material (Acero)
rho = 7800.0   # Densidad kg/m^3
G = 79.3e9     # Módulo de corte (Pa)
c_s = np.sqrt(G / rho) # Velocidad de onda de corte

# Dominio
L = 10.0
nx = 200
dx = L / nx
x = np.linspace(0, L, nx)

dt = 0.8 * dx / c_s # Condición CFL < 1
nt = 300

# Desplazamiento transversal u(x,t)
u = np.zeros(nx)
u_prev = np.zeros(nx)
u_next = np.zeros(nx)

# Perturbación inicial (Pulso Gaussiano)
u_prev = np.exp(-10 * (x - L/2)**2)
u = u_prev.copy()

alpha = (c_s * dt / dx)**2

# Simulación
history = [u.copy()]
for _ in range(nt):
    for i in range(1, nx - 1):
        u_next[i] = 2*u[i] - u_prev[i] + alpha * (u[i+1] - 2*u[i] + u[i-1])
    
    # Condiciones de borde libres
    u_next[0] = u_next[1]
    u_next[-1] = u_next[-2]
    
    u_prev[:] = u[:]
    u[:] = u_next[:]
    if _ % 50 == 0:
        history.append(u.copy())

plt.figure(figsize=(10, 5))
for idx, snap in enumerate(history):
    plt.plot(x, snap, label=f"t={idx*50*dt:.4f}s")
plt.title("Propagación de Onda de Corte en Medio Elástico")
plt.xlabel("Distancia (m)")
plt.ylabel("Desplazamiento transversal u(x)")
plt.legend()
plt.grid(True)
plt.show()
```
