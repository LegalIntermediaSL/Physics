# Flujo de Poiseuille

Perfil de velocidad parabólico de fluidos viscosos newtonianos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

r = np.linspace(-1, 1, 100)
v = 1 - r**2

plt.figure(figsize=(5,6))
plt.plot(v, r, 'k-', lw=2)
plt.fill_betweenx(r, 0, v, color='skyblue', alpha=0.5)
plt.title('Perfil de Velocidades (Tubería Cilíndrica)')
plt.xlabel('Velocidad')
plt.ylabel('Radio')
plt.grid()
plt.show()
```
