# Efecto Halofit en Cosmología de Estructuras

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = np.logspace(-2, 1, 100)
P_lin = k**1.2 * np.exp(-k/0.1)
P_nl = P_lin + 5*(k**2.5)*np.exp(-k/2) # Correccion no lineal ficticia

plt.loglog(k, P_lin, label='Lineal')
plt.loglog(k, P_nl, label='No Lineal (Halofit)')
plt.title('Espectro de Potencias de Materia a redshift $z=0$')
plt.xlabel('$k$')
plt.ylabel('$P(k)$')
plt.legend(); plt.grid(True)
plt.show()
```
