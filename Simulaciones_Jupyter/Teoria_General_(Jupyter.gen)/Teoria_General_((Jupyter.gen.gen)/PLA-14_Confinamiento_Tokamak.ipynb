# Confinamiento en Tokamak

Trayectorias helicoidales confinadas por campos magnéticos toroidales.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(0, 10*np.pi, 500)
U = t
V = 5*t
X = (3 + np.cos(V)) * np.cos(U)
Y = (3 + np.cos(V)) * np.sin(U)
Z = np.sin(V)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(X, Y, Z, color='red', lw=1)
plt.title('Trayectoria Helicoidal de Confinamiento de Plasma')
plt.axis('off')
plt.show()
```
