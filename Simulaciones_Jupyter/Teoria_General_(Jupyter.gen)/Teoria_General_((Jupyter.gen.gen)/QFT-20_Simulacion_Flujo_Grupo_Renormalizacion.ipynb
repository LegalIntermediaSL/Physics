# Flujo del Grupo de Renormalización (Funciones Beta)
Simulación interactiva avanzada.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Función Beta para QED (asintóticamente no libre, beta > 0)
def beta_qed(alpha, t):
    return (2/(3*np.pi)) * alpha**2

# Función Beta para QCD (libertad asintótica, beta < 0)
def beta_qcd(alpha, t, Nf=6):
    b0 = (11 - (2/3)*Nf) / (4*np.pi)
    return -b0 * alpha**2

log_mu = np.linspace(0, 10, 100)

alpha_qed_0 = 1/137.0
alpha_qed = odeint(beta_qed, alpha_qed_0, log_mu)

alpha_qcd_0 = 0.5
alpha_qcd = odeint(beta_qcd, alpha_qcd_0, log_mu)

plt.figure(figsize=(10, 6))
plt.plot(log_mu, alpha_qed, 'b-', label='Acoplamiento QED $\\alpha$ (Crece con Energía)')
plt.plot(log_mu, alpha_qcd, 'r-', label='Acoplamiento QCD $\\alpha_s$ (Libertad Asintótica)')
plt.xlabel('Escala de Energía log($\\mu$)')
plt.ylabel('Constante de Acoplamiento')
plt.title('Flujo del Grupo de Renormalización en QED vs QCD')
plt.legend()
plt.grid(True)
plt.show()
```
