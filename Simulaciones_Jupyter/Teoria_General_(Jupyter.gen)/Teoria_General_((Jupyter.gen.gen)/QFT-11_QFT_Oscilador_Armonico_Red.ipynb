# Integral de Camino en el Lattice (Oscilador 1D)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100 # Sitios en la red temporal
a = 0.1 # Lattice spacing
m = 1.0; omega = 1.0
x = np.zeros(N)
# Metrópolis simple para evaluar la Integral de Camino (Monte Carlo)
for step in range(5000):
    i = np.random.randint(N)
    dx = np.random.normal(0, 0.5)
    x_new = x.copy(); x_new[i] += dx
    # Accion discreta
    S_old = np.sum(0.5*m*((np.roll(x,-1)-x)/a)**2 + 0.5*m*omega**2*x**2) * a
    S_new = np.sum(0.5*m*((np.roll(x_new,-1)-x_new)/a)**2 + 0.5*m*omega**2*x_new**2) * a
    if np.exp(-(S_new - S_old)) > np.random.rand():
        x = x_new

plt.plot(x, marker='o', markersize=3, linestyle='-', lw=0.5)
plt.title('Fluctuación Cuántica en el Lattice (Path Integral MC)')
plt.xlabel('Tiempo Euclidiano $\tau$ (sitios)')
plt.ylabel('Posición $x(\tau)$')
plt.grid(True)
plt.show()
```
