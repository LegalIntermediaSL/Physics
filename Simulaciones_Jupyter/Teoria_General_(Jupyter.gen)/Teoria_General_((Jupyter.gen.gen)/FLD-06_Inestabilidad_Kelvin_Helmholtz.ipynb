# Inestabilidad Kelvin-Helmholtz

Formación algorítmica de vórtices por cizalladura en fluidos estratificados.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 200)
y = np.sin(x) * np.exp(-0.1*x)

plt.figure(figsize=(8,3))
plt.plot(x, y, color='blue', lw=2)
plt.fill_between(x, -2, y, color='blue', alpha=0.2)
plt.title('Perturbación en la Interfaz (Modo Kelvin-Helmholtz)')
plt.axis('off')
plt.show()
```
