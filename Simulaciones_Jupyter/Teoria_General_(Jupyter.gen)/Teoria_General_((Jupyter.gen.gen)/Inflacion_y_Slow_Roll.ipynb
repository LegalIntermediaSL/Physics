# Inflación Cósmica y el Campo del Inflatón
En los primeros instantes del universo ($10^{-35}$ s), el universo experimentó una expansión violentamente exponencial. Esto fue causado por un campo escalar cuántico llamado **Inflatón** rodando por su potencial $V(\phi)$.

Simularemos las ecuaciones acopladas de Klein-Gordon (rodadura del campo) y Friedmann (expansión). Veremos cómo el Inflatón cae lentamente (Slow-Roll) causando expansión exponencial, hasta que cae al mínimo y oscila, transfiriendo su energía a partículas (Recalentamiento).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parámetros del modelo (M=1)
m = 0.5  # Masa del Inflatón

# Sistema de EDOs acopladas
# y = [phi, dphi/dt, a]
def inflation_system(y, t):
    phi, phi_dot, a = y
    
    # Potencial del Inflatón V(phi) = 1/2 m^2 phi^2
    V = 0.5 * m**2 * phi**2
    
    # Ecuación de Friedmann: H^2 = (1/3) * rho_phi
    # rho_phi = 1/2 phi_dot^2 + V
    rho = 0.5 * phi_dot**2 + V
    H = np.sqrt(rho / 3.0)
    
    # Ecuación de Klein-Gordon (con fricción de Hubble)
    # d2phi/dt2 + 3H dphi/dt + dV/dphi = 0
    dV_dphi = m**2 * phi
    phi_ddot = -3 * H * phi_dot - dV_dphi
    
    a_dot = a * H
    return [phi_dot, phi_ddot, a_dot]

t = np.linspace(0, 100, 2000)
# Condiciones iniciales: campo alto, sin velocidad inicial, factor de escala 1
y0 = [15.0, 0.0, 1.0]
sol = odeint(inflation_system, y0, t)

phi = sol[:, 0]
a = sol[:, 2]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Evolución del Campo
ax1.plot(t, phi, color='blue', lw=2)
ax1.set_title("Campo del Inflatón $\\phi(t)$")
ax1.set_xlabel("Tiempo")
ax1.set_ylabel("$\\phi$")
ax1.grid(True)

# Evolución del Factor de Escala (Logarítmico para ver exponencial)
ax2.plot(t, np.log(a), color='red', lw=2)
ax2.set_title("Expansión Inflacionaria ($\ln(a)$ vs $t$)")
ax2.set_xlabel("Tiempo")
ax2.set_ylabel("$\ln(a)$ (e-folds)")
ax2.grid(True)

plt.tight_layout()
plt.show()
```
