# [QIC-05] VQE y Optimización

Simulación cuántica pura en Python.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Variational Quantum Eigensolver (VQE) - Emulación simplificada
# Buscamos minimizar un Hamiltoniano H usando un ansatz parametrizado.

# Supongamos un Hamiltoniano H para una molécula simple (matriz 2x2 hermítica)
H = np.array([[1.5, 0.2],
              [0.2, 0.5]])

# Ansatz Cuántico: Rotación Y
# |psi(theta)> = Ry(theta) |0>
def state(theta):
    return np.array([np.cos(theta/2), np.sin(theta/2)])

# Función de Coste: Valor Esperado <psi | H | psi>
def expectation_value(theta):
    psi = state(theta)
    return np.dot(psi.T, np.dot(H, psi))

# Optimización Clásica
initial_theta = 0.0
result = minimize(expectation_value, initial_theta, method='COBYLA')

optimal_theta = result.x[0]
min_energy = result.fun

# Visualización del panorama energético
thetas = np.linspace(-np.pi, np.pi, 100)
energies = [expectation_value(t) for t in thetas]

plt.figure(figsize=(10, 6))
plt.plot(thetas, energies, 'b-', label='<H>(θ)')
plt.plot(optimal_theta, min_energy, 'ro', markersize=10, label=f'Mínimo VQE: {min_energy:.4f}')
plt.title("Paisaje de Optimización VQE")
plt.xlabel("Parámetro Variacional (Theta)")
plt.ylabel("Energía Esperada")
plt.axhline(y=np.linalg.eigvals(H).min(), color='k', linestyle='--', label='Energía Exacta (Diagonalización)')
plt.legend()
plt.grid(True)
plt.show()

```
