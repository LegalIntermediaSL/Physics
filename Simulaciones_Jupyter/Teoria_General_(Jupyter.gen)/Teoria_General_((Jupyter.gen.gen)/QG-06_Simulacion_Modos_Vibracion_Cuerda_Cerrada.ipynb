# Modos de Vibración Cuánticos de una Cuerda Cerrada
Simulación interactiva avanzada.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

sigma = np.linspace(0, 2*np.pi, 100)
tau = np.linspace(0, 2*np.pi, 100)
Sigma, Tau = np.meshgrid(sigma, tau)

# Cuerda cerrada con modos n=2 y m=3
R = 2 + np.cos(2*Sigma - Tau) + 0.5*np.cos(3*Sigma + 2*Tau)
X = R * np.cos(Sigma)
Y = R * np.sin(Sigma)
Z = np.sin(2*Sigma - Tau)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none', alpha=0.9)
ax.set_title('Hoja de Mundo (Worldsheet) de una Supercuerda Vibrando')
ax.axis('off')
plt.show()
```
