# Dinámica de la Inflación Cósmica: Campo Lento

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Potencial de Linde V(phi) = 0.5 * m^2 * phi^2
def inflatondynamics(y, t, m, H0):
    phi, phidot = y
    H = np.sqrt(H0**2 + (4*np.pi/3)*(0.5*phidot**2 + 0.5*m**2*phi**2))
    phiddot = -3*H*phidot - m**2 * phi
    return [phidot, phiddot]

t = np.linspace(0, 100, 1000)
y0 = [20.0, -0.1]
sol = odeint(inflatondynamics, y0, t, args=(0.1, 0.05))
plt.plot(t, sol[:,0])
plt.title('Evolución del campo Inflatón $\\phi(t)$')
plt.xlabel('Tiempo cósmico'); plt.ylabel('Amplitud del campo'); plt.grid(True)
plt.show()
```
