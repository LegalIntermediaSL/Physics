# Efecto Hall Clásico

Dinámica de portadores de carga bajo campos eléctricos y magnéticos cruzados.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 500)
x = t - np.sin(t)
y = 1 - np.cos(t)

plt.figure(figsize=(8,4))
plt.plot(x, y, color='green')
plt.title('Trayectoria Cicloidal (Deriva ExB)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
