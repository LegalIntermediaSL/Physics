# Función de Masa de Halos (Press-Schechter)

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.logspace(10, 15, 100)
sigma = 10 * (M/1e12)**(-0.3) # Aproximacion varianza
dn_dM = np.sqrt(2/np.pi) * (0.3) * (M**(-2)) * (1.686/sigma) * np.exp(-1.686**2 / (2*sigma**2))

plt.loglog(M, dn_dM)
plt.title('Función de Masa de Halos de Materia Oscura (Press-Schechter)')
plt.xlabel('Masa del Halo $M/M_\odot$')
plt.ylabel('$dn/dM$')
plt.grid(True)
plt.show()
```
