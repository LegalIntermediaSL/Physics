# Interferómetro de Michelson

Modelado de franjas de interferencia 2D.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
R = X**2 + Y**2
I = np.cos(R)**2

plt.figure(figsize=(6,6))
plt.imshow(I, cmap='gray')
plt.title('Franjas de Interferencia Circulares (Michelson)')
plt.axis('off')
plt.show()
```
