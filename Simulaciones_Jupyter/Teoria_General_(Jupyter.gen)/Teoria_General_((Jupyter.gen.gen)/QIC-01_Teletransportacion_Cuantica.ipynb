# Protocolo de Teletransportación

Transmisión de estados cuánticos usando canales clásicos y EPR.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Diagrama de Bloques Teórico Simulado
fig, ax = plt.subplots(figsize=(6,4))
ax.text(0.1, 0.5, '|ψ> Alice', fontsize=12, bbox=dict(facecolor='red', alpha=0.5))
ax.text(0.5, 0.5, 'Canal EPR', fontsize=12)
ax.text(0.8, 0.5, '|ψ> Bob', fontsize=12, bbox=dict(facecolor='blue', alpha=0.5))
ax.annotate('', xy=(0.8, 0.52), xytext=(0.25, 0.52), arrowprops=dict(arrowstyle="->"))
plt.axis('off')
plt.title('Esquema Lógico de Teletransportación')
plt.show()
```
