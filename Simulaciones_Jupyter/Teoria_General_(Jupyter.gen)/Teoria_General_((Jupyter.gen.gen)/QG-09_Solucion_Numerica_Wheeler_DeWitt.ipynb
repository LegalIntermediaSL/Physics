# [QG-09] Wheeler-DeWitt y Big Bounce LQC

Simulación interactiva generada para el bloque de Gravedad Cuántica de Lazos (LQG).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Solución Numérica de Wheeler-DeWitt en LQC (Minisuperespacio)
# Simulamos cómo el Universo evita la singularidad del Big Bang y realiza un "Big Bounce".

# Ecuación de Friedmann Efectiva en LQC: H^2 = (8piG/3) * rho * (1 - rho/rho_crit)
G = 1.0
rho_crit = 1.0  # Densidad crítica de Planck
t = np.linspace(-10, 10, 500)

def lqc_friedmann(a, t):
    # rho proporcional a 1/a^3 (Polvo)
    rho = 0.1 / (a**3 + 1e-5)
    
    # Evitar raiz cuadrada negativa si rho > rho_crit
    if rho > rho_crit:
        rho = rho_crit
        
    H2 = (8 * np.pi * G / 3) * rho * (1 - rho / rho_crit)
    H = np.sqrt(max(0, H2))
    
    # Si t < 0 el universo colapsa (H < 0), si t > 0 se expande (H > 0)
    return -H * a if t < 0 else H * a

# Condición inicial en el rebote (Bounce) t=0
a0 = 0.5  # Factor de escala mínimo no nulo
a_t = odeint(lqc_friedmann, a0, t)

plt.figure(figsize=(10, 6))
plt.plot(t, a_t, color='blue', linewidth=3, label='Factor de Escala LQC')
plt.axvline(x=0, color='red', linestyle='--', label='Big Bounce (Rebote Cuántico)')

plt.title('Cosmología Cuántica de Lazos: El Gran Rebote (Big Bounce)')
plt.xlabel('Tiempo Cósmico (t)')
plt.ylabel('Factor de Escala a(t)')
plt.legend()
plt.grid(True)
plt.show()

```
