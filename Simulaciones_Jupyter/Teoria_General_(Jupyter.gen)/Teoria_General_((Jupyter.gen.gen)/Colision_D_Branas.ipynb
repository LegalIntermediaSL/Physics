# Colisión de D-Branas y Radiación de Cuerdas
Visualización conceptual de dos branas (superficies) que se atraen e intersecan. En la Teoría de Cuerdas, cuando una D-brana choca con una Anti-D-brana, el campo del taquión condensado (representando cuerdas estiradas entre ellas) las aniquila, liberando una lluvia de cuerdas cerradas masivas (gravitones).

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

def animate(i):
    ax.clear()
    # Distancia z variable entre las branas (chocan en i=30)
    z_sep = 5.0 - 0.16 * i
    if z_sep < 0: z_sep = 0
    
    Z1 = np.ones_like(X) * (z_sep / 2)
    Z2 = np.ones_like(X) * (-z_sep / 2)
    
    # Dibujar la D-brana superior
    ax.plot_surface(X, Y, Z1, color='cyan', alpha=0.4, edgecolor='b')
    # Dibujar la Anti-D-brana inferior
    ax.plot_surface(X, Y, Z2, color='magenta', alpha=0.4, edgecolor='r')
    
    # Dibujar "cuerdas taquiónicas" rojas conectándolas
    if z_sep > 0:
        for _ in range(10):
            cx = np.random.uniform(-4, 4)
            cy = np.random.uniform(-4, 4)
            ax.plot([cx, cx], [cy, cy], [Z2[0,0], Z1[0,0]], color='red', lw=1.5, alpha=0.6)
    else:
        # Explosión de gravitones (cuerdas cerradas verdes)
        theta = np.linspace(0, 2*np.pi, 20)
        for _ in range(15):
            cx = np.random.uniform(-4, 4)
            cy = np.random.uniform(-4, 4)
            cz = np.random.uniform(-1, 1)
            ax.plot(cx + 0.5*np.cos(theta), cy + 0.5*np.sin(theta), cz, color='green', lw=2)

    ax.set_zlim(-3, 3)
    ax.set_title(f"Colisión Brana/Anti-Brana (Paso {i})")
    ax.axis('off')

anim = FuncAnimation(fig, animate, frames=60, interval=60)
plt.close()
HTML(anim.to_jshtml())
```
