# Evolución de Estado Entrelazado (EPR)

Dinámica de dos partículas entrelazadas propagándose en direcciones opuestas conservando su correlación.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)
t = 3.0
# Paquetes de ondas viajando en direcciones opuestas
psi_A = np.exp(-(x + t)**2)
psi_B = np.exp(-(x - t)**2)

plt.figure(figsize=(8,4))
plt.plot(x, psi_A, 'b-', label='Partícula A (Alice)')
plt.plot(x, psi_B, 'r-', label='Partícula B (Bob)')
plt.title('Propagación Espacial de un Par EPR Entrelazado')
plt.xlabel('Posición x')
plt.legend()
plt.grid()
plt.show()
```
