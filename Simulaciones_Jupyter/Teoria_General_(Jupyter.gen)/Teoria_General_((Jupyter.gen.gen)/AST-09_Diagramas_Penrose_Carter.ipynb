# Topología Conforme: Diagramas de Penrose

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-10, 10, 200)
r = np.linspace(0, 10, 200)
# Coordenadas compactificadas arctan()
U = np.arctan(t - r)
V = np.arctan(t + r)

plt.plot(U, V)
plt.title('Coordenadas Conformes del Infinito Nulo (Esbozo 1D)')
plt.xlabel('Coordenada Retardada Conforme $U$')
plt.ylabel('Coordenada Avanzada Conforme $V$')
plt.grid(True)
plt.show()
```
