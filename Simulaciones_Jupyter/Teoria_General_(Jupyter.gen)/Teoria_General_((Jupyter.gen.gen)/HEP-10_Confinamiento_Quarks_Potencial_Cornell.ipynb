# Potencial QCD de Cornell

Asintotismo libre y confinamiento lineal de los quarks.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

r = np.linspace(0.1, 3, 100)
V = -4/3 * (1/r) + 2.0 * r # Coulómbico + Lineal

plt.figure(figsize=(8,5))
plt.plot(r, V, 'k-', lw=2)
plt.title('Potencial de Cornell para Quarks (Confinamiento por Color)')
plt.xlabel('Distancia r')
plt.ylabel('Potencial Efectivo V(r)')
plt.grid()
plt.show()
```
