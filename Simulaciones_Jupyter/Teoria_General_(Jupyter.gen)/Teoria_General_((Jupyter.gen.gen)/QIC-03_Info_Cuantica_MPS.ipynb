# Investigación Avanzada: Info Cuantica MPS

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

def create_mps_site(d, D):
    """
    Create a random MPS site tensor.
    d: physical dimension
    D: bond dimension
    """
    A = np.random.randn(d, D, D) + 1j * np.random.randn(d, D, D)
    return A

def contract_mps(A_list):
    """
    Contract a list of MPS tensors to get the full state vector.
    """
    N = len(A_list)
    state = A_list[0]
    for i in range(1, N):
        # state is shape (d, ..., d, D), next A is (d, D, D')
        state = np.tensordot(state, A_list[i], axes=(-1, 1))
    return state

if __name__ == "__main__":
    N = 4  # Number of sites
    d = 2  # Physical dimension (qubit)
    D = 2  # Bond dimension
    
    # Open boundary conditions: first and last tensors have D=1 on outside
    A1 = np.random.randn(d, 1, D)
    A2 = np.random.randn(d, D, D)
    A3 = np.random.randn(d, D, D)
    A4 = np.random.randn(d, D, 1)
    
    A_list = [A1, A2, A3, A4]
    
    full_tensor = contract_mps(A_list)
    state_vector = full_tensor.flatten()
    state_vector /= np.linalg.norm(state_vector)
    
    print(f"MPS constructed state vector of size {len(state_vector)}")
    print(f"Norm: {np.linalg.norm(state_vector)}")
    
    plt.plot(np.abs(state_vector)**2, marker='o', linestyle='-', color='purple')
    plt.title(f'Amplitudes of {N}-site MPS (D={D})')
    plt.xlabel('Basis State Index')
    plt.ylabel('Probability')
    plt.show()
```
