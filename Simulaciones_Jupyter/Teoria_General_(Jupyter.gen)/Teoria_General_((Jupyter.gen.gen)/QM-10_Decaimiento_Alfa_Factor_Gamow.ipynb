# Efecto Túnel: Factor de Gamow

Decaimiento Alfa modelado como tunelaje cuántico a través de la fuerte barrera de repulsión de Coulomb.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

r = np.linspace(1, 10, 300)
R_nuc = 2.0
V_coulomb = np.where(r > R_nuc, 10 / r, -15)
E_alpha = 2.0

plt.figure(figsize=(8,5))
plt.plot(r, V_coulomb, 'k-', lw=2, label='Potencial Nuclear + Coulomb')
plt.axhline(E_alpha, color='r', linestyle='--', label='Energía de la partícula $\\alpha$')
plt.fill_between(r, E_alpha, V_coulomb, where=(V_coulomb > E_alpha), color='red', alpha=0.3, label='Barrera de Gamow (Túnel)')
plt.title('Decaimiento Alfa y Barrera de Coulomb')
plt.xlabel('Distancia Radial r')
plt.ylabel('Potencial V(r)')
plt.legend()
plt.grid()
plt.show()
```
