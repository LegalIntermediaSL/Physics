# Precesión de un Giróscopo

Modelo vectorial del torque y la conservación del momento angular.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 500)
L_x = np.cos(t)
L_y = np.sin(t)

plt.figure(figsize=(6,6))
plt.plot(L_x, L_y, label='Vector L(t)')
plt.title('Precesión del Vector Momento Angular en el Plano XY')
plt.legend()
plt.axis('equal')
plt.grid()
plt.show()
```
