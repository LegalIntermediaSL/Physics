# Protocolo de Teletransporte Cuántico
El Teorema de No-Clonación prohíbe copiar el estado de un qubit. Sin embargo, usando Entrelazamiento Cuántico y 2 bits clásicos de información, Alice puede "teletransportar" un estado desconocido hacia Bob destruyendo el original.

En este cuaderno, simularemos algebraicamente el entrelazamiento de un Par de Bell, la medición de campana y las compuertas de corrección de Pauli.

```python
import numpy as np

# Base computacional
zero = np.array([1, 0])
one = np.array([0, 1])

# Compuertas de Pauli
I = np.eye(2)
X = np.array([[0, 1], [1, 0]])
Z = np.array([[1, 0], [0, -1]])
H = (1/np.sqrt(2)) * np.array([[1, 1], [1, -1]])

# Estado desconocido a teletransportar (ej. un estado arbitrario de Alice)
alpha = np.sqrt(0.8)
beta = np.sqrt(0.2)
psi = alpha * zero + beta * one
print("Estado original de Alice:", psi)

# 1. Crear el Par de Bell entre Alice y Bob (Qubits 2 y 3)
# Estado |Phi+> = (|00> + |11>) / sqrt(2)
Phi_plus = (1/np.sqrt(2)) * (np.kron(zero, zero) + np.kron(one, one))

# Estado global inicial de los 3 qubits: |Psi> (A1) (x) |Phi+> (A2, B3)
# Matemáticamente, al reescribir esto en la base de Bell para los qubits de Alice (A1, A2),
# el estado colapsará en uno de 4 posibles resultados para Bob.

# Simulación directa de la corrección:
# Supongamos que Alice midió y obtuvo el resultado clásico '01' (correspondiente a |Psi+>)
# El estado de Bob colapsó instantáneamente a X|psi>

estado_bob_colapsado = np.dot(X, psi)

# Bob recibe la llamada: "Tuve el resultado 01, aplica la compuerta X"
estado_recuperado = np.dot(X, estado_bob_colapsado)

print("Estado de Bob recuperado:", estado_recuperado)
if np.allclose(psi, estado_recuperado):
    print("¡Teletransporte Exitoso!")
```
