# Colapso Esférico: Métrica Lemaître-Tolman-Bondi

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 0.99, 100)
R = (1 - t)**(2/3) # Escala de colapso esférico simplificada

plt.plot(t, R, color='black')
plt.title('Radio fìsico en un colapso de polvo esférico (Modelo LTB)')
plt.xlabel('Tiempo $t/t_{colapso}$')
plt.ylabel('Factor de escala radial $R(t, r)$')
plt.grid(True)
plt.show()
```
