# Optimización Computacional y Colapso Cuántico

Universo Simulado vs Matemáticas de Copenhague: Computar una matriz probabilística difusa gigante (caro) frente a fijar (colapsar) una posición aleatoria solo cuando es observada (barato).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
# Nube de probabilidad (Superposición no observada)
psi_squared = np.exp(-0.2*x**2) * (np.cos(5*x)**2)
psi_squared = psi_squared / np.sum(psi_squared) # Normalización

plt.figure(figsize=(10,4))
plt.plot(x, psi_squared, 'b-', label='Estado no observado (Bajo coste de memoria: Función Analítica)')
plt.fill_between(x, psi_squared, color='blue', alpha=0.2)

# Momento de Observación (Renderizado Diferido / Colapso)
observed_pos = np.random.choice(x, p=psi_squared)
plt.axvline(observed_pos, color='r', lw=3, label=f'Colapso al medir en x={observed_pos:.2f} (Renderizado Puntual)')

plt.title('Colapso de la Función de Onda como "Occlusion Culling" Computacional')
plt.xlabel('Posición (x)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid()
plt.show()
```
