# Diagramas de Minkowski

Transformaciones de Lorentz y visualización de conos de luz.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
t1 = x
t2 = -x

plt.figure(figsize=(6,6))
plt.plot(x, t1, 'k--', label='Rayo de Luz (v=c)')
plt.plot(x, t2, 'k--')
plt.fill_between(x, np.abs(x), 5, color='yellow', alpha=0.3, label='Futuro Causal')
plt.fill_between(x, -5, -np.abs(x), color='gray', alpha=0.3, label='Pasado Causal')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title('Cono de Luz en Espaciotiempo de Minkowski')
plt.legend()
plt.grid()
plt.show()
```
