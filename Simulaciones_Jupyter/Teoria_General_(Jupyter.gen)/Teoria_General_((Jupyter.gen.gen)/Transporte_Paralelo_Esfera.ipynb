# Transporte Paralelo en la Esfera

Demostración de cómo un vector rotado a lo largo de un circuito cerrado en un espacio curvo no coincide con el original.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

u = np.linspace(0, 2*np.pi, 50)
v = np.linspace(0, np.pi, 50)
U, V = np.meshgrid(u, v)
X, Y, Z = np.sin(V)*np.cos(U), np.sin(V)*np.sin(U), np.cos(V)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.3, edgecolor='k', lw=0.1)

# Camino cerrado
theta = np.linspace(0, np.pi/2, 20)
ax.plot(np.sin(theta), np.zeros_like(theta), np.cos(theta), 'r-', lw=3, label='Camino Geodésico')
ax.plot(np.cos(theta), np.sin(theta), np.zeros_like(theta), 'b-', lw=3)
ax.plot(np.zeros_like(theta), np.cos(theta)[::-1], np.sin(theta)[::-1], 'g-', lw=3)

plt.title('Transporte Paralelo: Defecto de Ángulo')
plt.legend()
plt.axis('off')
plt.show()
```
