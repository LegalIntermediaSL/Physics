# Cálculo Exterior de Cartan

Representación de formas diferenciales y el teorema de Stokes generalizado.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)

# 1-forma w = -y dx + x dy
U = -Y
V = X

plt.figure(figsize=(8,8))
plt.quiver(X, Y, U, V, color='purple')
plt.title('Representación de una 1-forma Diferencial en $R^2$')
plt.grid(True)
plt.show()
```
