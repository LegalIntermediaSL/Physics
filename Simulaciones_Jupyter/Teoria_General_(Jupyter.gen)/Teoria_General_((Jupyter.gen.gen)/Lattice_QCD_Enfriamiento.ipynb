# Lattice QCD: Enfriamiento de Monte Carlo (Modelo de Ising Espín-Lattice)
En QCD real, los gluones se modelan con matrices de SU(3) en los enlaces de un retículo (Lattice). Para propósitos pedagógicos, simularemos una versión abeliana simplificada (teoría Z2 / Modelo de Ising 2D) que exhibe transición de fase y termodinámica análoga al confinamiento del color.

El algoritmo de Metropolis-Hastings se utiliza para 'enfriar' el sistema y llevarlo a su vacío de mínima energía.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Parámetros del retículo
N = 50       # Tamaño de la cuadrícula NxN
T = 2.269    # Temperatura crítica analítica de Onsager
beta = 1.0/T

# Inicialización aleatoria del retículo
lattice = 2 * np.random.randint(2, size=(N, N)) - 1

def mcmc_step(config, beta):
    for i in range(N):
        for j in range(N):
            a = np.random.randint(0, N)
            b = np.random.randint(0, N)
            s = config[a, b]
            
            # Interacción con vecinos más cercanos (Condiciones periódicas)
            nb = config[(a+1)%N,b] + config[a,(b+1)%N] + config[(a-1)%N,b] + config[a,(b-1)%N]
            dE = 2 * s * nb
            
            # Regla de Metropolis
            if dE < 0 or np.random.rand() < np.exp(-dE * beta):
                config[a, b] *= -1
    return config

# Simular
fig, ax = plt.subplots(figsize=(6,6))
img = ax.imshow(lattice, cmap='coolwarm', interpolation='nearest')
ax.set_title("Simulación Lattice (Metropolis MCMC)")
plt.axis('off')

def animate(frame):
    global lattice
    # Hacer múltiples pasos por frame para acelerar la convergencia
    for _ in range(5):
        lattice = mcmc_step(lattice, beta)
    img.set_data(lattice)
    return [img]

anim = FuncAnimation(fig, animate, frames=60, interval=50, blit=True)
plt.close()
HTML(anim.to_jshtml())
```
