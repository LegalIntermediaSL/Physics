# Superficie de Riemann de un Polo

Visualización tridimensional del comportamiento divergente en un polo complejo simple.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

r = np.linspace(0.1, 2, 100)
theta = np.linspace(0, 2*np.pi, 100)
R, Theta = np.meshgrid(r, theta)
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

# Función 1/z
Z = 1 / R

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
plt.title('Singularidad 1/|z| en el Origen Complejo')
plt.axis('off')
plt.show()
```
