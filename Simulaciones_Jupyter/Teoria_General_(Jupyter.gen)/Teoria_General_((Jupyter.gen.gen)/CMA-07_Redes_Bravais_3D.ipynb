# Redes de Bravais 3D

Representación cristalográfica tridimensional del empaquetamiento atómico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x, y, z = np.meshgrid([0,1,2], [0,1,2], [0,1,2])

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='r', s=100)
plt.title('Red Cristalina Cúbica Simple (SC)')
plt.axis('off')
plt.show()
```
