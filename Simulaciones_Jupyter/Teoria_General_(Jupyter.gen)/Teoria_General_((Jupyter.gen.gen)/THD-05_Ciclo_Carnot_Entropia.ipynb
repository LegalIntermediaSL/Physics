# Ciclo de Carnot en Diagrama P-V

El ciclo termodinámico ideal y la limitación entrópica.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

V1 = np.linspace(1, 2, 50)
P1 = 10 / V1 # Isoterma alta
V2 = np.linspace(2, 3, 50)
P2 = P1[-1] * (V1[-1]/V2)**1.4 # Adiabatica

plt.figure(figsize=(6,5))
plt.plot(V1, P1, 'r-', label='Expansión Isoterma (T_H)')
plt.plot(V2, P2, 'k-', label='Expansión Adiabática')
plt.title('Diagrama P-V del Ciclo de Carnot (Parcial)')
plt.xlabel('Volumen')
plt.ylabel('Presión')
plt.legend()
plt.grid()
plt.show()
```
