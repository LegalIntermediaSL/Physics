# Transformada de Fourier Cuántica (El núcleo de Shor)
El Algoritmo de Shor quiebra la criptografía RSA en tiempo polinomial. Su secreto es usar la **Transformada de Fourier Cuántica (QFT)** para encontrar el período de una función modular dentro de una inmensa superposición de estados.

Aquí visualizaremos cómo la QFT rota las fases de los estados cuánticos para que las amplitudes interfieran constructivamente solo en la frecuencia (período) correcta.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el número de qubits n
n = 4
N = 2**n

# Función para aplicar QFT a un estado dado
def apply_qft(state):
    qft_matrix = np.zeros((N, N), dtype=complex)
    omega = np.exp(2.0 * np.pi * 1j / N)
    
    for j in range(N):
        for k in range(N):
            qft_matrix[j, k] = (1.0 / np.sqrt(N)) * (omega ** (j * k))
            
    return np.dot(qft_matrix, state)

# Supongamos que preparamos una superposición periódica (período r=4)
# Solo los estados |0>, |4>, |8>, |12> tienen amplitud, el resto 0
state_in = np.zeros(N, dtype=complex)
r = 4
for i in range(0, N, r):
    state_in[i] = 1.0
# Normalizar
state_in = state_in / np.linalg.norm(state_in)

# Aplicar QFT
state_out = apply_qft(state_in)

# Probabilidades
prob_in = np.abs(state_in)**2
prob_out = np.abs(state_out)**2

# Visualización
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.bar(range(N), prob_in, color='blue')
ax1.set_title("Estado de Entrada (Periódico, $r=4$)")
ax1.set_xticks(range(N))

ax2.bar(range(N), prob_out, color='red')
ax2.set_title("Salida tras QFT (Picos en $N/r = 16/4 = 4$)")
ax2.set_xticks(range(N))

plt.tight_layout()
plt.show()

print("La QFT ha revelado instantáneamente la frecuencia del período.")
```
