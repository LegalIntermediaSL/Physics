# Pixelación de Planck

Visualización del límite matemático donde un espacio continuo infinito es imposible si la memoria (información) es finita, resultando en una resolución máxima (Longitud de Planck).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Simulación de acercamiento extremo a una función continua
x_continuous = np.linspace(0, 1, 10000)
y_continuous = np.sin(50 * x_continuous)

# El Universo Simulado: La malla subyacente de resolución de Planck
planck_length = 0.05 # Tamaño del pixel hipotético
x_discrete = np.arange(0, 1, planck_length)
y_discrete = np.sin(50 * x_discrete)

plt.figure(figsize=(10, 4))
plt.plot(x_continuous, y_continuous, 'k--', alpha=0.3, label='Realidad "Continua" Matemática Infinita')
plt.step(x_discrete, y_discrete, 'r-', where='mid', lw=2, label='Realidad Física Simulada (Cuantizada / Píxeles de Planck)')

plt.title('El Espaciotiempo Cuantizado: Límite de Información Física')
plt.xlim(0.2, 0.5)
plt.legend()
plt.grid(True)
plt.show()
```
