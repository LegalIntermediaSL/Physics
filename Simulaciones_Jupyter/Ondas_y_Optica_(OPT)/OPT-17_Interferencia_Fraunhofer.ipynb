# Patrones de Fraunhofer

Difracción por múltiples rendijas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(-0.1, 0.1, 500)
I = (np.sin(50*theta)/(50*theta + 1e-6))**2

plt.figure(figsize=(8,4))
plt.plot(theta, I, color='indigo')
plt.title('Intensidad de Difracción (Rendija Única)')
plt.xlabel('Ángulo Theta')
plt.ylabel('Intensidad')
plt.grid()
plt.show()
```
