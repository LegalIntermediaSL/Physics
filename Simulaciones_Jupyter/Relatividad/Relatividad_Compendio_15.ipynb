# Compendio Subtemático: Relatividad Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def light_deflection():
    # Toy model for light deflection
    x = np.linspace(-10, 10, 500)
    b = 2.0 # impact parameter
    
    # Newtonian deflection (straight line)
    y_newton = np.full_like(x, b)
    
    # Relativistic deflection approximation
    y_rel = b - 4 * 1.0 / np.sqrt(x**2 + b**2) + 4/b
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y_newton, '--', label='No gravity (straight)', color='gray')
    plt.plot(x, y_rel, label='Relativistic Path', color='orange')
    plt.plot(0, 0, 'ko', markersize=15, label='Massive Object')
    
    plt.xlabel('x coordinate')
    plt.ylabel('y coordinate')
    plt.title('Deflection of Light by a Massive Object')
    plt.grid(True)
    plt.legend()
    plt.axis('equal')
    plt.savefig('light_deflection.png')
    print('Plot saved as light_deflection.png')

if __name__ == '__main__':
    light_deflection()
```
