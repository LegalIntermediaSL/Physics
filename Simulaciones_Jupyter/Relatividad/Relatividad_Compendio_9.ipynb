# Compendio Subtemático: Relatividad Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def minkowski_diagram():
    c = 1.0
    
    plt.figure(figsize=(6, 6))
    
    # Light lines (x = ct and x = -ct)
    x = np.linspace(-5, 5, 100)
    plt.plot(x, x, 'y--', label='Photon (Right)')
    plt.plot(x, -x, 'y--', label='Photon (Left)')
    
    # Stationary object
    plt.plot(np.zeros(100), np.linspace(0, 5, 100), 'b-', label='Stationary')
    
    # Moving object (v = 0.5c)
    v = 0.5
    plt.plot(x[x>=0], x[x>=0]/v, 'r-', label='Moving ($v=0.5c$)')
    
    plt.xlim(-5, 5)
    plt.ylim(0, 5)
    plt.xlabel('Space (x)')
    plt.ylabel('Time (ct)')
    plt.title('Minkowski Spacetime Diagram')
    plt.grid(True)
    plt.legend()
    plt.savefig('minkowski_diagram.png')
    print('Plot saved as minkowski_diagram.png')

if __name__ == '__main__':
    minkowski_diagram()
```
