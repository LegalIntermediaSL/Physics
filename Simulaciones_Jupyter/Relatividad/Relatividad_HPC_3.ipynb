# Supercomputación: Relatividad HPC 3

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 1D Special Relativistic Hydrodynamics (Toy Model - Advection of density)
N = 200
x = np.linspace(0, 1, N)
dx = x[1] - x[0]
rho = np.ones(N)
rho[(x > 0.4) & (x < 0.6)] = 2.0
v = np.full(N, 0.9) # v/c = 0.9

W = 1 / np.sqrt(1 - v**2)
D = rho * W

dt = dx * 0.5
for _ in range(100):
    F = D * v
    D[1:-1] -= (dt/dx) * (F[1:-1] - F[:-2]) # Upwind

rho = D / W

plt.plot(x, rho)
plt.title('1D SRHD Toy Advection v=0.9c')
plt.savefig('Relatividad_HPC_3_output.png')
```
