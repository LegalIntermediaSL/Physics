# Compendio Subtemático: Relatividad Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def kinetic_energy():
    c = 1.0
    m = 1.0
    v = np.linspace(0, 0.99*c, 500)
    
    KE_classical = 0.5 * m * v**2
    gamma = 1 / np.sqrt(1 - v**2/c**2)
    KE_rel = (gamma - 1) * m * c**2
    
    plt.figure(figsize=(8, 5))
    plt.plot(v/c, KE_classical, '--', label=r'Classical ($1/2 mv^2$)', color='blue')
    plt.plot(v/c, KE_rel, label=r'Relativistic ($(\gamma-1)mc^2$)', color='red')
    plt.xlabel('Velocity (v/c)')
    plt.ylabel('Kinetic Energy')
    plt.title('Classical vs Relativistic Kinetic Energy')
    plt.ylim(0, 5)
    plt.grid(True)
    plt.legend()
    plt.savefig('kinetic_energy.png')
    print('Plot saved as kinetic_energy.png')

if __name__ == '__main__':
    kinetic_energy()
```
