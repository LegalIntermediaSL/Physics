# Compendio Subtemático: Relatividad Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def frame_dragging():
    # Omega_ZAMO ~ 2*G*J / (c^2 * r^3)
    r = np.linspace(2, 10, 500)
    J = 1.0 # Angular momentum factor
    
    omega = 2 * J / r**3
    
    plt.figure(figsize=(8, 5))
    plt.plot(r, omega, color='purple')
    plt.xlabel('Distance from Kerr Black Hole ($r$)')
    plt.ylabel('Angular Velocity of ZAMO ($\Omega$)')
    plt.title('Frame Dragging (Lense-Thirring Effect)')
    plt.grid(True)
    plt.savefig('frame_dragging.png')
    print('Plot saved as frame_dragging.png')

if __name__ == '__main__':
    frame_dragging()
```
