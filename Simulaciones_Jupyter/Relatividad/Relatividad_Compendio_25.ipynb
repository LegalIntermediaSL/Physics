# Compendio Subtemático: Relatividad Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def four_velocity():
    v = np.linspace(0, 0.95, 500)
    gamma = 1 / np.sqrt(1 - v**2)
    
    u_0 = gamma
    u_1 = gamma * v
    
    plt.figure(figsize=(8, 5))
    plt.plot(v, u_0, label='Time component ($u^0 = \gamma c$)', color='purple')
    plt.plot(v, u_1, label='Spatial component ($u^1 = \gamma v$)', color='teal')
    plt.xlabel('3-Velocity (v/c)')
    plt.ylabel('4-Velocity Component')
    plt.title('Components of 4-Velocity')
    plt.grid(True)
    plt.legend()
    plt.savefig('four_velocity.png')
    print('Plot saved as four_velocity.png')

if __name__ == '__main__':
    four_velocity()
```
