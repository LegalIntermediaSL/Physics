# Compendio Subtemático: Relatividad Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def shapiro_delay():
    # Delay ~ ln(4 r_e r_p / b^2)
    b = np.linspace(1, 10, 500) # Impact parameter
    delay = -np.log(b)
    
    plt.figure(figsize=(8, 5))
    plt.plot(b, delay - np.min(delay), color='navy')
    plt.xlabel('Impact Parameter $b$')
    plt.ylabel('Time Delay (arbitrary units)')
    plt.title('Shapiro Time Delay')
    plt.grid(True)
    plt.savefig('shapiro_delay.png')
    print('Plot saved as shapiro_delay.png')

if __name__ == '__main__':
    shapiro_delay()
```
