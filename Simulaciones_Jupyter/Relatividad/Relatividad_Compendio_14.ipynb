# Compendio Subtemático: Relatividad Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def bh_effective_potential():
    r = np.linspace(1.5, 20, 500) # In units of Schwarzschild radius
    L = np.array([3.5, 4.0, 4.5]) # Angular momentum in units of mM
    
    plt.figure(figsize=(8, 5))
    
    for l in L:
        # Effective potential V(r) = (1 - 1/r)(1 + l^2/r^2)
        V = (1 - 1/r) * (1 + l**2 / r**2)
        plt.plot(r, V, label=f'$L = {l} mM$')
        
    plt.xlabel('Radius ($r/R_s$)')
    plt.ylabel('Effective Potential ($V_{eff}$)')
    plt.title('Effective Potential for Massive Particles around a Black Hole')
    plt.ylim(0.8, 1.2)
    plt.grid(True)
    plt.legend()
    plt.savefig('bh_eff_potential.png')
    print('Plot saved as bh_eff_potential.png')

if __name__ == '__main__':
    bh_effective_potential()
```
