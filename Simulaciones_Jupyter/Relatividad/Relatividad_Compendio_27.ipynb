# Compendio Subtemático: Relatividad Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def relativistic_rocket():
    mass_ratio = np.logspace(0, 3, 500) # M_i / M_f
    Isp = 1.0 # Exhaust velocity v_e = c
    
    # v/c = ( (M_i/M_f)^(2 v_e/c) - 1 ) / ( (M_i/M_f)^(2 v_e/c) + 1 )
    term = mass_ratio**(2 * Isp)
    v = (term - 1) / (term + 1)
    
    plt.figure(figsize=(8, 5))
    plt.semilogx(mass_ratio, v, color='black')
    plt.xlabel('Mass Ratio ($M_i / M_f$)')
    plt.ylabel('Final Velocity ($v/c$)')
    plt.title('Relativistic Rocket Equation')
    plt.grid(True, which="both", ls="--")
    plt.savefig('relativistic_rocket.png')
    print('Plot saved as relativistic_rocket.png')

if __name__ == '__main__':
    relativistic_rocket()
```
