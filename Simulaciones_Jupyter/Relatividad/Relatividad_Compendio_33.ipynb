# Compendio Subtemático: Relatividad Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def kerr_ergosphere():
    theta = np.linspace(0, 2*np.pi, 500)
    M = 1.0
    a = 0.9 # Spin parameter (a = J/M, must be <= M)
    
    # Event horizon (outer): r_H = M + sqrt(M^2 - a^2)
    r_H = M + np.sqrt(M**2 - a**2)
    
    # Ergosphere (stationary limit surface): r_E = M + sqrt(M^2 - a^2 cos^2(theta))
    r_E = M + np.sqrt(M**2 - a**2 * np.cos(theta)**2)
    
    # Convert to Cartesian for plotting
    x_H = r_H * np.sin(theta)
    y_H = r_H * np.cos(theta)
    
    x_E = r_E * np.sin(theta)
    y_E = r_E * np.cos(theta)
    
    plt.figure(figsize=(6, 6))
    plt.plot(x_H, y_H, 'k-', linewidth=2, label='Event Horizon')
    plt.plot(x_E, y_E, 'b--', linewidth=2, label='Ergosphere')
    plt.fill_between(x_E, y_E, color='lightblue', alpha=0.3)
    plt.fill_between(x_H, y_H, color='black')
    
    plt.xlabel('x')
    plt.ylabel('z (axis of rotation)')
    plt.title('Kerr Black Hole: Event Horizon and Ergosphere')
    plt.legend()
    plt.axis('equal')
    plt.grid(True)
    plt.savefig('kerr_ergosphere.png')
    print('Plot saved as kerr_ergosphere.png')

if __name__ == '__main__':
    kerr_ergosphere()
```
