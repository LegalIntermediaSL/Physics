# Compendio Subtemático: Relatividad Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def spherical_geodesic():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Sphere
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = np.outer(np.cos(u), np.sin(v))
    y = np.outer(np.sin(u), np.sin(v))
    z = np.outer(np.ones(np.size(u)), np.cos(v))
    ax.plot_surface(x, y, z, color='cyan', alpha=0.3)
    
    # Geodesic (Great Circle)
    t = np.linspace(0, 2*np.pi, 100)
    gx = np.cos(t)
    gy = np.sin(t)
    gz = np.zeros_like(t)
    
    ax.plot(gx, gy, gz, color='red', linewidth=3, label='Geodesic')
    
    ax.set_title('Geodesic on a Curved Surface (Sphere)')
    plt.legend()
    plt.savefig('spherical_geodesic.png')
    print('Plot saved as spherical_geodesic.png')

if __name__ == '__main__':
    spherical_geodesic()
```
