# Compendio Subtemático: Relatividad Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def mass_energy():
    m0 = 1.0 # Rest mass in kg
    c = 3e8 # m/s
    v = np.linspace(0, 0.99*c, 500)
    
    E0 = m0 * c**2
    E_total = m0 * c**2 / np.sqrt(1 - (v/c)**2)
    KE = E_total - E0
    
    plt.figure(figsize=(8, 5))
    plt.plot(v/c, E_total / E0, label='Total Energy ($E/E_0$)', color='green')
    plt.plot(v/c, KE / E0, label='Kinetic Energy ($K/E_0$)', color='orange')
    plt.axhline(1, linestyle=':', color='black', label='Rest Energy ($E_0$)')
    plt.xlabel('Velocity (v/c)')
    plt.ylabel('Energy (Multiples of Rest Energy)')
    plt.title('Mass-Energy Equivalence ($E = \gamma mc^2$)')
    plt.ylim(0, 10)
    plt.grid(True)
    plt.legend()
    plt.savefig('mass_energy.png')
    print('Plot saved as mass_energy.png')

if __name__ == '__main__':
    mass_energy()
```
