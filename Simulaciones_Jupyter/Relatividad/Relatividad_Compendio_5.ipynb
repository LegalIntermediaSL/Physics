# Compendio Subtemático: Relatividad Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def twin_paradox():
    v = 0.8 # 80% speed of light
    c = 1.0
    gamma = 1 / np.sqrt(1 - (v/c)**2)
    
    time_earth = np.linspace(0, 10, 100) # Years
    
    # Earth twin ages normally
    age_earth = time_earth
    
    # Traveling twin ages slower
    age_traveler = time_earth / gamma
    
    plt.figure(figsize=(8, 5))
    plt.plot(time_earth, age_earth, label='Earth Twin Age', color='blue')
    plt.plot(time_earth, age_traveler, label='Traveling Twin Age (v=0.8c)', color='red')
    plt.xlabel('Earth Time (years)')
    plt.ylabel('Twin Age (years)')
    plt.title('Twin Paradox: Proper Time')
    plt.grid(True)
    plt.legend()
    plt.savefig('twin_paradox.png')
    print('Plot saved as twin_paradox.png')

if __name__ == '__main__':
    twin_paradox()
```
