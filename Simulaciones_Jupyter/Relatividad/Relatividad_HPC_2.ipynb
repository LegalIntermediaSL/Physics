# Supercomputación: Relatividad HPC 2

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Simplified equatorial geodesic of an extreme Kerr BH (a = M)
M = 1.0
a = M # Extreme Kerr

# Effective potential for equatorial geodesics
def V_eff(r, E, L):
    delta = r**2 - 2*M*r + a**2
    return -(M/r) + L**2/(2*r**2) - (M*(L-a*E)**2)/(r**3) - (E**2-1)/2

r = np.linspace(1.1*M, 10*M, 500)
E = 0.95
L = 3.0

v = V_eff(r, E, L)

plt.plot(r, v)
plt.axhline(0, color='r', linestyle='--')
plt.title('Effective Potential - Extreme Kerr Equatorial')
plt.xlabel('r')
plt.ylabel('V_eff')
plt.savefig('Relatividad_HPC_2_output.png')
```
