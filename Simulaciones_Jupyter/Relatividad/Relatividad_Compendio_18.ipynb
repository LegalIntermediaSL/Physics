# Compendio Subtemático: Relatividad Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def hubbles_law():
    H0 = 70.0 # km/s/Mpc
    d = np.linspace(0, 100, 100) # Distance in Mpc
    
    v = H0 * d
    
    plt.figure(figsize=(8, 5))
    plt.plot(d, v, label='$v = H_0 d$', color='blue')
    plt.xlabel('Distance (Mpc)')
    plt.ylabel('Recession Velocity (km/s)')
    plt.title("Hubble's Law: Expanding Universe")
    plt.grid(True)
    plt.legend()
    plt.savefig('hubbles_law.png')
    print('Plot saved as hubbles_law.png')

if __name__ == '__main__':
    hubbles_law()
```
