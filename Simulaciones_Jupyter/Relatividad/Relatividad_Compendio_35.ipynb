# Compendio Subtemático: Relatividad Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def relativistic_aberration():
    theta_true = np.linspace(0, np.pi, 500)
    v = 0.8 # v/c
    
    # cos(theta_obs) = (cos(theta_true) + v) / (1 + v*cos(theta_true))
    cos_obs = (np.cos(theta_true) + v) / (1 + v * np.cos(theta_true))
    theta_obs = np.arccos(cos_obs)
    
    plt.figure(figsize=(8, 5))
    plt.plot(np.degrees(theta_true), np.degrees(theta_obs), color='green')
    plt.plot([0, 180], [0, 180], 'k--', label='No aberration ($v=0$)')
    plt.xlabel('True Angle (degrees)')
    plt.ylabel('Observed Angle (degrees)')
    plt.title('Relativistic Aberration of Light ($v=0.8c$)')
    plt.grid(True)
    plt.legend()
    plt.savefig('relativistic_aberration.png')
    print('Plot saved as relativistic_aberration.png')

if __name__ == '__main__':
    relativistic_aberration()
```
