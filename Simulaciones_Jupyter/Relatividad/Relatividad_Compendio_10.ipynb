# Compendio Subtemático: Relatividad Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def light_cone():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    t = np.linspace(-5, 5, 50)
    theta = np.linspace(0, 2*np.pi, 50)
    T, Theta = np.meshgrid(t, theta)
    
    R = np.abs(T)
    X = R * np.cos(Theta)
    Y = R * np.sin(Theta)
    
    ax.plot_surface(X, Y, T, cmap='viridis', alpha=0.6)
    ax.plot([0,0], [0,0], [-5,5], color='red', label='Worldline of stationary particle')
    
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('Time (ct)')
    ax.set_title('Spacetime Light Cone')
    plt.legend()
    plt.savefig('light_cone.png')
    print('Plot saved as light_cone.png')

if __name__ == '__main__':
    light_cone()
```
