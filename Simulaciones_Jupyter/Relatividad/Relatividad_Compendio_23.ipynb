# Compendio Subtemático: Relatividad Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def thomas_precession():
    v = np.linspace(0, 0.99, 500)
    gamma = 1 / np.sqrt(1 - v**2)
    
    # Precession rate omega_T ~ (gamma^2 / (gamma + 1)) * (a x v)
    # We plot the factor gamma^2 / (gamma + 1)
    factor = gamma**2 / (gamma + 1)
    
    plt.figure(figsize=(8, 5))
    plt.plot(v, factor, color='darkorange')
    plt.xlabel('Velocity (v/c)')
    plt.ylabel(r'Thomas Precession Factor $\frac{\gamma^2}{\gamma + 1}$')
    plt.title('Thomas Precession Kinematic Factor')
    plt.grid(True)
    plt.savefig('thomas_precession.png')
    print('Plot saved as thomas_precession.png')

if __name__ == '__main__':
    thomas_precession()
```
