# Compendio Subtemático: Relatividad Compendio 22

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def relativistic_beaming():
    theta_prime = np.linspace(0, np.pi, 500)
    v = 0.9 # v/c
    beta = v
    
    # cos(theta) = (cos(theta') + beta) / (1 + beta*cos(theta'))
    cos_theta = (np.cos(theta_prime) + beta) / (1 + beta * np.cos(theta_prime))
    theta = np.arccos(cos_theta)
    
    plt.figure(figsize=(6, 6), subplot_kw={'projection': 'polar'})
    plt.plot(theta_prime, np.ones_like(theta_prime), '--', label='Rest Frame Isotropic', color='gray')
    plt.plot(theta, np.ones_like(theta), label='Moving Frame ($v=0.9c$)', color='red')
    plt.title('Relativistic Beaming (Headlight Effect)')
    plt.legend(loc='lower right')
    plt.savefig('relativistic_beaming.png')
    print('Plot saved as relativistic_beaming.png')

if __name__ == '__main__':
    relativistic_beaming()
```
