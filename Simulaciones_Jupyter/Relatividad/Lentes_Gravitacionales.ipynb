# Lentes Gravitacionales (Anillo de Einstein)

Deformación óptica del espaciotiempo masivo.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 200)
r = 2.0
x, y = r*np.cos(theta), r*np.sin(theta)

plt.figure(figsize=(6,6))
plt.plot(x, y, color='blue', lw=3, label='Anillo de Einstein')
plt.scatter([0], [0], color='black', s=100, label='Masa Lente')
plt.axis('equal')
plt.title('Lente Gravitacional Fuerte')
plt.legend()
plt.grid(False)
plt.show()
```
