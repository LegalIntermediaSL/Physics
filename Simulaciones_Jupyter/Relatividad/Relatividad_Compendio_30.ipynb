# Compendio Subtemático: Relatividad Compendio 30

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def equivalence_principle():
    t = np.linspace(0, 2, 100)
    g = 9.8
    
    # Drop an apple in a uniform gravitational field
    y_grav = -0.5 * g * t**2
    
    # Drop an apple in an accelerating rocket (no gravity)
    # Rocket accelerates up with a, apple stays still, so relative to rocket it accelerates down
    y_rocket = -0.5 * g * t**2
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    
    ax1.plot(t, y_grav, color='blue')
    ax1.set_title('Stationary in Gravity $g$')
    ax1.set_xlabel('Time (s)')
    ax1.set_ylabel('Height (m)')
    
    ax2.plot(t, y_rocket, color='red')
    ax2.set_title('Accelerating at $a=g$ in Space')
    ax2.set_xlabel('Time (s)')
    
    plt.suptitle('Einstein Equivalence Principle')
    plt.savefig('equivalence_principle.png')
    print('Plot saved as equivalence_principle.png')

if __name__ == '__main__':
    equivalence_principle()
```
