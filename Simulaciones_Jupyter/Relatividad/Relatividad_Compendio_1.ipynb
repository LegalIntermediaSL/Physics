# Compendio Subtemático: Relatividad Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def time_dilation():
    c = 3e8
    v = np.linspace(0, 0.99*c, 500)
    gamma = 1 / np.sqrt(1 - (v**2 / c**2))
    
    plt.figure(figsize=(8, 5))
    plt.plot(v/c, gamma, label=r'$\gamma = \frac{1}{\sqrt{1 - v^2/c^2}}$', color='blue')
    plt.xlabel('Velocity (v/c)')
    plt.ylabel('Lorentz Factor ($\gamma$)')
    plt.title('Relativistic Time Dilation')
    plt.grid(True)
    plt.legend()
    plt.savefig('time_dilation.png')
    print('Plot saved as time_dilation.png')

if __name__ == '__main__':
    time_dilation()
```
