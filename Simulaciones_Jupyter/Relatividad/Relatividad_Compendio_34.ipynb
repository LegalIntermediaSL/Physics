# Compendio Subtemático: Relatividad Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def cmb_dipole():
    theta = np.linspace(0, np.pi, 100)
    phi = np.linspace(0, 2*np.pi, 100)
    Theta, Phi = np.meshgrid(theta, phi)
    
    T_0 = 2.725 # CMB rest temperature
    v = 0.0012 # v/c for Earth relative to CMB
    
    # Relativistic Doppler shift for temperature
    # T(theta) = T_0 * sqrt(1 - v^2) / (1 - v*cos(theta))
    # Approximation for small v: T = T_0 (1 + v*cos(theta))
    
    gamma = 1 / np.sqrt(1 - v**2)
    T_obs = T_0 * gamma / (1 - v * np.cos(Theta))
    
    delta_T = T_obs - T_0
    
    plt.figure(figsize=(8, 5))
    plt.contourf(Phi, Theta, delta_T, levels=50, cmap='coolwarm')
    plt.colorbar(label='$\Delta T$ (K)')
    plt.xlabel(r'$\phi$ (Azimuthal Angle)')
    plt.ylabel(r'$\theta$ (Polar Angle)')
    plt.title('CMB Dipole Anisotropy due to Observer Motion')
    plt.savefig('cmb_dipole.png')
    print('Plot saved as cmb_dipole.png')

if __name__ == '__main__':
    cmb_dipole()
```
