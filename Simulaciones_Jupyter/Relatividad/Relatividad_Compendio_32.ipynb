# Compendio Subtemático: Relatividad Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def wormhole():
    # Ellis drainhole (simplest traversable wormhole)
    # r^2 = rho^2 + a^2, z(rho) ~ a * ln(rho/a + sqrt((rho/a)^2 - 1))
    # Wait, embedding formula for Ellis wormhole:
    # z(r) = int sqrt(r^2 / (r^2 - a^2) - 1) dr? No, z(r) = a * arccosh(r/a)
    
    a = 1.0 # Throat radius
    r = np.linspace(a, 5*a, 100)
    theta = np.linspace(0, 2*np.pi, 100)
    
    R, Theta = np.meshgrid(r, theta)
    X = R * np.cos(Theta)
    Y = R * np.sin(Theta)
    
    Z = a * np.arccosh(R/a)
    
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Upper half
    ax.plot_surface(X, Y, Z, cmap='plasma', alpha=0.8)
    # Lower half
    ax.plot_surface(X, Y, -Z, cmap='plasma', alpha=0.8)
    
    ax.set_title('Embedding Diagram of a Wormhole')
    plt.axis('off')
    plt.savefig('wormhole.png')
    print('Plot saved as wormhole.png')

if __name__ == '__main__':
    wormhole()
```
