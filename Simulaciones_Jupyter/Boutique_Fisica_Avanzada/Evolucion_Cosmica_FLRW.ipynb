# Evolución Cósmica y Energía Oscura (Métrica FLRW)
La Ecuación de Friedmann dicta cómo el universo se expande (o se contrae) basado en su contenido de energía y materia. 

En esta simulación, resolveremos numéricamente la evolución del factor de escala $a(t)$ bajo tres escenarios: 
1. Universo dominado por Materia ($a \propto t^{2/3}$)
2. Universo dominado por Radiación ($a \propto t^{1/2}$)
3. Nuestro universo actual: dominado por la Energía Oscura (Constante Cosmológica $\Lambda$), produciendo una expansión acelerada exponencial tipo de Sitter.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parámetros cosmológicos actuales (Planck)
H0 = 1.0  # Hubble normalizado
Omega_M = 0.3  # Materia
Omega_R = 0.0  # Radiación (despreciable hoy)
Omega_Lambda = 0.7  # Energía Oscura

# Ecuación diferencial de Friedmann: da/dt = a * H0 * sqrt(O_m/a^3 + O_r/a^4 + O_L)
def friedmann(a, t, Om, Or, OL):
    # Evitar divisiones por cero en el Big Bang (a=0)
    if a < 1e-5: a = 1e-5
    return a * H0 * np.sqrt(Om/a**3 + Or/a**4 + OL)

t_span = np.linspace(0, 3, 500)
a_init = 1e-3  # Un poco después del Big Bang

# Escenarios
a_mat = odeint(friedmann, a_init, t_span, args=(1.0, 0.0, 0.0))
a_rad = odeint(friedmann, a_init, t_span, args=(0.0, 1.0, 0.0))
a_dark = odeint(friedmann, a_init, t_span, args=(Omega_M, Omega_R, Omega_Lambda))

plt.figure(figsize=(8, 5))
plt.plot(t_span, a_mat, label='Dominado por Materia ($\Omega_M=1$)', linestyle='--')
plt.plot(t_span, a_rad, label='Dominado por Radiación ($\Omega_R=1$)', linestyle='-.')
plt.plot(t_span, a_dark, label='Con Energía Oscura ($\Omega_\\Lambda=0.7$)', color='purple', lw=2)

plt.title("Evolución del Factor de Escala Cósmico $a(t)$")
plt.xlabel("Tiempo Cósmico (Unidades Normalizadas)")
plt.ylabel("Factor de Escala $a(t)$")
plt.legend()
plt.grid(True)
plt.show()
```
