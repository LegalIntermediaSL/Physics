# Evaporación de Agujeros Negros (Radiación de Hawking)
Un Agujero Negro no es negro a nivel cuántico. Emite radiación de cuerpo negro debido a fluctuaciones del vacío en el horizonte de sucesos (Radiación de Hawking). Su temperatura es inversamente proporcional a su masa ($T \propto 1/M$).

A medida que irradia fotones, pierde masa ($E=mc^2$). A medida que pierde masa, se calienta más, irradiando más rápido. Esto crea un ciclo descontrolado (runaway) que culmina en una explosión final de Rayos Gamma. ¡Resolvamos la ecuación diferencial de la evaporación!

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Ecuación diferencial simplificada de pérdida de masa
# dM/dt = - K / M^2 (donde K agrupa constantes G, hbar, c)
K = 1.0

def evaporacion(t, M):
    if M[0] <= 0: return [0]
    return [-K / (M[0]**2)]

# Evento para detener la integración cuando M llega a cero
def masa_cero(t, M):
    return M[0] - 1e-3
masa_cero.terminal = True

# Integración
M_inicial = 5.0
sol = solve_ivp(evaporacion, [0, 100], [M_inicial], events=masa_cero, max_step=0.1)

t = sol.t
M = sol.y[0]
# Temperatura es 1/M (normalizada)
T = 1.0 / M

fig, ax1 = plt.subplots(figsize=(8, 5))

# Gráfica Masa
color1 = 'tab:blue'
ax1.set_xlabel('Tiempo de Vida')
ax1.set_ylabel('Masa del Agujero Negro (M)', color=color1)
ax1.plot(t, M, color=color1, lw=3)
ax1.tick_params(axis='y', labelcolor=color1)

# Gráfica Temperatura
ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Temperatura de Hawking (T)', color=color2)
ax2.plot(t, T, color=color2, linestyle='--')
ax2.tick_params(axis='y', labelcolor=color2)
ax2.set_ylim(0, 10)

plt.title("Evaporación Explosiva de Hawking")
fig.tight_layout()
plt.grid(True)
plt.show()
```
