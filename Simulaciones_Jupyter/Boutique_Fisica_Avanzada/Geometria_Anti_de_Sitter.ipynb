# Geometría Anti-de Sitter (AdS) y el Disco de Poincaré
El espacio Anti-de Sitter (AdS) es una geometría hiperbólica con curvatura negativa constante. Es el teatro donde ocurre la dualidad holográfica (Maldacena). Una rebanada espacial en 2D de AdS se representa famosamente como el **Disco de Poincaré**.

En esta simulación, observaremos cómo la 'distancia física' diverge al acercarse al borde. Las líneas rectas en AdS parecen arcos curvos (geodésicas) desde nuestra perspectiva euclídea plana.

```python
import numpy as np
import matplotlib.pyplot as plt

# Geodésicas en el disco de Poincaré son arcos circulares perpendiculares al borde
theta = np.linspace(0, 2*np.pi, 200)

fig, ax = plt.subplots(figsize=(7,7))
# El borde conforme (infinito espacial en AdS)
ax.plot(np.cos(theta), np.sin(theta), 'k-', lw=3, label="Borde Conforme (CFT)")

# Trazar varias geodésicas (luz viajando por AdS)
for R_center in [1.2, 1.5, 2.0]:
    for angle in np.linspace(0, 2*np.pi, 6, endpoint=False):
        cx = R_center * np.cos(angle)
        cy = R_center * np.sin(angle)
        R_circle = np.sqrt(R_center**2 - 1)
        
        t = np.linspace(0, 2*np.pi, 100)
        x_c = cx + R_circle * np.cos(t)
        y_c = cy + R_circle * np.sin(t)
        
        # Recortar el círculo fuera del disco de Poincaré
        mask = (x_c**2 + y_c**2) <= 1
        ax.plot(x_c[mask], y_c[mask], color='blue', alpha=0.5)

ax.set_aspect('equal')
ax.axis('off')
ax.set_title("Geodésicas en Geometría Hiperbólica (AdS Espacial)")
ax.legend(loc='upper right')
plt.show()
```
