# Materia Oscura: Perfil de Navarro-Frenk-White (NFW)
La evidencia física más aplastante de la Materia Oscura proviene de las Curvas de Rotación Galáctica. Las estrellas en los bordes de las galaxias orbitan ridículamente rápido, en lugar de frenarse según la ley de Kepler $v \propto 1/\sqrt{r}$.

Para que la velocidad se mantenga plana, la galaxia debe estar envuelta en un colosal halo esférico de Materia Oscura. Las simulaciones cuánticas N-cuerpos demuestran que este halo sigue el **Perfil de Densidad de Navarro-Frenk-White (NFW)**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes arbitrarias para visualización
G = 1.0
M_visible = 10.0  # Masa bariónica central concentrada
rho_0 = 5.0       # Densidad central del halo oscuro
R_s = 3.0         # Radio de escala del halo oscuro

r = np.linspace(0.1, 15, 500)

# 1. Velocidad Kepleriana Clásica (Solo materia visible puntual)
# v = sqrt(GM/r)
v_kepler = np.sqrt(G * M_visible / r)

# 2. Masa encerrada por el Halo NFW
# M(r) = 4*pi*rho_0*R_s^3 * [ln((R_s+r)/R_s) - r/(R_s+r)]
M_halo = 4 * np.pi * rho_0 * R_s**3 * (np.log((R_s + r)/R_s) - r/(R_s + r))

# Velocidad generada por el Halo NFW
v_halo = np.sqrt(G * M_halo / r)

# Velocidad Total Observada (Suma en cuadratura)
v_total = np.sqrt(v_kepler**2 + v_halo**2)

plt.figure(figsize=(8, 5))
plt.plot(r, v_kepler, '--', color='grey', label='Esperado (Solo Materia Visible)')
plt.plot(r, v_halo, '-.', color='purple', label='Halo NFW (Materia Oscura)')
plt.plot(r, v_total, '-', color='red', lw=3, label='Observado (Rotación Plana)')

plt.title("Curva de Rotación Galáctica y Materia Oscura")
plt.xlabel("Distancia al Centro Galáctico (r)")
plt.ylabel("Velocidad Orbital (v)")
plt.legend()
plt.grid(True)
plt.show()
```
