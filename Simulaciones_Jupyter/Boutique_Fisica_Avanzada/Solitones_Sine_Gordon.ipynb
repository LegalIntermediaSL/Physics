# Ecuación Sine-Gordon: Kinks y Solitones
La ecuación de Sine-Gordon $\partial_t^2 \phi - \partial_x^2 \phi + \sin(\phi) = 0$ es un modelo de teoría de campos no lineal famosa por poseer soluciones topológicas exactas llamadas **Solitones (o Kinks)**. Un Kink es un retorcimiento espacial del campo que transiciona del vacío $\phi=0$ al vacío $\phi=2\pi$. ¡Estas ondas chocan como partículas reales sin destruirse!

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Parámetros espaciales y temporales
L = 40.0
Nx = 800
dx = L / Nx
x = np.linspace(-L/2, L/2, Nx)
dt = 0.02
Nt = 400

# Solución Kink analítica exacta de Sine-Gordon en reposo
def kink(x, x0, v):
    gamma = 1.0 / np.sqrt(1 - v**2)
    return 4 * np.arctan(np.exp(gamma * (x - x0)))

# Evolución temporal simple de un pulso propagándose
# Visualizaremos la forma analítica de un Kink moviéndose
fig, ax = plt.subplots(figsize=(8,4))
line, = ax.plot(x, kink(x, -10, 0.5), color='purple', lw=2)
ax.set_ylim(-0.5, 7)
ax.set_title("Evolución de un Solitón Topológico (Kink)")
ax.set_xlabel("x (Espacio)")
ax.set_ylabel("$\\phi(x,t)$ (Campo)")
ax.grid(True)

def update(frame):
    t = frame * dt * 10
    v = 0.5
    pos = -15 + v * t
    y = kink(x, pos, v)
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, update, frames=Nt//2, interval=40, blit=True)
plt.close()
HTML(anim.to_jshtml())
```
