# Transición de Fase: Simulación de Monte Carlo del Modelo de Ising
El **Modelo de Ising** es el paradigma central de la mecánica estadística y la materia condensada para estudiar el Ferromagnetismo y las transiciones de fase.

En esta simulación bidimensional, usamos el **Algoritmo de Metrópolis-Hastings** para iterar estocásticamente una red de espines ($\pm 1$).
Observaremos gráficamente cómo, por debajo de la temperatura crítica de Curie ($T_C \approx 2.269 J/k_B$), los espines rompen la simetría y se agrupan en dominios magnéticos gigantescos (orden espontáneo). Por encima de $T_C$, el ruido térmico destruye brutalmente el orden (caos paramagnético).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del modelo
L = 50          # Tamaño de la red (L x L)
J = 1.0         # Constante de intercambio (Ferromagnético si J>0)
kB = 1.0        # Constante de Boltzmann (unidades reducidas)

# Temperaturas a simular (T_critica teórica ~ 2.269)
temps = [1.5, 2.269, 3.5]
steps = 100000  # Pasos de Monte Carlo

def init_grid(L):
    """Inicializa una red 2D de espines aleatorios (+1 o -1)"""
    return np.random.choice([-1, 1], size=(L, L))

def metropolis_step(grid, T):
    """Ejecuta un paso de Monte Carlo usando el algoritmo de Metrópolis"""
    for _ in range(L * L):
        # Elegir un espín al azar
        x, y = np.random.randint(0, L, 2)
        s = grid[x, y]
        
        # Calcular la energía de interacción con los 4 vecinos (condiciones de frontera periódicas)
        neighbors = grid[(x+1)%L, y] + grid[x, (y+1)%L] + grid[(x-1)%L, y] + grid[x, (y-1)%L]
        dE = 2 * J * s * neighbors
        
        # Aceptar o rechazar el cambio (flip)
        if dE < 0:
            grid[x, y] *= -1
        elif np.random.rand() < np.exp(-dE / (kB * T)):
            grid[x, y] *= -1
    return grid

# Simulación y visualización
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, T in enumerate(temps):
    grid = init_grid(L)
    # Termalización del sistema
    for _ in range(300):
        grid = metropolis_step(grid, T)
        
    ax = axes[i]
    ax.imshow(grid, cmap='coolwarm', interpolation='nearest')
    ax.set_title(f'T = {T} (T_c ~ 2.269)')
    ax.axis('off')

plt.suptitle("Modelo de Ising 2D: Transición de Fase Ferromagnética (Monte Carlo)", fontsize=16)
plt.tight_layout()
plt.show()

print("Azul: Espín Abajo (-1). Rojo: Espín Arriba (+1).")
print("A baja T: Dominios ordenados gigantes. En T_c: Fluctuaciones críticas fractales. A alta T: Ruido blanco paramagnético.")
```
