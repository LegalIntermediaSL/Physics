# Estructura Fina del Hidrógeno

Ruptura de la degeneración energética debido al acoplamiento espín-órbita y correcciones cinemáticas relativistas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Diagrama de niveles esquemático para n=2
fig, ax = plt.subplots(figsize=(6,6))
ax.axhline(0, xmin=0.1, xmax=0.3, color='k', lw=3, label='Bohr (Sin estructura fina)')

# Separación relativista y espín-órbita
ax.axhline(-0.5, xmin=0.5, xmax=0.7, color='r', lw=2, label='2P_{1/2} / 2S_{1/2}')
ax.axhline(0.2, xmin=0.5, xmax=0.7, color='b', lw=2, label='2P_{3/2}')

# Lamb Shift
ax.axhline(-0.45, xmin=0.8, xmax=0.95, color='g', lw=2, label='2S_{1/2} (Lamb Shift)')
ax.axhline(-0.55, xmin=0.8, xmax=0.95, color='orange', lw=2, label='2P_{1/2}')

plt.title('Diagrama de Niveles: Estructura Fina y Lamb Shift (n=2)')
plt.ylabel('Desplazamiento Energético')
plt.xticks([])
plt.legend()
plt.show()
```
