# Oscilaciones de Rabi

Interacción de un sistema cuántico de dos niveles con un campo electromagnético monocromático (ej. láser).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 500)
Omega = 2.0 # Frecuencia de Rabi
Delta = 0.5 # Detuning (desacople)

Omega_eff = np.sqrt(Omega**2 + Delta**2)
P_excited = (Omega**2 / Omega_eff**2) * np.sin(Omega_eff * t / 2)**2
P_ground = 1 - P_excited

plt.figure(figsize=(8,4))
plt.plot(t, P_excited, 'r-', label='Probabilidad Estado Excitado |e>')
plt.plot(t, P_ground, 'b--', label='Probabilidad Estado Base |g>')
plt.title('Oscilaciones Cuánticas de Rabi')
plt.xlabel('Tiempo')
plt.legend()
plt.grid()
plt.show()
```
