# Efecto Aharonov-Bohm

Demostración geométrica del cambio de fase topológico adquirido por un electrón rodeando un solenoide, pese a que $B=0$ en su trayectoria.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 200)
r = 2.0
x, y = r*np.cos(theta), r*np.sin(theta)

# Fase adquirida dependiente del flujo magnético interior
Phi = 1.5 * np.pi # Flujo
fase = np.exp(1j * Phi * (theta/(2*np.pi)))

plt.figure(figsize=(6,6))
sc = plt.scatter(x, y, c=np.angle(fase), cmap='hsv', s=50)
plt.scatter([0], [0], color='black', marker='x', s=100, label='Solenoide (B interno)')
plt.title('Cambio de Fase de Aharonov-Bohm')
plt.colorbar(sc, label='Fase de la Función de Onda')
plt.axis('equal')
plt.legend()
plt.show()
```
