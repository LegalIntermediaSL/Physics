# [QM-07] SUSY Cuántica: Potenciales Isospectrales

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh

# Mecánica Cuántica Supersimétrica (SUSY QM): Potenciales Compañeros
# Demostramos que dos universos con reglas (Potenciales) distintos pueden ser Isospectrales
# (mismos niveles de energía) mediante Álgebra de Factorización.

N_x = 500
L = 10.0
x = np.linspace(-L/2, L/2, N_x)
dx = x[1] - x[0]

# Operador de segunda derivada discreta
D2 = (np.diag(-2 * np.ones(N_x)) + np.diag(np.ones(N_x-1), 1) + np.diag(np.ones(N_x-1), -1)) / dx**2

# Elegimos un "Superpotencial" W(x). Usaremos el Oscilador Armónico como base: W(x) = x
W = x
dW = np.ones_like(x) # Derivada de W(x)

# Construcción de los Potenciales Compañeros (Ecuaciones de Riccati en SUSY)
# V_1 = W^2 - W'
# V_2 = W^2 + W'
V1 = W**2 - dW
V2 = W**2 + dW

# Construimos los Hamiltonianos de ambos universos (H = -d^2/dx^2 + V(x))
H1 = -D2 + np.diag(V1)
H2 = -D2 + np.diag(V2)

# Diagonalizamos matemáticamente para encontrar los niveles de Energía
eigvals_1, eigvecs_1 = eigh(H1)
eigvals_2, eigvecs_2 = eigh(H2)

plt.figure(figsize=(10, 6))

plt.plot(x, V1, 'b-', lw=3, label="Potencial SUSY 1 ($V_1 = x^2 - 1$)")
plt.plot(x, V2, 'r--', lw=3, label="Potencial SUSY 2 ($V_2 = x^2 + 1$)")

# Dibujar las líneas de energía cuántica horizontales
for n in range(5): # Primeros 5 estados
    # El Universo 1 tiene un Estado Base de energía cero E=0 (Aniquilado por SUSY)
    plt.axhline(eigvals_1[n], color='blue', alpha=0.5, linestyle='-', lw=1.5)
    
    # El Universo 2 tiene EXACTAMENTE las mismas energías a partir del n=1 del primer universo
    plt.axhline(eigvals_2[n], color='red', alpha=0.5, linestyle=':', lw=3)

plt.title("Isospectralidad en SUSY QM (Niveles Cuánticos Emparejados)", fontsize=14)
plt.xlabel("Posición Espacial $x$")
plt.ylabel("Energía Potencial / Eigenvalores")
plt.legend()
plt.ylim(-2, 10)
plt.xlim(-4, 4)
plt.grid(True)
plt.show()

# FÍSICA: El Álgebra de Supersimetría actúa como un milagro. V1 y V2 son parábolas
# desplazadas verticalmente por 2 unidades. Matemáticamente, el primer nivel excitado 
# de V1 tiene energía EXACTAMENTE IGUAL al estado fundamental de V2, demostrando
# el emparejamiento perfecto Fermión-Bosón unidimensional de la Teoría SUSY de Cuerdas.

```
