# Investigación Avanzada: Mecanica Cuantica Jaynes Cummings

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

def main():
    # Jaynes-Cummings Model (Atom + Cavity Mode)
    omega_a = 1.0 # Atom transition freq
    omega_c = 1.0 # Cavity freq
    g = 0.1 # Coupling strength
    N_max = 20 # Max photon number
    
    dim = 2 * N_max
    H = np.zeros((dim, dim))
    
    for n in range(N_max):
        # Basis: |g, n> -> 2n, |e, n> -> 2n+1
        # Cavity energy
        H[2*n, 2*n] = n * omega_c
        H[2*n+1, 2*n+1] = n * omega_c + omega_a
        
        # Interaction g(a+ sigma_- + a sigma_+)
        if n > 0:
            H[2*n, 2*n-1] = g * np.sqrt(n)
            H[2*n-1, 2*n] = g * np.sqrt(n)
            
    # Initial state: Atom in excited state, cavity in vacuum |e, 0> = index 1
    psi0 = np.zeros(dim)
    psi0[1] = 1.0
    
    t_vals = np.linspace(0, 100, 1000)
    evals, evecs = la.eigh(H)
    
    prob_e = np.zeros_like(t_vals)
    for i, t in enumerate(t_vals):
        U = evecs @ np.diag(np.exp(-1j * evals * t)) @ evecs.conj().T
        psi_t = U @ psi0
        # Prob of being excited is sum of abs(psi[2n+1])**2
        prob_e[i] = np.sum(np.abs(psi_t[1::2])**2)
        
    plt.figure(figsize=(8,5))
    plt.plot(t_vals, prob_e, 'r', label=r'$P_e(t)$')
    plt.title("Rabi Oscillations in Jaynes-Cummings Model")
    plt.xlabel("Time")
    plt.ylabel("Excited State Probability")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
