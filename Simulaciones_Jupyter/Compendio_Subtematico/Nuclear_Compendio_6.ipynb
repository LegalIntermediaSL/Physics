# Compendio Subtemático: Nuclear Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    E = np.linspace(0, 2.5, 100)
    Q = 2.5
    # Simplified Fermi distribution for beta decay spectrum
    spectrum = np.sqrt(E) * (Q - E)**2
    
    plt.plot(E, spectrum, color='blue')
    plt.title('Beta Decay Energy Spectrum (Simplified)')
    plt.xlabel('Kinetic Energy of Electron')
    plt.ylabel('Intensity / Probability')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
