# Compendio Subtemático: Astrofisica Compendio 22

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def doppler_shift():
    wav0 = 500 # nm
    v = np.linspace(-30000, 30000, 100) # km/s
    c = 3e5 # km/s
    wav_obs = wav0 * np.sqrt((1 + v/c)/(1 - v/c))
    
    plt.figure(figsize=(8, 6))
    plt.plot(v, wav_obs - wav0, 'g-')
    plt.title("Relativistic Doppler Shift")
    plt.xlabel("Velocity (km/s) [>0 is recession]")
    plt.ylabel("$\Delta \lambda$ (nm)")
    plt.axhline(0, color='k', linestyle='--')
    plt.axvline(0, color='k', linestyle='--')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    doppler_shift()
```
