# Compendio Subtemático: Fluidos Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def couette():
    y = np.linspace(0, 0.1, 100) # m
    h = 0.1 # m
    U = 2 # m/s
    u = U * (y / h)
    plt.plot(u, y)
    plt.xlabel('Velocity u (m/s)')
    plt.ylabel('Distance y (m)')
    plt.title('Couette Flow Profile')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    couette()
```
