# Compendio Subtemático: Info Cuantica Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    messages = ['00', '01', '10', '11']
    transmitted_bits = [1, 1, 1, 1] # 1 qubit transmits 2 classical bits
    
    plt.bar(messages, transmitted_bits, color='orange')
    plt.title('Superdense Coding (1 Qubit -> 2 Classical Bits)')
    plt.xlabel('Classical Message')
    plt.ylabel('Qubits Sent')
    plt.show()

if __name__ == '__main__':
    main()
```
