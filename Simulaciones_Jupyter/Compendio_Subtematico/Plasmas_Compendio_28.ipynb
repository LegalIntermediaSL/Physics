# Compendio Subtemático: Plasmas Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def coulomb_collisions():
    x = np.linspace(-5, 5, 100)
    b = 1.0
    y = np.sqrt(b**2 + (1/x)**2)
    plt.plot(x, y)
    plt.title("Rutherford Scattering (Toy trajectory)")
    plt.grid(True); plt.savefig("coulomb.png")
if __name__ == "__main__": coulomb_collisions()
```
