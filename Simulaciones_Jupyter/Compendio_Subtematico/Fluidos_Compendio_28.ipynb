# Compendio Subtemático: Fluidos Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def sound_speed():
    K = np.linspace(1e9, 3e9, 100) # Bulk modulus Pa
    rho = 1000
    c = np.sqrt(K / rho)
    plt.plot(K/1e9, c)
    plt.xlabel('Bulk Modulus (GPa)')
    plt.ylabel('Speed of Sound (m/s)')
    plt.title('Speed of Sound in Liquid vs Bulk Modulus')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    sound_speed()
```
