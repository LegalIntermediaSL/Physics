# Compendio Subtemático: Materia Condensada Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Bloch's Theorem: Wavefunction in a periodic potential
x = np.linspace(0, 4 * np.pi, 500)
k = 0.5 # Bloch wavevector

# Periodic function u_k(x) = u_k(x + a)
u_x = 1.0 + 0.3 * np.cos(2 * x) # Period a = pi

# Bloch wavefunction psi_k(x) = e^{ikx} u_k(x)
psi_x = np.exp(1j * k * x) * u_x

# Probability density
prob_density = np.abs(psi_x)**2

plt.figure(figsize=(8, 5))
plt.plot(x, np.real(psi_x), label='Re($\psi(x)$)', color='blue')
plt.plot(x, np.imag(psi_x), label='Im($\psi(x)$)', color='red', linestyle='--')
plt.plot(x, prob_density, label='$|\psi(x)|^2$', color='black', lw=2)
plt.title("Bloch Wavefunction in 1D Periodic Potential")
plt.xlabel("Position x")
plt.ylabel("Amplitude / Probability Density")
plt.legend()
plt.grid(True)
plt.savefig("bloch_theorem.png")
```
