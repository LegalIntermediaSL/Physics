# Compendio Subtemático: Fisica Atomica Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def photoelectric_effect(work_function):
    # h in eV*s, c in m/s
    hc = 1240 # eV*nm
    
    wavelengths = np.linspace(100, 1000, 200) # nm
    freq = 3e8 / (wavelengths * 1e-9)
    
    energy_photons = hc / wavelengths
    
    K_max = energy_photons - work_function
    K_max = np.maximum(K_max, 0)
    
    plt.figure(figsize=(8, 6))
    plt.plot(freq, K_max, 'b-', linewidth=2)
    
    threshold_freq = (work_function / hc) * 3e8 * 1e9
    plt.axvline(threshold_freq, color='r', linestyle='--', label=f'Threshold frequency $f_0$')
    
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Max Kinetic Energy $K_{max}$ (eV)')
    plt.title(f'Photoelectric Effect (Work function = {work_function} eV)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    photoelectric_effect(2.28) # Sodium
```
