# Compendio Subtemático: Termodinamica Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

V = np.linspace(1, 10, 100)
T_vals = [300, 400, 500]
R = 8.314
n = 1

for T in T_vals:
    P = n * R * T / V
    plt.plot(V, P, label=f'T = {T} K')

plt.title('Ideal Gas Law: Isotherms')
plt.xlabel('Volume (m^3)')
plt.ylabel('Pressure (Pa)')
plt.legend()
plt.grid(True)
plt.savefig('ideal_gas.png')
```
