# Compendio Subtemático: Fisica Atomica Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def natural_broadening(tau):
    # Natural line width Delta nu = 1 / (2 * pi * tau)
    delta_nu_N = 1 / (2 * np.pi * tau)
    
    nu = np.linspace(-5 * delta_nu_N, 5 * delta_nu_N, 500)
    
    # Lorentzian profile
    L = (delta_nu_N / (2 * np.pi)) / (nu**2 + (delta_nu_N / 2)**2)
    
    plt.figure(figsize=(8, 6))
    plt.plot(nu, L, 'r-', label=f'$\\tau$ = {tau} s')
    plt.xlabel('Frequency detuning $\\nu - \\nu_0$ (Hz)')
    plt.ylabel('Intensity (normalized)')
    plt.title(f'Natural Line Broadening (Lorentzian Profile)')
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    natural_broadening(1e-8) # 10 ns typical lifetime
```
