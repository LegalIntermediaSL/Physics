# Compendio Subtemático: Mecanica Cuantica Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
theta = np.linspace(-np.pi/2, np.pi/2, 500)
I = np.cos(theta)**2 * (1 + np.cos(2*np.pi + theta))
plt.plot(theta, I)
plt.title('Aharonov-Bohm Interference Pattern')
plt.xlabel('Observation Angle')
plt.grid()
plt.show()
```
