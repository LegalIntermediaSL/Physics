# Compendio Subtemático: Mecanica Cuantica Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
E = np.linspace(0.01, 3, 500)
V0 = 1.0
R = np.where(E > V0, ((np.sqrt(E) - np.sqrt(E - V0)) / (np.sqrt(E) + np.sqrt(E - V0)))**2, 1.0)
T = 1 - R
plt.plot(E, R, label='Reflection (R)')
plt.plot(E, T, label='Transmission (T)')
plt.axvline(V0, color='r', linestyle='--', label='V0')
plt.xlabel('Energy')
plt.legend()
plt.title('Step Potential R and T')
plt.grid()
plt.show()
```
