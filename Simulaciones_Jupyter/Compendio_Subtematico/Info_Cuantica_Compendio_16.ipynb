# Compendio Subtemático: Info Cuantica Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    theta = np.linspace(0, 2*np.pi, 50)
    S = 2 * np.sqrt(2) * np.cos(theta) # Simplified CHSH correlation
    
    plt.plot(theta, S, label='Quantum Correlation S')
    plt.axhline(2, color='r', linestyle='--', label='Classical Limit (2)')
    plt.axhline(-2, color='r', linestyle='--')
    plt.title('CHSH Inequality Violation')
    plt.xlabel('Measurement Angle')
    plt.ylabel('CHSH Value (S)')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
