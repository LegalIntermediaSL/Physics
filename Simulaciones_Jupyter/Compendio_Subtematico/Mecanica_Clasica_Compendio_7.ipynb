# Compendio Subtemático: Mecanica Clasica Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 100)
    m = 1.0
    k = 10.0
    omega = np.sqrt(k/m)
    x = np.cos(omega * t)
    v = -omega * np.sin(omega * t)
    K = 0.5 * m * v**2
    U = 0.5 * k * x**2
    E = K + U
    plt.plot(t, K, label='Kinetic Energy')
    plt.plot(t, U, label='Potential Energy')
    plt.plot(t, E, label='Total Energy', linestyle='--')
    plt.title('Conservation of Energy (Spring)')
    plt.xlabel('Time (s)')
    plt.ylabel('Energy (J)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
