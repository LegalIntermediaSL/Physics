# Compendio Subtemático: Optica Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def single_slit():
    theta = np.linspace(-0.1, 0.1, 500)
    lam, a = 500e-9, 0.05e-3
    beta = np.pi*a*np.sin(theta)/lam
    I = np.sinc(beta/np.pi)**2
    plt.plot(theta, I)
    plt.title("Single Slit Diffraction")
    plt.xlabel('Angle (rad)')
    plt.ylabel('Intensity')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    single_slit()
```
