# Compendio Subtemático: Electromagnetismo Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 1000)
    q = 1.0
    m = 1.0
    B = 1.0
    v0 = 2.0
    omega = q*B/m
    x = (v0/omega) * np.sin(omega*t)
    y = (v0/omega) * (1 - np.cos(omega*t))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("24. Lorentz Force - Cyclotron Motion")
    plt.axis('equal')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
