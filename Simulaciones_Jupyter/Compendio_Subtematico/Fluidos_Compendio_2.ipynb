# Compendio Subtemático: Fluidos Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def archimedes():
    volume_submerged = np.linspace(0, 10, 100) # m^3
    rho_fluid = 1000 # kg/m^3
    g = 9.81 # m/s^2
    F_b = rho_fluid * g * volume_submerged
    plt.plot(volume_submerged, F_b)
    plt.xlabel('Submerged Volume (m^3)')
    plt.ylabel('Buoyant Force (N)')
    plt.title('Archimedes Principle')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    archimedes()
```
