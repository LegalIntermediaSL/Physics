# Compendio Subtemático: Fisica Atomica Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def zeeman_splitting(B, l):
    # Bohr magneton in eV/T
    mu_B = 5.788e-5
    ml = np.arange(-l, l + 1)
    delta_E = mu_B * B * ml
    return ml, delta_E

def plot_zeeman():
    B_values = np.linspace(0, 5, 100)
    l = 2 # d-orbital
    
    plt.figure(figsize=(8, 6))
    for ml in np.arange(-l, l + 1):
        delta_E = 5.788e-5 * B_values * ml
        plt.plot(B_values, delta_E, label=f'$m_l$ = {ml}')
        
    plt.xlabel('Magnetic Field B (T)')
    plt.ylabel('Energy Shift $\Delta E$ (eV)')
    plt.title(f'Normal Zeeman Effect for l={l}')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_zeeman()
```
