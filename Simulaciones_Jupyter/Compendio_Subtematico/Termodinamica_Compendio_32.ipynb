# Compendio Subtemático: Termodinamica Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

M = np.linspace(0.1, 3.0, 100) # Mach number
gamma = 1.4

T_ratio = 1 / (1 + (gamma - 1)/2 * M**2)
P_ratio = (1 + (gamma - 1)/2 * M**2)**(-gamma / (gamma - 1))

plt.plot(M, T_ratio, label='T/T0')
plt.plot(M, P_ratio, label='P/P0')

plt.title('Isentropic Flow Relations')
plt.xlabel('Mach Number (M)')
plt.ylabel('Ratio to Stagnation')
plt.legend()
plt.grid(True)
plt.savefig('isentropic_nozzle.png')
```
