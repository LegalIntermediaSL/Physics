# Compendio Subtemático: Fisica Atomica Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def lande_g(s, l, j):
    if j == 0: return 0
    return 1 + (j*(j+1) - l*(l+1) + s*(s+1)) / (2*j*(j+1))

def anomalous_zeeman_splitting(B, s, l, j):
    mu_B = 5.788e-5 # eV/T
    g = lande_g(s, l, j)
    mj = np.arange(-j, j + 1, 1)
    return mj, g * mu_B * B * mj

def plot_anomalous_zeeman():
    B = np.linspace(0, 2, 100)
    
    # Example for Sodium D lines: 2P_1/2 and 2P_3/2
    s = 0.5
    l = 1
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # 2P_1/2
    j1 = 0.5
    mj1, E1 = anomalous_zeeman_splitting(1, s, l, j1)
    for m in mj1:
        ax1.plot(B, lande_g(s,l,j1) * 5.788e-5 * B * m, label=f'$m_j$={m}')
    ax1.set_title('Splitting of $^2P_{1/2}$')
    ax1.set_xlabel('B (T)')
    ax1.set_ylabel('$\Delta E$ (eV)')
    ax1.legend()
    
    # 2P_3/2
    j2 = 1.5
    mj2, E2 = anomalous_zeeman_splitting(1, s, l, j2)
    for m in mj2:
        ax2.plot(B, lande_g(s,l,j2) * 5.788e-5 * B * m, label=f'$m_j$={m}')
    ax2.set_title('Splitting of $^2P_{3/2}$')
    ax2.set_xlabel('B (T)')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    plot_anomalous_zeeman()
```
