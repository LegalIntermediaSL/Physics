# Compendio Subtemático: Plasmas Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def cyclotron_motion():
    q = 1.6e-19
    m = 1.67e-27
    B = np.array([0, 0, 1.0])
    v0 = np.array([1e5, 0, 0])
    r0 = np.array([0.0, 0.0, 0.0])
    dt = 1e-10
    t = np.arange(0, 1e-6, dt)
    r = np.zeros((len(t), 3))
    v = np.zeros((len(t), 3))
    r[0] = r0
    v[0] = v0
    for i in range(1, len(t)):
        v[i] = v[i-1] + (q/m)*np.cross(v[i-1], B)*dt
        r[i] = r[i-1] + v[i]*dt
    plt.plot(r[:, 0], r[:, 1])
    plt.title("Cyclotron Motion")
    plt.xlabel("x"); plt.ylabel("y")
    plt.grid(True)
    plt.savefig("cyclotron.png")

if __name__ == "__main__":
    cyclotron_motion()
```
