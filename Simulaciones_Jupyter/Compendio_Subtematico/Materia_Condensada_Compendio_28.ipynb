# Compendio Subtemático: Materia Condensada Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Heat Diffusion in a 1D Rod (Fundamental Solution to Diffusion Equation)
x = np.linspace(-10, 10, 300)
D = 1.0 # Diffusion coefficient
times = [0.1, 1.0, 3.0, 10.0]

plt.figure(figsize=(8, 5))

for t in times:
    # Gaussian solution to heat equation
    T = (1 / np.sqrt(4 * np.pi * D * t)) * np.exp(-x**2 / (4 * D * t))
    plt.plot(x, T, lw=2, label=f't = {t}')

plt.title("Heat Diffusion in 1D")
plt.xlabel("Position x")
plt.ylabel("Temperature T(x, t)")
plt.legend()
plt.grid(True)
plt.savefig("heat_diffusion.png")
```
