# Compendio Subtemático: Relatividad Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def schwarzschild_radius():
    G = 6.674e-11
    c = 3e8
    
    masses_solar = np.logspace(-1, 10, 100)
    masses_kg = masses_solar * 1.989e30
    
    Rs = 2 * G * masses_kg / c**2
    
    plt.figure(figsize=(8, 5))
    plt.loglog(masses_solar, Rs, color='black')
    plt.xlabel('Mass (Solar Masses)')
    plt.ylabel('Schwarzschild Radius (m)')
    plt.title('Black Hole Event Horizon Radius vs Mass')
    plt.grid(True, which="both", ls="--")
    plt.savefig('schwarzschild_radius.png')
    print('Plot saved as schwarzschild_radius.png')

if __name__ == '__main__':
    schwarzschild_radius()
```
