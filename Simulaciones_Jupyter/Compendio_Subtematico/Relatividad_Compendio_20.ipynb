# Compendio Subtemático: Relatividad Compendio 20

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gravitational_waves():
    t = np.linspace(0, 2, 1000)
    # Simple chirp waveform for binary inspiral
    f = 10 + 20 * t**2
    A = 1 / (2.1 - t) # Amplitude increasing
    
    h_plus = A * np.cos(2 * np.pi * f * t)
    
    plt.figure(figsize=(10, 4))
    plt.plot(t, h_plus, color='green')
    plt.xlabel('Time (s)')
    plt.ylabel('Strain $h_+(t)$')
    plt.title('Gravitational Wave Chirp from Binary Inspiral')
    plt.grid(True)
    plt.savefig('gravitational_waves.png')
    print('Plot saved as gravitational_waves.png')

if __name__ == '__main__':
    gravitational_waves()
```
