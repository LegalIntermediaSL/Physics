# Compendio Subtemático: Materia Condensada Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# London Equations: Magnetic Field Penetration
x = np.linspace(0, 5, 200) # Distance into superconductor in units of London penetration depth (lambda_L)
B0 = 1.0 # External magnetic field at surface

B_x = B0 * np.exp(-x)

plt.figure(figsize=(8, 5))
plt.plot(x, B_x, lw=2, color='crimson')
plt.fill_between(x, 0, B_x, color='crimson', alpha=0.2)
plt.title("Meissner Effect: Magnetic Field Penetration")
plt.xlabel(r"Distance $x/\lambda_L$")
plt.ylabel(r"Magnetic Field $B(x)$")
plt.grid(True)
plt.savefig("london_equations.png")
```
