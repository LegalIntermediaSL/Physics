# Compendio Subtemático: Astrofisica Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def grav_lensing():
    x = np.linspace(-5, 5, 400)
    y = np.linspace(-5, 5, 400)
    X, Y = np.meshgrid(x, y)
    R = np.sqrt(X**2 + Y**2)
    
    # Einstein radius
    R_E = 1.5
    
    # Deflection angle magnitude (simplified)
    alpha = R_E**2 / (R + 1e-6)
    
    # Mapping source plane
    beta_x = X - alpha * (X / R)
    beta_y = Y - alpha * (Y / R)
    
    # Source light profile (Gaussian)
    source = np.exp(-((beta_x - 0.5)**2 + (beta_y - 0.5)**2)/0.5)
    
    plt.figure(figsize=(8, 8))
    plt.imshow(source, extent=(-5, 5, -5, 5), origin='lower', cmap='hot')
    plt.title("Gravitational Lensing (Einstein Ring)")
    plt.xlabel("x (arcsec)")
    plt.ylabel("y (arcsec)")
    plt.colorbar(label='Intensity')
    plt.show()

if __name__ == '__main__':
    grav_lensing()
```
