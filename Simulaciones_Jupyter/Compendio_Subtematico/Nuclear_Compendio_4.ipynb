# Compendio Subtemático: Nuclear Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    theta = np.linspace(np.pi/18, np.pi, 100)
    # Differential cross section proportional to 1/sin^4(theta/2)
    cross_section = 1 / np.sin(theta/2)**4
    
    plt.plot(np.degrees(theta), cross_section, color='green')
    plt.yscale('log')
    plt.title('Rutherford Scattering Cross Section')
    plt.xlabel('Scattering Angle (Degrees)')
    plt.ylabel('Differential Cross Section (log scale)')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
