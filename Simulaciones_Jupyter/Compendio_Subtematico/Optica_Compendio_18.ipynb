# Compendio Subtemático: Optica Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def doppler():
    v = np.linspace(0, 0.99, 100) * 3e8
    lam0 = 500e-9
    c = 3e8
    lam_obs = lam0 * np.sqrt((1+v/c)/(1-v/c))
    plt.plot(v/c, lam_obs*1e9)
    plt.title("Relativistic Doppler Effect")
    plt.xlabel('v/c')
    plt.ylabel('Observed Wavelength (nm)')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    doppler()
```
