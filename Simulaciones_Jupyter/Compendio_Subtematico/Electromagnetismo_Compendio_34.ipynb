# Compendio Subtemático: Electromagnetismo Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    theta = np.linspace(-0.1, 0.1, 1000)
    lam = 500e-9
    a = 20e-6
    alpha = np.pi * a * np.sin(theta) / lam
    # avoid div by zero
    alpha[alpha==0] = 1e-10
    I = (np.sin(alpha) / alpha)**2
    plt.plot(theta, I)
    plt.xlabel('Angle (rad)')
    plt.ylabel('Intensity')
    plt.title("34. Single Slit Diffraction")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
