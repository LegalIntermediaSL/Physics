# Compendio Subtemático: Astrofisica Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def beaming():
    theta = np.linspace(0, np.pi, 200)
    beta = 0.9 # v/c
    gamma = 1 / np.sqrt(1 - beta**2)
    # Doppler factor
    delta = 1 / (gamma * (1 - beta * np.cos(theta)))
    
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(8, 8))
    ax.plot(theta, delta**3, 'r')
    ax.plot(-theta, delta**3, 'r')
    plt.title("Relativistic Beaming (Intensity Pattern)")
    plt.show()

if __name__ == '__main__':
    beaming()
```
