# Compendio Subtemático: Astrofisica Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def hubble_law():
    distances = np.linspace(10, 1000, 50) # Mpc
    H0 = 70 # km/s/Mpc
    velocities = H0 * distances
    noise = np.random.normal(0, 2000, size=distances.shape)
    obs_velocities = velocities + noise
    
    plt.figure(figsize=(10, 6))
    plt.scatter(distances, obs_velocities, label='Simulated Galaxies')
    plt.plot(distances, H0 * distances, 'r--', label=f'Hubble Law ($H_0$={H0})')
    plt.title("Hubble's Law: Universe Expansion")
    plt.xlabel("Distance (Mpc)")
    plt.ylabel("Recessional Velocity (km/s)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    hubble_law()
```
