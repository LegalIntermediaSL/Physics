# Compendio Subtemático: Electromagnetismo Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    z = np.linspace(-3, 3, 200)
    R = 1.0
    d = R  # Helmholtz spacing
    mu0 = 4*np.pi*1e-7
    I = 1.0
    N = 100
    B1 = (mu0 * N * I * R**2) / (2 * ((z - d/2)**2 + R**2)**1.5)
    B2 = (mu0 * N * I * R**2) / (2 * ((z + d/2)**2 + R**2)**1.5)
    B = B1 + B2
    plt.plot(z, B)
    plt.xlabel('Distance along axis z')
    plt.ylabel('Magnetic Field B_z')
    plt.title("23. Helmholtz Coil Magnetic Field")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
