# Compendio Subtemático: Fisica Atomica Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import genlaguerre
import math

def hydrogen_radial_wavefunction(n, l, r):
    a0 = 1.0 # Bohr radius
    rho = 2 * r / (n * a0)
    norm = np.sqrt((2 / (n * a0))**3 * math.factorial(n - l - 1) / (2 * n * math.factorial(n + l)))
    L = genlaguerre(n - l - 1, 2 * l + 1)(rho)
    R = norm * np.exp(-rho / 2) * (rho**l) * L
    return R

def plot_wavefunctions():
    r = np.linspace(0, 30, 1000)
    states = [(1, 0), (2, 0), (2, 1), (3, 0), (3, 1), (3, 2)]
    
    plt.figure(figsize=(10, 6))
    for n, l in states:
        R = hydrogen_radial_wavefunction(n, l, r)
        P = r**2 * R**2 # Radial probability density
        plt.plot(r, P, label=f'n={n}, l={l}')
        
    plt.xlabel('r (Bohr radii)')
    plt.ylabel('Radial Probability Density $P(r)$')
    plt.title('Hydrogen Atom Radial Wavefunctions')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_wavefunctions()
```
