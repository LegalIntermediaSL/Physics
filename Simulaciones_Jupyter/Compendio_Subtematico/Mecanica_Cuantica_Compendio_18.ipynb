# Compendio Subtemático: Mecanica Cuantica Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
alpha_a = np.linspace(0.01, 4*np.pi, 1000)
P = 10
rhs = P * np.sin(alpha_a) / alpha_a + np.cos(alpha_a)
plt.plot(alpha_a, rhs)
plt.axhline(1, color='r', linestyle='--')
plt.axhline(-1, color='r', linestyle='--')
plt.title('Kronig-Penney Model RHS')
plt.ylim(-3, 3)
plt.grid()
plt.show()
```
