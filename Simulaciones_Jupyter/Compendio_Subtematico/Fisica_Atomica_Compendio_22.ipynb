# Compendio Subtemático: Fisica Atomica Compendio 22

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def collisional_broadening(pressure, T):
    # Pressure broadening width is proportional to pressure P / sqrt(T)
    # Just qualitative demonstration
    
    delta_nu_C = 0.5 * pressure / np.sqrt(T) # Arbitrary scaling
    
    nu = np.linspace(-10, 10, 500)
    
    # Lorentzian profile for collisional broadening
    L = (delta_nu_C / (2 * np.pi)) / (nu**2 + (delta_nu_C / 2)**2)
    
    plt.figure(figsize=(8, 6))
    plt.plot(nu, L, 'g-', label=f'P={pressure} atm, T={T} K')
    plt.xlabel('Frequency detuning (arbitrary units)')
    plt.ylabel('Intensity')
    plt.title('Collisional (Pressure) Broadening')
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    collisional_broadening(1.0, 300)
```
