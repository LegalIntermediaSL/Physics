# Compendio Subtemático: Optica Compendio 31

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def faraday():
    B = np.linspace(0, 10, 100)
    V = 0.05
    d = 0.1
    theta = V * B * d
    plt.plot(B, theta)
    plt.title("Faraday Effect")
    plt.xlabel('Magnetic Field (T)')
    plt.ylabel('Rotation (rad)')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    faraday()
```
