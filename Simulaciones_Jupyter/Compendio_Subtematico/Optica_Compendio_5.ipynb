# Compendio Subtemático: Optica Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def convex_lens():
    f, do, ho = 10, 20, 5
    di = 1/(1/f - 1/do)
    hi = -ho*di/do
    plt.plot([-do, 0, di], [ho, ho, hi], 'r-', label='Ray 1')
    plt.plot([-do, 0, di], [ho, 0, hi], 'g-', label='Ray 2')
    plt.plot([-do, 0, di], [ho, -ho, hi], 'b-', label='Ray 3')
    plt.axvline(0, color='k', linestyle='--')
    plt.legend()
    plt.grid()
    plt.title("Convex Lens Ray Tracing")
    plt.show()

if __name__ == '__main__': 
    convex_lens()
```
