# Compendio Subtemático: Fluidos Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def bernoulli():
    v1 = 2 # m/s
    P1 = 150000 # Pa
    rho = 1000 # kg/m^3
    v2 = np.linspace(2, 10, 100) # m/s
    P2 = P1 + 0.5 * rho * (v1**2 - v2**2)
    plt.plot(v2, P2/1000)
    plt.xlabel('Velocity 2 (m/s)')
    plt.ylabel('Pressure 2 (kPa)')
    plt.title('Bernoulli\\'s Equation (Venturi Effect)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    bernoulli()
```
