# Compendio Subtemático: Optica Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def brewster():
    theta = np.linspace(0, np.pi/2, 500)
    n1, n2 = 1.0, 1.5
    
    term = n1/n2*np.sin(theta)
    Rs = np.abs((n1*np.cos(theta) - n2*np.sqrt(1-term**2)) / (n1*np.cos(theta) + n2*np.sqrt(1-term**2)))**2
    Rp = np.abs((n1*np.sqrt(1-term**2) - n2*np.cos(theta)) / (n1*np.sqrt(1-term**2) + n2*np.cos(theta)))**2
    
    plt.plot(np.degrees(theta), Rs, label='s-polarization')
    plt.plot(np.degrees(theta), Rp, label='p-polarization')
    plt.legend()
    plt.title("Brewster's Angle")
    plt.xlabel('Angle of Incidence (degrees)')
    plt.ylabel('Reflectance')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    brewster()
```
