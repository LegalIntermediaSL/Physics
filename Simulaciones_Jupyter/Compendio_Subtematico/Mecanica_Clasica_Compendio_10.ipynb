# Compendio Subtemático: Mecanica Clasica Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    m1 = 2.0
    m2 = 3.0
    v1i = 5.0
    v2i = -2.0
    v1f = ((m1 - m2)*v1i + 2*m2*v2i) / (m1 + m2)
    v2f = (2*m1*v1i + (m2 - m1)*v2i) / (m1 + m2)
    labels = ['Initial', 'Final']
    v1_bars = [v1i, v1f]
    v2_bars = [v2i, v2f]
    x = np.arange(len(labels))
    width = 0.35
    plt.bar(x - width/2, v1_bars, width, label='Mass 1')
    plt.bar(x + width/2, v2_bars, width, label='Mass 2')
    plt.xticks(x, labels)
    plt.ylabel('Velocity (m/s)')
    plt.title('1D Elastic Collision Velocities')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
