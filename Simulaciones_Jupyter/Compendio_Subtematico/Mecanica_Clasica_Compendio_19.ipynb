# Compendio Subtemático: Mecanica Clasica Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Rolling without slipping: v_cm = R * omega, K = 0.5*m*v^2 + 0.5*I*w^2
    m = 2.0
    R = 0.5
    I = 0.5 * m * R**2 # Solid cylinder
    
    v = np.linspace(0, 10, 100)
    omega = v / R
    
    K_trans = 0.5 * m * v**2
    K_rot = 0.5 * I * omega**2
    K_tot = K_trans + K_rot
    
    plt.plot(v, K_trans, label='Translational KE')
    plt.plot(v, K_rot, label='Rotational KE')
    plt.plot(v, K_tot, label='Total KE')
    plt.title('Kinetic Energy of a Rolling Cylinder')
    plt.xlabel('Center of Mass Velocity (m/s)')
    plt.ylabel('Energy (J)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
