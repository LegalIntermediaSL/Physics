# Compendio Subtemático: Materia Condensada Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Ginzburg-Landau Order Parameter Profile at a surface
x = np.linspace(0, 5, 200) # distance from boundary in units of coherence length xi
# The solution for an isolated boundary where psi(0) = 0 is psi(x) = tanh(x / sqrt(2))
psi = np.tanh(x / np.sqrt(2))

plt.figure(figsize=(8, 5))
plt.plot(x, psi, lw=2, color='indigo')
plt.title("Ginzburg-Landau: Order Parameter near a Surface")
plt.xlabel(r"Distance from surface $x/\xi$")
plt.ylabel(r"Normalized Order Parameter $\psi/\psi_\infty$")
plt.grid(True)
plt.savefig("gl_order_parameter.png")
```
