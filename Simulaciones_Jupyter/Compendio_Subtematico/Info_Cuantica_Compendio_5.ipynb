# Compendio Subtemático: Info Cuantica Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Simplified teleportation success probabilities
    states = ['Alice prepared', 'Bob received']
    fidelity = [1.0, 1.0] # Ideal teleportation
    
    plt.plot(states, fidelity, marker='o', linestyle='-', color='red')
    plt.title('Quantum Teleportation Fidelity (Ideal)')
    plt.ylim(0, 1.1)
    plt.ylabel('Fidelity')
    plt.show()

if __name__ == '__main__':
    main()
```
