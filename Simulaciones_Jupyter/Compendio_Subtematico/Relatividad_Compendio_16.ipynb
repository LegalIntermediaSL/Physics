# Compendio Subtemático: Relatividad Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def mercury_precession():
    theta = np.linspace(0, 10 * 2*np.pi, 1000)
    # Simple toy model: r(theta) = p / (1 + e*cos((1 - k)*theta))
    p = 1.0
    e = 0.2
    k = 0.05 # Precession factor
    
    r = p / (1 + e * np.cos((1 - k) * theta))
    
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    
    plt.figure(figsize=(6, 6))
    plt.plot(x, y, label='Precessing Orbit', color='orange')
    plt.plot(0, 0, 'yo', markersize=10, label='Sun')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Precession of Mercury's Perihelion (Exaggerated)")
    plt.grid(True)
    plt.legend()
    plt.axis('equal')
    plt.savefig('mercury_precession.png')
    print('Plot saved as mercury_precession.png')

if __name__ == '__main__':
    mercury_precession()
```
