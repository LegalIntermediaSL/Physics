# Compendio Subtemático: Mecanica Cuantica Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
k = np.linspace(0, 10, 100)
omega = 0.5 * k**2
v_phase = omega / k
v_group = k
plt.plot(k, v_phase, label='Phase Velocity')
plt.plot(k, v_group, label='Group Velocity')
plt.title('Free Particle Velocities')
plt.legend()
plt.grid()
plt.show()
```
