# Compendio Subtemático: Relatividad Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def penrose_diagram():
    # Mapping u = t - x, v = t + x
    # U = arctan(u), V = arctan(v)
    # T = U + V, X = V - U
    
    x = np.linspace(-10, 10, 20)
    t = np.linspace(-10, 10, 20)
    
    plt.figure(figsize=(6, 6))
    
    # Lines of constant t
    for t_val in t:
        X = np.arctan(t_val + x) - np.arctan(t_val - x)
        T = np.arctan(t_val + x) + np.arctan(t_val - x)
        plt.plot(X, T, color='blue', alpha=0.3)
        
    # Lines of constant x
    for x_val in x:
        X = np.arctan(t + x_val) - np.arctan(t - x_val)
        T = np.arctan(t + x_val) + np.arctan(t - x_val)
        plt.plot(X, T, color='red', alpha=0.3)
        
    plt.title('Penrose Diagram of Minkowski Space')
    plt.axis('equal')
    plt.axis('off')
    plt.savefig('penrose_diagram.png')
    print('Plot saved as penrose_diagram.png')

if __name__ == '__main__':
    penrose_diagram()
```
