# Compendio Subtemático: Fluidos Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def mach():
    v = np.linspace(0, 1000, 100)
    T = 293 # K
    gamma = 1.4
    R = 287
    a = np.sqrt(gamma * R * T)
    Ma = v / a
    plt.plot(v, Ma)
    plt.axhline(1, color='r', linestyle='--', label='Sonic (Ma=1)')
    plt.xlabel('Velocity (m/s)')
    plt.ylabel('Mach Number')
    plt.title('Mach Number vs Velocity in Air')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    mach()
```
