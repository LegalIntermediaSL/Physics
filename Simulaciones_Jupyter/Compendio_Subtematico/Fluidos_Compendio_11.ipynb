# Compendio Subtemático: Fluidos Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def stokes():
    r = np.linspace(0.001, 0.05, 100) # m
    eta = 0.001 # Pa.s (water)
    v = 1 # m/s
    Fd = 6 * np.pi * eta * r * v
    plt.plot(r, Fd)
    plt.xlabel('Radius (m)')
    plt.ylabel('Stokes Drag Force (N)')
    plt.title('Stokes\\' Law')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    stokes()
```
