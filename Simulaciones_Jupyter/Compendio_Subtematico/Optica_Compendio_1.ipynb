# Compendio Subtemático: Optica Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def snells_law():
    n1, n2 = 1.0, 1.5
    theta1_deg = np.linspace(0, 90, 100)
    theta1_rad = np.radians(theta1_deg)
    
    sin_theta2 = (n1 / n2) * np.sin(theta1_rad)
    theta2_deg = np.degrees(np.arcsin(np.clip(sin_theta2, -1, 1)))
    
    plt.plot(theta1_deg, theta2_deg, label=f'n1={n1}, n2={n2}')
    plt.xlabel('Angle of Incidence (degrees)')
    plt.ylabel('Angle of Refraction (degrees)')
    plt.title("Snell's Law")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    snells_law()
```
