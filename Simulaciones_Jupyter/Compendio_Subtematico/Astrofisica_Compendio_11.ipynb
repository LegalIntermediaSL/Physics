# Compendio Subtemático: Astrofisica Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def white_dwarf():
    mass = np.linspace(0.1, 1.44, 100) # Solar masses
    # M^(-1/3) relation roughly
    radius = 0.01 * (mass**(-1/3) * (1 - (mass/1.44)**(4/3))**(1/2))
    
    plt.figure(figsize=(8, 6))
    plt.plot(mass, radius, 'k-')
    plt.axvline(x=1.44, color='r', linestyle='--', label='Chandrasekhar Limit')
    plt.title("White Dwarf Mass-Radius Relation")
    plt.xlabel("Mass ($M_\odot$)")
    plt.ylabel("Radius ($R_\odot$)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    white_dwarf()
```
