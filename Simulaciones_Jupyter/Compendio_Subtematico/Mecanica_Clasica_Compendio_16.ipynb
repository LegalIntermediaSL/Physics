# Compendio Subtemático: Mecanica Clasica Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Moment of inertia for a solid disk: I = 0.5 * m * r^2
    m = np.linspace(1, 10, 50)
    r = 2.0
    I = 0.5 * m * r**2
    plt.plot(m, I)
    plt.title('Moment of Inertia of Solid Disk vs Mass')
    plt.xlabel('Mass (kg)')
    plt.ylabel('Moment of Inertia (kg m^2)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
