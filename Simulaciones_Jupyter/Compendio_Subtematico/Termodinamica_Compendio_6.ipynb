# Compendio Subtemático: Termodinamica Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

mass = 1.0 # kg
c = 4184 # J/(kg*K) for water
dT = np.linspace(0, 100, 100)

Q = mass * c * dT

plt.plot(dT, Q)
plt.title('Sensible Heat (Water)')
plt.xlabel('Temperature Change (K)')
plt.ylabel('Heat Added (J)')
plt.grid(True)
plt.savefig('sensible_heat.png')
```
