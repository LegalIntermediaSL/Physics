# Compendio Subtemático: Fluidos Compendio 30

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def pump_curve():
    Q = np.linspace(0, 100, 100) # Flow rate
    H0 = 50 # shutoff head
    k = 0.003
    H = H0 - k * Q**2
    plt.plot(Q, H)
    plt.xlabel('Flow Rate Q')
    plt.ylabel('Pump Head H')
    plt.title('Centrifugal Pump Curve')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    pump_curve()
```
