# Compendio Subtemático: Relatividad Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def doppler_effect():
    f0 = 1000 # Source frequency in Hz
    c = 1.0
    v = np.linspace(-0.9, 0.9, 500) # Negative means receding, positive approaching
    
    # Relativistic Doppler effect formula
    f_obs = f0 * np.sqrt((1 + v/c) / (1 - v/c))
    
    plt.figure(figsize=(8, 5))
    plt.plot(v/c, f_obs, label='Observed Frequency', color='teal')
    plt.axvline(0, color='black', linestyle='--')
    plt.axhline(f0, color='black', linestyle='--', label='Source Frequency')
    plt.xlabel('Velocity (v/c, + is approaching)')
    plt.ylabel('Frequency (Hz)')
    plt.title('Relativistic Doppler Effect')
    plt.grid(True)
    plt.legend()
    plt.savefig('doppler_effect.png')
    print('Plot saved as doppler_effect.png')

if __name__ == '__main__':
    doppler_effect()
```
