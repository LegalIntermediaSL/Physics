# Compendio Subtemático: Materia Condensada Compendio 30

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Simple Cubic (SC) Lattice Generation
def generate_sc(n):
    points = []
    for x in range(n):
        for y in range(n):
            for z in range(n):
                points.append([x, y, z])
    return np.array(points)

pts = generate_sc(3) # 3x3x3 lattice

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pts[:,0], pts[:,1], pts[:,2], c='blue', marker='o', s=100)
ax.set_title("Simple Cubic (SC) Lattice")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.savefig("sc_lattice.png")
```
