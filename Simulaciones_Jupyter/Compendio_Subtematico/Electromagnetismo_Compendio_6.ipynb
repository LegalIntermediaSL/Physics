# Compendio Subtemático: Electromagnetismo Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)
    d = 1.0
    R1 = np.sqrt((X-d)**2 + Y**2)
    R2 = np.sqrt((X+d)**2 + Y**2)
    R1[R1 == 0] = 1e-10
    R2[R2 == 0] = 1e-10
    V = 1.0/R1 - 1.0/R2
    plt.contourf(X, Y, V, levels=50, cmap='coolwarm')
    plt.colorbar(label='Potential (V)')
    plt.title("6. Electric Potential of a Dipole")
    plt.show()

if __name__ == "__main__":
    main()
```
