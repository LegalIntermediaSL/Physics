# Compendio Subtemático: Plasmas Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def exb_drift():
    q, m = 1.6e-19, 1.67e-27
    B = np.array([0, 0, 1.0])
    E = np.array([0, 1000.0, 0])
    v0, r0 = np.array([1e4, 0, 0]), np.zeros(3)
    dt, t_end = 1e-10, 5e-7
    t = np.arange(0, t_end, dt)
    r, v = np.zeros((len(t), 3)), np.zeros((len(t), 3))
    r[0], v[0] = r0, v0
    for i in range(1, len(t)):
        v[i] = v[i-1] + (q/m)*(E + np.cross(v[i-1], B))*dt
        r[i] = r[i-1] + v[i]*dt
    plt.plot(r[:, 0], r[:, 1])
    plt.title("E x B Drift")
    plt.grid(True); plt.savefig("exb.png")

if __name__ == "__main__":
    exb_drift()
```
