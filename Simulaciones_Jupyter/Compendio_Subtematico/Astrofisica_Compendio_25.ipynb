# Compendio Subtemático: Astrofisica Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def extinction():
    z = np.linspace(0, 80, 100) # Zenith angle in degrees
    airmass = 1 / np.cos(np.radians(z))
    transmission = np.exp(-0.2 * airmass)
    
    plt.figure(figsize=(8, 6))
    plt.plot(z, transmission, 'c-')
    plt.title("Atmospheric Transmission vs Zenith Angle")
    plt.xlabel("Zenith Angle (degrees)")
    plt.ylabel("Transmission Fraction")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    extinction()
```
