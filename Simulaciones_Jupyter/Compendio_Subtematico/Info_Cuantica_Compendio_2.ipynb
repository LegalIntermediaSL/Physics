# Compendio Subtemático: Info Cuantica Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    state_0 = np.array([1, 0])
    X_gate = np.array([[0, 1], [1, 0]])
    state_1 = X_gate.dot(state_0)
    
    labels = ['|0>', '|1>']
    probs = np.abs(state_1)**2
    
    plt.bar(labels, probs, color='blue')
    plt.title('Pauli X Gate Action on |0>')
    plt.ylabel('Probability')
    plt.show()

if __name__ == '__main__':
    main()
```
