# Compendio Subtemático: Fisica Atomica Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def plancks_law(T, lambdas):
    # h in J*s, c in m/s, k_B in J/K, lambdas in m
    h = 6.626e-34
    c = 3e8
    k_B = 1.38e-23
    
    # Spectral radiance B_lambda (W / (m^2 sr m))
    exponent = (h * c) / (lambdas * k_B * T)
    B = (2 * h * c**2 / lambdas**5) / (np.exp(exponent) - 1)
    return B

def plot_blackbody():
    lambdas_nm = np.linspace(100, 3000, 1000)
    lambdas_m = lambdas_nm * 1e-9
    
    temperatures = [3000, 4000, 5000, 6000] # Kelvin
    
    plt.figure(figsize=(10, 6))
    
    for T in temperatures:
        B = plancks_law(T, lambdas_m)
        plt.plot(lambdas_nm, B * 1e-9, label=f'T = {T} K') # plot in W / (m^2 sr nm)
        
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Spectral Radiance (W / m$^2$ sr nm)')
    plt.title("Planck's Law of Blackbody Radiation")
    plt.legend()
    plt.grid(True)
    plt.xlim(0, 3000)
    plt.show()

if __name__ == "__main__":
    plot_blackbody()
```
