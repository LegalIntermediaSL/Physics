# Compendio Subtemático: Mecanica Clasica Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 100)
    m = 2.0
    F = 10.0
    a = F / m
    v = a * t
    plt.plot(t, v)
    plt.title('Newtons 2nd Law - Velocity vs Time')
    plt.xlabel('Time (s)')
    plt.ylabel('Velocity (m/s)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
