# Compendio Subtemático: Relatividad Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def relativistic_momentum():
    m = 1.0 # kg
    c = 3e8 # m/s
    v = np.linspace(0, 0.95*c, 500)
    
    p_classical = m * v
    p_rel = m * v / np.sqrt(1 - (v/c)**2)
    
    plt.figure(figsize=(8, 5))
    plt.plot(v/c, p_classical, '--', label='Classical Momentum ($p = mv$)', color='grey')
    plt.plot(v/c, p_rel, label='Relativistic Momentum ($p = \gamma mv$)', color='purple')
    plt.xlabel('Velocity (v/c)')
    plt.ylabel('Momentum (kg m/s)')
    plt.title('Classical vs Relativistic Momentum')
    plt.grid(True)
    plt.legend()
    plt.savefig('relativistic_momentum.png')
    print('Plot saved as relativistic_momentum.png')

if __name__ == '__main__':
    relativistic_momentum()
```
