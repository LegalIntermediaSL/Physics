# Compendio Subtemático: Materia Condensada Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 2D Ising Model - Onsager Exact Solution for Magnetization
Tc = 2.269 # Critical temperature in units of J/kB
T = np.linspace(1.0, 3.5, 500)

# Exact magnetization formula
def magnetization(T):
    if T < Tc:
        return (1 - np.sinh(2/T)**-4)**(1/8)
    else:
        return 0.0

M = np.array([magnetization(t) for t in T])

plt.figure(figsize=(8, 5))
plt.plot(T, M, lw=2, color='purple')
plt.axvline(Tc, color='r', linestyle='--', label=f'$T_c \\approx {Tc}$')
plt.title("2D Ising Model: Exact Magnetization (Onsager)")
plt.xlabel(r"Temperature $T$ ($J/k_B$)")
plt.ylabel("Spontaneous Magnetization $M$")
plt.legend()
plt.grid(True)
plt.savefig("ising_onsager.png")
```
