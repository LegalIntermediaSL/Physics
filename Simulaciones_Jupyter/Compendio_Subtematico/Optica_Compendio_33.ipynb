# Compendio Subtemático: Optica Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def blackbody():
    lam = np.linspace(100e-9, 3000e-9, 500)
    h = 6.626e-34
    c = 3e8
    k = 1.38e-23
    T = 5000
    
    I = (2*h*c**2 / lam**5) / (np.exp(h*c/(lam*k*T)) - 1)
    plt.plot(lam*1e9, I)
    plt.title(f"Planck's Law (Blackbody) at T={T}K")
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Spectral Radiance')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    blackbody()
```
