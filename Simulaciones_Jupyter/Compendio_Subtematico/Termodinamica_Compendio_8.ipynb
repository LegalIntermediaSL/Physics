# Compendio Subtemático: Termodinamica Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Carnot Cycle P-V
V1, V2, V3, V4 = 1, 2, 4, 2
# simplified curves
v_iso_1 = np.linspace(1, 2, 50)
v_adia_1 = np.linspace(2, 4, 50)
v_iso_2 = np.linspace(4, 2.5, 50)
v_adia_2 = np.linspace(2.5, 1, 50)

p_iso_1 = 10 / v_iso_1
p_adia_1 = (10/2) * (2/v_adia_1)**1.4
p_iso_2 = 1.9 * (4/v_iso_2)
p_adia_2 = 1.9 * (2.5/v_adia_2)**1.4

plt.plot(v_iso_1, p_iso_1, 'r', label='Isothermal Exp')
plt.plot(v_adia_1, p_adia_1, 'g', label='Adiabatic Exp')
plt.plot(v_iso_2, p_iso_2, 'b', label='Isothermal Comp')
plt.plot(v_adia_2, p_adia_2, 'y', label='Adiabatic Comp')

plt.title('Carnot Cycle')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.savefig('carnot.png')
```
