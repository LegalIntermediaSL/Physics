# Compendio Subtemático: Optica Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def opt_act():
    lam = np.linspace(400, 700, 100)
    rotation = 1e5 / lam**2
    plt.plot(lam, rotation)
    plt.title("Optical Activity Rotation")
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Angle (deg)')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    opt_act()
```
