# Compendio Subtemático: Mecanica Clasica Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 100)
    # Simple perfectly inelastic collision where they stick together
    m1 = 2.0
    m2 = 3.0
    v1i = 5.0
    v2i = -2.0
    vf = (m1*v1i + m2*v2i) / (m1 + m2)
    labels = ['Initial m1', 'Initial m2', 'Final Combined']
    velocities = [v1i, v2i, vf]
    plt.bar(labels, velocities)
    plt.ylabel('Velocity (m/s)')
    plt.title('1D Inelastic Collision')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
