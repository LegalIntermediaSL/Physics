# Compendio Subtemático: Materia Condensada Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 1D Monoatomic chain phonon dispersion
a = 1.0 # lattice constant
C = 1.0 # force constant
M = 1.0 # mass

k = np.linspace(-np.pi/a, np.pi/a, 200)
omega = 2 * np.sqrt(C / M) * np.abs(np.sin(k * a / 2))

plt.figure(figsize=(8, 5))
plt.plot(k, omega, lw=2, color='teal')
plt.title("Phonon Dispersion: 1D Monoatomic Chain")
plt.xlabel("Wavevector k")
plt.ylabel(r"Angular Frequency $\omega$")
plt.xticks([-np.pi/a, 0, np.pi/a], [r'$-\pi/a$', '0', r'$\pi/a$'])
plt.grid(True)
plt.savefig("monoatomic_phonon.png")
```
