# Compendio Subtemático: Termodinamica Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Simplified Rankine Cycle T-s diagram
s = [1, 1, 6, 6, 1]
T = [300, 600, 600, 300, 300]

# Saturation dome approx
s_dome = np.linspace(0.5, 7.5, 100)
T_dome = 650 - 20*(s_dome - 4)**2

plt.plot(s_dome, T_dome, 'k--', label='Saturation Dome')
plt.plot(s, T, 'b-', label='Rankine Cycle')

plt.title('Ideal Rankine Cycle')
plt.xlabel('Entropy (s)')
plt.ylabel('Temperature (T)')
plt.legend()
plt.grid(True)
plt.savefig('rankine.png')
```
