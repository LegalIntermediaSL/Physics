# Compendio Subtemático: Materia Condensada Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# AC Josephson Effect
time = np.linspace(0, 10, 500)
V = 1.0 # Constant voltage applied
# I(t) = I_c * sin(phi(0) + (2e/hbar) * V * t)
# Normalized frequencies
omega_J = 2.0 * V 
I_c = 1.0
I_t = I_c * np.sin(omega_J * time)

plt.figure(figsize=(8, 5))
plt.plot(time, I_t, lw=2, color='forestgreen')
plt.title("AC Josephson Effect")
plt.xlabel("Time (normalized)")
plt.ylabel("Supercurrent $I(t)$")
plt.grid(True)
plt.savefig("josephson.png")
```
