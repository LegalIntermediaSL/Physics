# Compendio Subtemático: Astrofisica Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def dark_energy():
    z = np.linspace(0, 2, 100)
    w0 = -1.0
    wa = 0.1
    w_z = w0 + wa * (z / (1 + z))
    
    plt.figure(figsize=(8, 6))
    plt.plot(z, w_z, 'r-', linewidth=2)
    plt.axhline(-1, color='k', linestyle='--', label='Cosmological Constant')
    plt.title("Dark Energy Equation of State (CPL parameterization)")
    plt.xlabel("Redshift z")
    plt.ylabel("w(z)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    dark_energy()
```
