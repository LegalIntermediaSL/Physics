# Compendio Subtemático: Plasmas Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def drifting_maxwellian():
    v = np.linspace(-10, 10, 200)
    vd = 3.0
    vth = 1.0
    f = (1/(np.sqrt(np.pi)*vth)) * np.exp(-(v-vd)**2 / vth**2)
    plt.plot(v, f)
    plt.title("Drifting Maxwellian")
    plt.grid(True); plt.savefig("drifting.png")
if __name__ == "__main__": drifting_maxwellian()
```
