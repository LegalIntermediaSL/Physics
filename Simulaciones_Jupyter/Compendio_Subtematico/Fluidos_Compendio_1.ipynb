# Compendio Subtemático: Fluidos Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def hydrostatics():
    depth = np.linspace(0, 100, 100) # m
    rho = 1000 # kg/m^3
    g = 9.81 # m/s^2
    P0 = 101325 # Pa
    P = P0 + rho * g * depth
    plt.plot(depth, P/1000)
    plt.xlabel('Depth (m)')
    plt.ylabel('Pressure (kPa)')
    plt.title('Hydrostatic Pressure vs Depth')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    hydrostatics()
```
