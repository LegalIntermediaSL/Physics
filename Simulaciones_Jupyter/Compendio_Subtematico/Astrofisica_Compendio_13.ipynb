# Compendio Subtemático: Astrofisica Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def black_hole():
    theta = np.linspace(0, 2*np.pi, 100)
    rs = 1.0
    r_isco = 3.0 * rs
    r_photon = 1.5 * rs
    
    plt.figure(figsize=(6, 6))
    plt.plot(rs*np.cos(theta), rs*np.sin(theta), 'k', label='Event Horizon')
    plt.plot(r_photon*np.cos(theta), r_photon*np.sin(theta), 'r--', label='Photon Sphere')
    plt.plot(r_isco*np.cos(theta), r_isco*np.sin(theta), 'b-.', label='ISCO')
    
    plt.title("Schwarzschild Black Hole Orbits")
    plt.axis('equal')
    plt.legend()
    plt.show()

if __name__ == '__main__':
    black_hole()
```
