# Compendio Subtemático: Termodinamica Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

V = np.linspace(1, 10, 100)
k_vals = [1000, 2000, 3000]

for k in k_vals:
    P = k / V
    plt.plot(V, P, label=f'k = {k}')

plt.title("Boyle's Law: Isothermal Process")
plt.xlabel('Volume (m^3)')
plt.ylabel('Pressure (Pa)')
plt.legend()
plt.grid(True)
plt.savefig('boyles_law.png')
```
