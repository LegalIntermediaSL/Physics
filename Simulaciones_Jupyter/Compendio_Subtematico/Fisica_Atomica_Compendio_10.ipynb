# Compendio Subtemático: Fisica Atomica Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import matplotlib.pyplot as plt

def jj_coupling_terms(l1, l2, s1=0.5, s2=0.5):
    j1_values = [abs(l1 - s1), l1 + s1]
    j2_values = [abs(l2 - s2), l2 + s2]
    
    terms = []
    for j1 in j1_values:
        for j2 in j2_values:
            J_values = [J/2.0 for J in range(int(2*abs(j1 - j2)), int(2*(j1 + j2) + 2), 2)]
            for J in J_values:
                terms.append((j1, j2, J))
    return terms

def plot_jj_terms(l1, l2):
    terms = jj_coupling_terms(l1, l2)
    
    plt.figure(figsize=(10, 6))
    
    x = 0
    for j1, j2, J in terms:
        term_symbol = f"({j1}, {j2})$_{{{J}}}$"
        
        # approximate energy order
        energy = j1 * 5 + j2 * 5 + J * 0.5
        
        plt.plot(x, energy, 'bo', markersize=10)
        plt.text(x+0.1, energy, term_symbol, fontsize=12)
        x += 1
        
    plt.title(f'jj Coupling States for $l_1={l1}, l_2={l2}$')
    plt.ylabel('Relative Energy (arbitrary units)')
    plt.xticks([])
    plt.grid(True, alpha=0.3)
    plt.show()

if __name__ == "__main__":
    plot_jj_terms(1, 1)
```
