# Compendio Subtemático: Nuclear Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    products = ['Fragment 1', 'Fragment 2', 'Neutrons', 'Prompt Gamma', 'Beta', 'Antineutrino']
    energy = [165, 0, 5, 7, 7, 9] # Approximate MeV distribution for U-235 fission
    
    plt.bar(products, energy, color='teal')
    plt.title('Energy Distribution in U-235 Fission')
    plt.ylabel('Energy (MeV)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    main()
```
