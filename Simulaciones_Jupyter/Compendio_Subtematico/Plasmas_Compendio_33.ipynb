# Compendio Subtemático: Plasmas Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def faraday_rotation():
    z = np.linspace(0, 10, 100)
    theta = 0.5 * z # rotation angle increases with distance
    plt.plot(z, theta)
    plt.title("Faraday Rotation Angle vs Distance")
    plt.xlabel("z"); plt.ylabel("Angle")
    plt.grid(True); plt.savefig("faraday.png")
if __name__ == "__main__": faraday_rotation()
```
