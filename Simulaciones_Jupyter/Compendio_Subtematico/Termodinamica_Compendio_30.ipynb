# Compendio Subtemático: Termodinamica Compendio 30

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Ericsson Cycle P-V
v_iso_1 = np.linspace(1, 4, 50)
p_iso_1 = 10 / v_iso_1

p_isob_1 = np.full(50, p_iso_1[-1])
v_isob_1 = np.linspace(4, 2, 50)

v_iso_2 = np.linspace(2, 0.5, 50)
p_iso_2 = 5 / v_iso_2

p_isob_2 = np.full(50, p_iso_2[-1])
v_isob_2 = np.linspace(0.5, 1, 50)

plt.plot(v_iso_1, p_iso_1, 'r', label='Isothermal Exp')
plt.plot(v_isob_1, p_isob_1, 'g', label='Isobaric Cooling')
plt.plot(v_iso_2, p_iso_2, 'b', label='Isothermal Comp')
plt.plot(v_isob_2, p_isob_2, 'y', label='Isobaric Heating')

plt.title('Ericsson Cycle')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.savefig('ericsson.png')
```
