# Compendio Subtemático: Fisica Atomica Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def bloch_equations(M, t, gamma, B0, B1, omega, T1, T2):
    Mx, My, Mz = M
    
    # Effective magnetic field in rotating frame with frequency omega
    delta_omega = gamma * B0 - omega
    omega_1 = gamma * B1
    
    dMx_dt = delta_omega * My - Mx / T2
    dMy_dt = -delta_omega * Mx + omega_1 * Mz - My / T2
    dMz_dt = -omega_1 * My - (Mz - 1.0) / T1
    
    return [dMx_dt, dMy_dt, dMz_dt]

def simulate_nmr():
    T1 = 1.0
    T2 = 0.1
    gamma = 1.0
    B0 = 10.0
    B1 = 1.0
    omega = gamma * B0 # On resonance
    
    M0 = [0.0, 0.0, 1.0] # Start in equilibrium
    
    t = np.linspace(0, 5, 500)
    sol = odeint(bloch_equations, M0, t, args=(gamma, B0, B1, omega, T1, T2))
    
    plt.figure(figsize=(8, 6))
    plt.plot(t, sol[:, 0], label='$M_x$')
    plt.plot(t, sol[:, 1], label='$M_y$')
    plt.plot(t, sol[:, 2], label='$M_z$')
    
    plt.xlabel('Time')
    plt.ylabel('Magnetization Components')
    plt.title('Bloch Equations (NMR/EPR)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    simulate_nmr()
```
