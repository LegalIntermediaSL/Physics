# Compendio Subtemático: Info Cuantica Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    state_0 = np.array([1, 0])
    H_gate = 1/np.sqrt(2) * np.array([[1, 1], [1, -1]])
    state_superpos = H_gate.dot(state_0)
    
    labels = ['|0>', '|1>']
    probs = np.abs(state_superpos)**2
    
    plt.bar(labels, probs, color='green')
    plt.title('Hadamard Gate Action on |0> (Superposition)')
    plt.ylabel('Probability')
    plt.show()

if __name__ == '__main__':
    main()
```
