# Compendio Subtemático: Relatividad Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def velocity_addition():
    c = 1.0
    v_frame = 0.5 # Frame velocity
    u_prime = np.linspace(-1.0, 1.0, 100) # Velocity in moving frame
    
    # Classical addition
    u_classical = u_prime + v_frame
    
    # Relativistic addition
    u_rel = (u_prime + v_frame) / (1 + (u_prime * v_frame) / c**2)
    
    plt.figure(figsize=(8, 5))
    plt.plot(u_prime, u_classical, '--', label='Galilean Addition', color='gray')
    plt.plot(u_prime, u_rel, label='Einstein Addition', color='darkgreen')
    plt.axhline(1.0, color='red', linestyle=':', label='Speed of Light')
    plt.axhline(-1.0, color='red', linestyle=':')
    plt.xlabel("Velocity in moving frame (u')")
    plt.ylabel('Observed Velocity (u)')
    plt.title('Addition of Velocities ($v_{frame} = 0.5c$)')
    plt.grid(True)
    plt.legend()
    plt.savefig('velocity_addition.png')
    print('Plot saved as velocity_addition.png')

if __name__ == '__main__':
    velocity_addition()
```
