# Compendio Subtemático: Mecanica Clasica Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 100)
    v0 = 0
    a = 2.0
    x = v0 * t + 0.5 * a * t**2
    plt.plot(t, x)
    plt.title('1D Motion with Constant Acceleration')
    plt.xlabel('Time (s)')
    plt.ylabel('Position (m)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
