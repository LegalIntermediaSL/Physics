# Compendio Subtemático: Mecanica Cuantica Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
sigma_x = np.linspace(0.1, 5, 100)
sigma_p = 1 / (2 * sigma_x)
plt.plot(sigma_x, sigma_x * sigma_p, label='dx * dp = hbar/2')
plt.plot(sigma_x, sigma_p, label='dp vs dx')
plt.title('Uncertainty Principle (Gaussian Packet)')
plt.legend()
plt.grid()
plt.show()
```
