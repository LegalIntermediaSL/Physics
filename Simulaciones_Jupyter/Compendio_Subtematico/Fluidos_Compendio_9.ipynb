# Compendio Subtemático: Fluidos Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def reynolds():
    v = np.linspace(0.1, 5, 100) # m/s
    D = 0.05 # m
    rho = 1000
    mu = 0.001
    Re = (rho * v * D) / mu
    plt.plot(v, Re)
    plt.axhline(2000, color='r', linestyle='--', label='Transition to turbulent')
    plt.xlabel('Velocity (m/s)')
    plt.ylabel('Reynolds Number')
    plt.title('Reynolds Number vs Velocity')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    reynolds()
```
