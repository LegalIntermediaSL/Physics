# Compendio Subtemático: Fisica Atomica Compendio 20

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def doppler_broadening(T, M_amu, lambda_0):
    # T in Kelvin, M in amu, lambda_0 in nm
    c = 3e8
    k_B = 1.38e-23
    M = M_amu * 1.66e-27
    
    # Frequency corresponding to lambda_0
    nu_0 = c / (lambda_0 * 1e-9)
    
    # Doppler width (FWHM)
    delta_nu_D = 2 * nu_0 * np.sqrt(2 * np.log(2) * k_B * T / (M * c**2))
    
    nu = np.linspace(nu_0 - 3*delta_nu_D, nu_0 + 3*delta_nu_D, 500)
    
    # Gaussian profile
    I = np.exp(-4 * np.log(2) * ((nu - nu_0) / delta_nu_D)**2)
    
    plt.figure(figsize=(8, 6))
    plt.plot((nu - nu_0)*1e-9, I, 'b-', label=f'T = {T} K')
    plt.xlabel('Frequency detuning (GHz)')
    plt.ylabel('Intensity (normalized)')
    plt.title(f'Doppler Broadening (M={M_amu} amu, $\lambda_0$={lambda_0} nm)')
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    doppler_broadening(300, 23, 589) # Sodium D line at room temp
```
