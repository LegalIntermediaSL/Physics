# Compendio Subtemático: Relatividad Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def cosmological_redshift():
    a = np.linspace(0.1, 1.0, 500) # Scale factor, a=1 is today
    
    # Redshift z = 1/a - 1
    z = 1/a - 1
    
    plt.figure(figsize=(8, 5))
    plt.plot(a, z, color='red')
    plt.xlabel('Scale Factor ($a$)')
    plt.ylabel('Cosmological Redshift ($z$)')
    plt.title('Redshift vs Scale Factor of the Universe')
    plt.grid(True)
    plt.gca().invert_xaxis()
    plt.savefig('cosmological_redshift.png')
    print('Plot saved as cosmological_redshift.png')

if __name__ == '__main__':
    cosmological_redshift()
```
