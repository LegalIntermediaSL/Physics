# Compendio Subtemático: Plasmas Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def magnetic_mirror():
    q, m = 1.6e-19, 1.67e-27
    v0, r0 = np.array([1e5, 1e5, 0]), np.zeros(3)
    dt = 1e-10
    t = np.arange(0, 1e-5, dt)
    r, v = np.zeros((len(t), 3)), np.zeros((len(t), 3))
    r[0], v[0] = r0, v0
    for i in range(1, len(t)):
        z = r[i-1, 2]
        B = np.array([-0.1*r[i-1,0]*z, -0.1*r[i-1,1]*z, 1.0 + 0.1*z**2])
        v[i] = v[i-1] + (q/m)*np.cross(v[i-1], B)*dt
        r[i] = r[i-1] + v[i]*dt
    plt.plot(t, r[:, 2])
    plt.title("Magnetic Mirror (z vs t)")
    plt.grid(True); plt.savefig("mirror.png")

if __name__ == "__main__":
    magnetic_mirror()
```
