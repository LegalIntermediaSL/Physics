# Compendio Subtemático: Astrofisica Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def roche_limit():
    r = np.linspace(0, 2, 100)
    F_grav = 1 / (1 - r)**2
    F_tidal = 2 * r # Linear approx close to body
    
    plt.figure(figsize=(8, 6))
    plt.plot(r, F_grav, label='Self Gravity')
    plt.plot(r, F_tidal, label='Tidal Force')
    plt.ylim(0, 5)
    plt.title("Tidal Forces and the Roche Limit")
    plt.xlabel("Distance from center of satellite")
    plt.ylabel("Force (Arbitrary Units)")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    roche_limit()
```
