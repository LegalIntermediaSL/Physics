# Compendio Subtemático: Fluidos Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def continuity():
    A1 = 0.1 # m^2
    v1 = 5 # m/s
    A2 = np.linspace(0.01, 0.2, 100) # m^2
    v2 = (A1 * v1) / A2
    plt.plot(A2, v2)
    plt.xlabel('Area 2 (m^2)')
    plt.ylabel('Velocity 2 (m/s)')
    plt.title('Equation of Continuity (Q = A1*v1 = 0.5 m^3/s)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    continuity()
```
