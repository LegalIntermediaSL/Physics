# Compendio Subtemático: Nuclear Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    E = np.linspace(0.1, 10, 200)
    # Breit-Wigner resonance simple model
    E0 = 5.0
    gamma = 0.5
    cross_section = (1 / ((E - E0)**2 + (gamma/2)**2))
    
    plt.plot(E, cross_section, color='purple')
    plt.title('Nuclear Resonance Cross Section (Breit-Wigner)')
    plt.xlabel('Energy (MeV)')
    plt.ylabel('Cross Section (barns)')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
