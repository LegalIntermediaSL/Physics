# Compendio Subtemático: Mecanica Clasica Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Fluid Pressure P = P0 + rho*g*h
    h = np.linspace(0, 100, 100) # depth
    rho = 1000 # water density
    g = 9.81
    P0 = 101325 # atmospheric pressure
    
    P = P0 + rho * g * h
    
    plt.plot(P / 1000, h) # P in kPa
    plt.gca().invert_yaxis()
    plt.title('Hydrostatic Pressure vs Depth')
    plt.xlabel('Pressure (kPa)')
    plt.ylabel('Depth (m)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
