# Compendio Subtemático: Plasmas Compendio 31

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def ponderomotive_force():
    x = np.linspace(-5, 5, 100)
    E = np.exp(-x**2)
    grad_E2 = -4*x * np.exp(-2*x**2)
    plt.plot(x, E, label='Electric Field Env')
    plt.plot(x, -grad_E2, label='Ponderomotive Force (propto -grad(E^2))')
    plt.legend()
    plt.title("Ponderomotive Force")
    plt.grid(True); plt.savefig("ponderomotive.png")
if __name__ == "__main__": ponderomotive_force()
```
