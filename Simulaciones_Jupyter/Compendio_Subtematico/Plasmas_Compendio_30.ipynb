# Compendio Subtemático: Plasmas Compendio 30

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def child_langmuir():
    V = np.linspace(0, 100, 100)
    d = 0.01
    J = (4/9) * 8.85e-12 * np.sqrt(2*1.6e-19/9.1e-31) * V**1.5 / d**2
    plt.plot(V, J)
    plt.title("Child-Langmuir Law")
    plt.xlabel("V")
    plt.ylabel("J")
    plt.grid(True); plt.savefig("child_langmuir.png")
if __name__ == "__main__": child_langmuir()
```
