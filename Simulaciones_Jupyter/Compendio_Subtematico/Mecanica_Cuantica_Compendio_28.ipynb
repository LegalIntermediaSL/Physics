# Compendio Subtemático: Mecanica Cuantica Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
E_field = np.linspace(0, 1, 100)
E0 = -13.6
dE = 3 * E_field 
plt.plot(E_field, E0/4 + dE, label='+3ea0 E')
plt.plot(E_field, E0/4 - dE, label='-3ea0 E')
plt.plot(E_field, E0/4 * np.ones_like(E_field), label='Unshifted')
plt.title('Linear Stark Effect (Hydrogen n=2)')
plt.legend()
plt.grid()
plt.show()
```
