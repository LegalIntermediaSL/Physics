# Compendio Subtemático: Termodinamica Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

Th = np.linspace(300, 1500, 100)
Tc = 300

eta_carnot = 1 - (Tc / Th)

plt.plot(Th, eta_carnot * 100, 'b')
plt.title('Carnot Efficiency vs Hot Reservoir Temp')
plt.xlabel('T_h (K)')
plt.ylabel('Efficiency (%)')
plt.grid(True)
plt.savefig('carnot_efficiency.png')
```
