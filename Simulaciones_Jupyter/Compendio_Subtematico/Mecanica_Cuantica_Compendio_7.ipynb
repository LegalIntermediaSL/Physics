# Compendio Subtemático: Mecanica Cuantica Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import genlaguerre
import math

def R_nl(r, n, l):
    a0 = 1.0
    rho = 2.0 * r / (n * a0)
    L_poly = genlaguerre(n - l - 1, 2 * l + 1)(rho)
    norm = np.sqrt((2.0/(n*a0))**3 * math.factorial(n-l-1)/(2.0*n*math.factorial(n+l)))
    return norm * np.exp(-rho/2.0) * (rho**l) * L_poly

r = np.linspace(0, 20, 1000)
plt.plot(r, R_nl(r, 1, 0)**2 * r**2, label='1s')
plt.plot(r, R_nl(r, 2, 0)**2 * r**2, label='2s')
plt.plot(r, R_nl(r, 2, 1)**2 * r**2, label='2p')
plt.title('Hydrogen Radial Probability Densities')
plt.xlabel('r (Bohr radii)')
plt.ylabel('P(r)')
plt.legend()
plt.grid(True)
plt.show()
```
