# Compendio Subtemático: Electromagnetismo Compendio 26

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(-5, 5, 500)
    v = 1.0
    z = v * t
    R = 1.0
    # simplified flux ~ 1 / (z^2 + R^2)^{3/2}
    flux = 1.0 / (z**2 + R**2)**1.5
    emf = -np.gradient(flux, t)
    plt.plot(t, emf)
    plt.xlabel('Time')
    plt.ylabel('EMF')
    plt.title("26. Faraday's Law - EMF of Magnet Falling Through Loop")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
