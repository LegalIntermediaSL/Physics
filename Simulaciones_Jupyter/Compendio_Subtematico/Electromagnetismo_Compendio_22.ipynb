# Compendio Subtemático: Electromagnetismo Compendio 22

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    z = np.linspace(-10, 10, 200)
    L = 5.0
    R = 1.0
    mu0 = 4*np.pi*1e-7
    n = 1000
    I = 1.0
    B = (mu0*n*I/2) * ((z + L/2)/np.sqrt((z + L/2)**2 + R**2) - (z - L/2)/np.sqrt((z - L/2)**2 + R**2))
    plt.plot(z, B)
    plt.xlabel('Distance along axis z')
    plt.ylabel('Magnetic Field B_z')
    plt.title("22. Magnetic Field on Axis of a Solenoid")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
