# Compendio Subtemático: Electromagnetismo Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 500)
    R = 5.0
    L = 1.0
    C = 1.0
    alpha = R / (2*L)
    omega0 = 1.0 / np.sqrt(L*C)
    s1 = -alpha + np.sqrt(alpha**2 - omega0**2)
    s2 = -alpha - np.sqrt(alpha**2 - omega0**2)
    Q0 = 1.0
    A = Q0 / (s1 - s2)
    Q = A * (s1*np.exp(s2*t) - s2*np.exp(s1*t))
    plt.plot(t, Q)
    plt.xlabel('Time (s)')
    plt.ylabel('Charge (C)')
    plt.title("13. RLC Circuit Overdamped")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
