# Compendio Subtemático: Mecanica Clasica Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # F = -kx - bv
    t = np.linspace(0, 10, 500)
    m = 1.0
    k = 10.0
    b = 0.5 # Damping coefficient
    omega0 = np.sqrt(k/m)
    gamma = b / (2*m)
    omega = np.sqrt(omega0**2 - gamma**2)
    
    A = 1.0
    x = A * np.exp(-gamma * t) * np.cos(omega * t)
    
    plt.plot(t, x)
    plt.title('Damped Harmonic Oscillator')
    plt.xlabel('Time (s)')
    plt.ylabel('Displacement (m)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
