# Compendio Subtemático: Termodinamica Compendio 26

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

P = np.linspace(1, 1000, 100) # atm
# empirical approximation for Z
Z_T1 = 1 - 0.5 * (P/500) + 0.5 * (P/500)**2
Z_T2 = 1 + 0.2 * (P/500)

plt.plot(P, Z_T1, label='Low T')
plt.plot(P, Z_T2, label='High T')
plt.axhline(1, color='k', linestyle='--', label='Ideal Gas')

plt.title('Compressibility Factor Z vs Pressure')
plt.xlabel('Pressure (atm)')
plt.ylabel('Z (PV/RT)')
plt.legend()
plt.grid(True)
plt.savefig('compressibility.png')
```
