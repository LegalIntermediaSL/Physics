# Compendio Subtemático: Relatividad Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gravitational_redshift():
    G = 6.674e-11
    M = 1.989e30 # Mass of Sun
    c = 3e8
    Rs = 2 * G * M / c**2
    R_sun = 6.96e8
    
    r = np.linspace(R_sun, 5*R_sun, 500)
    
    # Redshift z = (1 - Rs/r)^(-1/2) - 1
    z = 1 / np.sqrt(1 - Rs/r) - 1
    
    plt.figure(figsize=(8, 5))
    plt.plot(r/R_sun, z, color='maroon')
    plt.xlabel('Distance from Sun Center ($R/R_{sun}$)')
    plt.ylabel('Gravitational Redshift ($z$)')
    plt.title('Gravitational Redshift from the Sun')
    plt.grid(True)
    plt.savefig('grav_redshift.png')
    print('Plot saved as grav_redshift.png')

if __name__ == '__main__':
    gravitational_redshift()
```
