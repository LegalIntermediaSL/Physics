# Compendio Subtemático: Mecanica Clasica Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    h = np.linspace(0, 100, 100)
    m = 2.0
    g = 9.81
    U = m * g * h
    plt.plot(h, U)
    plt.title('Gravitational Potential Energy')
    plt.xlabel('Height (m)')
    plt.ylabel('Potential Energy (J)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
