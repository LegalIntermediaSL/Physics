# Compendio Subtemático: Nuclear Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    E = np.linspace(0.01, 0.5, 100)
    # Photoelectric cross-section roughly proportional to Z^n / E^3
    cross_section = 1 / E**3
    
    plt.plot(E, cross_section, color='red')
    plt.yscale('log')
    plt.title('Photoelectric Effect Cross Section vs Energy')
    plt.xlabel('Photon Energy (MeV)')
    plt.ylabel('Cross Section (log scale)')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
