# Compendio Subtemático: Electromagnetismo Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    theta = np.linspace(-0.1, 0.1, 1000)
    lam = 500e-9
    d = 20e-6
    I = np.cos(np.pi * d * np.sin(theta) / lam)**2
    plt.plot(theta, I)
    plt.xlabel('Angle (rad)')
    plt.ylabel('Intensity')
    plt.title("33. Double Slit Interference")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
