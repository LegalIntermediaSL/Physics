# Compendio Subtemático: Termodinamica Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 60, 100)
T_env = 20
T_init = 100
k = 0.05 # cooling constant

T = T_env + (T_init - T_env) * np.exp(-k * t)

plt.plot(t, T, 'b')
plt.title("Newton's Law of Cooling")
plt.xlabel('Time (min)')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.savefig('newton_cooling.png')
```
