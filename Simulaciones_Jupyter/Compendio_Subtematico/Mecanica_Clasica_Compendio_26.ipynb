# Compendio Subtemático: Mecanica Clasica Compendio 26

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 100)
    L = 2.0
    g = 9.81
    theta0 = 0.2 # Small angle assumption
    omega = np.sqrt(g/L)
    
    theta = theta0 * np.cos(omega * t)
    
    plt.plot(t, theta)
    plt.title('Simple Pendulum (Small Angle Approximation)')
    plt.xlabel('Time (s)')
    plt.ylabel('Angle (rad)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
