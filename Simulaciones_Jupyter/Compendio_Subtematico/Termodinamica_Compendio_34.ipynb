# Compendio Subtemático: Termodinamica Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

V = np.linspace(1, 10, 100)
T = 300
R = 8.314
# Ideal gas F(V) - F(V0) = -nRT ln(V/V0)
F = -R * T * np.log(V / 1.0)

plt.plot(V, F, 'k')
plt.title('Helmholtz Free Energy vs Volume (Isothermal)')
plt.xlabel('Volume (L)')
plt.ylabel('Delta F (J/mol)')
plt.grid(True)
plt.savefig('helmholtz.png')
```
