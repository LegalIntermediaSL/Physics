# Compendio Subtemático: Electromagnetismo Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 20, 500)
    R = 0.5
    L = 1.0
    C = 1.0
    alpha = R / (2*L)
    omega0 = 1.0 / np.sqrt(L*C)
    omega = np.sqrt(omega0**2 - alpha**2)
    Q0 = 1.0
    Q = Q0 * np.exp(-alpha*t) * np.cos(omega * t)
    plt.plot(t, Q)
    plt.xlabel('Time (s)')
    plt.ylabel('Charge (C)')
    plt.title("11. RLC Circuit Damped Oscillations")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
