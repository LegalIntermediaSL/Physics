# Compendio Subtemático: Mecanica Cuantica Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite
import math
x = np.linspace(-5, 5, 500)
n = 5
Hn = hermite(n)
psi = (1.0 / np.sqrt((2**n) * math.factorial(n) * np.sqrt(np.pi))) * np.exp(-x**2 / 2.0) * Hn(x)
plt.plot(x, psi**2)
plt.title('Quantum Harmonic Oscillator Probability (n=5)')
plt.grid()
plt.show()
```
