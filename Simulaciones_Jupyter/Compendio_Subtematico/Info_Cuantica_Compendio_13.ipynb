# Compendio Subtemático: Info Cuantica Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 5, 100)
    T2 = 1.5
    coherence = np.exp(-t / T2)
    
    plt.plot(t, coherence, color='black', lw=2)
    plt.title('Qubit Decoherence over Time')
    plt.xlabel('Time (us)')
    plt.ylabel('Off-diagonal Element Magnitude |rho_01|')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
