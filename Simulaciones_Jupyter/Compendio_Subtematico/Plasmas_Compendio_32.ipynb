# Compendio Subtemático: Plasmas Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def whistler_waves():
    k = np.linspace(0.1, 5, 100)
    omega = k**2
    plt.plot(k, omega)
    plt.title("Whistler Waves Dispersion (Toy)")
    plt.xlabel("k"); plt.ylabel("Omega")
    plt.grid(True); plt.savefig("whistler.png")
if __name__ == "__main__": whistler_waves()
```
