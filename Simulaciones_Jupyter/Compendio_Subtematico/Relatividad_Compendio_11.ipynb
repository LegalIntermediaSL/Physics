# Compendio Subtemático: Relatividad Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gravitational_time_dilation():
    G = 6.674e-11
    M = 5.972e24 # Mass of Earth
    c = 3e8
    R = 6.371e6 # Radius of Earth
    
    r = np.linspace(R, R + 20000e3, 500) # Up to 20000 km altitude
    Rs = 2 * G * M / c**2
    
    # Ratio of proper time to coordinate time
    dilation = np.sqrt(1 - Rs/r)
    
    plt.figure(figsize=(8, 5))
    plt.plot((r-R)/1000, 1 - dilation, color='navy')
    plt.xlabel('Altitude (km)')
    plt.ylabel('Fractional Time Difference ($1 - t_0/t_f$)')
    plt.title('Gravitational Time Dilation near Earth')
    plt.grid(True)
    plt.savefig('grav_time_dilation.png')
    print('Plot saved as grav_time_dilation.png')

if __name__ == '__main__':
    gravitational_time_dilation()
```
