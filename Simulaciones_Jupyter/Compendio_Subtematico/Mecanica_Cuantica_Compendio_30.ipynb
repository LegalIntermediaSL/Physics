# Compendio Subtemático: Mecanica Cuantica Compendio 30

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
t = np.linspace(0, 50, 500)
g = 1.0 
P_e = np.zeros_like(t)
for n in range(10):
    pn = np.exp(-5) * (5**n) / math.factorial(n)
    P_e += pn * np.cos(g * np.sqrt(n+1) * t)**2
plt.plot(t, P_e)
plt.title('Jaynes-Cummings Collapse and Revival')
plt.grid()
plt.show()
```
