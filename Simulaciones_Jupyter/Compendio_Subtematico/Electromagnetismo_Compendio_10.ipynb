# Compendio Subtemático: Electromagnetismo Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 500)
    L = 1.0
    C = 1.0
    omega = 1.0 / np.sqrt(L*C)
    Q0 = 1.0
    Q = Q0 * np.cos(omega * t)
    I = -Q0 * omega * np.sin(omega * t)
    plt.plot(t, Q, label='Charge')
    plt.plot(t, I, label='Current')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title("10. LC Circuit Oscillations")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
