# Compendio Subtemático: Nuclear Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    A = np.arange(10, 250)
    Z = A / 2.0  # Simplification for stable nuclei approximation
    aV, aS, aC, aA = 15.75, 17.8, 0.711, 23.7
    
    BE = aV * A - aS * A**(2/3) - aC * Z**2 / A**(1/3) - aA * (A - 2*Z)**2 / A
    BE_per_A = BE / A
    
    plt.plot(A, BE_per_A, color='purple')
    plt.title('Binding Energy per Nucleon (Liquid Drop Model)')
    plt.xlabel('Mass Number A')
    plt.ylabel('BE/A (MeV)')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
