# Compendio Subtemático: Electromagnetismo Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    theta1 = np.linspace(0, np.pi/2, 500)
    n1, n2 = 1.5, 1.0
    sin_theta2 = n1/n2 * np.sin(theta1)
    theta2 = np.arcsin(np.clip(sin_theta2, -1, 1))
    theta2[sin_theta2 > 1] = np.nan
    plt.plot(np.degrees(theta1), np.degrees(theta2))
    plt.axvline(np.degrees(np.arcsin(n2/n1)), color='r', linestyle='--', label='Critical Angle')
    plt.xlabel('Incident Angle (deg)')
    plt.ylabel('Refracted Angle (deg)')
    plt.title("32. Total Internal Reflection")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
