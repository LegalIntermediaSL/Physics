# Compendio Subtemático: Materia Condensada Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Density of States of Graphene (Linear near Dirac point)
E = np.linspace(-3.0, 3.0, 500) # Energy in eV
# Near the Dirac point (E=0), DOS is proportional to |E|
dos_graphene = np.abs(E)

# For a normal 2D electron gas, DOS is constant
dos_2deg = np.ones_like(E)

plt.figure(figsize=(8, 5))
plt.plot(E, dos_graphene, lw=2, label="Graphene DOS ($\propto |E|$)", color='darkblue')
plt.plot(E, dos_2deg, lw=2, label="2DEG DOS (Constant)", color='orange', linestyle='--')
plt.title("Density of States: Graphene vs Normal 2D Electron Gas")
plt.xlabel("Energy $E$ (eV)")
plt.ylabel("Density of States $g(E)$")
plt.legend()
plt.grid(True)
plt.savefig("graphene_dos.png")
```
