# Compendio Subtemático: Fluidos Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rankine_vortex():
    r = np.linspace(0, 5, 100)
    R = 1 # core radius
    Gamma = 10
    v = np.piecewise(r, [r < R, r >= R], [lambda r: Gamma * r / (2 * np.pi * R**2), lambda r: Gamma / (2 * np.pi * r)])
    plt.plot(r, v)
    plt.xlabel('Radius r (m)')
    plt.ylabel('Tangential Velocity v (m/s)')
    plt.title('Rankine Vortex')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    rankine_vortex()
```
