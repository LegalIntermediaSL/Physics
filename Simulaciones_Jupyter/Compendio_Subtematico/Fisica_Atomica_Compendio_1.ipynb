# Compendio Subtemático: Fisica Atomica Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def bohr_model(n_max):
    n = np.arange(1, n_max + 1)
    # Energy in eV
    E = -13.6 / (n**2)
    # Radius in Angstroms
    r = 0.529 * (n**2)
    return n, E, r

def plot_bohr(n, E, r):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Energy levels
    for i, e in enumerate(E):
        ax1.hlines(e, 0, 1, colors='b')
        ax1.text(1.05, e, f'n={n[i]}, E={e:.2f} eV', va='center')
    ax1.set_xlim(0, 1.5)
    ax1.set_ylabel('Energy (eV)')
    ax1.set_title('Bohr Model Energy Levels')
    ax1.axis('off')
    
    # Orbits
    theta = np.linspace(0, 2*np.pi, 100)
    for rad in r:
        ax2.plot(rad * np.cos(theta), rad * np.sin(theta), label=f'r={rad:.2f} $\AA$')
    ax2.plot(0, 0, 'ro', markersize=10, label='Nucleus')
    ax2.set_aspect('equal')
    ax2.set_xlabel('x ($\AA$)')
    ax2.set_ylabel('y ($\AA$)')
    ax2.set_title('Bohr Orbits')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    n, E, r = bohr_model(5)
    plot_bohr(n, E, r)
```
