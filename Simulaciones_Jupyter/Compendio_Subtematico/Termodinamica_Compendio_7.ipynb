# Compendio Subtemático: Termodinamica Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Simulate water heating from ice to steam
time = np.linspace(0, 100, 500)
temp = np.piecewise(time, 
    [time < 20, (time >= 20) & (time < 40), (time >= 40) & (time < 60), (time >= 60) & (time < 80), time >= 80],
    [lambda t: -20 + t, 0, lambda t: (t-40)*5, 100, lambda t: 100 + (t-80)*2]
)

plt.plot(time, temp)
plt.title('Phase Change Temperature Profile (Water)')
plt.xlabel('Time / Heat added (arb. units)')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.savefig('phase_change.png')
```
