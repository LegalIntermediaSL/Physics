# Compendio Subtemático: Optica Compendio 30

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def kerr():
    E = np.linspace(0, 1e6, 100)
    K = 1e-10
    dn = K * E**2
    plt.plot(E, dn)
    plt.title("Kerr Effect (Quadratic Electro-optic)")
    plt.xlabel('Electric Field (V/m)')
    plt.ylabel('Change in Refractive Index (dn)')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    kerr()
```
