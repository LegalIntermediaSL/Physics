# Compendio Subtemático: Fluidos Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def flow_cylinder():
    r = np.linspace(1, 5, 100)
    theta = np.pi / 2 # top of cylinder
    U = 10 # m/s
    R = 1 # m
    u_theta = -U * (1 + (R**2 / r**2)) * np.sin(theta)
    plt.plot(r, np.abs(u_theta))
    plt.xlabel('Radial distance r (m)')
    plt.ylabel('Tangential Velocity (m/s)')
    plt.title('Ideal Flow over Cylinder')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    flow_cylinder()
```
