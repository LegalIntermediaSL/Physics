# Compendio Subtemático: Mecanica Cuantica Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

E = np.linspace(0.1, 2, 500)
V0 = 1.0
L = 2.0
m = 1.0
hbar = 1.0

T = np.zeros_like(E)
for i, e in enumerate(E):
    if e < V0:
        kappa = np.sqrt(2*m*(V0 - e)) / hbar
        T[i] = 1.0 / (1.0 + (V0**2 * np.sinh(kappa*L)**2) / (4*e*(V0 - e)))
    else:
        k = np.sqrt(2*m*(e - V0)) / hbar
        T[i] = 1.0 / (1.0 + (V0**2 * np.sin(k*L)**2) / (4*e*(e - V0)))

plt.plot(E, T)
plt.axvline(V0, color='r', linestyle='--', label='V0')
plt.title('Quantum Tunneling Probability vs Energy')
plt.xlabel('Energy (E)')
plt.ylabel('Transmission Probability (T)')
plt.legend()
plt.grid(True)
plt.show()
```
