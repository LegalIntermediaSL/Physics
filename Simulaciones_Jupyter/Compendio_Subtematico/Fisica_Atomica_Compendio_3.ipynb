# Compendio Subtemático: Fisica Atomica Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rydberg_formula(n1, n2_max):
    R_H = 1.097e7 # Rydberg constant in m^-1
    n2 = np.arange(n1 + 1, n2_max + 1)
    wavelengths = 1 / (R_H * (1 / n1**2 - 1 / n2**2))
    return n2, wavelengths * 1e9 # Return in nm

def plot_spectra():
    series = {'Lyman (n1=1)': 1, 'Balmer (n1=2)': 2, 'Paschen (n1=3)': 3}
    
    fig, ax = plt.subplots(figsize=(12, 4))
    
    colors = ['b', 'g', 'r']
    for (name, n1), color in zip(series.items(), colors):
        n2, wl = rydberg_formula(n1, n1 + 10)
        for w in wl:
            ax.vlines(w, 0, 1, colors=color, label=name if w == wl[0] else "")
            
    ax.set_xlabel('Wavelength (nm)')
    ax.set_yticks([])
    ax.set_title('Hydrogen Emission Spectrum')
    ax.legend()
    ax.set_xscale('log')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    plot_spectra()
```
