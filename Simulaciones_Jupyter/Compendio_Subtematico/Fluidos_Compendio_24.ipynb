# Compendio Subtemático: Fluidos Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def corner_flow():
    x = np.linspace(0.1, 2, 20)
    y = np.linspace(0.1, 2, 20)
    X, Y = np.meshgrid(x, y)
    A = 1
    psi = A * X * Y
    plt.contour(X, Y, psi, 20)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Streamlines in a 90-degree Corner')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    corner_flow()
```
