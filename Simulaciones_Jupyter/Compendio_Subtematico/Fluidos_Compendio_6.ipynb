# Compendio Subtemático: Fluidos Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def torricelli():
    h = np.linspace(0, 10, 100) # m
    g = 9.81
    v = np.sqrt(2 * g * h)
    plt.plot(h, v)
    plt.xlabel('Height of fluid (m)')
    plt.ylabel('Efflux Velocity (m/s)')
    plt.title('Torricelli\\'s Law')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    torricelli()
```
