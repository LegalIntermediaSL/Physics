# Compendio Subtemático: Relatividad Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def lorentz_transform():
    c = 1.0
    v = 0.5 * c
    gamma = 1 / np.sqrt(1 - (v/c)**2)
    
    # Grid in rest frame S
    x = np.linspace(-5, 5, 11)
    t = np.linspace(-5, 5, 11)
    
    X, T = np.meshgrid(x, t)
    
    # Transform to moving frame S'
    X_prime = gamma * (X - v * T)
    T_prime = gamma * (T - v * X / c**2)
    
    plt.figure(figsize=(6, 6))
    for i in range(len(x)):
        plt.plot(X_prime[:, i], T_prime[:, i], color='blue', alpha=0.5)
        plt.plot(X_prime[i, :], T_prime[i, :], color='red', alpha=0.5)
        
    plt.xlabel("x'")
    plt.ylabel("t'")
    plt.title('Lorentz Transformation Grid ($v = 0.5c$)')
    plt.grid(True)
    plt.axis('equal')
    plt.savefig('lorentz_transform.png')
    print('Plot saved as lorentz_transform.png')

if __name__ == '__main__':
    lorentz_transform()
```
