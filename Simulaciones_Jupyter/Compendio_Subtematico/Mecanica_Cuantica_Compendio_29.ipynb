# Compendio Subtemático: Mecanica Cuantica Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
B = np.linspace(0, 5, 100)
g = 2.0
mu_b = 5.788e-5 
plt.plot(B, g * mu_b * B * 0.5, label='Spin up')
plt.plot(B, -g * mu_b * B * 0.5, label='Spin down')
plt.title('Normal Zeeman Effect Splitting')
plt.legend()
plt.grid()
plt.show()
```
