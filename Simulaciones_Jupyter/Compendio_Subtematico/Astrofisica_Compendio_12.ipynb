# Compendio Subtemático: Astrofisica Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def pulsar():
    t = np.linspace(0, 5, 1000)
    period = 1.0
    signal = np.exp(-((t % period - 0.5)**2) / 0.005)
    noise = np.random.normal(0, 0.1, len(t))
    
    plt.figure(figsize=(10, 5))
    plt.plot(t, signal + noise, 'c')
    plt.title("Pulsar Radio Emission")
    plt.xlabel("Time (s)")
    plt.ylabel("Intensity")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    pulsar()
```
