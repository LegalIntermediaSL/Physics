# Compendio Subtemático: Fisica Atomica Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def moseleys_law():
    # K-alpha lines
    # f = c / lambda = A * (Z - b)^2
    # For K-alpha, b ~ 1, A ~ (3/4)*c*R_H
    
    Z = np.arange(10, 60, 2)
    b = 1
    R_H = 1.097e7
    c = 3e8
    A = (3/4) * c * R_H
    
    f = A * (Z - b)**2
    sqrt_f = np.sqrt(f)
    
    plt.figure(figsize=(8, 6))
    plt.plot(Z, sqrt_f, 'bo-', label='K-alpha lines')
    plt.xlabel('Atomic Number Z')
    plt.ylabel('$\sqrt{f}$ ($\sqrt{Hz}$)')
    plt.title("Moseley's Law for K-alpha X-rays")
    plt.grid(True)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    moseleys_law()
```
