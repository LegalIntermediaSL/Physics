# Compendio Subtemático: Astrofisica Compendio 31

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def protoplanetary_disk():
    r = np.linspace(0.1, 50, 200) # AU
    sigma = 1000 * r**(-1.5) # Surface density
    
    plt.figure(figsize=(8, 6))
    plt.plot(r, sigma, 'k-', linewidth=2)
    plt.yscale('log')
    plt.xscale('log')
    plt.title("Protoplanetary Disk Surface Density")
    plt.xlabel("Radius (AU)")
    plt.ylabel("$\Sigma$ (g/cm$^2$)")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    protoplanetary_disk()
```
