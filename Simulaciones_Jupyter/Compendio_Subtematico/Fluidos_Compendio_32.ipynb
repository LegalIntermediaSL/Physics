# Compendio Subtemático: Fluidos Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def sloshing():
    t = np.linspace(0, 10, 200)
    omega = 2.0 # natural frequency
    zeta = 0.05 # damping ratio
    amp = np.exp(-zeta * omega * t) * np.cos(omega * np.sqrt(1 - zeta**2) * t)
    plt.plot(t, amp)
    plt.xlabel('Time (s)')
    plt.ylabel('Wave amplitude (m)')
    plt.title('Damped Fluid Sloshing in a Tank')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    sloshing()
```
