# Compendio Subtemático: Astrofisica Compendio 20

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def scale_factor():
    t = np.linspace(0.1, 20, 200) # Gyr
    a_matter = (t/13.8)**(2/3) # Matter dominated
    a_rad = (t/13.8)**(1/2) # Rad dominated
    a_lambda = np.exp(0.1*(t-13.8)) # Dark energy dominated
    
    plt.figure(figsize=(10, 6))
    plt.plot(t, a_matter, label='Matter-dominated')
    plt.plot(t, a_rad, label='Radiation-dominated')
    plt.plot(t, a_lambda, label='Lambda-dominated')
    plt.title("Evolution of the Scale Factor $a(t)$")
    plt.xlabel("Time (Gyr)")
    plt.ylabel("Scale Factor $a$")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    scale_factor()
```
