# Compendio Subtemático: Fluidos Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def stokes_first_problem():
    y = np.linspace(0, 0.05, 100)
    nu = 1e-6 # kinematic viscosity of water
    U0 = 1.0 # plate velocity
    t_vals = [0.1, 0.5, 1.0, 2.0]
    for t in t_vals:
        from scipy.special import erfc
        u = U0 * erfc(y / (2 * np.sqrt(nu * t)))
        plt.plot(u, y, label=f't = {t}s')
    plt.xlabel('Velocity u (m/s)')
    plt.ylabel('Distance y (m)')
    plt.title('Stokes\\' First Problem (Impulsively Started Plate)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    stokes_first_problem()
```
