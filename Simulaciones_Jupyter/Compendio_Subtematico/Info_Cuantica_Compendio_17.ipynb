# Compendio Subtemático: Info Cuantica Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # 3-qubit bit flip code correction rate vs physical error rate
    p = np.linspace(0, 0.5, 100)
    p_logical = 3 * p**2 - 2 * p**3
    
    plt.plot(p, p, 'k--', label='Unencoded (p)')
    plt.plot(p, p_logical, 'b-', label='Encoded (3-qubit)')
    plt.title('Quantum Error Correction Performance')
    plt.xlabel('Physical Error Rate (p)')
    plt.ylabel('Logical Error Rate')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
