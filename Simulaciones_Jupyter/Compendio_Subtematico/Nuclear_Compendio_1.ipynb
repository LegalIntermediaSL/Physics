# Compendio Subtemático: Nuclear Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 100)
    N0 = 1000
    lambda_decay = 0.5
    N_t = N0 * np.exp(-lambda_decay * t)
    
    plt.plot(t, N_t, 'r-', lw=2)
    plt.title('Radioactive Decay (Exponential)')
    plt.xlabel('Time')
    plt.ylabel('Number of Nuclei N(t)')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
