# Compendio Subtemático: Electromagnetismo Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 4*np.pi, 500)
    Ex = np.cos(t)
    Ey = np.cos(t)
    plt.plot(Ex, Ey)
    plt.xlabel('E_x')
    plt.ylabel('E_y')
    plt.title("29. Linear Polarization of EM Wave")
    plt.axis('equal')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
