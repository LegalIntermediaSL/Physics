# Compendio Subtemático: Mecanica Clasica Compendio 31

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x = np.linspace(0, np.pi, 100)
    # Standing wave y = 2A * sin(kx) * cos(wt)
    # Plotting envelopes and some snapshots
    plt.plot(x, np.sin(x), 'k--', label='Envelope')
    plt.plot(x, -np.sin(x), 'k--')
    
    for t in [0, 0.25, 0.5, 0.75, 1.0]:
        y = np.sin(x) * np.cos(2*np.pi * t)
        plt.plot(x, y, label=f't={t}T')
        
    plt.title('Standing Wave on a String (Fundamental)')
    plt.xlabel('Position x (m)')
    plt.ylabel('Displacement y')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
