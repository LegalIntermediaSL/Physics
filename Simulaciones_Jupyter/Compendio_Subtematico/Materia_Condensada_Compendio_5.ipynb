# Compendio Subtemático: Materia Condensada Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Tight binding model 1D
a = 1.0 # lattice constant
k = np.linspace(-np.pi/a, np.pi/a, 200)
E0 = 0.0 # onsite energy
t = 1.0 # hopping parameter

E_k = E0 - 2 * t * np.cos(k * a)

plt.figure(figsize=(8, 5))
plt.plot(k, E_k, lw=2, color='purple')
plt.title("1D Tight Binding Model: Dispersion Relation")
plt.xlabel("Wavevector k")
plt.ylabel("Energy E(k)")
plt.xticks([-np.pi/a, 0, np.pi/a], [r'$-\pi/a$', '0', r'$\pi/a$'])
plt.axhline(E0, color='gray', linestyle='--', label="Onsite Energy $E_0$")
plt.legend()
plt.grid(True)
plt.savefig("tight_binding_1d.png")
```
