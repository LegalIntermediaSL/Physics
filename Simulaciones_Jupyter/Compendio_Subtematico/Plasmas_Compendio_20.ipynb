# Compendio Subtemático: Plasmas Compendio 20

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def resistive_diffusion():
    x = np.linspace(-1, 1, 100)
    t_vals = [0.01, 0.1, 0.5]
    for t in t_vals:
        B = np.exp(-x**2 / (4*0.1*t)) / np.sqrt(4*np.pi*0.1*t)
        plt.plot(x, B, label=f't={t}')
    plt.legend()
    plt.title("Resistive Diffusion of Magnetic Field")
    plt.grid(True); plt.savefig("diffusion.png")
if __name__ == "__main__": resistive_diffusion()
```
