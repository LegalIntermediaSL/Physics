# Compendio Subtemático: Plasmas Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def magnetosonic_waves():
    k = np.linspace(0, 10, 100)
    Va, Cs = 5000.0, 3000.0
    omega_fast = k * np.sqrt(Va**2 + Cs**2)
    plt.plot(k, omega_fast)
    plt.title("Fast Magnetosonic Wave")
    plt.grid(True); plt.savefig("magnetosonic.png")
if __name__ == "__main__": magnetosonic_waves()
```
