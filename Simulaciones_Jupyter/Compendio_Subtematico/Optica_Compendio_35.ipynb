# Compendio Subtemático: Optica Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def zernike():
    n = 100
    x = np.linspace(-1, 1, n)
    X, Y = np.meshgrid(x, x)
    R = np.sqrt(X**2 + Y**2)
    Z = (2*R**2 - 1) * (R <= 1)
    
    plt.imshow(Z, cmap='viridis')
    plt.title("Zernike Defocus Mode")
    plt.axis('off')
    plt.colorbar()
    plt.show()

if __name__ == '__main__': 
    zernike()
```
