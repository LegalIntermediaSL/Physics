# Compendio Subtemático: Fluidos Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def water_hammer():
    v_stop = np.linspace(0, 5, 100) # m/s
    rho = 1000
    a = 1482 # speed of sound in water
    delta_P = rho * a * v_stop
    plt.plot(v_stop, delta_P/1e6)
    plt.xlabel('Stopped Velocity (m/s)')
    plt.ylabel('Pressure Spike (MPa)')
    plt.title('Water Hammer Effect')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    water_hammer()
```
