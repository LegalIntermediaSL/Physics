# Compendio Subtemático: Relatividad Compendio 26

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def hyperbolic_motion():
    # Constant proper acceleration a
    a = 1.0
    tau = np.linspace(0, 5, 100) # Proper time
    
    # x(tau) = (c^2/a) * cosh(a*tau/c) - c^2/a
    # t(tau) = (c/a) * sinh(a*tau/c)
    x = np.cosh(tau) - 1
    t = np.sinh(tau)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, t, label='Worldline (a=1)', color='magenta')
    plt.plot(x, x, '--', color='gray', label='Light speed asymptote')
    plt.xlabel('Distance $x$')
    plt.ylabel('Time $t$')
    plt.title('Hyperbolic Motion (Constant Proper Acceleration)')
    plt.grid(True)
    plt.legend()
    plt.savefig('hyperbolic_motion.png')
    print('Plot saved as hyperbolic_motion.png')

if __name__ == '__main__':
    hyperbolic_motion()
```
