# Compendio Subtemático: Mecanica Clasica Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Bernoulli's principle simplified: P + 0.5*rho*v^2 = const
    v = np.linspace(1, 20, 100)
    rho = 1.225 # air density
    const = 101325 + 0.5 * rho * 5**2 # P0 + initial KE
    
    P = const - 0.5 * rho * v**2
    
    plt.plot(v, P)
    plt.title('Bernoullis Principle: Pressure vs Velocity')
    plt.xlabel('Fluid Velocity (m/s)')
    plt.ylabel('Pressure (Pa)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
