# Compendio Subtemático: Plasmas Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def z_pinch():
    r = np.linspace(0, 1, 100)
    Jz = 1.0 - r**2
    B_theta = r * (1 - 0.5*r**2)
    plt.plot(r, B_theta, label='B_theta')
    plt.plot(r, Jz, label='Jz')
    plt.legend()
    plt.title("Z-Pinch Equilibrium")
    plt.grid(True); plt.savefig("zpinch.png")
if __name__ == "__main__": z_pinch()
```
