# Compendio Subtemático: Mecanica Cuantica Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite
import math

def psi_ho(x, n):
    Hn = hermite(n)
    return (1.0 / np.sqrt((2**n) * math.factorial(n) * np.sqrt(np.pi))) * np.exp(-x**2 / 2.0) * Hn(x)

x = np.linspace(-4, 4, 1000)
for n in range(4):
    plt.plot(x, psi_ho(x, n) + n + 0.5, label=f'n={n}')
plt.title('Quantum Harmonic Oscillator Wavefunctions')
plt.xlabel('x')
plt.ylabel(r'$\psi_n(x) + E_n$')
plt.legend()
plt.grid(True)
plt.show()
```
