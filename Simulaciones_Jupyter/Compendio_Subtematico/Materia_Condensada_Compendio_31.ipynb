# Compendio Subtemático: Materia Condensada Compendio 31

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# X-Ray Diffraction (Bragg's Law) Simulation
theta_deg = np.linspace(10, 80, 1000)
theta_rad = np.radians(theta_deg)

wavelength = 1.54 # Angstroms (Cu K-alpha)
d_spacing = 2.5 # Interplanar spacing in Angstroms

# Simulated diffraction intensity (using a Lorentzian profile for peaks)
def intensity(theta, n, d, wl, width=0.5):
    # Bragg condition: n * wl = 2 * d * sin(theta_bragg)
    sin_theta_bragg = (n * wl) / (2 * d)
    if sin_theta_bragg > 1: return np.zeros_like(theta)
    theta_bragg = np.arcsin(sin_theta_bragg)
    
    # Peak profile
    return 1.0 / (1.0 + ((theta - theta_bragg) / np.radians(width))**2)

I_total = np.zeros_like(theta_rad)
for n in [1, 2, 3]:
    I_total += intensity(theta_rad, n, d_spacing, wavelength)

plt.figure(figsize=(8, 5))
plt.plot(theta_deg, I_total, color='black', lw=1.5)
plt.title(f"X-Ray Diffraction Pattern (d = {d_spacing} $\AA$, $\lambda$ = {wavelength} $\AA$)")
plt.xlabel(r"Diffraction Angle $\theta$ (degrees)")
plt.ylabel("Intensity (arb. units)")
plt.grid(True, alpha=0.5)
plt.savefig("xray_diffraction.png")
```
