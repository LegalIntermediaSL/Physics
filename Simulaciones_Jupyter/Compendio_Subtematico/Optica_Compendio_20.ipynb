# Compendio Subtemático: Optica Compendio 20

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def airy_fft():
    n = 256
    x = np.linspace(-1, 1, n)
    X, Y = np.meshgrid(x, x)
    aperture = (X**2 + Y**2) <= 0.1**2
    I = np.abs(np.fft.fftshift(np.fft.fft2(aperture)))**2
    
    plt.imshow(I[n//2-20:n//2+20, n//2-20:n//2+20], cmap='hot')
    plt.title("Airy Disk via FFT")
    plt.axis('off')
    plt.show()

if __name__ == '__main__': 
    airy_fft()
```
