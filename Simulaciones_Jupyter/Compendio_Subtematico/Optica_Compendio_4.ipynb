# Compendio Subtemático: Optica Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def thin_lens():
    f = 10
    do = np.linspace(11, 50, 100)
    di = 1 / (1/f - 1/do)
    plt.plot(do, di)
    plt.xlabel('Object Distance')
    plt.ylabel('Image Distance')
    plt.title('Thin Lens Equation')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    thin_lens()
```
