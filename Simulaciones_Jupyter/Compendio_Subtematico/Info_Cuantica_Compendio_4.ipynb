# Compendio Subtemático: Info Cuantica Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    bell_state = 1/np.sqrt(2) * np.array([1, 0, 0, 1])
    labels = ['|00>', '|01>', '|10>', '|11>']
    probs = np.abs(bell_state)**2
    
    plt.bar(labels, probs, color='purple')
    plt.title('Quantum Entanglement: Bell State |Phi+>')
    plt.ylabel('Probability')
    plt.show()

if __name__ == '__main__':
    main()
```
