# Compendio Subtemático: Fluidos Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def kelvin_helmholtz():
    k = np.linspace(0.1, 10, 100)
    U = 5.0 # velocity difference
    rho1 = 1000
    rho2 = 1.2
    g = 9.81
    # Simplified growth rate for KH instability without surface tension
    term = (rho1 * rho2) / (rho1 + rho2)**2
    omega_i = k * U * np.sqrt(term)
    plt.plot(k, omega_i)
    plt.xlabel('Wavenumber k (1/m)')
    plt.ylabel('Growth Rate (1/s)')
    plt.title('Kelvin-Helmholtz Instability Growth Rate')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    kelvin_helmholtz()
```
