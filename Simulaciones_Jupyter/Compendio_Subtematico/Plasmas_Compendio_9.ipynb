# Compendio Subtemático: Plasmas Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def ion_acoustic_waves():
    k = np.linspace(0.1, 10, 100)
    Cs = 3000.0 # m/s
    lambda_D = 0.01
    omega = k * Cs / np.sqrt(1 + k**2 * lambda_D**2)
    plt.plot(k, omega)
    plt.title("Ion Acoustic Wave Dispersion")
    plt.xlabel("k")
    plt.ylabel("omega")
    plt.grid(True); plt.savefig("ion_acoustic.png")

if __name__ == "__main__":
    ion_acoustic_waves()
```
