# Compendio Subtemático: Mecanica Clasica Compendio 28

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 10, 500)
    # y = A1*sin(w1*t) + A2*sin(w2*t)
    y1 = np.sin(2 * np.pi * 1.0 * t)
    y2 = np.sin(2 * np.pi * 1.1 * t)
    y = y1 + y2
    
    plt.plot(t, y)
    plt.title('Superposition and Beats')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
