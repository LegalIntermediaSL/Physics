# Compendio Subtemático: Info Cuantica Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x = np.linspace(-5, 5, 100)
    psi_0 = np.pi**(-0.25) * np.exp(-x**2 / 2) # Ground state
    
    plt.plot(x, np.abs(psi_0)**2, label='|psi_0|^2', color='brown')
    plt.title('Quantum Harmonic Oscillator (Ground State)')
    plt.xlabel('Position (x)')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
