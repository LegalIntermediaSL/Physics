# Compendio Subtemático: Termodinamica Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

dT = np.linspace(0, 100, 100)
L0 = 1.0 # m
alpha = 2.4e-5 # Aluminum

dL = L0 * alpha * dT
L = L0 + dL

plt.plot(dT, L * 1000)
plt.title('Linear Thermal Expansion (Aluminum)')
plt.xlabel('Temperature Change (K)')
plt.ylabel('Length (mm)')
plt.grid(True)
plt.savefig('thermal_expansion.png')
```
