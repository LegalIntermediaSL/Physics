# Compendio Subtemático: Nuclear Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    T = np.linspace(1e6, 1e8, 100)
    # Gamow peak approximation: convolution of Maxwell-Boltzmann and Tunneling prob
    MB = np.exp(-1e7 / T)
    tunnel = np.exp(-np.sqrt(1e9 / T))
    gamow = MB * tunnel
    
    plt.plot(T, gamow, color='crimson')
    plt.title('Gamow Peak for Nuclear Fusion')
    plt.xlabel('Temperature (K)')
    plt.ylabel('Reaction Probability Rate')
    plt.xscale('log')
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
