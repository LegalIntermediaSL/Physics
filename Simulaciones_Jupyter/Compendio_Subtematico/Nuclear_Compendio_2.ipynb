# Compendio Subtemático: Nuclear Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 20, 200)
    l1, l2 = 0.3, 0.1
    N0 = 1000
    
    N1 = N0 * np.exp(-l1 * t)
    N2 = N0 * (l1 / (l2 - l1)) * (np.exp(-l1 * t) - np.exp(-l2 * t))
    N3 = N0 - N1 - N2
    
    plt.plot(t, N1, label='Parent (N1)')
    plt.plot(t, N2, label='Daughter (N2)')
    plt.plot(t, N3, label='Stable (N3)')
    plt.title('Bateman Equations (Radioactive Decay Chain)')
    plt.xlabel('Time')
    plt.ylabel('Number of Nuclei')
    plt.legend()
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
