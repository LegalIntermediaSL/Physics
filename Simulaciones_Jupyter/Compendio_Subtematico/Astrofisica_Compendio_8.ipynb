# Compendio Subtemático: Astrofisica Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def two_body():
    t = np.linspace(0, 10, 1000)
    r1 = 1.0 * np.array([np.cos(t), np.sin(t)])
    r2 = -0.5 * np.array([np.cos(t), np.sin(t)])
    
    plt.figure(figsize=(6, 6))
    plt.plot(r1[0], r1[1], label='Mass 1')
    plt.plot(r2[0], r2[1], label='Mass 2')
    plt.scatter([0], [0], c='k', marker='x', label='Center of Mass')
    plt.title("2-Body Problem")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.axis('equal')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    two_body()
```
