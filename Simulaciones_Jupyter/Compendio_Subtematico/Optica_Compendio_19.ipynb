# Compendio Subtemático: Optica Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def rayleigh():
    lam = np.linspace(400, 700, 100)
    I = 1 / lam**4
    plt.plot(lam, I)
    plt.title("Rayleigh Scattering")
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Scattering Intensity (arb. units)')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    rayleigh()
```
