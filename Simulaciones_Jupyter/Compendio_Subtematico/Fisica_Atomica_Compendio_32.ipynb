# Compendio Subtemático: Fisica Atomica Compendio 32

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def optical_pumping(y, t, pump_rate, decay_rate):
    # Simplified 3-level optical pumping
    # y = [N_g1, N_g2, N_e]
    # Pump from g1 -> e, then decay to g1 and g2
    Ng1, Ng2, Ne = y
    
    dNg1_dt = -pump_rate * Ng1 + 0.5 * decay_rate * Ne
    dNg2_dt = 0.5 * decay_rate * Ne
    dNe_dt = pump_rate * Ng1 - decay_rate * Ne
    
    return [dNg1_dt, dNg2_dt, dNe_dt]

def simulate_optical_pumping():
    y0 = [0.5, 0.5, 0.0] # start equally distributed in ground states
    t = np.linspace(0, 10, 200)
    
    pump = 1.0
    decay = 5.0
    
    sol = odeint(optical_pumping, y0, t, args=(pump, decay))
    
    plt.figure(figsize=(8, 6))
    plt.plot(t, sol[:, 0], label='State 1 (Pumped)')
    plt.plot(t, sol[:, 1], label='State 2 (Dark state)')
    plt.plot(t, sol[:, 2], label='Excited State')
    
    plt.xlabel('Time')
    plt.ylabel('Population fraction')
    plt.title('Optical Pumping into a Dark State')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    simulate_optical_pumping()
```
