# Compendio Subtemático: Fisica Atomica Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def excited_state_decay(tau, N0=1000):
    t = np.linspace(0, 5 * tau, 100)
    N = N0 * np.exp(-t / tau)
    
    # simulate discrete radioactive decay for comparison
    times = np.random.exponential(tau, N0)
    
    plt.figure(figsize=(8, 6))
    plt.plot(t, N, 'r-', linewidth=2, label='Theoretical: $N(t) = N_0 e^{-t/\\tau}$')
    
    # Histogram of simulated decay times
    plt.hist(times, bins=30, alpha=0.5, density=False, weights=np.ones_like(times) * (30 / (5*tau)) * (N0/1000), label='Simulated Counts')
    
    plt.xlabel('Time')
    plt.ylabel('Number of Atoms in Excited State')
    plt.title(f'Spontaneous Emission Decay (Lifetime $\\tau$ = {tau} ns)')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    excited_state_decay(10.0) # 10 ns typical for allowed transitions
```
