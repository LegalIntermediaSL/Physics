# Compendio Subtemático: Materia Condensada Compendio 1

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def drude_velocity(t, E, tau, m, q):
    # v(t) = v_d(1 - exp(-t/tau))
    v_d = (q * E * tau) / m
    return v_d * (1 - np.exp(-t / tau))

t = np.linspace(0, 5e-14, 500)
E_field = 1e4 # V/m
tau_scatter = 1e-14 # s
m_e = 9.11e-31 # kg
q_e = 1.602e-19 # C

v = drude_velocity(t, E_field, tau_scatter, m_e, q_e)

plt.figure(figsize=(8, 5))
plt.plot(t * 1e15, v, label="Electron velocity")
plt.axhline((q_e * E_field * tau_scatter) / m_e, color='r', linestyle='--', label="Drift velocity")
plt.title("Drude Model: Electron Velocity vs Time")
plt.xlabel("Time (fs)")
plt.ylabel("Velocity (m/s)")
plt.legend()
plt.grid(True)
plt.savefig("drude_model.png")
```
