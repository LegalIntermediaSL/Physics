# Compendio Subtemático: Plasmas Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def theta_pinch():
    r = np.linspace(0, 1, 100)
    Bz = 1.0 + r**2
    J_theta = 2*r
    plt.plot(r, Bz, label='Bz')
    plt.plot(r, J_theta, label='J_theta')
    plt.legend()
    plt.title("Theta-Pinch Equilibrium")
    plt.grid(True); plt.savefig("thetapinch.png")
if __name__ == "__main__": theta_pinch()
```
