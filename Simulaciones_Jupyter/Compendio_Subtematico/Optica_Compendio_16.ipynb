# Compendio Subtemático: Optica Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def michelson():
    dL = np.linspace(0, 2e-6, 500)
    lam = 633e-9
    I = np.cos(2*np.pi*dL/lam)**2
    plt.plot(dL*1e9, I)
    plt.title("Michelson Interferometer")
    plt.xlabel('Path Difference (nm)')
    plt.ylabel('Intensity')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    michelson()
```
