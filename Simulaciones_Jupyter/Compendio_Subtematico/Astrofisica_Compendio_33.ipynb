# Compendio Subtemático: Astrofisica Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def redshift_distance():
    z = np.linspace(0, 10, 100)
    # Approximation of luminosity distance
    D_L = 4280 * z * (1 + z/2) # Mpc
    
    plt.figure(figsize=(8, 6))
    plt.plot(z, D_L, 'b-', linewidth=2)
    plt.title("Luminosity Distance vs Redshift")
    plt.xlabel("Redshift z")
    plt.ylabel("Distance (Mpc)")
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    redshift_distance()
```
