# Compendio Subtemático: Plasmas Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def alfven_waves():
    k = np.linspace(0, 10, 100)
    Va = 5000.0
    omega = k * Va
    plt.plot(k, omega)
    plt.title("Alfvén Wave Dispersion")
    plt.xlabel("k")
    plt.ylabel("omega")
    plt.grid(True); plt.savefig("alfven.png")

if __name__ == "__main__":
    alfven_waves()
```
