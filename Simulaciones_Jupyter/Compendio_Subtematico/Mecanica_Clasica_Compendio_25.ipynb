# Compendio Subtemático: Mecanica Clasica Compendio 25

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Driven harmonic oscillator resonance curve
    omega = np.linspace(0.5, 3.0, 200)
    omega0 = 1.5
    gamma = 0.1
    F0 = 1.0
    m = 1.0
    
    A = (F0/m) / np.sqrt((omega0**2 - omega**2)**2 + (2*gamma*omega)**2)
    
    plt.plot(omega, A)
    plt.title('Resonance Curve of Driven Oscillator')
    plt.xlabel('Driving Frequency (rad/s)')
    plt.ylabel('Amplitude (m)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
