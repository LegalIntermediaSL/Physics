# Compendio Subtemático: Mecanica Cuantica Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 200)
s_param = 0.5
psi_sq = (np.pi * s_param)**(-0.25) * np.exp(-x**2 / (2 * s_param))
plt.plot(x, psi_sq**2, label=f'Squeezed (s={s_param})')
plt.title('Squeezed State Probability Density')
plt.legend()
plt.grid()
plt.show()
```
