# Compendio Subtemático: Materia Condensada Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def fermi_dirac(E, EF, T):
    k_B = 8.617e-5 # eV/K
    return 1 / (np.exp((E - EF) / (k_B * T)) + 1)

E = np.linspace(0, 2, 500)
EF = 1.0 # eV
T_vals = [0, 300, 1000, 3000] # K

plt.figure(figsize=(8, 5))
for T in T_vals:
    if T == 0:
        f_E = np.heaviside(EF - E, 0.5)
        plt.plot(E, f_E, label=f'T = {T} K')
    else:
        plt.plot(E, fermi_dirac(E, EF, T), label=f'T = {T} K')

plt.title("Fermi-Dirac Distribution")
plt.xlabel("Energy E (eV)")
plt.ylabel("Occupation Probability f(E)")
plt.axvline(EF, color='k', linestyle='--', label=f'Fermi Energy ($E_F$ = {EF} eV)')
plt.legend()
plt.grid(True)
plt.savefig("fermi_dirac.png")
```
