# Compendio Subtemático: Astrofisica Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def planck_law(wav, T):
    h = 6.626e-34
    c = 3.0e8
    k = 1.38e-23
    a = 2.0 * h * c**2
    b = h * c / (wav * k * T)
    intensity = a / ((wav**5) * (np.exp(b) - 1.0))
    return intensity

def blackbody():
    wavelengths = np.linspace(1e-9, 3e-6, 1000)
    temps = [3000, 4000, 5000, 6000]
    
    plt.figure(figsize=(10, 6))
    for T in temps:
        intensity = planck_law(wavelengths, T)
        plt.plot(wavelengths * 1e9, intensity, label=f'{T} K')
        
    plt.title("Blackbody Radiation (Planck's Law)")
    plt.xlabel("Wavelength (nm)")
    plt.ylabel("Spectral Radiance")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    blackbody()
```
