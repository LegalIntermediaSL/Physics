# Compendio Subtemático: Electromagnetismo Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    freqs = np.linspace(10, 1000, 500)
    omega = 2*np.pi*freqs
    R, L, C = 10.0, 0.1, 1e-4
    Z = np.sqrt(R**2 + (omega*L - 1/(omega*C))**2)
    I0 = 10.0 / Z
    plt.plot(freqs, I0)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Current Amplitude (A)')
    plt.title("17. AC Series RLC Resonance")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
