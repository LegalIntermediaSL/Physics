# Compendio Subtemático: Mecanica Clasica Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # v_escape = sqrt(2GM/r)
    r = np.linspace(6.371e6, 2e7, 100)
    G = 6.674e-11
    M = 5.972e24
    v_esc = np.sqrt(2 * G * M / r)
    
    plt.plot(r, v_esc)
    plt.title('Escape Velocity vs Distance from Earth Center')
    plt.xlabel('Distance r (m)')
    plt.ylabel('Escape Velocity (m/s)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
