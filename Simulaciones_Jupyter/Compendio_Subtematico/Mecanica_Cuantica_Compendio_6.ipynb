# Compendio Subtemático: Mecanica Cuantica Compendio 6

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

u = np.linspace(0.01, 10, 1000)
v0 = 10.0

v_even = u * np.tan(u)
v_odd = -u / np.tan(u)
circle = np.sqrt(np.maximum(v0**2 - u**2, 0))

plt.plot(u, v_even, label='u tan(u)')
plt.plot(u, v_odd, label='-u cot(u)')
plt.plot(u, circle, 'r--', label='Circle radius v0')
plt.ylim(0, v0+2)
plt.xlim(0, v0+1)
plt.title('Finite Well Graphical Solution')
plt.xlabel('u')
plt.legend()
plt.grid(True)
plt.show()
```
