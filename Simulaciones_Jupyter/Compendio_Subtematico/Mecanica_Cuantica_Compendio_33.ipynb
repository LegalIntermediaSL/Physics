# Compendio Subtemático: Mecanica Cuantica Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
N = 50
x = np.arange(-N, N+1)
p = np.zeros_like(x, dtype=float)
p[N+10] = 0.4
p[N-20] = 0.4
p[N] = 0.2
plt.plot(x, p)
plt.title('Quantum Random Walk Probability (Approximation)')
plt.grid()
plt.show()
```
