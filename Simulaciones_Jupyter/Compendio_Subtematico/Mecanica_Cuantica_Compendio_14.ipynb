# Compendio Subtemático: Mecanica Cuantica Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
psi = np.array([1, 1j]) / np.sqrt(2)
rho = np.outer(psi, np.conj(psi))
plt.imshow(np.abs(rho), cmap='Blues')
plt.colorbar()
plt.title('Absolute value of Density Matrix (Pure State)')
plt.xticks([0,1], ['|0>', '|1>'])
plt.yticks([0,1], ['|0>', '|1>'])
plt.show()
```
