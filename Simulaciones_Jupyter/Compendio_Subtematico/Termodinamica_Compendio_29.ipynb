# Compendio Subtemático: Termodinamica Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Stirling Cycle P-V
v_iso_1 = np.linspace(1, 4, 50)
p_iso_1 = 10 / v_iso_1

v_isoch_1 = np.full(50, 4)
p_isoch_1 = np.linspace(p_iso_1[-1], 2, 50)

v_iso_2 = np.linspace(4, 1, 50)
p_iso_2 = 2 * (4/v_iso_2) # wait, 8 / v_iso_2

v_isoch_2 = np.full(50, 1)
p_isoch_2 = np.linspace(8, 10, 50)

plt.plot(v_iso_1, p_iso_1, 'r', label='Isothermal Exp')
plt.plot(v_isoch_1, p_isoch_1, 'g', label='Isochoric Cooling')
plt.plot(v_iso_2, 8/v_iso_2, 'b', label='Isothermal Comp')
plt.plot(v_isoch_2, p_isoch_2, 'y', label='Isochoric Heating')

plt.title('Stirling Cycle')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.savefig('stirling.png')
```
