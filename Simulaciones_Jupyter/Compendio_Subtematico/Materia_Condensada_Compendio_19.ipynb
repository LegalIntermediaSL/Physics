# Compendio Subtemático: Materia Condensada Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Quantum Hall Effect / Landau Levels
B = np.linspace(0.1, 5, 200) # Magnetic field

plt.figure(figsize=(8, 5))
for n in range(5):
    E_n = (n + 0.5) * B # Energy proportional to B (ignoring constants hbar * e / m)
    plt.plot(B, E_n, lw=2, label=f'n={n}')

plt.title("Landau Levels in a Magnetic Field")
plt.xlabel("Magnetic Field B")
plt.ylabel("Energy E_n")
plt.legend()
plt.grid(True)
plt.savefig("landau_levels.png")
```
