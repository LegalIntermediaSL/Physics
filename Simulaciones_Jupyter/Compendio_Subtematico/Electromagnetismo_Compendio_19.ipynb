# Compendio Subtemático: Electromagnetismo Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    x, y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
    d = 2.0
    r1_2 = (x-d)**2 + y**2
    r2_2 = (x+d)**2 + y**2
    r1_2[r1_2 == 0] = 1e-10
    r2_2[r2_2 == 0] = 1e-10
    Bx = -y/r1_2 - y/r2_2
    By = (x-d)/r1_2 + (x+d)/r2_2
    plt.quiver(x, y, Bx, By, color='cyan')
    plt.title("19. Magnetic Field of Two Parallel Wires (Same Current)")
    plt.show()

if __name__ == "__main__":
    main()
```
