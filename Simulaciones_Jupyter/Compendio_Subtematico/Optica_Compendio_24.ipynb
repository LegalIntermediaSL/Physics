# Compendio Subtemático: Optica Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def sph_ab():
    h = np.linspace(0, 5, 100)
    dL = 0.1 * h**2
    plt.plot(h, dL)
    plt.title("Spherical Aberration")
    plt.xlabel('Ray Height')
    plt.ylabel('Longitudinal Focal Shift')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    sph_ab()
```
