# Compendio Subtemático: Mecanica Clasica Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    a = 1.5e11 # semi-major axis of Earth
    e = 0.0167 # eccentricity
    theta = np.linspace(0, 2*np.pi, 100)
    r = a * (1 - e**2) / (1 + e * np.cos(theta))
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    
    plt.plot(x, y)
    plt.scatter(0, 0, color='yellow', s=100, label='Sun')
    plt.title('Kepler First Law: Elliptical Orbit')
    plt.xlabel('X position (m)')
    plt.ylabel('Y position (m)')
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
