# Compendio Subtemático: Electromagnetismo Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 5, 100)
    R = 1.0
    C = 1.0
    tau = R*C
    V0 = 5.0
    Vc = V0 * np.exp(-t/tau)
    plt.plot(t, Vc, label='Capacitor Voltage')
    plt.xlabel('Time (s)')
    plt.ylabel('Voltage (V)')
    plt.title("8. RC Circuit Discharging")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
