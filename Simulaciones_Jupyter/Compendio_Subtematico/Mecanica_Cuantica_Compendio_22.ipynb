# Compendio Subtemático: Mecanica Cuantica Compendio 22

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(1, 10)
E0 = n**2
dE = 0.5 * np.ones_like(n)
plt.plot(n, E0, 'o-', label='Unperturbed E')
plt.plot(n, E0 + dE, 's--', label='Perturbed E (1st order)')
plt.title('Perturbation Theory (Slant Potential)')
plt.legend()
plt.grid()
plt.show()
```
