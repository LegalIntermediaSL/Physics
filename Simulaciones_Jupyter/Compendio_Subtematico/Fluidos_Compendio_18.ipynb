# Compendio Subtemático: Fluidos Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def hydraulic_jump():
    y1 = 1.0 # m
    F1 = np.linspace(1.1, 5, 100) # Froude number
    y2 = (y1 / 2) * (np.sqrt(1 + 8 * F1**2) - 1)
    plt.plot(F1, y2)
    plt.xlabel('Froude Number (Fr1)')
    plt.ylabel('Downstream Depth y2 (m)')
    plt.title('Hydraulic Jump')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    hydraulic_jump()
```
