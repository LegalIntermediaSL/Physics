# Compendio Subtemático: Optica Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def double_slit():
    theta = np.linspace(-0.1, 0.1, 500)
    lam, d = 500e-9, 0.1e-3
    I = np.cos(np.pi*d*np.sin(theta)/lam)**2
    plt.plot(theta, I)
    plt.title("Young's Double Slit Interference")
    plt.xlabel('Angle (rad)')
    plt.ylabel('Intensity')
    plt.grid()
    plt.show()

if __name__ == '__main__': 
    double_slit()
```
