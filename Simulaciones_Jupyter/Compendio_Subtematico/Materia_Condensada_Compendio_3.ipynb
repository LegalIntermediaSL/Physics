# Compendio Subtemático: Materia Condensada Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def debye_cv(T, TD):
    if T == 0:
        return 0
    # C_v proportional to (T/TD)^3 \int_0^{TD/T} x^4 e^x / (e^x - 1)^2 dx
    def integrand(x):
        return (x**4 * np.exp(x)) / (np.exp(x) - 1)**2
    
    integral, _ = quad(integrand, 1e-8, TD / T)
    return 9 * (T / TD)**3 * integral

T_vals = np.linspace(1, 500, 100)
TD = 300 # Debye temperature in K
cv_vals = [debye_cv(T, TD) for T in T_vals]

plt.figure(figsize=(8, 5))
plt.plot(T_vals, cv_vals, 'k-', lw=2)
plt.title("Debye Model: Specific Heat vs Temperature")
plt.xlabel("Temperature (K)")
plt.ylabel("C_V / N k_B")
plt.axvline(TD, color='r', linestyle='--', label="Debye Temperature")
plt.legend()
plt.grid(True)
plt.savefig("debye_model.png")
```
