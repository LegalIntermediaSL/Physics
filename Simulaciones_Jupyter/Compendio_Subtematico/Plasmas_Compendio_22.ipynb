# Compendio Subtemático: Plasmas Compendio 22

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def saha_equation():
    T = np.logspace(3, 5, 100) # Kelvin
    chi = 13.6 * 1.6e-19 # Hydrogen ionization energy
    k = 1.38e-23
    n = 1e20 # Density
    frac = np.exp(-chi / (k * T)) * T**1.5
    frac = frac / (frac + n*1e-20)
    plt.plot(T, frac)
    plt.xscale('log')
    plt.title("Saha Ionization Fraction (Toy)")
    plt.grid(True); plt.savefig("saha.png")
if __name__ == "__main__": saha_equation()
```
