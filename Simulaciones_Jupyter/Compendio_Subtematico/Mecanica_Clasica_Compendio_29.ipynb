# Compendio Subtemático: Mecanica Clasica Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    # v = sqrt(T/mu) for a wave on a string
    mu = 0.05
    T = np.linspace(10, 100, 100)
    v = np.sqrt(T / mu)
    
    plt.plot(T, v)
    plt.title('Wave Speed on a String vs Tension')
    plt.xlabel('Tension T (N)')
    plt.ylabel('Wave Speed v (m/s)')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
