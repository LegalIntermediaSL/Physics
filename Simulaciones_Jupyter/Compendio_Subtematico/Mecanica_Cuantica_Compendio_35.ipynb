# Compendio Subtemático: Mecanica Cuantica Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
E = np.linspace(0.01, 2, 200)
T_vals = [0.2, 0.5, 1.0]
for T in T_vals:
    f = 1 / (np.exp(E / T) - 1)
    plt.plot(E, f, label=f'T={T}')
plt.title('Bose-Einstein Distribution')
plt.ylim(0, 10)
plt.legend()
plt.grid()
plt.show()
```
