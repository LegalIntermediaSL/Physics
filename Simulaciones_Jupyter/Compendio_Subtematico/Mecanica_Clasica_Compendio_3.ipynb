# Compendio Subtemático: Mecanica Clasica Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def main():
    t = np.linspace(0, 2*np.pi, 100)
    r = 5
    x = r * np.cos(t)
    y = r * np.sin(t)
    plt.plot(x, y)
    plt.title('Uniform Circular Motion')
    plt.xlabel('X (m)')
    plt.ylabel('Y (m)')
    plt.gca().set_aspect('equal')
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
