# Compendio Subtemático: Plasmas Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def two_stream():
    v = np.linspace(-5, 5, 100)
    f = np.exp(-(v-2)**2) + np.exp(-(v+2)**2)
    plt.plot(v, f)
    plt.title("Two-Stream Instability Dist")
    plt.grid(True); plt.savefig("two_stream.png")
if __name__ == "__main__": two_stream()
```
