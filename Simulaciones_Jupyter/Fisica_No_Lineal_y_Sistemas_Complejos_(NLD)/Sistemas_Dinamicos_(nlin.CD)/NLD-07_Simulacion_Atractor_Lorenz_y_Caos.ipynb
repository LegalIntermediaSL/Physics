# [NLD-07] Física del Caos: Atractor de Lorenz

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Dinámica No Lineal: El Atractor Extraño de Lorenz (1963)
# Simulamos un modelo meteorológico hiper-simplificado donde descubrimos
# el Caos Determinista: El mítico "Efecto Mariposa".

# Parámetros del Sistema de Lorenz
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

def lorenz_deriv(t, state):
    x, y, z = state
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return [dx, dy, dz]

# Simulamos DOS estados iniciales separados por una millonésima (1e-6)
# Representan dos universos atmosféricos casi idénticos.
state0_A = [1.0, 1.0, 1.0]
state0_B = [1.0 + 1e-6, 1.0, 1.0]

t_span = (0, 30)
t_eval = np.linspace(*t_span, 10000)

sol_A = solve_ivp(lorenz_deriv, t_span, state0_A, t_eval=t_eval)
sol_B = solve_ivp(lorenz_deriv, t_span, state0_B, t_eval=t_eval)

fig = plt.figure(figsize=(12, 6))

# Subplot 1: El Atractor en el Espacio de Fases 3D (La Mariposa)
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot(sol_A.y[0], sol_A.y[1], sol_A.y[2], lw=0.5, color='blue', alpha=0.8)
ax1.set_title("Espacio de Fases 3D (Atractor)", fontsize=14)
ax1.set_xlabel("X (Convección)")
ax1.set_ylabel("Y (Temp Horiz)")
ax1.set_zlabel("Z (Temp Vert)")

# Subplot 2: La Divergencia Exponencial en el Tiempo (Efecto Mariposa)
ax2 = fig.add_subplot(122)
ax2.plot(sol_A.t, sol_A.y[0], lw=1, color='blue', label="Universo A (x)")
ax2.plot(sol_B.t, sol_B.y[0], lw=1, color='red', linestyle='--', label="Universo B (x + 1e-6)")
ax2.set_title("Evolución de X en el Tiempo (Sensibilidad)", fontsize=14)
ax2.set_xlabel("Tiempo")
ax2.set_ylabel("Variable X")
ax2.legend()

plt.tight_layout()
plt.show()

# FÍSICA Y CAOS: En el gráfico 3D vemos el Atractor Extraño. En el gráfico 2D vemos
# cómo ambos universos (Azul y Rojo) son idénticos hasta T=15. Después de ese horizonte,
# la pequeñísima diferencia microscópica crece exponencialmente y los climas se separan
# por completo. Esto hace imposible predecir el clima meteorológico a más de 2 semanas vista.

```
