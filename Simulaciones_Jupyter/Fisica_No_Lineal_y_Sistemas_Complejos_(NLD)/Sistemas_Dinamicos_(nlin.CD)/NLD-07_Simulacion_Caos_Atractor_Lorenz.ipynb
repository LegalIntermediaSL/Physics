# [NLD-07] Caos Determinista y el Efecto Mariposa

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from mpl_toolkits.mplot3d import Axes3D

# Sistema de Lorenz: Modelo simplificado de convección atmosférica
# Famoso por acuñar el "Efecto Mariposa"

sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

def lorenz_system(t, state):
    x, y, z = state
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return [dx_dt, dy_dt, dz_dt]

# Integramos dos trayectorias con una diferencia minúscula en la condición inicial (1e-5)
t_span = (0, 40)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

initial_state_1 = [1.0, 1.0, 1.0]
initial_state_2 = [1.0 + 1e-5, 1.0, 1.0]

sol1 = solve_ivp(lorenz_system, t_span, initial_state_1, t_eval=t_eval)
sol2 = solve_ivp(lorenz_system, t_span, initial_state_2, t_eval=t_eval)

fig = plt.figure(figsize=(12, 6))

# Gráfico 3D del Atractor Extraño
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot(sol1.y[0], sol1.y[1], sol1.y[2], lw=0.5, color='teal')
ax1.set_title("Atractor Extraño de Lorenz (Espacio de Fases)")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Z")

# Gráfico de Divergencia Exponencial (Sensibilidad a Condiciones Iniciales)
ax2 = fig.add_subplot(122)
distance = np.sqrt((sol1.y[0]-sol2.y[0])**2 + (sol1.y[1]-sol2.y[1])**2 + (sol1.y[2]-sol2.y[2])**2)
ax2.plot(sol1.t, distance, 'darkred', lw=1.5)
ax2.set_yscale('log')
ax2.set_title("Divergencia de Trayectorias (Efecto Mariposa)")
ax2.set_xlabel("Tiempo t")
ax2.set_ylabel("Distancia Espacial $\Delta$ (log)")
ax2.grid(True)

plt.tight_layout()
plt.show()

```
