# [NLD-06] Embudo Energético en Plegamiento

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación del Energy Landscape Funnel (Plegamiento de Proteínas)
# La coordenada de reacción (Q) es la fracción de contactos nativos (0 a 1).
# Generamos un paisaje energético rugoso V(Q).

Q = np.linspace(0.01, 1, 500)

# Potencial macroscópico liso: Embudo termodinámico (Entalpía vs Entropía)
# La energía desciende a medida que Q se acerca a 1.
V_funnel = -10.0 * Q**2 - 5.0 * Q

# Rugosidad local: Mínimos locales / estados intermedios trampa
roughness = 1.2 * np.sin(50 * Q) * (1 - Q)

# Energía Libre Total
F = V_funnel + roughness

plt.figure(figsize=(10, 6))
plt.plot(Q, F, 'darkorange', lw=2)
plt.fill_between(Q, F, max(F), color='navajowhite', alpha=0.3)

# Pintamos el estado desplegado y el nativo
plt.scatter([Q[0]], [F[0]], color='blue', s=100, zorder=5, label="Estado Desplegado (Random Coil)")
plt.scatter([Q[-1]], [F[-1]], color='red', s=150, zorder=5, label="Estado Nativo (Folded)")

plt.title("Embudo Energético (Energy Landscape) de una Proteína")
plt.xlabel("Fracción de Contactos Nativos (Q)")
plt.ylabel("Energía Libre (F)")
plt.legend()
plt.grid(True)
plt.show()

```
