# Oscilaciones Acústicas de Bariones (BAO)

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = np.logspace(-3, 0, 500)
P_k = k**1.2 * np.exp(-k/0.1) * (1 + 0.05 * np.sin(2*np.pi*k/0.05))

plt.loglog(k, P_k)
plt.title('Espectro de Potencias de Materia con wiggles BAO')
plt.xlabel('Número de onda $k$ (h/Mpc)')
plt.ylabel('Espectro $P(k)$')
plt.grid(True)
plt.show()
```
