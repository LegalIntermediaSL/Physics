# Compendio Subtemático: Termodinamica Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(100, 500, 100)
V_vals = [1, 2, 5]
R = 8.314
n = 1

for V in V_vals:
    P = n * R * T / V
    plt.plot(T, P, label=f'V = {V} m^3')

plt.title("Gay-Lussac's Law: Isochoric Process")
plt.xlabel('Temperature (K)')
plt.ylabel('Pressure (Pa)')
plt.legend()
plt.grid(True)
plt.savefig('gay_lussac.png')
```
