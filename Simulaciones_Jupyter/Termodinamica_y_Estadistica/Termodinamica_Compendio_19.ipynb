# Compendio Subtemático: Termodinamica Compendio 19

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Joule-Thomson coefficient sign change approx
T = np.linspace(100, 800, 100)
P = 400 - (T - 400)**2 / 400

plt.plot(T, P, 'b-', label='Inversion Curve')
plt.fill_between(T, 0, P, where=(P>0), alpha=0.3, color='blue', label='Cooling (μ>0)')
plt.fill_between(T, P, 500, alpha=0.3, color='red', label='Heating (μ<0)')

plt.title('Joule-Thomson Inversion Curve')
plt.xlabel('Temperature (K)')
plt.ylabel('Pressure (atm)')
plt.ylim(0, 500)
plt.legend()
plt.grid(True)
plt.savefig('joule_thomson.png')
```
