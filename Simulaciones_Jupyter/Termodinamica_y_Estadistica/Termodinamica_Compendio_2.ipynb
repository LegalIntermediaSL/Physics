# Compendio Subtemático: Termodinamica Compendio 2

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(100, 500, 100)
P_vals = [101325, 202650]
R = 8.314
n = 1

for P in P_vals:
    V = n * R * T / P
    plt.plot(T, V, label=f'P = {P} Pa')

plt.title("Charles's Law: Isobaric Expansion")
plt.xlabel('Temperature (K)')
plt.ylabel('Volume (m^3)')
plt.legend()
plt.grid(True)
plt.savefig('charles_law.png')
```
