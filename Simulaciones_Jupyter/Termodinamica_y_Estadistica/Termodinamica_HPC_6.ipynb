# Supercomputación: Termodinamica HPC 6

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 128
L = 10.0
x = np.linspace(-L/2, L/2, N)
y = np.linspace(-L/2, L/2, N)
X, Y = np.meshgrid(x, y)
V = 0.5 * (X**2 + Y**2)
dt = -0.01j
g = 100.0

kx = np.fft.fftfreq(N, d=L/N) * 2 * np.pi
KX, KY = np.meshgrid(kx, kx)
K = 0.5 * (KX**2 + KY**2)

psi = np.exp(-(X**2 + Y**2))
psi /= np.linalg.norm(psi)

for _ in range(500):
    psi = np.exp(-dt * (V + g * np.abs(psi)**2)) * psi
    psi_k = np.fft.fft2(psi)
    psi_k = np.exp(-dt * K) * psi_k
    psi = np.fft.ifft2(psi_k)
    psi /= np.linalg.norm(psi)

plt.imshow(np.abs(psi)**2, extent=[-L/2, L/2, -L/2, L/2], cmap='magma')
plt.title('Bose-Einstein Condensate (Gross-Pitaevskii)')
plt.show()
```
