# Compendio Subtemático: Termodinamica Compendio 20

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

V = np.linspace(1, 10, 100)
P_iso = 10 / V
P_adia = 10 / V**1.4

plt.plot(V, P_iso, 'r', label='Isothermal (PV = const)')
plt.plot(V, P_adia, 'b', label='Adiabatic (PV^gamma = const)')

plt.title('Isothermal vs Adiabatic Expansion')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.savefig('iso_vs_adia.png')
```
