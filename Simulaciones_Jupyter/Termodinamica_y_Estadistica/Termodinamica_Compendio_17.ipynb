# Compendio Subtemático: Termodinamica Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

v = np.linspace(0.06, 0.5, 500)
T_vals = [273, 300, 350]
R = 0.08206
a = 3.59
b = 0.0427

for T in T_vals:
    P = (R * T) / (v - b) - a / (v**2)
    plt.plot(v, P, label=f'T = {T} K')

plt.title('Van der Waals Isotherms (CO2 approx)')
plt.xlabel('Molar Volume (L/mol)')
plt.ylabel('Pressure (atm)')
plt.ylim(0, 150)
plt.legend()
plt.grid(True)
plt.savefig('van_der_waals.png')
```
