# Supercomputación: Termodinamica HPC 4

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy_stable

n_steps = 50000
alpha, beta = 1.2, 0.0
x = levy_stable.rvs(alpha, beta, size=n_steps)
y = levy_stable.rvs(alpha, beta, size=n_steps)
path_x = np.cumsum(x)
path_y = np.cumsum(y)

plt.plot(path_x, path_y, lw=0.3, color='crimson')
plt.title(f"Levy Flight (Anomalous Diffusion), alpha={alpha}")
plt.axis('equal')
plt.show()
```
