# Compendio Subtemático: Termodinamica Compendio 31

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

P = np.linspace(1, 100000, 100) # Pa
T = 300 # K
k = 1.38e-23
d = 3e-10 # collision diameter (m)

lambda_mfp = (k * T) / (np.sqrt(2) * np.pi * d**2 * P)

plt.plot(P, lambda_mfp * 1e9, 'c')
plt.title('Mean Free Path vs Pressure')
plt.xlabel('Pressure (Pa)')
plt.ylabel('Mean Free Path (nm)')
plt.yscale('log')
plt.grid(True)
plt.savefig('mean_free_path.png')
```
