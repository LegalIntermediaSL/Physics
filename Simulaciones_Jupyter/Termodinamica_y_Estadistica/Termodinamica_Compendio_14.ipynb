# Compendio Subtemático: Termodinamica Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 0.1, 100) # 10 cm wall
k = 205 # Aluminum
A = 1
T_hot = 100
T_cold = 20

# Steady state 1D T(x) = T_hot - (T_hot - T_cold) * x / L
T = T_hot - (T_hot - T_cold) * (x / 0.1)

plt.plot(x, T, 'r')
plt.title("Fourier's Law: 1D Steady Conduction")
plt.xlabel('Position (m)')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.savefig('fourier.png')
```
