# Investigación Avanzada: Termodinamica Ising 2D

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def initialize_lattice(N):
    """ Initialize a 2D spin lattice of size N x N with random spins (+1 or -1) """
    return np.random.choice([-1, 1], size=(N, N))

def mcmc_step(lattice, beta, J=1.0):
    """ Perform one Monte Carlo sweep over the lattice using Metropolis-Hastings """
    N = lattice.shape[0]
    for _ in range(N * N):
        x = np.random.randint(0, N)
        y = np.random.randint(0, N)
        
        spin = lattice[x, y]
        # Periodic boundary conditions
        neighbors = lattice[(x+1)%N, y] + lattice[(x-1)%N, y] + \
                    lattice[x, (y+1)%N] + lattice[x, (y-1)%N]
        
        dE = 2 * J * spin * neighbors
        
        # Metropolis acceptance criterion
        if dE <= 0 or np.random.rand() < np.exp(-beta * dE):
            lattice[x, y] *= -1
            
    return lattice

def simulate_ising():
    N = 64
    temperatures = [1.5, 2.269, 3.5] # Below Tc, at Tc (approx 2.269), Above Tc
    steps = 1000
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    for i, T in enumerate(temperatures):
        beta = 1.0 / T
        lattice = initialize_lattice(N)
        
        # Equilibration
        for _ in range(steps):
            lattice = mcmc_step(lattice, beta)
            
        axes[i].imshow(lattice, cmap='coolwarm', interpolation='nearest')
        axes[i].set_title(f'Ising Model at T={T}')
        axes[i].axis('off')
        
    plt.tight_layout()
    plt.savefig('Ising_2D_Phase_Transition.png', dpi=300)
    print("Ising 2D simulation complete. Image saved.")

if __name__ == '__main__':
    simulate_ising()
```
