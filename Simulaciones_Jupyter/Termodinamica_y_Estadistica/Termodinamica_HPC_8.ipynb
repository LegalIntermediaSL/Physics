# Supercomputación: Termodinamica HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

N = 128
L = 100.0
x = np.linspace(0, L, N, endpoint=False)
k = np.fft.fftfreq(N, d=L/(2*np.pi*N))

def ks_rhs(t, u):
    u_k = np.fft.fft(u)
    lin_k = (k**2 - k**4) * u_k
    nonlin_k = -0.5 * 1j * k * np.fft.fft(u**2)
    return np.fft.ifft(lin_k + nonlin_k).real

u0 = np.cos(2*np.pi*x/L) * (1 + 0.1*np.sin(2*np.pi*x/L))
sol = solve_ivp(ks_rhs, [0, 50], u0, t_eval=np.linspace(0, 50, 200))

plt.imshow(sol.y, aspect='auto', cmap='jet', extent=[0, 50, 0, L])
plt.title('Kuramoto-Sivashinsky Equation (Spatiotemporal Chaos)')
plt.ylabel('x')
plt.xlabel('t')
plt.show()
```
