# Compendio Subtemático: Termodinamica Compendio 10

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Diesel Cycle
v_comp = np.linspace(15, 1, 50)
p_comp = 1 * (15/v_comp)**1.4

v_comb = np.linspace(1, 3, 50)
p_comb = np.full_like(v_comb, p_comp[-1])

v_exp = np.linspace(3, 15, 50)
p_exp = p_comb[-1] * (3/v_exp)**1.4

plt.plot(v_comp, p_comp, 'b', label='Compression')
plt.plot(v_comb, p_comb, 'r', label='Combustion (Isobaric)')
plt.plot(v_exp, p_exp, 'g', label='Expansion')
plt.plot([15, 15], [p_exp[-1], p_comp[0]], 'y', label='Heat Rejection')

plt.title('Diesel Cycle')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.savefig('diesel.png')
```
