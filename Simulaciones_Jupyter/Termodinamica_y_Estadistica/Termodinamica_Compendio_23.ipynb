# Compendio Subtemático: Termodinamica Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(273.15, 373.15, 100)
L = 40.65e3 # J/mol
R = 8.314
P0 = 101325
T0 = 373.15

P = P0 * np.exp(-L/R * (1/T - 1/T0))

plt.plot(T, P, 'b')
plt.title('Clausius-Clapeyron Phase Boundary (Water)')
plt.xlabel('Temperature (K)')
plt.ylabel('Vapor Pressure (Pa)')
plt.grid(True)
plt.savefig('clausius_clapeyron.png')
```
