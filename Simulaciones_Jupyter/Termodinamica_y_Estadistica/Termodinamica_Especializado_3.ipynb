# Laboratorio Especializado: Termodinamica Especializado 3

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def ising_step(lattice, beta):
    N = lattice.shape[0]
    for _ in range(N*N):
        i, j = np.random.randint(0, N, 2)
        S = lattice[i, j]
        nb_sum = lattice[(i+1)%N, j] + lattice[(i-1)%N, j] + lattice[i, (j+1)%N] + lattice[i, (j-1)%N]
        dE = 2 * S * nb_sum
        if dE < 0 or np.random.rand() < np.exp(-beta * dE):
            lattice[i, j] *= -1

N = 50
T = 2.0
beta = 1.0 / T
lattice = np.random.choice([-1, 1], size=(N, N))

for step in range(100):
    ising_step(lattice, beta)

plt.figure(figsize=(6,6))
plt.imshow(lattice, cmap='gray')
plt.title(f'Ising Model at T={T}')
plt.axis('off')
plt.savefig('ising_model.png')
```
