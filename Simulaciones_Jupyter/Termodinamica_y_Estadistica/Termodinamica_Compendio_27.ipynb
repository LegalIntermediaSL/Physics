# Compendio Subtemático: Termodinamica Compendio 27

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(10, 500, 100)
T_E = 200 # Einstein temperature
R = 8.314

x = T_E / T
C_v = 3 * R * (x**2 * np.exp(x)) / (np.exp(x) - 1)**2

plt.plot(T, C_v, 'g')
plt.axhline(3*R, color='k', linestyle='--', label='Dulong-Petit limit')
plt.title('Einstein Model for Specific Heat of Solids')
plt.xlabel('Temperature (K)')
plt.ylabel('C_v (J/mol K)')
plt.legend()
plt.grid(True)
plt.savefig('einstein_solid.png')
```
