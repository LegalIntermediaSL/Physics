# Compendio Subtemático: Termodinamica Compendio 13

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

v = np.linspace(0, 2000, 200)
m = 4.65e-26 # N2 molecule mass
k = 1.38e-23
T_vals = [300, 600, 900]

for T in T_vals:
    f = 4 * np.pi * (m / (2 * np.pi * k * T))**1.5 * v**2 * np.exp(-m * v**2 / (2 * k * T))
    plt.plot(v, f, label=f'T = {T} K')

plt.title('Maxwell-Boltzmann Velocity Distribution')
plt.xlabel('Velocity (m/s)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.savefig('maxwell_boltzmann.png')
```
