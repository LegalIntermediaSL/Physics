# Compendio Subtemático: Termodinamica Compendio 22

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(300, 1000, 100)
T0 = 300
C_p = 29.1 # diatomic ideal gas
H0 = 0

H = H0 + C_p * (T - T0)

plt.plot(T, H, 'g')
plt.title('Enthalpy vs Temperature (Ideal Gas)')
plt.xlabel('Temperature (K)')
plt.ylabel('Enthalpy (J/mol)')
plt.grid(True)
plt.savefig('enthalpy.png')
```
