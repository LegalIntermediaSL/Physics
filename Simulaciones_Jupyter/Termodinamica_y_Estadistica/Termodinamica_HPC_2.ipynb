# Supercomputación: Termodinamica HPC 2

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

Nx, Ny = 100, 50
tau = 0.6
nul = (tau - 0.5) / 3.0
w = np.array([4/9] + [1/9]*4 + [1/36]*4)
cx = np.array([0, 1, 0, -1, 0, 1, -1, -1, 1])
cy = np.array([0, 0, 1, 0, -1, 1, 1, -1, -1])

f = np.ones((9, Ny, Nx)) / 9.0
obstacle = np.zeros((Ny, Nx), dtype=bool)
obstacle[Ny//2-5:Ny//2+5, Nx//4-5:Nx//4+5] = True

for _ in range(500):
    rho = np.sum(f, axis=0)
    ux = np.sum(f * cx[:, np.newaxis, np.newaxis], axis=0) / rho
    uy = np.sum(f * cy[:, np.newaxis, np.newaxis], axis=0) / rho
    ux[:, 0] = 0.1
    for i in range(9):
        cu = 3 * (cx[i]*ux + cy[i]*uy)
        feq = rho * w[i] * (1 + cu + 0.5*cu**2 - 1.5*(ux**2 + uy**2))
        f[i] += -(f[i] - feq) / tau
    
    for i in range(9):
        f[i, obstacle] = f[[0, 3, 4, 1, 2, 7, 8, 5, 6][i], obstacle]
    
    for i in range(9):
        f[i] = np.roll(np.roll(f[i], cx[i], axis=1), cy[i], axis=0)

plt.imshow(np.sqrt(ux**2 + uy**2), cmap='viridis')
plt.title('Lattice Boltzmann flow past cylinder')
plt.show()
```
