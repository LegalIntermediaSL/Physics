# Compendio Subtemático: Termodinamica Compendio 11

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Brayton Cycle (Ideal)
v_comp = np.linspace(5, 1, 50)
p_comp = 1 * (5/v_comp)**1.4

v_comb = np.linspace(1, 2.5, 50)
p_comb = np.full_like(v_comb, p_comp[-1])

v_exp = np.linspace(2.5, 12.5, 50)
p_exp = p_comb[-1] * (2.5/v_exp)**1.4

plt.plot(v_comp, p_comp, 'b', label='Isentropic Comp')
plt.plot(v_comb, p_comb, 'r', label='Isobaric Heat Add')
plt.plot(v_exp, p_exp, 'g', label='Isentropic Exp')

plt.title('Brayton Cycle (Gas Turbine)')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.savefig('brayton.png')
```
