# Compendio Subtemático: Termodinamica Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(10, 1000, 100)
M_He = 4e-3 / 6.022e23
M_N2 = 28e-3 / 6.022e23
k = 1.38e-23

v_rms_He = np.sqrt(3 * k * T / M_He)
v_rms_N2 = np.sqrt(3 * k * T / M_N2)

plt.plot(T, v_rms_He, label='He')
plt.plot(T, v_rms_N2, label='N2')

plt.title('RMS Velocity vs Temperature')
plt.xlabel('Temperature (K)')
plt.ylabel('V_rms (m/s)')
plt.legend()
plt.grid(True)
plt.savefig('v_rms.png')
```
