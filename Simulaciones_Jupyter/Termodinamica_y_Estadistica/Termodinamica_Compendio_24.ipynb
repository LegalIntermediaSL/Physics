# Compendio Subtemático: Termodinamica Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(200, 400, 100)
H = -285.8e3 # J/mol
S = -163.3 # J/mol K

G = H - T * S

plt.plot(T, G/1000, 'r')
plt.title('Gibbs Free Energy vs Temperature')
plt.xlabel('Temperature (K)')
plt.ylabel('Gibbs Free Energy (kJ/mol)')
plt.grid(True)
plt.savefig('gibbs_free_energy.png')
```
