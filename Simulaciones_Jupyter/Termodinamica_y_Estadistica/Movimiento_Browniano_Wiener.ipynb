# Movimiento Browniano (Proceso de Wiener)

Simulación estocástica de trayectorias aleatorias en 2D.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

n_steps = 1000
dx = np.random.randn(n_steps)
dy = np.random.randn(n_steps)
x = np.cumsum(dx)
y = np.cumsum(dy)

plt.figure(figsize=(6,6))
plt.plot(x, y, lw=0.5, color='orange')
plt.title('Trayectoria de Partícula en Movimiento Browniano')
plt.axis('equal')
plt.grid(True)
plt.show()
```
