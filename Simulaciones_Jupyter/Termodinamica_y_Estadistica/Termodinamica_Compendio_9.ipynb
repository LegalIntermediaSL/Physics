# Compendio Subtemático: Termodinamica Compendio 9

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Otto Cycle P-V
v_comp = np.linspace(10, 1, 50)
p_comp = 1 * (10/v_comp)**1.4

v_exp = np.linspace(1, 10, 50)
p_exp = 50 * (1/v_exp)**1.4

plt.plot(v_comp, p_comp, 'b', label='Compression')
plt.plot([1, 1], [p_comp[-1], p_exp[0]], 'r', label='Combustion')
plt.plot(v_exp, p_exp, 'g', label='Expansion')
plt.plot([10, 10], [p_exp[-1], p_comp[0]], 'y', label='Heat Rejection')

plt.title('Otto Cycle')
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.savefig('otto.png')
```
