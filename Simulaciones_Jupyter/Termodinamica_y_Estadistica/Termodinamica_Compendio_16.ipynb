# Compendio Subtemático: Termodinamica Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

T = np.linspace(300, 1000, 100)
sigma = 5.67e-8
epsilon = 0.9

P = epsilon * sigma * T**4

plt.plot(T, P, 'k')
plt.title('Stefan-Boltzmann Law (Radiation Power)')
plt.xlabel('Temperature (K)')
plt.ylabel('Power Emitted (W/m^2)')
plt.grid(True)
plt.savefig('stefan_boltzmann.png')
```
