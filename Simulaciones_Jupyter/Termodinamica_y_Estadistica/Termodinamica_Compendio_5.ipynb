# Compendio Subtemático: Termodinamica Compendio 5

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(0.1, 5, 100)
T, P = 300, 101325
R = 8.314

V = n * R * T / P

plt.plot(n, V)
plt.title("Avogadro's Law")
plt.xlabel('Number of moles (n)')
plt.ylabel('Volume (m^3)')
plt.grid(True)
plt.savefig('avogadro.png')
```
