# Compendio Subtemático: Termodinamica Compendio 33

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 1D composite wall thermal resistance
L = np.array([0.1, 0.05, 0.2])
k = np.array([1.5, 0.04, 0.8])
R = L / k

T_hot, T_cold = 100, 20
Q = (T_hot - T_cold) / np.sum(R)

T_int1 = T_hot - Q * R[0]
T_int2 = T_int1 - Q * R[1]

x = [0, L[0], L[0]+L[1], L[0]+L[1]+L[2]]
T = [T_hot, T_int1, T_int2, T_cold]

plt.plot(x, T, 'mo-')
plt.title('Composite Wall Temp Profile (Thermal Resistance)')
plt.xlabel('Position x (m)')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.savefig('thermal_resistance.png')
```
