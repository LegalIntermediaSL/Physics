# Compendio Subtemático: Termodinamica Compendio 18

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

wl = np.linspace(100e-9, 3000e-9, 500)
T_vals = [3000, 4000, 5000]
h = 6.626e-34
c = 3.0e8
k = 1.38e-23

for T in T_vals:
    B = (2 * h * c**2) / (wl**5) / (np.exp(h * c / (wl * k * T)) - 1)
    plt.plot(wl * 1e9, B, label=f'T = {T} K')

plt.title("Planck's Law: Black Body Radiation")
plt.xlabel('Wavelength (nm)')
plt.ylabel('Spectral Radiance (W/m^3/sr)')
plt.legend()
plt.grid(True)
plt.savefig('plancks_law.png')
```
