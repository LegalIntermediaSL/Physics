# Constante Fuerte Running Coupling ($alpha_s$)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Q = np.logspace(0.5, 3, 100) # Momento transferido GeV
Lambda_QCD = 0.2 # GeV
beta_0 = 11 - (2/3)*6 # 6 sabores de quarks
alpha_s = 4 * np.pi / (beta_0 * np.log(Q**2 / Lambda_QCD**2))

plt.semilogx(Q, alpha_s, color='red', lw=2)
plt.title('Libertad Asintótica en Cromodinámica Cuántica (QCD)')
plt.xlabel('$Q$ (GeV)')
plt.ylabel('$\alpha_s(Q)$')
plt.grid(True)
plt.show()
```
