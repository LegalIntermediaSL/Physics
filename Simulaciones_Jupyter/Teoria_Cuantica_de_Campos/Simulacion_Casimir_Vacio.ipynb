# Efecto Casimir (Energía de Punto Cero)
Simulación interactiva avanzada.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fuerza de Casimir por unidad de área: F/A = - (pi^2 hbar c) / (240 a^4)
a = np.linspace(1e-9, 10e-9, 100) # Distancia en metros
hbar_c = 3.16e-26 # J m

F_A = - (np.pi**2 * hbar_c) / (240 * a**4)

plt.figure(figsize=(8, 5))
plt.plot(a * 1e9, F_A, 'r-')
plt.title('Efecto Casimir: Presión del Vacío Cuántico vs Distancia')
plt.xlabel('Separación de placas $a$ (nm)')
plt.ylabel('Presión Atractiva $F/A$ (Pa)')
plt.grid(True)
plt.show()
```
