# Anomalía del Triángulo (Adler-Bell-Jackiw) en decaimiento $\pi^0$

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m_pi = np.linspace(130, 140, 100) # MeV
f_pi = 93 # MeV
alpha = 1/137.036
Gamma = (alpha**2 * m_pi**3) / (64 * np.pi**3 * f_pi**2)

plt.plot(m_pi, Gamma * 1e6)
plt.title('Anchura de decaimiento $\\pi^0 \to 2\\gamma$ (Anomalía Quiral ABJ)')
plt.xlabel('Masa del Pión $m_{\\pi^0}$ (MeV)')
plt.ylabel('Tasa de Decaimiento $\\Gamma$ (eV)')
plt.grid(True)
plt.show()
```
