# Fórmula de Masas de Gell-Mann–Okubo (Bariones)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
isospin = [0.5, 1.0, 0.5, 0.0] # N, Sigma, Xi, Lambda
strangeness = [0, -1, -2, -1]
# Formula GMO paramétrica
m0, m1, m2 = 1115, -190, 40
mass = [m0 + m1*S + m2*(I*(I+1) - 0.25*S**2) for I, S in zip(isospin, strangeness)]
labels = ['N', '$\\Sigma$', '$\Xi$', '$\\Lambda$']

plt.bar(labels, mass, color=['blue', 'red', 'green', 'purple'])
plt.title('Espectro de Masas del Octeto Bariónico SU(3)')
plt.ylabel('Masa (MeV)')
plt.grid(axis='y')
plt.show()
```
