# Corrección de Schwinger al Momento Magnético ($g-2$)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha_inv = np.linspace(137.035, 137.037, 100)
alpha = 1 / alpha_inv
# Aproximación de Schwinger a un loop (alfa/2pi)
a_e = alpha / (2*np.pi)

plt.plot(alpha_inv, a_e * 1e3)
plt.title('Momento Magnético Anómalo del Electrón $a_e = (g-2)/2$')
plt.xlabel(r'Inversa de Constante de Estructura Fina $lpha^{-1}$')
plt.ylabel('$a_e \times 10^3$')
plt.grid(True)
plt.show()
```
