# Representación Matricial Truncada del Espacio de Fock

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10 # Nivel maximo del espacio de Fock
a = np.zeros((N,N))
for n in range(1, N):
    a[n-1, n] = np.sqrt(n)
a_dag = a.T
num_op = np.dot(a_dag, a)

plt.imshow(num_op, cmap='viridis')
plt.colorbar(label='Eigenvalue $n$')
plt.title('Matriz del Operador Número $\\hat{N} = \\hat{a}^\dagger \\hat{a}$ en Base de Fock')
plt.xlabel(r'$|mangle$'); plt.ylabel(r'$\\langle n|$')
plt.show()
```
