# Teoría de Campos Térmica: Restauración de Simetría

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
phi = np.linspace(-3, 3, 200)
lambda_ = 1.0
m2 = -1.0 # Taquiónico a T=0
T_vals = [0.0, 0.5, 1.0, 1.5]

plt.figure(figsize=(8,5))
for T in T_vals:
    m_eff2 = m2 + (lambda_ * T**2)/24
    V = 0.5*m_eff2*phi**2 + (lambda_/24)*phi**4
    plt.plot(phi, V, label=f'T = {T}')

plt.title('Restauración Térmica de la Simetría (Universo Temprano)')
plt.xlabel('Campo de Higgs $\\phi$')
plt.ylabel('Potencial Efectivo Termal $V(\\phi, T)$')
plt.legend(); plt.grid(True)
plt.show()
```
