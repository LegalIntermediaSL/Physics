# Fórmula de Klein-Nishina (Efecto Compton)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parametrizacion del angulo y el cociente epsilon = E_gamma / m_e c^2
theta = np.linspace(0, np.pi, 200)
epsilons = [0.1, 1.0, 10.0]

plt.figure(figsize=(8,6))
for eps in epsilons:
    ratio = 1 / (1 + eps*(1 - np.cos(theta)))
    # Klein-Nishina diferencial cross section d(sigma)/d(Omega)
    dsig = 0.5 * ratio**2 * (ratio + 1/ratio - np.sin(theta)**2)
    plt.plot(theta, dsig, label=f'$\epsilon = {eps}$')

plt.title('Sección Eficaz Diferencial de Klein-Nishina (QED)')
plt.xlabel('Ángulo de dispersión $\theta$ (rad)')
plt.ylabel(r'$d\sigma / d\Omega$ (Unidades relativas)')
plt.yscale('log')
plt.legend(); plt.grid(True)
plt.show()
```
