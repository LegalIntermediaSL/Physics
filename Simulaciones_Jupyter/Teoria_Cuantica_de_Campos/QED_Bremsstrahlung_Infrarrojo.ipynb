# Divergencia Infrarroja y Corte Térmico (Bremsstrahlung)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega = np.logspace(-5, 0, 200)
omega_cutoff = 1e-3
# Probabilidad diferencial de emision blanda ~ 1/omega
dP_domega = 1 / omega
dP_domega[omega < omega_cutoff] = 0 # Imposición de masa fotonica o corte IR

plt.loglog(omega, dP_domega, color='red')
plt.axvline(omega_cutoff, color='k', linestyle='--', label='Corte IR $\\mu$')
plt.title('Espectro Bremsstrahlung Blando (QED)')
plt.xlabel('Energía del Fotón $\omega$')
plt.ylabel('Probabilidad $dP / d\omega$')
plt.legend(); plt.grid(True)
plt.show()
```
