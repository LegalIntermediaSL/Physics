# Tasa de Producción de Pares de Schwinger

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
E_field = np.logspace(16, 19, 100) # V/m
E_c = 1.32e18 # Campo crítico de Schwinger
rate = (E_field**2) * np.exp(-np.pi * E_c / E_field)

plt.plot(E_field, rate, color='purple', lw=2)
plt.title('Creación de Pares en el Vacío (Efecto Schwinger)')
plt.xlabel('Intensidad del Campo Eléctrico $E$ (V/m)')
plt.ylabel('Tasa de Producción $\\Gamma$')
plt.xscale('log')
plt.axvline(E_c, color='gray', linestyle=':', label='$E_c$ (Crítico)')
plt.legend(); plt.grid(True)
plt.show()
```
