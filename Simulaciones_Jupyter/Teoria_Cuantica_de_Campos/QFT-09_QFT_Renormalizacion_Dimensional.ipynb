# Flujo RG de Callan-Symanzik

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = np.logspace(0, 3, 100) # Escala de energia
# Funciones Beta aproximadas
alpha_QED = 1/137.0 / (1 - (1/137.0)*(1/(3*np.pi))*np.log(mu))
alpha_QCD = 0.1 / (1 + 0.1*(7/(4*np.pi))*np.log(mu)) # Libertad asintotica

plt.semilogx(mu, alpha_QED, label='$\alpha_{QED}$ (Polo Landau)')
plt.semilogx(mu, alpha_QCD, label='$\alpha_s$ (QCD Libertad Asintótica)')
plt.title('Flujo del Grupo de Renormalización')
plt.xlabel('Escala de Energía $\\mu$ (GeV)')
plt.ylabel('Constante de Acoplamiento')
plt.legend(); plt.grid(True)
plt.show()
```
