# Anchuras de Decaimiento Electrodébil (Bosón Z)

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sin2_theta_W = np.linspace(0.2, 0.25, 50)
G_F = 1.166e-5 # GeV^-2
M_Z = 91.1876 # GeV
Gamma_0 = G_F * M_Z**3 / (24 * np.sqrt(2) * np.pi)
# Neutrinos (invisible)
Gamma_inv = 3 * Gamma_0
# Leptones cargados
Gamma_l = 3 * Gamma_0 * (( -0.5 + 2*sin2_theta_W )**2 + 0.25)

plt.plot(sin2_theta_W, np.ones_like(sin2_theta_W) * Gamma_inv, label='$\Gamma(Z \to \nu\bar{\nu})$')
plt.plot(sin2_theta_W, Gamma_l, label='$\Gamma(Z \to l^+ l^-)$')
plt.title('Anchuras de decaimiento del Bosón Z')
plt.xlabel('Ángulo de mezcla de Weinberg $\sin^2(\theta_W)$')
plt.ylabel('$\Gamma$ (GeV)')
plt.legend(); plt.grid(True)
plt.show()
```
