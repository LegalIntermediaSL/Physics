# Oscilación de Neutrinos y Matriz PMNS

Simulación cuántica de campos avanzada y rigurosa.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = np.linspace(0, 1000, 500) # km
E = 1.0 # GeV
Delta_m2 = 2.5e-3 # eV^2
theta = np.pi / 4 # Mezcla maximal bifásica simplificada

P_surv = 1 - np.sin(2*theta)**2 * np.sin(1.27 * Delta_m2 * L / E)**2

plt.plot(L, P_surv, color='darkcyan')
plt.title('Probabilidad de Supervivencia de Sabor $P(\nu_\\mu \to \nu_\\mu)$')
plt.xlabel('Distancia de línea base $L$ (km)')
plt.ylabel('Probabilidad')
plt.grid(True)
plt.show()
```
