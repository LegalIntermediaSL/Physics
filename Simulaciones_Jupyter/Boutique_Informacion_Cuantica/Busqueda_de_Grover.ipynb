# Algoritmo de Grover: Búsqueda Cuántica Cuadrática
El Algoritmo de Grover busca un elemento en una base de datos desordenada de tamaño $N$ en $\mathcal{O}(\sqrt{N})$ pasos, en comparación con los $\mathcal{O}(N)$ clásicos.

Utiliza la **Inversión sobre la Media** (Amplificación de Amplitud). Veremos iteración por iteración cómo la probabilidad del estado "marcado" (nuestro objetivo) crece chupando la probabilidad de todos los demás estados.

```python
import numpy as np
import matplotlib.pyplot as plt

# Tamaño de la base de datos N (N = 2^n)
n = 6
N = 2**n

# Estado inicial: Superposición uniforme
state = (1/np.sqrt(N)) * np.ones(N)

# Elemento objetivo (w) que buscamos
w = 42

# Número óptimo de iteraciones de Grover: ~ (pi/4) * sqrt(N)
num_iters = int(np.floor(np.pi/4 * np.sqrt(N)))

probs_w = []

# Algoritmo de Grover iterativo
for i in range(num_iters + 1):
    probs_w.append(np.abs(state[w])**2)
    
    # 1. Oráculo: Invierte la fase del estado objetivo |w>
    state[w] = -state[w]
    
    # 2. Difusión (Inversión sobre la Media)
    mean = np.mean(state)
    state = 2 * mean - state

# Visualización
plt.figure(figsize=(8, 5))
plt.plot(range(num_iters + 1), probs_w, marker='o', color='purple', lw=2)
plt.axhline(1.0, color='grey', linestyle='--')
plt.title(f"Amplificación de Amplitud de Grover (Buscando elemento {w} de {N})")
plt.xlabel("Número de Iteraciones")
plt.ylabel("Probabilidad de Medir el Objetivo P(w)")
plt.grid(True)
plt.show()

print(f"Probabilidad final tras {num_iters} iteraciones: {probs_w[-1]*100:.2f}%")
print("¡El estado correcto ha absorbido casi toda la probabilidad del universo!")
```
