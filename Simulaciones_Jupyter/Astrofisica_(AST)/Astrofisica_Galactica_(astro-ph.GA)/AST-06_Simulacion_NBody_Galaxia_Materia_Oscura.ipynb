# [AST-06] Astrofísica: Dinámica Galáctica N-Body

```python
import numpy as np
import matplotlib.pyplot as plt

# Astrofísica Dinámica Estelar: Simulador Leapfrog de Galaxia Espiral
# Modelo cinemático N-cuerpos en un Pozo de Potencial de Materia Oscura (NFW)

N_stars = 500
dt = 0.01
steps = 200

# Masa del Agujero Negro Central Supermasivo
M_SMBH = 1000.0 
# Factor del Halo de Materia Oscura (Mantiene la curva de rotación plana y no Kepleriana)
M_DarkMatter = 5000.0 

np.random.seed(42)
# Distribución radial log-normal de estrellas en el disco
r = np.random.lognormal(mean=1.5, sigma=0.5, size=N_stars)
theta = np.random.uniform(0, 2*np.pi, N_stars)

x = r * np.cos(theta)
y = r * np.sin(theta)

# Velocidad orbital para sostener equilibrio centrífugo
# v = sqrt(G*M(r)/r). En Kepler cae como 1/sqrt(r), pero añadimos la Materia Oscura (M_DM*r).
v_orbital = np.sqrt(M_SMBH / r + M_DarkMatter * (r / (r**2 + 10.0)))
vx = -v_orbital * np.sin(theta)
vy =  v_orbital * np.cos(theta)

plt.figure(figsize=(8, 8))
plt.style.use('dark_background')

# Animamos (ploteando estados inicial e intermedios como capas alfa)
for s in range(steps):
    # Aceleración gravitacional
    r_vec = np.sqrt(x**2 + y**2)
    # Fuerza central (Agujero Negro + Halo DM integrado)
    a_central = -(M_SMBH / r_vec**2 + M_DarkMatter * r_vec / (r_vec**2 + 10.0)**2) 
    
    ax = a_central * (x / r_vec)
    ay = a_central * (y / r_vec)
    
    # Integrador Leapfrog
    vx += ax * dt
    vy += ay * dt
    x += vx * dt
    y += vy * dt
    
    # Renderizado estroboscópico de brazos espirales
    if s % 50 == 0:
        plt.scatter(x, y, s=1, color='cyan', alpha=0.3 + 0.6*(s/steps))

plt.scatter([0], [0], color='white', s=20, label="Agujero Negro Central")
plt.title("Dinámica Galáctica N-Cuerpos embebida en Materia Oscura", color='white', fontsize=14)
plt.xlabel("X (Kpc)")
plt.ylabel("Y (Kpc)")
plt.legend()
plt.xlim(-25, 25)
plt.ylim(-25, 25)
plt.show()

# FÍSICA: Sin la inyección analítica de la gravedad del "Halo de Materia Oscura", 
# las estrellas de la periferia saldrían disparadas al espacio por fuerza centrífuga 
# o rotarían extremadamente lentas (Decaimiento de Kepler), despedazando el disco galáctico.

```
