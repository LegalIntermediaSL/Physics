# Fondo Estocástico de Ondas Gravitacionales

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.logspace(-9, -7, 100) # PTA band
Omega_gw = 1e-9 * (f/1e-8)**(2/3)

plt.loglog(f, Omega_gw, color='darkgreen', lw=2)
plt.title('Fondo Estocástico GWB (Pulsar Timing Arrays)')
plt.xlabel('Frecuencia $f$ (Hz)')
plt.ylabel('$\Omega_{GW}(f)$')
plt.grid(True)
plt.show()
```
