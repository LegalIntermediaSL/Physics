# Evolución Química y Problema de Enanas G

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Z = np.linspace(0.001, 0.02, 100)
N_Z_closed = 1 - np.exp(-Z/0.005)  # Modelo de caja cerrada simple

plt.plot(Z, N_Z_closed)
plt.title('Distribución de Metalicidad (Modelo Caja Cerrada)')
plt.xlabel('Metalicidad $Z$')
plt.ylabel('Fracción de Estrellas $N(<Z)$')
plt.grid(True)
plt.show()
```
