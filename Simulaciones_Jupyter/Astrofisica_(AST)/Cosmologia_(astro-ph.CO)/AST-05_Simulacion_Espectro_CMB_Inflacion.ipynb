# [AST-05] Cosmología de Precisión: Espectro CMB

```python
import numpy as np
import matplotlib.pyplot as plt

# Astrofísica Cosmológica: Espectro de Potencia del CMB
# Simulador simplificado de los picos acústicos de la radiación reliquia del Big Bang.
# (Emula el software cosmológico real CAMB / CLASS)

# Momento Multipolar L (inverso del tamaño angular en el cielo)
# L = 200 corresponde a ~1 grado angular
multipolos_l = np.arange(2, 2500)

# Modelo fenomenológico simplificado (oscilador armónico amortiguado para el plasma)
# Pico 1: Geometría del universo (Aplastamiento / Flatness)
# Pico 2: Densidad de bariones (Materia normal)
# Pico 3: Densidad de Materia Oscura
def CMB_power_spectrum(l, omega_m=0.3, omega_b=0.05, h=0.7):
    # Frecuencia acústica (escala del horizonte sónico)
    k = l / 220.0 
    
    # Amplitud decae por difusión térmica (Silk damping)
    damping = np.exp(-(l / 1200.0)**1.5)
    
    # Oscilaciones acústicas de Bariones (BAO)
    # Ajustamos la amplitud de los picos pares vs impares usando omega_b
    baryon_shift = 0.5 * omega_b / 0.05 
    
    # Modelo analítico (Aproximación de Hu)
    spectrum = (1.0 + 3.0 * np.cos(np.pi * k)**2 - baryon_shift * np.sin(np.pi * k)) * damping
    spectrum += 2.0 / (1.0 + (l/100)**2) # Efecto Sachs-Wolfe a bajos multipolos
    return spectrum * 1000

# Parámetros del Modelo Estándar (LCDM: Lambda Cold Dark Matter)
spectrum_lcdm = CMB_power_spectrum(multipolos_l)

plt.figure(figsize=(10, 6))
plt.plot(multipolos_l, spectrum_lcdm, 'k-', lw=3, label="Modelo LCDM (Misión Planck)")

# Añadimos un universo hipotético sin Materia Oscura, solo materia ordinaria
spectrum_no_dark = CMB_power_spectrum(multipolos_l, omega_m=0.05, omega_b=0.05)
plt.plot(multipolos_l, spectrum_no_dark, 'r--', lw=2, label="Universo sin Materia Oscura (Pico 3 destrozado)")

plt.title("Cosmología de Precisión: Espectro de Potencia del CMB", fontsize=14)
plt.xlabel("Momento Multipolar $\ell$ (Resolución angular más fina $\rightarrow$)")
plt.ylabel("Anisotropías de Temperatura $\Delta T^2$ [$\mu K^2$]")
plt.legend()
plt.grid(True)
plt.xlim(2, 2500)
plt.show()

# FÍSICA: En un universo sin materia oscura que comprima gravitacionalmente el plasma, 
# el tercer pico (L~800) desaparecería por completo. Las mediciones astronómicas 
# demuestran inequívocamente su existencia a través de estas huellas fósiles acústicas.

```
