# [AST-02] Astrofísica: Simulador Fotométrico de Tránsito

```python
import numpy as np
import matplotlib.pyplot as plt

# Astrofísica Exoplanetaria: El Método del Tránsito
# Simulador de la Curva de Luz de una Estrella cuando es eclipsada por un Exoplaneta.

# Parámetros (Unidades Solares / Jovianas)
R_star = 1.0       # Radio de la estrella (Ej. Sol)
R_planet = 0.1     # Radio del exoplaneta (Ej. Júpiter es ~0.1 R_sol)
a = 15.0           # Semieje mayor orbital escalado
period = 365.25    # Días
inclination = 90.0 # Grados (90º = Tránsito ecuatorial perfecto por el medio)

time = np.linspace(-10, 10, 1000) # Días alrededor del tránsito
# Posición proyectada X del planeta frente a la estrella
x_planet = a * np.sin(2 * np.pi * time / period)
y_planet = a * np.cos(inclination * np.pi/180) * np.cos(2 * np.pi * time / period)
distance = np.sqrt(x_planet**2 + y_planet**2)

# Modelo fotométrico puro geométrico + Oscurecimiento del Limbo (Limb Darkening)
def transit_flux(r_distance, r_star, r_planet, u_limb=0.6):
    flux = np.ones_like(r_distance)
    for i, d in enumerate(r_distance):
        if d < (r_star + r_planet):
            # Fracción geométrica de superposición (Simplificada)
            if d < (r_star - r_planet): overlap = np.pi * r_planet**2
            else: overlap = 0.5 * np.pi * r_planet**2 # Borde parcial
            
            # Oscurecimiento del limbo (La estrella es más tenue en los bordes)
            mu = np.sqrt(1 - min(1, (d/r_star)**2))
            intensity = 1.0 - u_limb * (1.0 - mu)
            
            # Área de la estrella (pi * R^2 = pi) normalizada
            flux[i] = 1.0 - (overlap / (np.pi * r_star**2)) * intensity
    return flux

light_curve = transit_flux(distance, R_star, R_planet)

plt.figure(figsize=(10, 5))
plt.plot(time, light_curve, 'k.-', markersize=3, label="Fotometría Simulada")
plt.axhline(1.0, color='r', linestyle='--', lw=1)

# Profundidad teórica del tránsito puro (R_p / R_s)^2
profundidad_max = 1.0 - (R_planet / R_star)**2
plt.axhline(profundidad_max, color='b', linestyle=':', lw=2, label=f"Profundidad Teórica ($\Delta F/F \approx {100*(1-profundidad_max):.1f}\%$)")

plt.title("Astrofísica: Detección de Exoplanetas por Tránsito", fontsize=14)
plt.xlabel("Tiempo desde el centro del eclipse (Días)")
plt.ylabel("Flujo Estelar Normalizado")
plt.legend()
plt.grid(True)
plt.show()

# FÍSICA: La forma de 'U' de la caída fotométrica confirma el radio del exoplaneta, 
# la redondez de los bordes confirma el grosor atmosférico estelar (Limb Darkening) 
# y la periodicidad del evento devela el Semieje Mayor vía las Leyes de Kepler.

```
