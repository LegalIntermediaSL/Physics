# Radiación Hawking y Evaporación de PBHs

Simulación avanzada de un fenómeno cosmológico/astrofísico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.logspace(10, 15, 100)
t_evap = M**3  # Tiempo de evaporacion proporcional al cubo de la masa

plt.loglog(M, t_evap, color='cyan')
plt.title('Tiempo de evaporación de Agujeros Negros Primordiales')
plt.xlabel('Masa (kg)')
plt.ylabel('Vida útil (s)')
plt.axhline(1e17, color='gray', linestyle='--', label='Edad del Universo')
plt.legend(); plt.grid(True)
plt.show()
```
