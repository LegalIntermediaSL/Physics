# Investigación Avanzada: Astrofisica Materia Oscura

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def dark_matter_halo():
    """Simulates Dark Matter Halo kinematics (Navarro-Frenk-White profile)."""
    r = np.linspace(0.1, 100, 500) # radius in kpc
    Rs = 10.0 # scale radius
    rho0 = 1e7 # central density parameter
    
    # NFW Density profile
    rho = rho0 / ((r/Rs) * (1 + r/Rs)**2)
    
    # Enclosed mass M(r) = 4*pi*rho0*Rs^3 * [ln(1+r/Rs) - (r/Rs)/(1+r/Rs)]
    M_r = 4 * np.pi * rho0 * Rs**3 * (np.log(1 + r/Rs) - (r/Rs)/(1 + r/Rs))
    
    # Circular velocity v(r) = sqrt(G*M(r)/r)
    G = 4.30091e-6 # kpc/M_sun (km/s)^2
    v_c = np.sqrt(G * M_r / r)
    
    fig, ax1 = plt.subplots(figsize=(10,6))
    ax1.plot(r, v_c, 'b-', label='Circular Velocity')
    ax1.set_xlabel('Radius $r$ [kpc]')
    ax1.set_ylabel('Velocity $v_c(r)$ [km/s]', color='b')
    ax1.tick_params('y', colors='b')
    
    ax2 = ax1.twinx()
    ax2.plot(r, rho, 'r--', label='NFW Density')
    ax2.set_ylabel('Density $\\rho(r)$ [$M_\odot/kpc^3$]', color='r')
    ax2.set_yscale('log')
    ax2.tick_params('y', colors='r')
    
    plt.title('Dark Matter Halo Kinematics (NFW Profile)')
    fig.tight_layout()
    plt.savefig('Dark_Matter_Halo.png')

if __name__ == '__main__':
    dark_matter_halo()
```
