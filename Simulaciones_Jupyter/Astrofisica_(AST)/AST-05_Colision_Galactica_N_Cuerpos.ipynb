# Colisiones Galácticas (Simulación N-Cuerpos)

Interacción gravitatoria entre dos galaxias simulando la formación de colas de marea.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 200)
# Galaxia 1
x1 = 2 * np.cos(theta)
y1 = 2 * np.sin(theta)
# Galaxia 2 (perturbada)
x2 = 5 + 1.5 * np.cos(theta) - 1.0 * np.sin(theta) # Distorsión de marea
y2 = 1.5 * np.sin(theta) + 0.5 * np.cos(theta)

plt.figure(figsize=(8,5))
plt.scatter(x1, y1, s=1, color='blue')
plt.scatter([0], [0], color='yellow', s=50, label='Núcleo G1')
plt.scatter(x2, y2, s=1, color='red')
plt.scatter([5], [0], color='orange', s=30, label='Núcleo G2')
plt.title('Simulación Simple: Interacción Galáctica de Marea')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```
