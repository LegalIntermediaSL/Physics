# [AST-10] Dinámica del Big Bang (Inflación Cósmica)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Simulación Cosmológica de la Inflación Cósmica (Slow-Roll del Inflatón)
# Ecuaciones Acopladas de Klein-Gordon (Campo Escalar phi) y Friedmann (Factor de Escala a).

# Unidades de Planck (8*pi*G = 1)
# Potencial del Inflatón: V(phi) = (1/2) * m^2 * phi^2
m = 1.0

def inflation_system(y, t):
    phi, phi_dot, a = y
    
    # Ecuación de Friedmann: H^2 = 1/3 * (0.5 * phi_dot^2 + V(phi))
    V = 0.5 * m**2 * phi**2
    H = np.sqrt((1.0 / 3.0) * (0.5 * phi_dot**2 + V))
    
    # Ecuación de Klein-Gordon: phi_ddot + 3H*phi_dot + V'(phi) = 0
    dV_dphi = m**2 * phi
    phi_ddot = -3 * H * phi_dot - dV_dphi
    
    # Factor de escala (Expansión Exponencial a(t))
    a_dot = a * H
    
    return [phi_dot, phi_ddot, a_dot]

# Condiciones iniciales (Universo primordial, campo desplazado del mínimo)
phi_0 = 15.0 # Valor alto del campo inflatón
phi_dot_0 = -0.1 # Slow-Roll (caída muy lenta)
a_0 = 1.0 # Factor de escala inicial

time = np.linspace(0, 50, 1000)
sol = odeint(inflation_system, [phi_0, phi_dot_0, a_0], time)

phi_evol = sol[:, 0]
a_evol = sol[:, 2]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Gráfico del Campo Inflatón
ax1.plot(time, phi_evol, 'b-', lw=3)
ax1.set_title("Dinámica del Inflatón $\phi(t)$ (Aproximación Slow-Roll)")
ax1.set_ylabel("Valor del Campo $\phi$")
ax1.grid()

# Gráfico del Factor de Escala (Escala Logarítmica para ver la exponenciación)
ax2.plot(time, a_evol, 'r-', lw=3)
ax2.set_yscale('log')
ax2.set_title("Expansión Inflacionaria (Factor de Escala $a(t)$)")
ax2.set_xlabel("Tiempo Cósmico")
ax2.set_ylabel("Factor de Escala $a$ (Log)")
ax2.grid(True, which="both", ls="--")

plt.tight_layout()
plt.show()

```
