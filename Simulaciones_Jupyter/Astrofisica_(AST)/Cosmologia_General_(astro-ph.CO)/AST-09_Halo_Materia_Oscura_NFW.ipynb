# [AST-09] Perfiles de Densidad de Materia Oscura

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación del Perfil de Densidad de Materia Oscura (Navarro-Frenk-White NFW)
# rho(r) = rho_0 / [ (r/Rs) * (1 + r/Rs)^2 ]

r = np.logspace(-2, 2, 200) # Distancia desde el centro galáctico (kpc)
Rs = 10.0 # Radio de escala (kpc)
rho_0 = 1.0 # Densidad característica

def nfw_profile(r):
    return rho_0 / ((r/Rs) * (1 + r/Rs)**2)

# Comparativa con un núcleo pseudo-isotérmico (núcleo plano)
def cored_profile(r):
    return rho_0 / (1 + (r/Rs)**2)

rho_nfw = nfw_profile(r)
rho_core = cored_profile(r)

plt.figure(figsize=(9, 6))
plt.plot(r, rho_nfw, 'b-', lw=3, label='Perfil NFW (Materia Oscura Fría CDM)')
plt.plot(r, rho_core, 'g--', lw=2, label='Perfil Isocore (Dark Matter con Interacción)')

plt.xscale('log')
plt.yscale('log')
plt.title("Densidad del Halo de Materia Oscura")
plt.xlabel("Radio r (kpc)")
plt.ylabel("Densidad $\rho(r)$")
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()

```
