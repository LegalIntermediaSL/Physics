# Compendio Subtemático: Plasmas Compendio 3

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def grad_b_drift():
    q, m = 1.6e-19, 1.67e-27
    v0, r0 = np.array([1e5, 0, 0]), np.zeros(3)
    dt = 1e-10
    t = np.arange(0, 2e-6, dt)
    r, v = np.zeros((len(t), 3)), np.zeros((len(t), 3))
    r[0], v[0] = r0, v0
    for i in range(1, len(t)):
        B = np.array([0, 0, 1.0 + 100 * r[i-1, 1]]) # Gradient in y
        v[i] = v[i-1] + (q/m)*np.cross(v[i-1], B)*dt
        r[i] = r[i-1] + v[i]*dt
    plt.plot(r[:, 0], r[:, 1])
    plt.title("Grad-B Drift")
    plt.grid(True); plt.savefig("grad_b.png")

if __name__ == "__main__":
    grad_b_drift()
```
