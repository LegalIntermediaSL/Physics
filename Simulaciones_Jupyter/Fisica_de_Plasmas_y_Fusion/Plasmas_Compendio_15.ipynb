# Compendio Subtemático: Plasmas Compendio 15

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def landau_damping():
    t = np.linspace(0, 20, 200)
    gamma = 0.2
    E = np.cos(5*t) * np.exp(-gamma*t)
    plt.plot(t, E)
    plt.title("Landau Damping")
    plt.grid(True); plt.savefig("landau.png")
if __name__ == "__main__": landau_damping()
```
