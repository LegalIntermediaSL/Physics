# Compendio Subtemático: Plasmas Compendio 4

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def curvature_drift():
    q, m = 1.6e-19, 1.67e-27
    # Simple simulated curved trajectory
    theta = np.linspace(0, np.pi/2, 1000)
    R = 2.0
    x = R * np.cos(theta)
    y = R * np.sin(theta)
    drift = np.linspace(0, 0.1, 1000)
    plt.plot(x, y + drift, label='Particle Path')
    plt.title("Curvature Drift (Conceptual)")
    plt.grid(True); plt.savefig("curvature.png")

if __name__ == "__main__":
    curvature_drift()
```
