# Compendio Subtemático: Plasmas Compendio 7

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def debye_shielding():
    r = np.linspace(0.01, 10, 100)
    lambda_D = 1.0
    phi_coulomb = 1/r
    phi_debye = (1/r) * np.exp(-r/lambda_D)
    plt.plot(r, phi_coulomb, label='Coulomb')
    plt.plot(r, phi_debye, label='Debye')
    plt.legend()
    plt.title("Debye Shielding Potential")
    plt.grid(True); plt.savefig("debye.png")

if __name__ == "__main__":
    debye_shielding()
```
