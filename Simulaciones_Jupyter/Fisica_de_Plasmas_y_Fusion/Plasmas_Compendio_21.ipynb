# Compendio Subtemático: Plasmas Compendio 21

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def magnetic_reconnection():
    x = np.linspace(-1, 1, 50)
    y = np.linspace(-1, 1, 50)
    X, Y = np.meshgrid(x, y)
    Bx = Y
    By = X
    plt.streamplot(X, Y, Bx, By)
    plt.title("Magnetic Reconnection (X-point)")
    plt.grid(True); plt.savefig("reconnection.png")
if __name__ == "__main__": magnetic_reconnection()
```
