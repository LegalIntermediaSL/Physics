# Plasmas Galactica 36

Simulación matemática generada paramétricamente para el compendio galáctico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x * 36)
plt.plot(x, y)
plt.title('Plasma Simulation 36')
```
