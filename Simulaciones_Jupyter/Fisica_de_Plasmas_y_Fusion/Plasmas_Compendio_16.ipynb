# Compendio Subtemático: Plasmas Compendio 16

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def plasma_beta():
    r = np.linspace(0, 1, 100)
    B = 1.0 + 0.5*r
    n = 1e20 * (1 - r**2)
    T = 1000 * (1 - r**2)
    mu0 = 4*np.pi*1e-7
    beta = (n * 1.6e-19 * T) / (B**2 / (2*mu0))
    plt.plot(r, beta)
    plt.title("Plasma Beta")
    plt.grid(True); plt.savefig("beta.png")
if __name__ == "__main__": plasma_beta()
```
