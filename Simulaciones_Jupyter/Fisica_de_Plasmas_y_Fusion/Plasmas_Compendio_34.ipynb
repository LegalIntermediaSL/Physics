# Compendio Subtemático: Plasmas Compendio 34

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def bremsstrahlung():
    nu = np.logspace(14, 19, 100)
    T = 1e7 # 1 keV
    h = 6.626e-34
    k = 1.38e-23
    spectrum = np.exp(-h*nu / (k*T))
    plt.loglog(nu, spectrum)
    plt.title("Bremsstrahlung Spectrum (Toy)")
    plt.xlabel("Frequency"); plt.ylabel("Intensity")
    plt.grid(True); plt.savefig("bremsstrahlung.png")
if __name__ == "__main__": bremsstrahlung()
```
