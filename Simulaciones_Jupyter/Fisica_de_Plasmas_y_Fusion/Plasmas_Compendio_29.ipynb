# Compendio Subtemático: Plasmas Compendio 29

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def lawson_criterion():
    T = np.logspace(0, 3, 100) # keV
    ntau = 1e20 * T**(-2) * np.exp(20/T) # Toy curve
    plt.loglog(T, ntau)
    plt.title("Lawson Criterion curve")
    plt.xlabel("T (keV)")
    plt.ylabel("n*tau")
    plt.grid(True); plt.savefig("lawson.png")
if __name__ == "__main__": lawson_criterion()
```
