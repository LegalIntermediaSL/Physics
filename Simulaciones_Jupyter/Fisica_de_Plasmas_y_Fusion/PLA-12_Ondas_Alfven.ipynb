# Ondas de Alfvén (MHD)

Perturbaciones en las líneas de campo magnético acopladas al fluido iónico.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 200)
B0 = np.ones_like(x)
delta_B = 0.2 * np.sin(x)

plt.figure(figsize=(8,4))
plt.plot(x, B0, 'k--', label='Campo Base $B_0$')
plt.plot(x, B0 + delta_B, 'b-', label='Línea de Campo Perturbada')
plt.title('Onda Transversal de Alfvén en un Plasma')
plt.legend()
plt.grid()
plt.show()
```
