# Laboratorio Especializado: Plasmas Especializado 2

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def exb_drift(t, y):
    q, m = 1.0, 1.0
    E = np.array([0, 1.0, 0])
    B = np.array([0, 0, 1.0])
    v = y[3:]
    a = (q/m) * (E + np.cross(v, B))
    return np.concatenate((v, a))

if __name__ == "__main__":
    sol = solve_ivp(exb_drift, [0, 20], np.zeros(6), max_step=0.1)
    plt.plot(sol.y[0], sol.y[1])
    plt.title('ExB Drift Trajectory')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.savefig('exb_drift.png')
```
