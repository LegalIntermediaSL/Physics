# Compendio Subtemático: Plasmas Compendio 24

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def weibel_instability():
    k = np.linspace(0.1, 5, 100)
    gamma = k * np.exp(-k) # toy growth rate
    plt.plot(k, gamma)
    plt.title("Weibel Instability Growth Rate")
    plt.xlabel("k"); plt.ylabel("Gamma")
    plt.grid(True); plt.savefig("weibel.png")
if __name__ == "__main__": weibel_instability()
```
