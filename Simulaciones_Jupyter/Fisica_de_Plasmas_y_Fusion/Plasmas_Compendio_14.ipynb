# Compendio Subtemático: Plasmas Compendio 14

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def vlasov_equation():
    x = np.linspace(0, 2*np.pi, 50)
    v = np.linspace(-5, 5, 50)
    X, V = np.meshgrid(x, v)
    f = np.exp(-V**2) * (1 + 0.1*np.cos(X))
    plt.contourf(X, V, f)
    plt.title("Vlasov Phase Space")
    plt.grid(True); plt.savefig("vlasov.png")
if __name__ == "__main__": vlasov_equation()
```
