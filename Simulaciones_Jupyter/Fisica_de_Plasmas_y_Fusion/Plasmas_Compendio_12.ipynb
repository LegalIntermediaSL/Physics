# Compendio Subtemático: Plasmas Compendio 12

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def langmuir_waves():
    k = np.linspace(0, 1, 100)
    wp = 1.0
    vth = 0.1
    omega = np.sqrt(wp**2 + 3 * k**2 * vth**2)
    plt.plot(k, omega)
    plt.title("Langmuir Waves")
    plt.grid(True); plt.savefig("langmuir.png")
if __name__ == "__main__": langmuir_waves()
```
