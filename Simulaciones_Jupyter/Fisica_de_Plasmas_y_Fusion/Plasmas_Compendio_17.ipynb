# Compendio Subtemático: Plasmas Compendio 17

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def tokamak_equilibrium():
    r = np.linspace(0.1, 1, 100)
    q = 1.0 + r**2
    plt.plot(r, q)
    plt.title("Tokamak Safety Factor q(r)")
    plt.grid(True); plt.savefig("tokamak.png")
if __name__ == "__main__": tokamak_equilibrium()
```
