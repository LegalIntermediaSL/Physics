# Supercomputación: Plasmas HPC 8

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np, matplotlib.pyplot as plt
def reconnection():
    B = np.meshgrid(np.linspace(-1,1,20), np.linspace(-1,1,20))
    plt.streamplot(B[0], B[1], B[1], B[0]); plt.savefig("recon.png")
if __name__=="__main__": reconnection()
```
