# Laboratorio Especializado: Plasmas Especializado 1

Simulación computacional ultra-avanzada correspondiente a la literatura científica extendida (Nivel Experto).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def vlasov_poisson_step(f, dt, dx, dv):
    # Simplified Vlasov-Poisson solver step
    rho = np.sum(f, axis=1) * dv
    rho = rho - np.mean(rho)
    E = np.cumsum(rho) * dx
    # Advection
    f_new = np.roll(f, 1, axis=0)
    return f_new, E

if __name__ == "__main__":
    Nx, Nv = 64, 64
    x = np.linspace(0, 10, Nx)
    v = np.linspace(-5, 5, Nv)
    X, V = np.meshgrid(x, v, indexing='ij')
    f = np.exp(-V**2 / 2) * (1 + 0.1 * np.cos(X))
    f, E = vlasov_poisson_step(f, 0.01, x[1]-x[0], v[1]-v[0])
    plt.plot(x, E)
    plt.title('Vlasov-Poisson: Electric Field')
    plt.savefig('vlasov_poisson.png')
```
