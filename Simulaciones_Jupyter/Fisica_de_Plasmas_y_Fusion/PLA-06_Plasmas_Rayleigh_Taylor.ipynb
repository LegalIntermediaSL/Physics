# Investigación Avanzada: Plasmas Rayleigh Taylor

Simulación numérica de vanguardia correspondiente a los Problemas Abiertos y Fronteras de Investigación (Nivel Doctorado).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# A simplified 2D linear Rayleigh-Taylor instability simulation
# Incompressible, inviscid, two fluids of different densities
# Heavy fluid on top of light fluid

nx, ny = 128, 128
Lx, Ly = 1.0, 2.0
dx = Lx / nx
dy = Ly / ny
x = np.linspace(0, Lx, nx)
y = np.linspace(0, Ly, ny)
X, Y = np.meshgrid(x, y)

# Densities
rho1 = 1.0  # Light fluid (bottom)
rho2 = 2.0  # Heavy fluid (top)
g = 9.81    # Gravity
A = (rho2 - rho1) / (rho2 + rho1) # Atwood number

# Interface perturbation
k = 2 * np.pi / Lx
eta0 = 0.05 * np.cos(k * X)

# Density profile
rho = np.zeros((ny, nx))
for i in range(ny):
    for j in range(nx):
        if Y[i, j] < Ly/2 + eta0[i, j]:
            rho[i, j] = rho1
        else:
            rho[i, j] = rho2

# Linear growth rate gamma = sqrt(A * g * k)
gamma = np.sqrt(A * g * k)
t = 0.5 # time
eta_t = eta0 * np.cosh(gamma * t)

rho_t = np.zeros((ny, nx))
for i in range(ny):
    for j in range(nx):
        if Y[i, j] < Ly/2 + eta_t[i, j]:
            rho_t[i, j] = rho1
        else:
            rho_t[i, j] = rho2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
c1 = ax1.contourf(X, Y, rho, levels=50, cmap='Blues')
ax1.set_title('Initial State (t=0)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

c2 = ax2.contourf(X, Y, rho_t, levels=50, cmap='Blues')
ax2.set_title(f'Linear Growth (t={t:.2f})')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.colorbar(c1, ax=ax1, orientation='horizontal', label='Density')
plt.colorbar(c2, ax=ax2, orientation='horizontal', label='Density')
plt.tight_layout()
plt.savefig('rayleigh_taylor.png')
```
