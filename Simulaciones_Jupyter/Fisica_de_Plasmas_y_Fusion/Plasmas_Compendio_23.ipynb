# Compendio Subtemático: Plasmas Compendio 23

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def pic_1d():
    x = np.random.rand(100) * 10
    v = np.random.randn(100)
    plt.scatter(x, v, s=5)
    plt.title("PIC 1D Phase Space Initialization")
    plt.xlabel("x"); plt.ylabel("v")
    plt.grid(True); plt.savefig("pic1d.png")
if __name__ == "__main__": pic_1d()
```
