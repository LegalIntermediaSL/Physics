# Compendio Subtemático: Plasmas Compendio 8

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def plasma_oscillations():
    t = np.linspace(0, 10, 1000)
    wp = 2 * np.pi * 1.0 # 1 Hz
    x = np.cos(wp * t) * np.exp(-0.1 * t)
    plt.plot(t, x)
    plt.title("Plasma Oscillations")
    plt.xlabel("Time")
    plt.ylabel("Displacement")
    plt.grid(True); plt.savefig("plasma_osc.png")

if __name__ == "__main__":
    plasma_oscillations()
```
