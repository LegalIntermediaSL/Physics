# Supercomputación: Plasmas HPC 1

Simulación computacional de altísimo rendimiento (HPC) orientada a sistemas ultracomplejos.

```python
%matplotlib inline
import numpy as np, matplotlib.pyplot as plt
def pic():
    x, v = np.random.rand(1000)*10, np.random.randn(1000)
    for _ in range(50): x = (x + v*0.1)%10
    plt.hist(x); plt.savefig("pic.png")
if __name__=="__main__": pic()
```
