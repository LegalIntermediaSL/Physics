# Compendio Subtemático: Plasmas Compendio 35

Simulación computacional para el compendio a escala industrial de físicas avanzadas.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def synchrotron():
    nu = np.logspace(9, 15, 100)
    nu_c = 1e12
    x = nu / nu_c
    spectrum = x**(1/3) * np.exp(-x)
    plt.loglog(nu, spectrum)
    plt.title("Synchrotron Spectrum (Toy)")
    plt.xlabel("Frequency"); plt.ylabel("Intensity")
    plt.grid(True); plt.savefig("synchrotron.png")
if __name__ == "__main__": synchrotron()
```
